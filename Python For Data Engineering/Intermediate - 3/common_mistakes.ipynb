```python

# ============================================================
# 1. NAMING FILE SAME AS BUILT-IN MODULE
# ============================================================

# ❌ Problem:
# If you name your file like "math.py", "random.py", "pickle.py",
# Python will import YOUR file instead of the actual library.

# Example (DO NOT RUN this, just understand):
# File name: math.py
# ------------------
# import math
# print(math.sqrt(16))
#
# ❗ This will fail because Python imports YOUR math.py, not the real one.

print("\n[1] Naming file same as built-in module")
print("👉 Avoid naming files like: math.py, random.py, pickle.py")

# ✅ Solution:
# Use unique names like:
# my_math_utils.py, custom_random.py

```

```python

# ============================================================
# 2. OVERRIDING IMPORTED FUNCTIONS
# ============================================================

print("\n[2] Overriding imported functions")

from math import sqrt

print("sqrt(16) =", sqrt(16))  # Works fine

# ❌ Mistake: overriding function name
sqrt = 25  # Now sqrt is no longer a function!

try:
    print(sqrt(16))  # This will crash
except Exception as e:
    print("❌ Error after overriding sqrt:", e)

# ✅ Solution:
# Always use meaningful variable names
# Avoid using same name as imported functions

# Fix:
from math import sqrt as real_sqrt
print("✅ Correct usage:", real_sqrt(16))

```

```python

# ============================================================
# 3. MUTABLE DEFAULT ARGUMENTS
# ============================================================

print("\n[3] Mutable Default Arguments")

# ❌ Dangerous function
def add_item_bad(item, my_list=[]):
    my_list.append(item)
    return my_list

print("First call:", add_item_bad(1))
print("Second call:", add_item_bad(2))  # Unexpected!
print("Third call:", add_item_bad(3))

# ⚠️ Why?
# Default list is created ONLY ONCE and reused!

# ✅ Correct way
def add_item_good(item, my_list=None):
    if my_list is None:
        my_list = []
    my_list.append(item)
    return my_list

print("\n✅ Fixed version:")
print("First call:", add_item_good(1))
print("Second call:", add_item_good(2))
print("Third call:", add_item_good(3))

```

```python

# ============================================================
# 4. IMPROPER USE OF GENERATORS (ITERATORS)
# ============================================================

print("\n[4] Improper use of generators")

numbers = [1, 2, 3, 4, 5]

# filter() returns an iterator (generator-like)
filtered = filter(lambda x: x % 2 == 0, numbers)

print("First use:", list(filtered))  # Works

# ❌ Second use (generator exhausted)
print("Second use:", list(filtered))  # Empty!

# ⚠️ Why?
# Generators can be used ONLY ONCE

# ✅ Solution:
filtered = filter(lambda x: x % 2 == 0, numbers)
print("Reset generator:", list(filtered))

# OR store result once
filtered = list(filter(lambda x: x % 2 == 0, numbers))
print("Store result:", filtered)
print("Reuse safely:", filtered)

```

```python

# ============================================================
# 5. SHALLOW COPY vs DEEP COPY
# ============================================================

print("\n[5] Shallow Copy vs Deep Copy")

import copy

original = [[1, 2], [3, 4]]

# ❌ Assignment (reference copy)
ref_copy = original

ref_copy[0][0] = 999
print("After modifying ref_copy:")
print("Original:", original)  # Changed!

# Reset
original = [[1, 2], [3, 4]]

# ❌ Shallow copy
shallow = copy.copy(original)

shallow[0][0] = 888
print("\nAfter modifying shallow copy:")
print("Original:", original)  # Still affected!

# Reset
original = [[1, 2], [3, 4]]

# ✅ Deep copy
deep = copy.deepcopy(original)

deep[0][0] = 777
print("\nAfter modifying deep copy:")
print("Original:", original)  # Safe!
print("Deep Copy:", deep)
```

```python


# ============================================================
# FINAL SUMMARY
# ============================================================

print("\n========== SUMMARY ==========")
print("""
1. Never name your file like built-in modules
2. Don't overwrite imported functions
3. Avoid mutable default arguments
4. Generators can be used only once
5. Use deepcopy for nested structures
""")

# ============================================================
# END OF FILE
# ============================================================
```

```python
# ============================================================
# QUESTION:
# Write a Python program for the following requirements:
#
# If inputs are:
#   start = 2
#   target = 512
#
# Perform repeated multiplication by 2 like:
#   2 * 2 = 4
#   4 * 2 = 8
#   8 * 2 = 16
#   ...
#   until the result becomes 512
#
# Output:
#   result = 512
#   count = 8   (number of multiplications performed)
#
# Example:
#   2 → 4 → 8 → 16 → 32 → 64 → 128 → 256 → 512
#   Total steps = 8
# ============================================================


# ============================================================
# EXPLANATION:
#
# 1. Start from initial value (start = 2)
# 2. Keep multiplying the current value by 2
# 3. After each multiplication:
#       - Print the step
#       - Increase the count
# 4. Stop when the current value becomes equal to target
# 5. Finally print:
#       - Final result
#       - Total number of steps (count)
#
# NOTE:
# - We use a while loop because we don't know beforehand
#   how many times we need to multiply.
# - The loop continues until current < target
# ============================================================


# -------------------------------
# INPUT VALUES
# -------------------------------
start = 2
target = 512


# -------------------------------
# INITIALIZATION
# -------------------------------
current = start   # This will store the current value
count = 0         # This will count number of multiplications


# -------------------------------
# PROCESS (LOOP)
# -------------------------------
print("Steps:")

while current < target:
    # Multiply current value by 2
    next_value = current * 2
    
    # Print the current step
    print(f"{current} * 2 = {next_value}")
    
    # Update current value
    current = next_value
    
    # Increase count
    count += 1


# -------------------------------
# OUTPUT
# -------------------------------
print("\nFinal Result =", current)
print("Total Steps (Count) =", count)


# ============================================================
# DRY RUN (IMPORTANT FOR INTERVIEWS):
#
# Iteration 1: current = 2   → 2*2 = 4     → count = 1
# Iteration 2: current = 4   → 4*2 = 8     → count = 2
# Iteration 3: current = 8   → 8*2 = 16    → count = 3
# Iteration 4: current = 16  → 16*2 = 32   → count = 4
# Iteration 5: current = 32  → 32*2 = 64   → count = 5
# Iteration 6: current = 64  → 64*2 = 128  → count = 6
# Iteration 7: current = 128 → 128*2 = 256 → count = 7
# Iteration 8: current = 256 → 256*2 = 512 → count = 8
#
# Loop stops because current == target
#
# Final Output:
#   Result = 512
#   Count = 8
# ============================================================


# ============================================================
# TIME COMPLEXITY:
# O(log₂(target))
# Because we are doubling each time.
#
# SPACE COMPLEXITY:
# O(1) (constant space)
# ============================================================
```
