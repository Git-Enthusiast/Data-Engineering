````python

# =========================================================
# 1. BUILT-IN NAMESPACE
# =========================================================

print("\n===== 1. BUILT-IN NAMESPACE =====")

print("len('Rajan'):", len("Rajan"))
print("id(len):", id(len))
print("id(print):", id(print))

print("Some built-ins:", dir(__builtins__)[:10])

````

````python

# =========================================================
# 2. GLOBAL (MODULE) NAMESPACE
# =========================================================

print("\n===== 2. GLOBAL NAMESPACE =====")

g = 100
print("Global g:", g)
print("id(g):", id(g))

def access_global():
    print("\nInside function:")
    print("g value:", g)
    print("id(g):", id(g))  # same as global

access_global()

print("Globals keys sample:", list(globals().keys())[:10])

````

````python

# =========================================================
# 3. LOCAL NAMESPACE
# =========================================================

print("\n===== 3. LOCAL NAMESPACE =====")

def local_demo():
    x = 50
    y = 60
    print("Local x:", x, "id:", id(x))
    print("Local y:", y, "id:", id(y))
    print("locals():", locals())

local_demo()

# print(x) ❌ ERROR (not accessible)

````

````python

# =========================================================
# 4. ENCLOSED (NON-LOCAL) NAMESPACE
# =========================================================

print("\n===== 4. ENCLOSED NAMESPACE =====")

def outer():
    y = 200
    print("Outer y id:", id(y))

    def inner():
        print("Inner accessing y:", y)
        print("Inner y id:", id(y))  # SAME

    inner()

outer()

````

````python

# =========================================================
# 5. LEGB RULE
# =========================================================

print("\n===== 5. LEGB RULE =====")

x = "Global X"  # Here x ia global variable
print("Global x id:", id(x))

def outer_legb():
    x = "Enclosed X"    # Here x is nonlocal variable
    print("Outer x id:", id(x))

    def inner_legb():
        x = "Local X"   # Here x is local variable
        print("Inner x id:", id(x))

    inner_legb()
    print("After inner, outer x:", x)

outer_legb()
print("Final global x:", x)

"""
LEGB Search Order:
Local → Enclosed → Global → Built-in
"""

````

````python

# =========================================================
# 6. GLOBAL KEYWORD
# =========================================================

print("\n===== 6. GLOBAL KEYWORD =====")

a = 10
print("Before change, id(a):", id(a))

def change_global():
    global a
    a = 99
    print("Inside function, id(a):", id(a))

change_global()

print("After change, id(a):", id(a))

````

````python

# =========================================================
# 7. NONLOCAL KEYWORD
# =========================================================

print("\n===== 7. NONLOCAL KEYWORD =====")

def outer_nonlocal():
    b = 5
    print("Outer initial id(b):", id(b))

    def inner_nonlocal():
        nonlocal b
        b = 20
        print("Inner changed id(b):", id(b))

    inner_nonlocal()
    print("Outer after change id(b):", id(b))

outer_nonlocal()

````

````python

# =========================================================
# 8. globals(), locals(), dir()
# =========================================================

print("\n===== 8. SPECIAL FUNCTIONS =====")

def inspect_names():
    m = 1
    n = 2
    print("locals():", locals())
    print("dir():", dir())

inspect_names()

print("globals() sample:", list(globals().keys())[:10])

````

````python

# =========================================================
# 9. MEMORY MODEL (REFERENCE BEHAVIOR)
# =========================================================

print("\n===== 9. MEMORY MODEL =====")

p = 10
q = p

print("p id:", id(p))
print("q id:", id(q))  # SAME (same object)

q = 20
print("After change:")
print("p id:", id(p))
print("q id:", id(q))  # DIFFERENT

````

````python

# =========================================================
# 10. MUTABLE vs IMMUTABLE
# =========================================================

print("\n===== 10. MUTABLE vs IMMUTABLE =====")

# Immutable Example
num = 10
print("num id before:", id(num))
num = num + 1
print("num id after :", id(num))  # NEW object

# Mutable Example
lst = [1, 2]
print("list id before:", id(lst))
lst.append(3)
print("list id after :", id(lst))  # SAME object

````

````python

# =========================================================
# 11. COMPLETE MEMORY FLOW VISUAL
# =========================================================

print("\n===== 11. MEMORY FLOW (CONCEPTUAL) =====")

"""
Memory Representation:

Local Namespace (Function)
   x ───► object

Enclosed Namespace (Outer Function)
   y ───► object

Global Namespace
   g ───► object

Built-in Namespace
   len ─► function

Search Order:
Local → Enclosed → Global → Built-in
"""

````

````python
"""
PROGRAM 1: UnboundLocalError with GLOBAL variable
"""

x = 10  # Global variable

def test_global():
    print("Trying to access x:", x)  # ❌ Error happens here
    x = 20  # Python treats x as LOCAL because of assignment

#test_global()  # IF we call this function will give UnboundLocalError
````

# 🧠 Why This Error Happens in Python

## 🔍 The Problem

When Python sees an assignment to a variable **inside a function**, it treats that variable as **local** by default.

### Example Code:

```python
x = 20   # This is a global variable

def my_function():
    print(x)        # ← Trying to read x
    x = 10          # ← Assigning to x (makes it local!)
    
my_function()

Output:

UnboundLocalError: cannot access local variable 'x' where it is not associated with a value

````python
# Fixing Using global keyword
x = 10

def test_global():
    global x
    print("Accessing global x:", x)
    x = 20

test_global()
````

# 🧠 Why This Error Happens in Python

## Same Logic – Nested Function Case

### Example Code:

```python
y = 200   # Global variable

def outer():
    def inner():
        print(y)     # ← Trying to read y
        y = 300      # ← Assignment makes y local to inner()
    
    inner()

outer()

output:

UnboundLocalError: cannot access local variable 'y' where it is not associated with a value

````python
"""
PROGRAM 2: UnboundLocalError with NONLOCAL (Enclosed) variable
"""

def outer():
    y = 100  # Enclosed variable

    def inner():
        print("Trying to access y:", y)  # ❌ Error
        y = 200  # Python treats y as LOCAL

    inner()

#outer() # IF we call this function will give UnboundLocalError
````

````python
def outer():
    y = 100

    def inner():
        nonlocal y
        print("Accessing enclosed y:", y)
        y = 200

    inner()
    print("Updated y:", y)

outer()
````

# LEGB Scope Hierarchy in Python

The **LEGB** rule defines the order in which Python searches for a variable name:

**L**ocal → **E**nclosed → **G**lobal → **B**uilt-in

```ascii
+----------------------------------------------------+
|                    Built-in Scope                  |
|   (Python's predefined functions, keywords)       |
|   e.g., len(), print(), range(), etc.             |
+----------------------------------------------------+
                 ↑
                 |
+----------------------------------------------------+
|                     Global Scope                   |
|   (Variables defined at the top level of module)  |
|   e.g., x = 100                                   |
+----------------------------------------------------+
                 ↑
                 |
+----------------------------------------------------+
|                   Enclosed Scope                   |
|   (Outer function’s local variables)              |
|   e.g., x = 20 in outer()                         |
+----------------------------------------------------+
                 ↑
                 |
+----------------------------------------------------+
|                     Local Scope                    |
|   (Inner function's variables)                    |
|   e.g., x = 200 in inner()                        |
|   Can access enclosed, global, built-in if needed |
+----------------------------------------------------+
