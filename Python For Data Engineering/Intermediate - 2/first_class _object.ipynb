```python
"""
first_class_objects_in_python


Topics Covered:
1. First-Class Objects
2. Properties with examples
3. Functions as first-class objects
4. Function aliasing
5. Storing functions in data structures
6. Passing functions as arguments
7. Returning functions from functions
8. Higher-order function examples
9. Strings and other types as objects
"""

def print_section(title):
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)
```

```python


# ----------------------------------------------------------
# 1. First-Class Objects in Python
# ----------------------------------------------------------
print_section("1. First-Class Objects in Python")

print("""
In Python, First-Class Objects are objects that can be:
- Assigned to variables
- Stored in data structures
- Passed as arguments to functions
- Returned from functions
- Operated on like any other object

Almost everything in Python is an object, including functions.
""")

```

```python

# ----------------------------------------------------------
# 2. Property: Assigned to Variables
# ----------------------------------------------------------
print_section("2. Assign Objects to Variables")

a = 10
b = a
print("a =", a)
print("b =", b)

```

```python

# ----------------------------------------------------------
# 3. Property: Stored in Data Structures
# ----------------------------------------------------------
print_section("3. Stored in Data Structures")

numbers = [1, 2, 3, 4]
print("List storing objects:", numbers)

person = {"name": "Rajan", "age": 24}
print("Dictionary storing objects:", person)

```

```python

# ----------------------------------------------------------
# 4. Property: Passed as Arguments
# ----------------------------------------------------------
print_section("4. Passed as Arguments")

def show_value(x):
    print("Received value:", x)

show_value(100)

```

```python

# ----------------------------------------------------------
# 5. Property: Returned from Functions
# ----------------------------------------------------------
print_section("5. Returned from Functions")

def create_message():
    return "Hello Python"

msg = create_message()
print("Returned value:", msg)

```

```python

# ----------------------------------------------------------
# 6. Property: Operations on Objects
# ----------------------------------------------------------
print_section("6. Operations on Objects")

x, y = 5, 7
result = x + y
print("x + y =", result)

```

```python

# ----------------------------------------------------------
# 7. Functions are First-Class Objects
# ----------------------------------------------------------
print_section("7. Functions as First-Class Objects")

def greet(name):
    return f"Hello {name}"

print("Function call:", greet("Rajan"))

```

```python

# ----------------------------------------------------------
# 8. Function Aliasing
# ----------------------------------------------------------
print_section("8. Function Aliasing")

shantanu = greet
print(shantanu("Python Learner"))
print("Here, 'shantanu' is an alias of the 'greet' function.")

```

```python

# ----------------------------------------------------------
# 9. Functions in Lists, Tuples, and Dictionaries
# ----------------------------------------------------------
print_section("9. Functions in Data Structures")

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

func_list = [add, multiply]
func_tuple = (add, multiply)
func_dict = {"add": add, "mul": multiply}

print("List - Add:", func_list[0](2, 3))
print("Tuple - Multiply:", func_tuple[1](2, 3))
print("Dictionary - Add:", func_dict["add"](5, 7))

```

```python

# ----------------------------------------------------------
# 10. Passing Functions as Arguments (Higher-Order Functions)
# ----------------------------------------------------------
print_section("10. Passing Functions as Arguments")

def demo(func, x, y):
    return func(x, y)

print("Higher-order function (addition):", demo(add, 3, 4))
print("Higher-order function (multiplication):", demo(multiply, 3, 4))

```

```python

# ----------------------------------------------------------
# 11. Returning Functions from Functions
# ----------------------------------------------------------
print_section("11. Returning Functions from Functions")

def choose_operation(op):
    def add(a, b):
        return a + b
    def subtract(a, b):
        return a - b
    return add if op == "add" else subtract

operation = choose_operation("add")
print("Result of returned function:", operation(10, 5))

```

```python
# ----------------------------------------------------------
# 11. Returning Functions from Functions (Using Lambda)
# ----------------------------------------------------------
print_section("11. Returning Functions from Functions")

def choose_operation(op):
    add = lambda a, b: a + b
    subtract = lambda a, b: a - b
    return add if op == "add" else subtract

operation = choose_operation("add")
print("Result of returned function:", operation(10, 5))
```

```python
# ----------------------------------------------------------
# 11. Returning Functions from Functions (Fully Lambda)
# ----------------------------------------------------------
print_section("11. Returning Functions from Functions")

choose_operation = lambda op: (lambda a, b: a + b) if op == "add" else (lambda a, b: a - b)

operation = choose_operation("add")
print("Result of returned function:", operation(10, 5))
```

```python
# ----------------------------------------------------------
# 11. Returning Functions from Functions (IIFE Version)
# ----------------------------------------------------------
print_section("11. Returning Functions from Functions")

# IIFE: choose and call the function immediately
result = (lambda op, a, b: (lambda x, y: x + y if op == "add" else x - y)(a, b))("add", 10, 5)

print("Result of IIFE function:", result)
```

```python
# ----------------------------------------------------------
# 11. Returning Functions from Functions (IIFE - Clean Version)
# ----------------------------------------------------------
print_section("11. Returning Functions from Functions")

# IIFE: function is created and immediately invoked
result = (lambda op: 
            (lambda a, b: a + b) if op == "add" 
            else (lambda a, b: a - b)
         )("add")(10, 5)

print("Result of IIFE function:", result)
```

```python

# ----------------------------------------------------------
# 12. Strings and Other Types are also Objects
# ----------------------------------------------------------
print_section("12. Strings and Other Types as Objects")

name = "Rajan"
print("String:", name)
print("Type:", type(name))
print("Uppercase method call on string:", name.upper())

number = 42
print("Integer:", number)
print("Type:", type(number))

```

```python
# ----------------------------------------------------------
# 1. What is a Higher-Order Function?
# ----------------------------------------------------------
print_section("1. Higher-Order Function Definition")

print("A Higher-Order Function is a function that:")
print("1. Takes another function as an argument OR")
print("2. Returns a function OR")
print("3. Does both")
```

```python


# ----------------------------------------------------------
# 2. Example: Function as Argument
# ----------------------------------------------------------
print_section("2. Function as Argument")

def apply_operation(func, a, b):
    return func(a, b)

def add(x, y):
    return x + y

def multiply(x, y):
    return x * y

print("Addition:", apply_operation(add, 5, 3))
print("Multiplication:", apply_operation(multiply, 5, 3))

```

```python

# ----------------------------------------------------------
# 3. Example: Returning a Function
# ----------------------------------------------------------
print_section("3. Returning a Function")

def choose_operation(op):
    def add(a, b):
        return a + b
    
    def subtract(a, b):
        return a - b
    
    return add if op == "add" else subtract

operation = choose_operation("subtract")
print("Result:", operation(10, 4))

```

```python

# ----------------------------------------------------------
# 4. Example: Both (Argument + Return)
# ----------------------------------------------------------
print_section("4. Both (Argument + Return)")

def process(func):
    def wrapper(x, y):
        print("Before operation")
        result = func(x, y)
        print("After operation")
        return result
    return wrapper

def add_numbers(a, b):
    return a + b

enhanced_add = process(add_numbers)
print("Final Result:", enhanced_add(3, 2))

```

```python

# ----------------------------------------------------------
# 13. Summary
# ----------------------------------------------------------
print_section("13. Summary")

print("""
Key Takeaways:

- First-Class Objects can be assigned, stored, passed, returned, and operated on.
- Functions in Python are first-class objects.
- You can alias functions, store them in lists/tuples/dicts, pass them as arguments, or return them from functions.
- Strings, integers, and almost all types in Python are objects and can be manipulated like any other first-class object.
- Higher-order functions and function returning functions rely on the first-class nature of functions.
""")
```
