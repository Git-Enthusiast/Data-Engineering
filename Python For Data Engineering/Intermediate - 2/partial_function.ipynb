```python
# ==========================================================
# Python Partial Function
# ==========================================================

from functools import partial

print("\n========== 1. WHAT IS PARTIAL FUNCTION ==========\n")

print("""
A partial function allows you to 'fix' certain arguments of a function
and generate a new function with fewer arguments.
It is part of the functools module.
""")
```

```python

# ==========================================================
# 2. BASIC EXAMPLE
# ==========================================================

def add(n1, n2, n3, n4):
    return n1 + n2 + n3 + n4

# Normally you call like this:
result = add(1, 2, 3, 4)
print("Normal add function result:", result)
```

```python

# ==========================================================
# 3. PARTIAL FUNCTION USAGE
# ==========================================================

# Fix n1=2 and n2=3
add_partial = partial(add, 2, 3)

# Now call only with remaining arguments n3, n4
result_partial = add_partial(4, 5)  # 2 + 3 + 4 + 5 = 14
print("Partial function result (n1=2, n2=3):", result_partial)
```

```python

# ==========================================================
# 4. USING KEYWORD ARGUMENTS
# ==========================================================

# Fix n1 and n4
add_partial_kw = partial(add, n1=10, n4=20)

# Provide remaining arguments n2 and n3
result_kw = add_partial_kw(n2=5, n3=15)  # 10 + 5 + 15 + 20 = 50
print("Partial function with keywords:", result_kw)
```

```python
# ==========================================================
# 5. REPLACING VALUES? NO
# ==========================================================

# Fixing n1 and n2 using keywords
add_fixed = partial(add, n1=1, n2=2)

# Must provide remaining arguments using keyword names
print("Fixed n1 and n2, remaining 3 and 4:", add_fixed(n3=3, n4=4))  # 1+2+3+4 = 10

# Trying to override fixed arguments still requires correct keywords
# (Passing n1 or n2 again would cause TypeError)
print("Trying to override n1=10, n2=20:", add_fixed(n3=3, n4=4))  # Still 10
```

```python

# ==========================================================
# 6. REAL-WORLD USE CASE
# ==========================================================

# Example: Fixed interest rate calculation
def calculate_interest(principal, rate, time):
    return principal * rate * time / 100

# Fix rate = 5% for all customers
fixed_rate_interest = partial(calculate_interest, rate=5)

# Now we only need to provide principal and time
interest = fixed_rate_interest(principal=1000, time=2)
print("Interest for principal 1000, time 2 years:", interest)
```

```python

# ==========================================================
# 7. USING PARTIAL WITH OTHER FUNCTIONS
# ==========================================================

# Example: Power function
def power(base, exponent):
    return base ** exponent

# Fix exponent=2 to create a square function
square = partial(power, exponent=2)
print("Square of 5:", square(5))  # 5^2 = 25

# Fix exponent=3 to create a cube function
cube = partial(power, exponent=3)
print("Cube of 3:", cube(3))  # 3^3 = 27
```

```python

# ==========================================================
# 8. ADVANTAGES OF PARTIAL FUNCTIONS
# ==========================================================

print("""
1. Avoid repeating arguments every time.
2. Clean and professional way to 'preset' arguments.
3. Can be combined with map, filter, reduce for functional programming.
4. Useful in real-world scenarios like:
   - Fixed interest rates
   - Default configurations
   - API calls with common parameters
""")
```

```python

# ==========================================================
# 9. COMBINE WITH MAP / FILTER
# ==========================================================

nums = [1, 2, 3, 4, 5]

# Partial function to multiply by 10
def multiply(a, b):
    return a * b

times_10 = partial(multiply, 10)

result_map = list(map(times_10, nums))
print("Multiply all numbers by 10 using partial + map:", result_map)
```
