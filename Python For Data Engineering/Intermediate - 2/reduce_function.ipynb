# 🧠 What is `reduce()` in Python?

`reduce()` is a **higher-order function** in Python used to process an iterable and return a **single aggregated value**.

---

## 👉 Purpose

- 🔹 Reduce a sequence to a **single value**
- 🔹 Perform **cumulative computation** on elements

---

## 🔹 Syntax

```python
reduce(function, iterable, initializer=None)

# ⚡ Core Idea

`reduce()` works step-by-step:

- 👉 Takes the **first two elements** and applies the function
- 👉 Takes the **result** and the **next element**, applies the function again
- 👉 Repeats the process until only **one final result** remains

---
- 👉 **function** → The function to apply (takes two arguments)
- 👉 **iterable** → The sequence (list, tuple, etc.)
- 👉 **initializer** → Optional starting value

---
## 🔹 How it works:

### If initializer is not provided:
**`reduce()` starts with the first two elements of the iterable.** Example:
```python
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)
print(result)
```

**Step by step:**
* Step 1 → 1 + 2 = 3
* Step 2 → 3 + 3 = 6
* Step 3 → 6 + 4 = 10

**Output:**
```text
10
```

### If initializer is provided:
**`reduce()` starts with the initializer and the first element of the iterable.**

Example:
```python
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers, 10)
print(result)
```

**Step by step:**
* Step 1 → 10 + 1 = 11
* Step 2 → 11 + 2 = 13
* Step 3 → 13 + 3 = 16
* Step 4 → 16 + 4 = 20

**Output:**
```text
20
```

---
✅ **Key Point:**
* **No initializer** → first element is used as the starting value
* **With initializer** → reduces starting with the initializer and the first element

````python
# ==========================================================
# Python reduce() Function
# ==========================================================

from functools import reduce
````

````python

print("\n========== 1. BASIC SUM EXAMPLE ==========\n")

nums = [1, 2, 3, 4, 5]

def add(x, y):
    return x + y

result = reduce(add, nums)

print("Sum:", result)

````

````python

print("\n========== 2. STEP-BY-STEP UNDERSTANDING ==========\n")

def debug_add(x, y):
    print(f"Processing: {x} + {y}")
    return x + y

result = reduce(debug_add, nums)

print("Final Result:", result)

````

````python

print("\n========== 3. USING LAMBDA ==========\n")

result = reduce(lambda x, y: x + y, nums)

print("Sum using lambda:", result)
````

````python


print("\n========== 4. FIND MAXIMUM VALUE ==========\n")

nums = [10, 25, 7, 99, 42]
 
def find_max(x, y):
    
    return x if x > y else y

max_value = reduce(find_max, nums)

print("Maximum value:", max_value)

````

````python

print("\n========== 5. USING LAMBDA FOR MAX ==========\n")

max_value = reduce(lambda x, y: x if x > y else y, nums)

print("Max using lambda:", max_value)

````

````python

print("\n========== 6. PRODUCT OF ELEMENTS ==========\n")

nums = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, nums)

print("Product:", product)

````

````python

print("\n========== 7. STRING CONCATENATION ==========\n")

words = ["Hello", " ", "Rajan", "!"]

sentence = reduce(lambda x, y: x + y, words)

print("Sentence:", sentence)

````

````python

print("\n========== 8. REDUCE WITH INITIAL VALUE ==========\n")

nums = [1, 2, 3]

# Initial value = 10
result = reduce(lambda x, y: x + y, nums, 10)

print("Sum with initial value:", result)

````

````python

print("\n========== 9. REAL-WORLD EXAMPLE ==========\n")

# Total price calculation
prices = [100, 200, 300, 400]

total_bill = reduce(lambda x, y: x + y, prices)

print("Total bill:", total_bill)

````

````python

print("\n========== 10. COMBINE MAP + FILTER + REDUCE ==========\n")

nums = [1, 2, 3, 4, 5, 6]

# Step 1: Filter even numbers
evens = filter(lambda x: x % 2 == 0, nums)

# Step 2: Square them
squares = map(lambda x: x**2, evens)

# Step 3: Sum them
total = reduce(lambda x, y: x + y, squares)

print("Final result (even squares sum):", total)

````

````python

print("\n========== 11. REDUCE VS BUILT-IN FUNCTIONS ==========\n")

nums = [1, 2, 3, 4]

# Reduce way
sum_reduce = reduce(lambda x, y: x + y, nums)

# Built-in way (preferred)
sum_builtin = sum(nums)

print("Reduce:", sum_reduce)
print("Built-in:", sum_builtin)

````

````python

print("\n========== IMPORTANT NOTES ==========\n")

print("""
1. reduce() returns a single value
2. It is in functools module
3. Used for cumulative operations
4. Takes function with 2 arguments
5. More complex than map/filter
6. Often replaced by built-in functions like sum(), max()
""")
````
