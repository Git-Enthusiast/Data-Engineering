```python
"""
==================== PYTHON globals() FUNCTION  ====================

This file explains EVERYTHING about globals() in Python in a structured,
detailed, and beginner-to-advanced friendly way.

------------------------------------------------------------------------------------
1. WHAT IS globals()?
------------------------------------------------------------------------------------

globals() is a built-in Python function that returns a dictionary representing
the CURRENT GLOBAL SYMBOL TABLE.

👉 In simple words:
- It gives you access to all GLOBAL variables in your program.
- It returns them in dictionary form.

SYNTAX:
    globals()

RETURN TYPE:
    dict

------------------------------------------------------------------------------------
2. WHAT IS A SYMBOL TABLE?
------------------------------------------------------------------------------------

A symbol table is a data structure used by Python to store information about
variables.

👉 There are mainly two types:
1. Global Symbol Table  → Stores global variables
2. Local Symbol Table   → Stores local variables (inside functions)

Example structure of globals():
    {
        'variable_name': value,
        'function_name': function_object,
        ...
    }
"""
```

```python

"""
------------------------------------------------------------------------------------
3. BASIC EXAMPLE OF globals()
------------------------------------------------------------------------------------
"""

# Global variable
x = 100

# Access global variables using globals()
global_dict = globals()

print("\n3. BASIC EXAMPLE")
print("Global dictionary:", global_dict)
print("Access x using globals():", global_dict['x'])

```

```python

"""
------------------------------------------------------------------------------------
4. ACCESSING GLOBAL VARIABLES INSIDE A FUNCTION
------------------------------------------------------------------------------------

Normally, functions cannot directly modify global variables without using the
'global' keyword. But with globals(), we can access them.
"""


def access_global():
    print("\n4. ACCESSING GLOBAL INSIDE FUNCTION")
    print("Value of x using globals():", globals()['x'])


access_global()

```

```python

"""
------------------------------------------------------------------------------------
5. MODIFYING GLOBAL VARIABLES USING globals()
------------------------------------------------------------------------------------

We can MODIFY global variables using the dictionary returned by globals().
"""


def modify_global():
    globals()['x'] = 500
    print("\n5. MODIFY GLOBAL")
    print("Updated x:", x)


modify_global()

```
