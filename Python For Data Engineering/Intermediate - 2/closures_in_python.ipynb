```python
"""
===========================================================
                CLOSURES IN PYTHON 
===========================================================

A closure is a function object that remembers values from its
enclosing (outer) scope even after the outer function has finished execution.

In simple words:
A closure = Inner function + Remembered outer variables

Closures are created when:
1. A function is defined inside another function
2. The inner function uses variables from the outer function
3. The outer function returns the inner function

-----------------------------------------------------------
IMPORTANT CONCEPT: LEGB RULE
-----------------------------------------------------------
L → Local
E → Enclosing
G → Global
B → Built-in

Closures mainly work because of the "Enclosing" scope.

-----------------------------------------------------------
BASIC EXAMPLE
-----------------------------------------------------------
"""
```

```python

def outer_function():
    x = 10   # Enclosing variable

    def inner_function():
        print("Value of x:", x)   # Accessing enclosing variable

    return inner_function


closure_func = outer_function()
closure_func()   # Output: Value of x: 10


"""
Explanation:
- outer_function() creates variable x = 10
- inner_function() uses x
- outer_function returns inner_function (not executed yet)
- Even after outer_function is finished, inner_function remembers x

This memory behavior is called a CLOSURE.
"""

print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
EXAMPLE 2: CLOSURE WITH PARAMETERS
-----------------------------------------------------------
"""

def multiplier(n):
    def multiply(x):
        return x * n   # n is remembered
    return multiply


double = multiplier(2)
triple = multiplier(3)

print(double(5))   # 10
print(triple(5))   # 15


"""
Explanation:
- multiplier(2) → creates closure with n = 2
- multiplier(3) → creates closure with n = 3
- Each closure has its own memory

This is very useful for creating customized functions.
"""

print("\n" + "="*50)

```

## 🔥 Key Concept: Compilation Phase (Static Analysis)

### Before running your function, Python does:

#### 👉 Step 1: Parse the whole function
#### 👉 Step 2: Detect all variable assignments
#### 👉 Step 3: Decide scope (local/global/enclosed)

```python

"""
-----------------------------------------------------------
EXAMPLE 3: CLOSURE WITH MODIFICATION (nonlocal)
-----------------------------------------------------------
"""

def counter():
    count = 0

    def increment():
        nonlocal count   # Required to modify enclosing variable
        count += 1
        return count

    return increment


c = counter()

print(c())  # 1
print(c())  # 2
print(c())  # 3


"""
Explanation:
- Without 'nonlocal', Python will throw UnboundLocalError
- 'nonlocal' tells Python to use the enclosing variable

IMPORTANT:
nonlocal → used for enclosing scope
global   → used for global scope
"""

print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
EXAMPLE 4: REAL-LIFE USE CASE (LOGGER)
-----------------------------------------------------------
"""

def logger(msg):
    def log_message():
        print("Log:", msg)
    return log_message


log1 = logger("User logged in")
log2 = logger("File uploaded")

log1()
log2()


"""
Closures are useful when:
- You want to preserve state
- You want lightweight function factories
- You don't want to use classes
"""

print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
EXAMPLE 5: FUNCTION FACTORY
-----------------------------------------------------------
"""

def power_function(power):
    def calculate(base):
        return base ** power
    return calculate


square = power_function(2)
cube = power_function(3)

print(square(4))   # 16
print(cube(4))     # 64


print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
EXAMPLE 6: CHECKING CLOSURE VARIABLES
-----------------------------------------------------------
"""

def outer():
    x = 100

    def inner():
        return x

    return inner


func = outer()

print(func.__closure__)  # Shows closure cells
print(func.__closure__[0].cell_contents)  # Access stored value


"""
Explanation:
- __closure__ stores the remembered variables
- Each variable is stored in a "cell object"
"""

print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
EXAMPLE 7: CLOSURE VS CLASS
-----------------------------------------------------------
"""

# Using Closure
def counter_closure():
    count = 0

    def increment():
        nonlocal count
        count += 1
        return count

    return increment


# Using Class
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1
        return self.count


c1 = counter_closure()
c2 = Counter()

print(c1(), c1())   # 1, 2
print(c2.increment(), c2.increment())  # 1, 2


"""
Difference:
Closure:
- Lightweight
- Functional style
- No class needed

Class:
- More structured
- Better for large systems
"""

print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
COMMON INTERVIEW QUESTIONS
-----------------------------------------------------------

Q1: Why use closures?
- Data hiding
- Avoid global variables
- Create function factories

Q2: Difference between closure and decorator?
- Closure: remembers variables
- Decorator: modifies function behavior (uses closures internally)

Q3: Can closures have multiple variables?
- Yes

"""

def outer_multi():
    a = 10
    b = 20

    def inner():
        return a + b

    return inner

f = outer_multi()
print(f())  # 30


print("\n" + "="*50)

```

```python

"""
-----------------------------------------------------------
IMPORTANT EDGE CASE (LATE BINDING ISSUE)
-----------------------------------------------------------
"""

functions = []

for i in range(3):
    def inner():
        return i
    functions.append(inner)

print([f() for f in functions])  # [2, 2, 2] ❌


"""
Problem:
- All closures reference SAME variable i
- Final value of i = 2

Solution:
Use default argument
"""

functions = []

for i in range(3):
    def inner(i=i):   # Fix
        return i
    functions.append(inner)

print([f() for f in functions])  # [0, 1, 2] ✅


print("\n" + "="*50)

```
