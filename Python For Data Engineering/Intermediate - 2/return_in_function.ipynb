```python

"""
COMPREHENSIVE GUIDE TO THE `return` STATEMENT IN PYTHON
======================================================

This script demonstrates BASIC to ADVANCED uses of the `return` statement.
You can run this file to see outputs for each example.

Author: Generated learning guide
"""

print("\n==============================")
print("1. BASIC RETURN")
print("==============================")

def add(a, b):
    return a + b

result = add(5, 7)
print("Addition Result:", result)

```

```python

print("\n==============================")
print("2. FUNCTION WITHOUT RETURN")
print("==============================")

def greet(name):
    print("Hello", name)

value = greet("Rajan")
print("Returned value:", value)   # None

```

```python


print("\n==============================")
print("3. RETURNING MULTIPLE VALUES")
print("==============================")

def calculate(a, b):
    return a+b, a-b, a*b

s, d, p = calculate(10, 4)
print("Sum:", s)
print("Difference:", d)
print("Product:", p)


print("\n==============================")
print("4. RETURNING LIST")
print("==============================")

def even_numbers(n):
    evens = []
    for i in range(n+1):
        if i % 2 == 0:
            evens.append(i)
    return evens

print(even_numbers(10))


print("\n==============================")
print("5. RETURNING DICTIONARY")
print("==============================")

def student():
    return {
        "name": "Rajan",
        "course": "CSE",
        "skills": ["Python", "Java", "SQL"]
    }

print(student())


print("\n==============================")
print("6. EARLY RETURN")
print("==============================")

def check_number(num):
    if num < 0:
        return "Negative"
    if num == 0:
        return "Zero"
    return "Positive"

print(check_number(-3))
print(check_number(0))
print(check_number(8))


print("\n==============================")
print("7. RETURN INSIDE LOOP")
print("==============================")

def first_even(nums):
    for n in nums:
        if n % 2 == 0:
            return n
    return None

print(first_even([1,3,5,9,11,14,15]))


print("\n==============================")
print("8. BOOLEAN RETURN")
print("==============================")

def is_even(n):
    return n % 2 == 0

print(is_even(6))
print(is_even(7))


print("\n==============================")
print("9. DEFAULT PARAMETERS")
print("==============================")

def power(base, exp=2):
    return base ** exp

print(power(5))
print(power(5,3))


print("\n==============================")
print("10. RETURN VS PRINT")
print("==============================")

def square_print(x):
    print("Square:", x*x)

def square_return(x):
    return x*x

square_print(4)
value = square_return(4)
print("Returned square:", value)


print("\n==============================")
print("11. RETURNING FUNCTION (HIGHER ORDER FUNCTION)")
print("==============================")

def outer():
    def inner():
        return "Hello from inner function"
    return inner

f = outer()
print(f())


print("\n==============================")
print("12. RETURNING MULTIPLE TYPES")
print("==============================")

def multiple():
    return 10, [1,2,3], {"a":1}

print(multiple())


print("\n==============================")
print("13. EXPLICIT NONE RETURN")
print("==============================")

def do_nothing():
    return None

print(do_nothing())


print("\n==============================")
print("ADVANCED PYTHON RETURN EXAMPLES")
print("==============================")


print("\n14. RETURN IN RECURSION")

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print("Factorial of 5:", factorial(5))


print("\n15. RETURN IN SEARCH ALGORITHM")

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

print(linear_search([10,20,30,40], 30))


print("\n16. RETURNING OBJECT FROM CLASS METHOD")

class Student:

    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name

s = Student("Rajan")
print(s.get_name())


print("\n17. RETURNING LAMBDA FUNCTION")

def multiplier(n):
    return lambda x: x * n

double = multiplier(2)
print(double(10))


print("\n18. RETURNING GENERATOR")

def count_up_to(n):
    for i in range(1, n+1):
        yield i

for num in count_up_to(5):
    print(num)


print("\n19. RETURN IN TRY/EXCEPT")

def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Division by zero not allowed"

print(divide(10,2))
print(divide(10,0))


print("\n20. RETURN IN DECORATORS")

def my_decorator(func):

    def wrapper():
        print("Before function call")
        value = func()
        print("After function call")
        return value

    return wrapper


@my_decorator
def say_hello():
    return "Hello World"

print(say_hello())


print("\n==============================")
print("KEY TAKEAWAYS")
print("==============================")

print("1. return immediately exits a function")
print("2. return sends value back to caller")
print("3. if no return -> Python returns None")
print("4. multiple values are returned as tuple")
print("5. return can return any Python object")
print("6. return is heavily used in recursion, APIs, algorithms")
print("7. return improves modular programming")

print("\nEND OF COMPREHENSIVE GUIDE")

```
