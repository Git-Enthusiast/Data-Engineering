# 1) What is an iterator?

An **iterator** is an object that gives you values **one by one**, on demand, instead of giving all values at once.

Think of it like this:

* A **list** is a box with all items already inside.
* An **iterator** is like a person handing you one item at a time.

Example:

```python
nums = [10, 20, 30]

it = iter(nums)   # make an iterator
print(next(it))   # 10
print(next(it))   # 20
print(next(it))   # 30
```

After the last item, there is nothing left. Then Python raises:

```python
StopIteration
```

---

# 2) Why iterators exist

Iterators are used because they are:

* **memory efficient**
* **lazy**: they compute values only when needed
* useful for **large data**
* useful for **infinite sequences**
* the backbone of `for` loops, generators, `map`, `filter`, `zip`, file reading, and more

Without iterators, Python would have to create everything upfront.

# 3) Iterable vs Iterator

## Iterable

An **iterable** is anything you can loop over.

An iterable can produce an iterator.

## Iterator

An **iterator** is the object that actually yields values one by one.

It must support:

* `__iter__()`

# 6) The iterator protocol

An object is an iterator if it implements these two methods:

```python
#   __iter__()
#   __next__()
```

### `__iter__()`

Returns the iterator object itself.

### `__next__()`

Returns the next item.
When items are exhausted, it raises `StopIteration`.

# 7) How `for` loop works internally

A `for` loop uses the iterator protocol.

This:

```python
for x in [1, 2, 3]:
    print(x)
```

is roughly like this:

```python
items = [1, 2, 3]
it = iter(items)

while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break
```

So `for` loops are built on top of iterators.

# 8) `iter()` function

`iter()` converts an iterable into an iterator.

```python
lst = [10, 20, 30]
it = iter(lst)

print(next(it))  # 10
print(next(it))  # 20
```

If you call `iter()` on an object that is already an iterator, it often returns itself.

```python
it2 = iter(it)
print(it2 is it)   # True for many iterators
```

# 9) `next()` function

`next()` asks an iterator for the next value.

```python
it = iter([1, 2, 3])

while True:
    try:
        print(next(it))
    except StopIteration:
        print("Iterator finished")
        break
```

You can also provide a default:

```python
it = iter([1])

print(next(it, "done"))  # 1
print(next(it, "done"))  # done
```

This prevents the error.

# 10) StopIteration

`StopIteration` is the signal that there are no more items.

Example:

```python
it = iter([1, 2, 3])

while True:
    try:
        print(next(it))
    except StopIteration:
        print("Stop Iteration")
        break
```

This is not a bug. It is how Python says:
“Iteration is finished.”

# 12) Iterator objects are usually one-time use

This is a common beginner confusion.

```python
nums = [1, 2, 3]
it = iter(nums)

print(next(it))  # 1
print(next(it))  # 2

print(list(it))  # [3]
```

Once an iterator is consumed, it moves forward. You cannot “rewind” it unless you create a new iterator.

```python
it = iter([1, 2, 3])
print(list(it))   # [1, 2, 3]
print(list(it))   # []
```

Second time is empty because the iterator is already exhausted.

# 13) Iterable can create a new iterator every time

This is a key difference.

```python
lst = [1, 2, 3]

it1 = iter(lst)
it2 = iter(lst)

print(next(it1))  # 1
print(next(it1))  # 2

print(next(it2))  # 1
```

A list is iterable, so every time you call `iter(lst)` you get a fresh iterator.

# 15) Custom iterator class

You can create your own iterator using a class.

## Example: count from 1 to n

```python
class CountUpTo:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max_value:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
```

Use it like this:

```python
counter = CountUpTo(5)

for num in counter:
    print(num)
```

Output:

```python
1
2
3
4
5
```

### What is happening?

* `__iter__()` returns the iterator itself
* `__next__()` gives current number
* once `current > max_value`, it stops

# 16) Step-by-step custom iterator behavior

This shows the internal nature clearly.

# 17) Generator functions are iterators

Generators are one of the most important iterator-related topics.

A function with `yield` produces a generator.

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

Use it:

```python
g = my_generator()

print(next(g))  # 1
print(next(g))  # 2
print(next(g))  # 3
```

A generator is an iterator.

---

# 18) Why generators are special

Generators are easier to write than custom iterator classes.

This:

```python
def count_up_to(n):
    current = 1
    while current <= n:
        yield current
        current += 1
```

is simpler than writing a full class.

Generators:

* remember their state
* pause at `yield`
* resume from where they left off

# 20) `yield` and lazy evaluation

Example:

```python
def squares(n):
    for i in range(n):
        yield i * i
```

Nothing is computed until you ask:

```python
g = squares(5)
print(next(g))  # 0
print(next(g))  # 1
print(next(g))  # 4
```

Values are produced lazily.

# 21) Infinite iterators

Iterators can even be infinite.

```python
def infinite_counter():
    num = 1
    while True:
        yield num
        num += 1
```

Use carefully:

```python
g = infinite_counter()
print(next(g))  # 1
print(next(g))  # 2
print(next(g))  # 3
```

Because it never stops, you must limit how much you consume.

This is where `itertools.islice()` becomes useful.

# 22) `itertools` and advanced iterator tools

Python has a module called `itertools` for powerful iterator operations.

Common ones:

## `count()`

Infinite counting iterator

```python
from itertools import count

c = count(10, 2)
print(next(c))  # 10
print(next(c))  # 12
print(next(c))  # 14
```

## `cycle()`

Repeats values forever

```python
from itertools import cycle

c = cycle(["A", "B"])
print(next(c))  # A
print(next(c))  # B
print(next(c))  # A
```

## `repeat()`

Repeats one value

```python
from itertools import repeat

r = repeat("hi", 3)
print(list(r))   # ['hi', 'hi', 'hi']
```

## `islice()`

Takes a slice from an iterator

```python
from itertools import islice, count

c = count(1)
print(list(islice(c, 5)))   # [1, 2, 3, 4, 5]
```

## `chain()`

Chains iterables together

```python
from itertools import chain

print(list(chain([1, 2], [3, 4])))  # [1, 2, 3, 4]
```

## `zip_longest()`

Pairs iterables, filling missing values

```python
from itertools import zip_longest

print(list(zip_longest([1, 2], [3, 4, 5], fillvalue=None)))
```

# 25) `filter(None, iterable)` special case

Since you recently asked about this, here is the internal logic.

```python
data = [0, 1, False, True, "", "Hello", None]
filtered = filter(None, data)
print(list(filtered))
```

This keeps only truthy values:

```python
[1, True, 'Hello']
```

When `None` is passed, Python behaves like:

```python
#   item for item in data if bool(item)
```

It removes falsy values such as:

* `0`
* `False`
* `""`
* `None`
* `[]`
* `{}`
* `set()`

# 26) Lazy evaluation

Iterators compute items only when needed.

Example:

```python
def numbers():
    print("Starting")
    yield 1
    print("After 1")
    yield 2
```

Use it:

```python
g = numbers()
print(next(g))
print(next(g))
```

Output:

```python
'''
Starting
1
After 1
2
'''
```

Notice the function does not run all at once. It pauses.

That is laziness.

# 27) Common mistakes with iterators

## Mistake 1: expecting an iterator to restart

```python
it = iter([1, 2, 3])
print(list(it))  # [1, 2, 3]
print(list(it))  # []
```

## Mistake 2: using `next()` after exhaustion

```python
it = iter([1])
print(next(it))
#print(next(it))  # StopIteration
```

## Mistake 3: confusing iterable with iterator

A list is iterable, but not itself an iterator.

```python
lst = [1, 2, 3]
print(hasattr(lst, "__next__"))  # False
print(hasattr(iter(lst), "__next__"))  # True
```

## Mistake 4: using iterator where you need reusable data

If you need to traverse multiple times, store data in a list or recreate the iterator.

# 29) Iterator from a custom iterable

You can make an object iterable by implementing `__iter__()`.

Example:

```python
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        current = self.start
        while current < self.end:
            yield current
            current += 1
```

Use:

```python
r = MyRange(2, 5)
for x in r:
    print(x)
```

Output:

```python
2
3
4
```

This object is iterable because `__iter__()` returns a generator iterator.

# 30) A class can be both iterable and iterator

This happens when:

* `__iter__()` returns `self`
* `__next__()` exists

Example:

```python
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max_value:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
```

This object is both iterable and iterator.

# 31) Important practical difference: iterator state changes

An iterator remembers its position.

```python
it = iter([10, 20, 30])

print(next(it))  # 10
print(next(it))  # 20
```

The next time you call `next(it)`, it continues from where it left off.

This statefulness is why iterators are powerful.

# 37) Advanced concept: iterator pipelines

You can chain lazy operations.

Example:

```python
nums = range(10)

result = map(
    lambda x: x * 2,
    filter(lambda x: x % 2 == 0, nums)
)

print(list(result))  # [0, 4, 8, 12, 16]
```

This works lazily:

* filter evens
* then double them
* values are produced only as needed

Once consumed by one operation, it is gone.

Example:

```python
nums = iter([1, 2, 3])

print(list(nums))  # [1, 2, 3]
print(list(nums))  # []
```

# 39) Peeking into an iterator

Sometimes you want to inspect the next value without consuming everything.

Python does not have a direct universal peek for all iterators, but you can use `itertools.tee()` or custom logic.

Example with `tee`:

```python
from itertools import tee

it1, it2 = tee(iter([1, 2, 3]))
print(next(it1))  # 1
print(next(it2))  # 1
```

This creates independent copies, but be careful with memory if the iterators are large.

# 45) Final practical examples

## Example 1: iterator from list

```python
items = [5, 6, 7]
it = iter(items)

print(next(it))  # 5
print(next(it))  # 6
print(next(it))  # 7
```

## Example 2: generator iterator

```python
def squares(n):
    for i in range(n):
        yield i * i

g = squares(4)
print(next(g))  # 0
print(next(g))  # 1
print(next(g))  # 4
print(next(g))  # 9
```

## Example 3: custom iterator

```python
class EvenNumbers:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 2
        return value

e = EvenNumbers(6)
for x in e:
    print(x)
```

Output:

```python
0
2
4
6
```
