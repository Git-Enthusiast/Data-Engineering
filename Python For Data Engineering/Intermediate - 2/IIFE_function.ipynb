```python
"""

iife_python_complete_guide

Complete tutorial on Immediately Invoked Function Expressions (IIFE)
and how to simulate them in Python using lambda functions.

This file demonstrates:
1. IIFE concept
2. JavaScript IIFE example
3. Why Python normal functions cannot be IIFE
4. Python lambda workaround
5. Passing arguments
6. Multiple return values
7. IIFE with loops
8. IIFE with conditions
9. IIFE with list comprehensions
10. Practical examples

"""

def print_section(title):
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)

```

```python

# ----------------------------------------------------------
# 1. What is IIFE?
# ----------------------------------------------------------
print_section("1. What is IIFE?")

print("IIFE = Immediately Invoked Function Expression")
print("It means defining and executing a function at the same time.")

```

```python

# ----------------------------------------------------------
# 2. JavaScript Example
# ----------------------------------------------------------
print_section("2. JavaScript IIFE Example")

print("""
JavaScript syntax:

(function(a, b){
    console.log(a + b);
})(3,4)

The function is:
1. Defined
2. Immediately executed
""")

```

```python

# ----------------------------------------------------------
# 3. Python does not support this syntax
# ----------------------------------------------------------
print_section("3. Python Normal Function Cannot Be IIFE")

print("Trying JavaScript style in Python would cause SyntaxError.")

```

```python

# ----------------------------------------------------------
# 4. Python IIFE using Lambda
# ----------------------------------------------------------
print_section("4. Python IIFE using Lambda")

result = (lambda a, b: a + b)(3, 4)

print("Expression: (lambda a, b: a + b)(3, 4)")
print("Result:", result)

```

```python

# ----------------------------------------------------------
# 5. IIFE with Single Argument
# ----------------------------------------------------------
print_section("5. IIFE with One Argument")

square = (lambda x: x * x)(6)

print("Expression: (lambda x: x*x)(6)")
print("Square:", square)

```

```python

# ----------------------------------------------------------
# 6. IIFE Returning Multiple Values
# ----------------------------------------------------------
print_section("6. IIFE Returning Multiple Values")

values = (lambda a, b: (a + b, a * b))(5, 2)
print(type(values))
print("Addition:", values[0])
print("Multiplication:", values[1])

```

```python

# ----------------------------------------------------------
# 7. IIFE with Conditional Logic
# ----------------------------------------------------------
print_section("7. IIFE with Conditions")

check = (lambda x: "Even" if x % 2 == 0 else "Odd")(int(input("Please Enter a number: ")))

print("Number:", number)
print("Result:", check)
```

```python


# ----------------------------------------------------------
# 8. IIFE with List Comprehension
# ----------------------------------------------------------
print_section("8. IIFE with List Comprehension")

squares = (lambda n: [i*i for i in range(n)])(5)

print("Squares up to 5:", squares)

```

```python

# ----------------------------------------------------------
# 9. IIFE with Loop-like Processing
# ----------------------------------------------------------
print_section("9. IIFE Processing a Loop")

numbers = [1,2,3,4,5]

double_numbers = (lambda nums: [n*2 for n in nums])(numbers)

print("Original:", numbers)
print("Doubled:", double_numbers)
```

```python


# ----------------------------------------------------------
# 10. Real World Example – Temporary Calculation
# ----------------------------------------------------------
print_section("10. Real World Example - Temporary Calculation")

total_price = (lambda price, tax: price + price*tax)(1000, 0.18)

print("Base price: 1000")
print("Price with tax:", total_price)
```

```python


# ----------------------------------------------------------
# 11. Real World Example – Data Processing
# ----------------------------------------------------------
print_section("11. Real World Example - Data Processing")

data = [10, 15, 20, 25, 30]

filtered = (lambda d: [x for x in d if x % 10 == 0])(data)

print("Original data:", data)
print("Filtered multiples of 10:", filtered)

```

```python

# ----------------------------------------------------------
# 12. Summary
# ----------------------------------------------------------
print_section("12. Summary")

print("""
JavaScript IIFE:
    (function(){ ... })()

Python Equivalent:
    (lambda ... : ...)(arguments)

Example:
    (lambda a, b: a + b)(3, 4)

Used when:
• Code needs to run only once
• No need to define a reusable function
• Temporary calculations
""")
```
