```python
# ---------------------------------------------
# Higher Order Functions in Python
# ---------------------------------------------

"""
Definition:
A Higher Order Function is a function that:
1. Takes another function as an argument, OR
2. Returns another function
"""

# ---------------------------------------------
# Higher Order Function
# ---------------------------------------------

def higher_order(func):
    
    # inner function (this will be returned)
    def wrapper(x):
        print("Before execution")
        result = func(x)   # using function passed as argument
        print("After execution")
        return result
    
    return wrapper   # returning function


# Normal function
def square(n):
    return n * n


# Using higher order function
new_function = higher_order(square)

print(new_function(5))
```

```python
"""
===========================================================
        PYTHON FILTER() FUNCTION
===========================================================

"""

# -----------------------------------------------------------
# 1. WHAT IS filter() ?
# -----------------------------------------------------------
"""
filter() is a built-in higher-order function in Python.

✔ It is used to FILTER elements from an iterable (like list, tuple, etc.)
✔ It keeps only those elements for which a condition is True
✔ It returns a FILTER OBJECT (not a list)

Syntax:
    filter(function, iterable)

Where:
- function → logic/condition (returns True or False)
- iterable → data (list, tuple, string, etc.)
"""

```

```python

# -----------------------------------------------------------
# 2. BASIC EXAMPLE
# -----------------------------------------------------------
print("---- Basic Example ----")

def is_even(num):
    return num % 2 == 0   # True for even numbers

data = [1, 2, 3, 4, 5, 6]

result = filter(is_even, data)

print("Filter object:", result)

# Convert to list to see values
print("Filtered (Even Numbers):", list(result))

```

```python

# -----------------------------------------------------------
# 3. IMPORTANT: FILTER OBJECT IS EXHAUSTIBLE
# -----------------------------------------------------------
print("\n---- Filter Object Exhaustion ----")

data = [1, 2, 3, 4, 5]

result = filter(is_even, data)

print(list(result))   # First use → works
print(list(result))   # Second use → EMPTY (already exhausted)

```

```python

# -----------------------------------------------------------
# 4. USING LOOP INSTEAD OF list()
# -----------------------------------------------------------
print("\n---- Using Loop ----")

data = [10, 15, 20, 25, 30]

result = filter(is_even, data)

for value in result:
    print(value)   # prints only even numbers

```

```python

# -----------------------------------------------------------
# 5. USING LAMBDA FUNCTION (MOST COMMON)
# -----------------------------------------------------------
print("\n---- Using Lambda ----")

data = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, data))

print("Even numbers using lambda:", even_numbers)

```

```python

# -----------------------------------------------------------
# 6. FILTERING STRINGS
# -----------------------------------------------------------
print("\n---- Filtering Strings ----")

names = ["Rajan", "", "Python", "", "AI"]

# Remove empty strings
non_empty = list(filter(lambda x: x != "", names))

print("Non-empty strings:", non_empty)

```

```python

# -----------------------------------------------------------
# 7. FILTERING BASED ON LENGTH
# -----------------------------------------------------------
print("\n---- Filter Based on Length ----")

words = ["cat", "elephant", "dog", "tiger"]

# Keep words with length > 3
long_words = list(filter(lambda w: len(w) > 3, words))

print("Words with length > 3:", long_words)

```

```python

# -----------------------------------------------------------
# 8. FILTER WITH NONE (SPECIAL CASE)
# -----------------------------------------------------------
print("\n---- Using None in filter() ----")

data = [0, 1, False, True, "", "Hello", None]

# Keeps only truthy values
filtered = list(filter(None, data))

print("Truthy values:", filtered)

```

```python

# -----------------------------------------------------------
# 9. REAL-WORLD EXAMPLE
# -----------------------------------------------------------
print("\n---- Real World Example ----")

students = [
    {"name": "A", "marks": 40},
    {"name": "B", "marks": 75},
    {"name": "C", "marks": 60},
    {"name": "D", "marks": 30}
]

# Select students with marks >= 50
passed_students = list(filter(lambda s: s["marks"] >= 50, students))

print("Passed Students:", passed_students)

```

```python
# -----------------------------------------------------------
# 11. Filter Out Vowel from string
# -----------------------------------------------------------
vowel_list = ["a","e","i","o","u"]
def vowels(ch):
    return ch in vowel_list
str = "Rajan Raj"
print(list(filter(vowels,str)))
```

```python

# -----------------------------------------------------------
# 11. KEY POINTS SUMMARY
# -----------------------------------------------------------
"""
✔ filter() is a HIGHER-ORDER FUNCTION (takes function as argument)
✔ It returns a FILTER OBJECT (not list)
✔ Use list() or loop to extract values
✔ It is EXHAUSTIBLE (usable only once)
✔ Lambda is commonly used for short conditions
✔ filter(None, iterable) removes falsy values
"""

```

```python

# -----------------------------------------------------------
# 12. COMPARISON WITH LIST COMPREHENSION
# -----------------------------------------------------------
print("\n---- filter() vs List Comprehension ----")

data = [1, 2, 3, 4, 5, 6]

# Using filter
result_filter = list(filter(lambda x: x % 2 == 0, data))

# Using list comprehension
result_list = [x for x in data if x % 2 == 0]

print("Using filter:", result_filter)
print("Using list comprehension:", result_list)

"""
✔ List comprehension is often more readable
✔ filter() is useful when function already exists
"""
```
