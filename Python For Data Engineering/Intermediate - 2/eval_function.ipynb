```python
"""
=========================================================
                    eval() IN PYTHON 
=========================================================

This file covers:
✔ Basics of eval()
✔ Syntax and arguments
✔ Examples (math, list, dict)
✔ globals and locals usage
✔ Dynamic input
✔ Security risks
✔ Restricting eval() using dictionaries (IMPORTANT)
✔ Whitelisting variables and functions
✔ Using math functions safely
✔ Function aliasing
✔ Safer alternatives

=========================================================
1. WHAT IS eval()?
=========================================================

eval() executes a string as a Python expression and returns the result.
"""

# Basic Example
print("1.", eval("2 + 3 * 4"))  # 14

```

```python

"""
=========================================================
2. USING VARIABLES
=========================================================
"""

a = 10
b = 20
print("2.", eval("a + b"))

```

```python

"""
=========================================================
3. DIFFERENT DATA TYPES
=========================================================
"""

print("3.", eval("[1, 2, 3]"))
print("4.", eval("{'name':'Rajan', 'age':22}"))
print("5.", eval("10 > 5"))

```

```python

"""
=========================================================
4. DYNAMIC USER INPUT (UNSAFE ⚠️)
=========================================================
"""

user_input = "5 * (2 + 3)"
print("6.", eval(user_input))

```

```python

"""
=========================================================
5. SECURITY RISK ⚠️
=========================================================
"""

# Dangerous Example
print("\n7. Running malicious code:")
eval("__import__('os').system('echo HACKED')")

```

```python

"""
=========================================================
6. RESTRICTING eval() USING DICTIONARIES 
=========================================================

We can control what eval() can access using:
- globals dictionary
- locals dictionary
"""

# Allow ONLY variable x
safe_globals = {"__builtins__": None}
safe_locals = {"x": 10}

print("\n8. Only x allowed:", eval("x + 5", safe_globals, safe_locals))

```

```python

"""
=========================================================
7. BLOCKING UNAUTHORIZED VARIABLES
=========================================================
"""

try:
    print(eval("y + 5", safe_globals, safe_locals))
except Exception as e:
    print("9. Error (y not allowed):", e)

```

```python

"""
=========================================================
8. ADDING MORE VARIABLES (x, y)
=========================================================
"""

safe_locals = {"x": 10, "y": 20}
print("10. x and y allowed:", eval("x + y", safe_globals, safe_locals))

```

```python

"""
=========================================================
9. RESTRICTING FUNCTIONS (WHITELIST APPROACH)
=========================================================
"""

import math

safe_globals = {
    "__builtins__": None,
    "sqrt": math.sqrt
}

print("11. sqrt allowed:", eval("sqrt(16)", safe_globals))

```

```python

"""
=========================================================
10. BLOCKING NON-WHITELISTED FUNCTIONS
=========================================================
"""

try:
    print(eval("pow(2,3)", safe_globals))
except Exception as e:
    print("12. pow blocked:", e)

```

```python

"""
=========================================================
11. ALLOWING MULTIPLE SAFE FUNCTIONS
=========================================================
"""

safe_globals = {
    "__builtins__": None,
    "sqrt": math.sqrt,
    "pow": pow
}

print("13. pow allowed:", eval("pow(2,3)", safe_globals))

```

```python

"""
=========================================================
12. FUNCTION ALIASING (ADVANCED SECURITY)
=========================================================

We can rename functions to hide actual names.
"""

safe_globals = {
    "__builtins__": None,
    "square_root": math.sqrt
}

print("14. Using alias:", eval("square_root(25)", safe_globals))

```

```python

"""
=========================================================
13. PREVENTING UNAUTHORIZED FUNCTION CALLS
=========================================================
"""

try:
    print(eval("__import__('os').system('dir')", safe_globals))
except Exception as e:
    print("15. Unauthorized blocked:", e)

```

```python

"""
=========================================================
14. CUSTOM SAFE ENVIRONMENT (BEST PRACTICE)
=========================================================
"""

safe_globals = {
    "__builtins__": None,
    "sqrt": math.sqrt,
    "pow": pow
}

safe_locals = {
    "x": 5,
    "y": 10
}

expression = "pow(x, 2) + sqrt(y)"
print("16. Custom safe eval:", eval(expression, safe_globals, safe_locals))

```

```python

"""
=========================================================
15. SAFER ALTERNATIVE: ast.literal_eval()
=========================================================
"""

import ast

print("17.", ast.literal_eval("[1,2,3]"))

# This will fail (safe behavior)
try:
    ast.literal_eval("2 + 3")
except Exception as e:
    print("18. literal_eval blocked expression:", e)

```

```python

"""
=========================================================
16. eval() vs exec()
=========================================================
"""

print("19. eval:", eval("5 + 5"))

code = "x = 10\nprint('20. exec x =', x)"
exec(code)

```

```python

"""
=========================================================
FINAL SUMMARY
=========================================================

✔ eval() executes string expressions
✔ Can use variables and functions
✔ globals & locals control execution environment

🔥 MOST IMPORTANT:
✔ Restrict eval using dictionaries
✔ Whitelist variables (x, y)
✔ Whitelist functions (sqrt, pow)
✔ Block everything else using "__builtins__": None
✔ Use aliasing for extra safety

❌ NEVER trust user input directly
❌ eval() can execute harmful system commands

👉 BEST PRACTICE:
- Avoid eval() when possible
- Use ast.literal_eval() for safety
- Always restrict environment if using eval()

=========================================================
"""
```
