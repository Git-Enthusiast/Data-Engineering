```python
"""
===========================================================
PYTHON FUNCTIONS - COMPLETE BEGINNER GUIDE
===========================================================

"""


# ---------------------------------------------------------
# 1. WHY FUNCTIONS ARE NEEDED
# ---------------------------------------------------------

"""
Imagine writing a bank application.

Without functions:

You might repeat the same code many times:
- deposit money
- withdraw money
- check balance
- calculate interest

This makes code:
❌ Long
❌ Hard to debug
❌ Difficult to maintain

Functions solve this problem.

Functions allow us to:
✔ Divide code into small tasks
✔ Reuse code
✔ Make programs organized
✔ Simplify testing
"""


# ---------------------------------------------------------
# 2. WHAT IS A FUNCTION?
# ---------------------------------------------------------

"""
A Function is a reusable block of code that performs
a specific task.

Example tasks:
- Calculate interest
- Validate login
- Convert temperature
- Process data

Syntax:

def function_name(parameters):
    code block
"""

```

```python

# ---------------------------------------------------------
# 3. BASIC FUNCTION EXAMPLE
# ---------------------------------------------------------

def greet():
    """This function prints a greeting message."""
    print("Hello! Welcome to Python Functions Tutorial.")


# calling the function
greet()


# ---------------------------------------------------------
# 4. FUNCTIONS WITH PARAMETERS
# ---------------------------------------------------------

"""
Parameters are variables defined in the function definition.
They act as placeholders for values.

"""

def greet_user(name):
    """Function that greets a specific user."""
    print("Hello", name, "Welcome!")


# calling the function
greet_user("Rajan")
greet_user("Amit")


# ---------------------------------------------------------
# 5. PARAMETERS VS ARGUMENTS
# ---------------------------------------------------------

"""
PARAMETER
---------
Variable used in function definition.

Example:
def add(a, b):
        ↑ ↑
   parameters


ARGUMENT
--------
Actual value passed to function.

Example:
add(10,20)
    ↑  ↑
 arguments
"""


def add_numbers(a, b):
    """Function to add two numbers."""
    result = a + b
    print("Sum =", result)


add_numbers(5, 7)


# ---------------------------------------------------------
# 6. FUNCTION WITH RETURN VALUE
# ---------------------------------------------------------

"""
Functions can return results using 'return'.

return sends the result back to the caller.
"""


def multiply(a, b):
    result = a * b
    return result


value = multiply(4, 5)

print("Multiplication Result:", value)


# ---------------------------------------------------------
# 7. PRACTICAL EXAMPLE - SIMPLE INTEREST
# ---------------------------------------------------------

"""
Formula:

Simple Interest = (P × R × T) / 100

Where
P = Principal
R = Rate of interest
T = Time
"""


def calculate_simple_interest(principal, rate, time):
    """
    Function to calculate simple interest.

    Parameters:
    principal : amount of money
    rate : interest rate
    time : duration
    """

    interest = (principal * rate * time) / 100

    return interest


# calling function
p = 10000
r = 5
t = 2

si = calculate_simple_interest(p, r, t)

print("Simple Interest =", si)

```

```python
""" 
In Functional Programming, there are four types of identifiers.
    1. Local Identifiers
    2. Global Identifiers
    3. Non-Local Identifiers
    4. Built-in Identifiers

"""

# Local Identifiers

def display(name):  # Here name is a parameter which is 
                    #  also a local variable to the diaplay function
    age = 22    # Local Variable
    print(f"{name} has age {age}")
    print(age)
    print(locals()) # locals() is a builtin funtion which 
                    #  returns a dict of local variables in current scope
    print(locals()["age"])  # use check whether a varibale exits or not in current scope
display("Rajan Raj")
```

```python

# ---------------------------------------------------------
# 8. VARIABLE SCOPE
# ---------------------------------------------------------

"""
Scope means the region where a variable is accessible.

Python uses LEGB rule.

L → Local
E → Enclosed
G → Global
B → Built-in
"""


# ---------------------------------------------------------
# LOCAL SCOPE EXAMPLE
# ---------------------------------------------------------

def example_scope():
    x = 10  # local variable
    print("Inside function:", x)


example_scope()

# print(x)   # This would cause an error because x is local

```

```python

# ---------------------------------------------------------
# GLOBAL VARIABLE EXAMPLE
# ---------------------------------------------------------

y = 100  # global variable


def show_global():
    global y
    print("Global variable inside function:", y)
    y =23
    print(f"The local variable: {y}")


show_global()

print("Global variable outside function:", y)
```

```python


# ---------------------------------------------------------
#  globals() FUNCTION
# ---------------------------------------------------------

"""
globals() returns a dictionary containing
all global variables in the current module.

Useful for:
- inspecting global variables
- debugging
- dynamically accessing global values
"""

# global variables
x = 100
name = "Rajan"


def globals_demo():
    y = 50  # local variable

    print("Access Number of global variable using globals():", len(globals()))
    print("All Global Variables:", globals())

print("Access Number of global variable using globals():", len(globals()))
globals_demo()
```

```python


# ---------------------------------------------------------
# 9. INTERNAL WORKING OF FUNCTIONS
# ---------------------------------------------------------

"""
When Python calls a function, several steps happen internally.

Step 1
-------
Python creates a new memory space (stack frame) for function .

Step 2
-------
Parameters are assigned values.

Step 3
-------
Local variables are created.

Step 4
-------
Function executes line by line.

Step 5
-------
Return value is sent back.

Step 6
-------
Local memory is deleted by Garbage Collector.
"""


def memory_demo(a, b):
    c = a + b
    print("Inside function:", c)
    return c


memory_demo(10, 20)

```

```python

# ---------------------------------------------------------
# 10. locals() FUNCTION
# ---------------------------------------------------------

"""
locals() returns a dictionary containing
all local variables inside a function.

Useful for:
- debugging
- checking variable existence
"""


def locals_demo():
    a = 10
    b = 20
    c = a + b

    print("Local Variables:", locals())


locals_demo()


# ---------------------------------------------------------
# CHECKING VARIABLE EXISTENCE USING locals()
# ---------------------------------------------------------

def check_variable():
    x = 5

    if "x" in locals():
        print("Variable x exists")

    if "y" not in locals():
        print("Variable y does not exist")


check_variable()

```

```python

# ---------------------------------------------------------
# 11. BANK APPLICATION ANALOGY
# ---------------------------------------------------------

"""
Imagine a banking system.

Instead of writing everything in one place,
we create different functions.
"""


def deposit(balance, amount):
    balance = balance + amount
    return balance


def withdraw(balance, amount):
    balance = balance - amount
    return balance


def check_balance(balance):
    print("Current Balance:", balance)


# simulation
balance = 1000

balance = deposit(balance, 500)
check_balance(balance)

balance = withdraw(balance, 200)
check_balance(balance)
```
