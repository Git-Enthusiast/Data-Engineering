```python
"""
=========================================================
    COMPLETE GUIDE TO COMMAND LINE ARGUMENTS IN PYTHON
                (USING sys MODULE)
=========================================================

This file covers:
✔ Basics of command line arguments
✔ sys.argv usage
✔ Accessing & iterating arguments
✔ Error handling
✔ Type conversion
✔ Practical calculator
✔ File handling using arguments (NEW - IMPORTANT)
✔ Real-world use cases
✔ IDE usage
✔ argparse (advanced)

=========================================================
1. WHAT ARE COMMAND LINE ARGUMENTS?
=========================================================

Arguments passed while running script from terminal.

Example:
    python script.py 10 20
"""

import sys

print("1. sys.argv:", sys.argv)


"""
=========================================================
2. UNDERSTANDING sys.argv
=========================================================
"""

print("\n2. Script Name:", sys.argv[0])

if len(sys.argv) > 1:
    print("3. First Argument:", sys.argv[1])
else:
    print("3. No argument provided")


"""
=========================================================
3. ITERATING ARGUMENTS
=========================================================
"""

print("\n4. All Arguments:")
for i, arg in enumerate(sys.argv):
    print(f"Index {i} → {arg}")


"""
=========================================================
4. HANDLING MISSING ARGUMENTS
=========================================================
"""

if len(sys.argv) < 3:
    print("\n5. ERROR: Provide at least 2 arguments")
else:
    print("5. Arguments OK")


"""
=========================================================
5. TYPE CONVERSION
=========================================================
"""

if len(sys.argv) >= 3:
    try:
        a = int(sys.argv[1])
        b = int(sys.argv[2])
        print("\n6. Sum:", a + b)
    except:
        print("6. Not numeric arguments")


"""
=========================================================
6. CALCULATOR USING COMMAND LINE
=========================================================
"""

if len(sys.argv) == 4:
    try:
        a = int(sys.argv[1])
        op = sys.argv[2]
        b = int(sys.argv[3])

        if op == "+":
            print("\n7. Result:", a + b)
        elif op == "-":
            print("7. Result:", a - b)
        elif op == "*":
            print("7. Result:", a * b)
        elif op == "/":
            print("7. Result:", a / b)
        else:
            print("7. Invalid operator")
    except Exception as e:
        print("7. Error:", e)


"""
=========================================================
7. FILE HANDLING USING COMMAND LINE ARGUMENTS (NEW 🔥)
=========================================================

👉 Read from one file and write to another

Usage:
    python script.py input.txt output.txt
"""

if len(sys.argv) == 3:
    source_file = sys.argv[1]
    destination_file = sys.argv[2]

    try:
        # Read from source file
        with open(source_file, "r") as f:
            data = f.read()

        # Write to destination file
        with open(destination_file, "w") as f:
            f.write(data)

        print("\n8. File copied successfully!")
        print(f"   From: {source_file}")
        print(f"   To: {destination_file}")

    except FileNotFoundError:
        print("8. ERROR: Source file not found")
    except Exception as e:
        print("8. Error:", e)


"""
=========================================================
8. IMPROVED FILE HANDLING (ADVANCED)
=========================================================
"""

if len(sys.argv) == 3:
    try:
        src = sys.argv[1]
        dest = sys.argv[2]

        with open(src, "r") as f:
            lines = f.readlines()

        # Example: Modify content before writing
        modified_lines = [line.upper() for line in lines]

        with open(dest, "w") as f:
            f.writelines(modified_lines)

        print("\n9. File processed & written (uppercase conversion)")

    except Exception as e:
        print("9. File processing error:", e)


"""
=========================================================
9. REAL-WORLD USE CASES
=========================================================

✔ File processing scripts
✔ Data pipelines
✔ Automation tools
✔ Passing credentials
✔ Batch jobs

Example:
    python app.py input.csv output.csv
"""


"""
=========================================================
10. DEFAULT VALUES
=========================================================
"""

name = sys.argv[1] if len(sys.argv) > 1 else "Guest"
print("\n10. Hello,", name)


"""
=========================================================
11. ERROR HANDLING BEST PRACTICE
=========================================================
"""

try:
    if len(sys.argv) < 2:
        raise Exception("No argument provided")
    print("11. Argument:", sys.argv[1])
except Exception as e:
    print("11. Exception:", e)


"""
=========================================================
12. HOW TO RUN
=========================================================

Terminal:
    python script.py input.txt output.txt

=========================================================
13. IDE USAGE
=========================================================

PyCharm:
Run → Edit Configurations → Parameters → enter:
    input.txt output.txt

VS Code:
launch.json → "args": ["input.txt", "output.txt"]
"""


"""
=========================================================
14. LIMITATIONS
=========================================================

❌ Arguments are strings
❌ Manual validation required
❌ Not scalable for complex apps

👉 Use argparse for production
"""


"""
=========================================================
15. argparse (ADVANCED)
=========================================================
"""

import argparse

parser = argparse.ArgumentParser(description="File Copy Tool")

parser.add_argument("source", help="Source file")
parser.add_argument("destination", help="Destination file")

# Uncomment to test
# args = parser.parse_args()
# print("Source:", args.source)
# print("Destination:", args.destination)


"""
=========================================================
FINAL SUMMARY
=========================================================

✔ sys.argv stores command line arguments
✔ sys.argv[0] → script name
✔ sys.argv[1:] → inputs

🔥 MOST IMPORTANT:
✔ Always validate arguments
✔ Handle missing inputs
✔ Convert types properly
✔ Use for file handling (real-world)

🚀 ADVANCED:
✔ Use argparse for scalable applications

=========================================================
"""
```
