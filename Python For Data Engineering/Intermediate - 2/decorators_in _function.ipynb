```python

# ===================== 1. FUNCTIONS ARE FIRST-CLASS OBJECTS =====================
# In Python, functions behave like variables

def greet():
    return "Hello!"

# Assign function to a variable
say_hello = greet
print("1.", say_hello())

# Pass function as argument

def execute_function(func):
    print("2.", func())

execute_function(greet)

# Return function from another function

def outer_function():
    def inner_function():
        return "Inner Function Called"
    return inner_function

returned_func = outer_function()
print("3.", returned_func())

```

```python

# ===================== 2. NESTED FUNCTIONS =====================
def outer():
    print("Outer function")

    def inner():
        print("Inner function")

    inner()

outer()

```

```python

# ===================== 3. CLOSURES =====================
# Closure = function remembers variables from enclosing scope

def outer_closure(x):
    def inner_closure(y):
        return x + y
    return inner_closure

closure_func = outer_closure(10)
print("4.", closure_func(5))  # 15

```

```python

# ===================== 4. WHAT IS A DECORATOR =====================
# A decorator is a function that wraps another function
# and adds extra functionality without modifying original function


def simple_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper


def say_hi():
    print("Hi!")

# Manual decoration
say_hi = simple_decorator(say_hi)
say_hi()

```

```python

# ===================== 5. DECORATOR USING @ SYNTAX =====================

def decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper

@decorator
def hello():
    print("Hello World")

hello()

```

```python

# ===================== 6. DECORATORS WITH ARGUMENTS =====================

def decorator_with_args(func):
    def wrapper(a, b):
        print("Before execution")
        result = func(a, b)
        print("After execution")
        return result
    return wrapper

@decorator_with_args
def add(a, b):
    return a + b

print("5.", add(3, 4))

```

```python

# ===================== 7. ADDING EXTRA FUNCTIONALITY =====================
# Example: Add third number automatically

def add_extra(func):
    def wrapper(a, b):
        c = 10
        print("Adding extra value:", c)
        return func(a, b) + c
    return wrapper

@add_extra
def add_two_numbers(a, b):
    return a + b

print("6.", add_two_numbers(5, 5))  # (5+5)+10 = 20

```

```python

# ===================== 8. USING *args AND **kwargs =====================
# Best practice: always use this in decorators

def flexible_decorator(func):
    def wrapper(*args, **kwargs):
        print("Arguments received:", args, kwargs)
        return func(*args, **kwargs)
    return wrapper

@flexible_decorator
def multiply(a, b):
    return a * b

print("7.", multiply(3, 5))

```

```python

# ===================== 9. MULTIPLE DECORATORS =====================

def decorator1(func):
    def wrapper(*args, **kwargs):
        print("Decorator 1")
        return func(*args, **kwargs)
    return wrapper


def decorator2(func):
    def wrapper(*args, **kwargs):
        print("Decorator 2")
        return func(*args, **kwargs)
    return wrapper

@decorator1
@decorator2
def test():
    print("Function executed")

# Execution order: decorator1(decorator2(test))
test()

```

```python

# ===================== 10. REAL-WORLD USE CASES =====================

# ---- Logging Decorator ----
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__}")
        return func(*args, **kwargs)
    return wrapper

@logger
def greet_user(name):
    return f"Hello {name}"

print("8.", greet_user("Rajan"))


# ---- Authentication Decorator ----
def require_login(func):
    def wrapper(user):
        if user != "admin":
            return "Access Denied"
        return func(user)
    return wrapper

@require_login
def dashboard(user):
    return "Welcome to dashboard"

print("9.", dashboard("admin"))
print("10.", dashboard("guest"))


# ---- Timing Decorator ----
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Execution Time: {end - start:.5f} seconds")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(1)
    return "Done"

print("11.", slow_function())


# ---- Simple Caching Decorator ----
def cache(func):
    memo = {}
    def wrapper(n):
        if n in memo:
            return memo[n]
        result = func(n)
        memo[n] = result
        return result
    return wrapper

@cache
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print("12.", fibonacci(10))

```

```python

# ===================== 11. CLASS DECORATORS =====================

def class_decorator(cls):
    cls.category = "Decorated Class"
    return cls

@class_decorator
class MyClass:
    pass

obj = MyClass()
print("13.", obj.category)

```

```python

# ===================== 12. BUILT-IN DECORATORS =====================

class Example:
    def __init__(self, value):
        self._value = value

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, new_value):
        self._value = new_value

    @staticmethod
    def static_method():
        return "I am static"

    @classmethod
    def class_method(cls):
        return f"Class is {cls.__name__}"

ex = Example(10)
print("14.", ex.value)
ex.value = 20
print("15.", ex.value)
print("16.", Example.static_method())
print("17.", Example.class_method())

```

```python

# ===================== 13. functools.wraps (VERY IMPORTANT) =====================
from functools import wraps


def better_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Wrapped function")
        return func(*args, **kwargs)
    return wrapper

@better_decorator
def sample_function():
    """This is original docstring"""
    return "Done"

print("18.", sample_function.__name__)  # preserves original name
print("19.", sample_function.__doc__)   # preserves docstring

```
