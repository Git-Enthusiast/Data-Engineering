```python

from functools import reduce


def print_section(title: str) -> None:
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)


def intro() -> None:
    print_section("1) What is a Lambda Function?")
    print(
        "A lambda function is a small anonymous function in Python.\n"
        "Anonymous means it usually has no name, and it is written in a single line.\n"
        "It is mainly used for short, simple operations."
    )

    print("\nKey idea:")
    print("  lambda arguments: expression")
    print("Example:")
    print("  add = lambda x, y: x + y")
    print("  print(add(3, 4))  ->  7")


def syntax_and_rules() -> None:
    print_section("2) Syntax and Rules")
    print(
        "General syntax:\n"
        "  lambda arguments: expression\n\n"
        "Rules:\n"
        "  - lambda can take any number of arguments\n"
        "  - lambda can contain only one expression\n"
        "  - it returns the result automatically\n"
        "  - you do NOT use the return keyword\n"
        "  - it is often used for small, one-time tasks"
    )

```

```python

def normal_vs_lambda() -> None:
    print_section("3) Normal Function vs Lambda Function")

    def normal_add(a, b):
        return a + b

    lambda_add = lambda a, b: a + b  # noqa: E731

    print("Normal function:")
    print("  def normal_add(a, b):")
    print("      return a + b")
    print("  normal_add(3, 4) =", normal_add(3, 4))

    print("\nLambda function:")
    print("  lambda_add = lambda a, b: a + b")
    print("  lambda_add(3, 4) =", lambda_add(3, 4))

    print(
        "\nDifference:\n"
        "  - normal function is written with def and has a name\n"
        "  - lambda function is usually shorter and anonymous\n"
        "  - use def for larger logic, use lambda for small expressions"
    )

```

```python

def basic_examples() -> None:
    print_section("4) Basic Examples")

    add = lambda x, y: x + y  # noqa: E731
    square = lambda x: x * x  # noqa: E731
    greet = lambda name: f"Hello, {name}!"  # noqa: E731

    print("Add two numbers:", add(10, 5))
    print("Square of 6:", square(6))
    print(greet("Rajan"))

    print("\nImmediate use without storing in a variable:")
    print((lambda x, y: x * y)(4, 5))
```

```python


def lambda_multiple_arguments() -> None:
    print_section("5) Lambda with Multiple Arguments")

    area = lambda length, width: length * width  # noqa: E731
    max_two = lambda n1, n2: n1 if n1 >= n2 else n2  # noqa: E731

    print("Area of rectangle (7 x 3):", area(7, 3))
    print("Maximum of 12 and 19:", max_two(12, 19))

    print(
        "\nThis shows that lambda can take multiple arguments,\n"
        "but the body must still be a single expression."
    )
```

```python


def lambda_returning_tuple() -> None:
    print_section("6) Lambda Returning a Tuple")

    pair = lambda x: (x, x * x)  # noqa: E731
    print("pair(5) ->", pair(5))
    print(
        "\nA tuple counts as one expression, so lambda can return it.\n"
        "This is useful when you want to return multiple values in a compact way."
    )
```

```python
print(lambda_returning_tuple())
```

```python


def lambda_with_if_else() -> None:
    print_section("7) Lambda with if-else")

    max_two = lambda a, b: a if a >= b else b  # noqa: E731
    min_two = lambda a, b: a if a <= b else b  # noqa: E731
    parity = lambda n: "Even" if n % 2 == 0 else "Odd"  # noqa: E731

    print("Max of 8 and 15:", max_two(8, 15))
    print("Min of 8 and 15:", min_two(8, 15))
    print("Parity of 11:", parity(11))

    print(
        "\nLambda works well with short conditional expressions.\n"
        "For big if-else blocks, use a normal function instead."
    )
```

```python


def map_filter_reduce_examples() -> None:
    print_section("8) Lambda with map(), filter(), and reduce()")

    numbers = [1, 2, 3, 4, 5, 6]

    squared = list(map(lambda x: x * x, numbers))  # noqa: E731
    evens = list(filter(lambda x: x % 2 == 0, numbers))  # noqa: E731
    product = reduce(lambda a, b: a * b, numbers)  # noqa: E731

    print("Numbers:", numbers)
    print("Squares using map:", squared)
    print("Even numbers using filter:", evens)
    print("Product of all numbers using reduce:", product)

    print(
        "\nThese are classic use-cases for lambda because the function is short\n"
        "and is passed directly as an argument to another function."
    )
```

```python


def sorted_examples() -> None:
    print_section("9) Lambda with sorted()")

    numbers = [45, 2, 88, 19, 7]
    words = ["banana", "apple", "kiwi", "pineapple", "fig"]
    names = ["John Doe", "Alice Brown", "Charlie Smith", "Bob Stone"]

    print("Original numbers:", numbers)
    print("sorted(numbers):", sorted(numbers))
    print("sorted(numbers, reverse=True):", sorted(numbers, reverse=True))

    print("\nSort by length of words using key=len:")
    print(sorted(words, key=len))

    print("\nSort words by last character using lambda:")
    print(sorted(words, key=lambda word: word[-1]))  # noqa: E731

    print("\nSort names by last name (default is first word order, not name logic):")
    print(sorted(names, key=lambda name: name.split()[-1]))  # noqa: E731

    print("\nSort names by first name:")
    print(sorted(names, key=lambda name: name.split()[0]))  # noqa: E731

    print(
        "\nWhat sorted() does:\n"
        "  - returns a NEW sorted list\n"
        "  - does not change the original list\n"
        "  - can use key= to tell Python how to compare items"
    )

    print("\nOriginal names list is still unchanged:")
    print(names)
```

```python


def nested_function_example() -> None:
    print_section("10) Nested Function Example and Lambda Equivalent")

    def outer():
        def add(num1, num2):
            return num1 + num2

        return add

    add_function = outer()
    print("Nested function demo: outer() returns inner function add()")
    print("outer()(3, 4) =", add_function(3, 4))

    print("\nA lambda equivalent for the inner add function:")
    outer_lambda = lambda: (lambda num1, num2: num1 + num2)  # noqa: E731
    add_lambda = outer_lambda()
    print("outer_lambda()(3, 4) =", add_lambda(3, 4))

    print(
        "\nThis is close to the idea shown in the screenshot/video:\n"
        "a function can return another function. Lambda can also be used\n"
        "to create compact function objects, but nested def is usually clearer."
    )
```

```python


def memory_and_performance_explanation() -> None:
    print_section("11) How do Lambda Functions save memory?")

    print(
        "Important concept:\n"
        "  Lambda functions can save memory in the sense that they are often\n"
        "  used inline without creating a separately named function definition.\n"
        "  However, the real memory benefit is usually small.\n\n"
        "The main advantage is simplicity and convenience for short tasks.\n"
        "You avoid writing an extra def function when you need a function only once.\n\n"
        "So, lambda is best thought of as a compact function tool,\n"
        "not a major memory optimization technique."
    )
```

```python


def limitations_and_best_practices() -> None:
    print_section("12) Limitations and Best Practices")

    print(
        "Limitations:\n"
        "  - only one expression is allowed\n"
        "  - no multiple statements\n"
        "  - no explicit return keyword\n"
        "  - complex logic becomes hard to read\n\n"
        "Best practices:\n"
        "  - use lambda for small, short operations\n"
        "  - use def for large, reusable, or readable code\n"
        "  - use lambda when passing a tiny function to map/filter/sorted/reduce"
    )

```

```python


def extra_examples() -> None:
    print_section("14) Extra Examples")

    # Lambda with list comprehension inside a normal function call
    nums = [1, 2, 3, 4, 5]
    squares_list_comp = (lambda lst: [x * x for x in lst])(nums)  # noqa: E731
    print("Squares using lambda + list comprehension:", squares_list_comp)

    # Lambda with string processing
    upper_name = lambda name: name.strip().upper()  # noqa: E731
    print("Uppercase name:", upper_name("   rajan raj   "))

    # Lambda inside list sorting by custom dictionary key
    students = [
        {"name": "Aman", "marks": 78},
        {"name": "Ravi", "marks": 92},
        {"name": "Zoya", "marks": 85},
    ]
    print("Students sorted by marks:")
    print(sorted(students, key=lambda student: student["marks"]))  # noqa: E731
```
