```python

"""
----------------------------------------------------------
🔹 3. RECURSION WITH BASE CONDITION
----------------------------------------------------------
"""

def count_down(n):
    if n == 0:
        print("Done!")
        return
    print(n)
    count_down(n - 1)
print(count_down(12))
```

```python

"""
----------------------------------------------------------
🔹 4. FACTORIAL
----------------------------------------------------------
"""

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(6))
```

```python

"""
----------------------------------------------------------
🔹 5. FIBONACCI
----------------------------------------------------------
"""

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(12))
```

```python
def fibonacci(n):
    a, b = 0, 1
    
    for i in range(n):
        a, b = b, a + b
    
    return a
print(fibonacci(0))
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(3))
print(fibonacci(4))
```

```python

"""
----------------------------------------------------------
🔹 7. TYPES OF RECURSION
----------------------------------------------------------
"""

# Direct Recursion
def direct_example(n):
    if n == 0:
        return
    print(n)
    direct_example(n - 1)
print(direct_example(10))
```

```python

# Indirect Recursion
def funcA(n):
    if n <= 0:
        return
    print("A:", n)
    funcB(n - 1)

def funcB(n):
    if n <= 0:
        return
    print("B:", n)
    funcA(n - 1)
print(funcA(12))
```

```python

"""
----------------------------------------------------------
🔹 8. MEMOIZATION (OPTIMIZED)
----------------------------------------------------------
"""

def fibonacci_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
    return memo[n]
```

```python

# EASY SOLUTIONS

def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n - 1)
    print(n)

def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n - 1)

def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n - 1)
```

```python

"""
----------------------------------------------------------
🟡 MEDIUM SOLUTIONS
----------------------------------------------------------
"""

def reverse_string(s):
    if s == "":
        return ""
    return reverse_string(s[1:]) + s[0]

def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])


def sum_of_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_of_digits(n // 10)
```

```python

"""
----------------------------------------------------------
🔴 HARD SOLUTIONS
----------------------------------------------------------
"""

# Tower of Hanoi
def tower_of_hanoi(n, source, helper, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    tower_of_hanoi(n - 1, source, destination, helper)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n - 1, helper, source, destination)

# Binary Search (Recursive)
def binary_search(arr, low, high, target):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif target < arr[mid]:
        return binary_search(arr, low, mid - 1, target)
    else:
        return binary_search(arr, mid + 1, high, target)
```
