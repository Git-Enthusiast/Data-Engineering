```python
# Nested function
# nested_function_example.py
# Example to demonstrate Nested Functions in Python

def outer_function(name):
    print("Outer function started")

    # Nested function (function inside another function)
    def inner_function():
        print("Hello", name)
        print("This is the inner function")

    # Calling the inner function
    inner_function()

    print("Outer function ended")


# Calling the outer function
outer_function("Rajan")
```

```python
# positional and keyword arguements 

# Function definition
def student_details(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)
    print("-------------------")


print("Using Positional Arguments")
# Positional Arguments (order matters)
student_details("Rajan", 22, "Computer Science")


print("Using Keyword Arguments")
# Keyword Arguments (order does not matter)
student_details(course="Computer Science", name="Rajan", age=22)
```

```python
# default_arguments_example.py
# Example of Default Arguments in Python

def greet(name, message="Welcome"):
    print(message, name)

# Calling function with both arguments
greet("Rajan", "Good Morning")

# Calling function with only one argument
# Default value of message will be used
greet("Rajan")
```

```python
# variable_arguments_example.py
# Example of *args and **kwargs

# Example of *args
def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum =", total)

# Calling function with different number of arguments
add_numbers(10, 20)
add_numbers(5, 10, 15, 20)


print("----------------------")

# Example of **kwargs
def student_info(**details):
    for key, value in details.items():
        print(key, ":", value)

# Calling function with keyword arguments
student_info(name="Rajan", age=22, course="CSE")
```

```python
# Mutable default arguements
# Example to demonstrate Mutable Default Arguments in Python

# Function with a mutable default argument (list)
def add_item(item, item_list=[]):
    item_list.append(item)
    return item_list

# Calling the function multiple times
print(add_item("Apple"))
print(add_item("Banana"))
print(add_item("Mango"))

'''

🔹 Explanation
The default argument item_list=[] is a mutable object (list).
In Python, default arguments are created only once when the function is defined.
So the same list is shared across multiple function calls.

That is why:

First call → ['Apple']
Second call → ['Apple', 'Banana']
Third call → ['Apple', 'Banana', 'Mango']

The list keeps growing instead of resetting.

''' 

#✅ Correct Way to Handle Mutable Default Arguments
#📄 Example: mutable_default_correct.py
# Correct way to avoid mutable default argument problem

def add_item(item, item_list=None):
    if item_list is None:
        item_list = []

    item_list.append(item)
    return item_list

print(add_item("Apple"))
print(add_item("Banana"))
print(add_item("Mango"))


''' 
📌 Key Points
Mutable objects: list, dictionary, set.
Default mutable arguments can cause unexpected behavior.
Best practice is to use None as default value and create a new object inside the function.
'''

```

```python
# python_functions_demo.py
# Demonstration of different types of functions and arguments in Python

print("----- 1. Nested Function Example -----")

def outer_function(name):
    def inner_function():
        print("Hello", name)
    inner_function()

outer_function("Rajan")


print("\n----- 2. Positional Arguments -----")

def student_info(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

student_info("Rajan", 22, "Computer Science")


print("\n----- 3. Keyword Arguments -----")

student_info(age=22, course="Computer Science", name="Rajan")


print("\n----- 4. Default Arguments -----")

def greet(name, message="Welcome"):
    print(message, name)

greet("Rajan")
greet("Rajan", "Good Morning")


print("\n----- 5. Variable Length Arguments (*args) -----")

def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum =", total)

add_numbers(10, 20)
add_numbers(5, 10, 15, 20)


print("\n----- 6. Variable Length Keyword Arguments (**kwargs) -----")

def display_details(**details):
    for key, value in details.items():
        print(key, ":", value)

display_details(name="Rajan", age=22, course="CSE")


print("\n----- 7. Mutable Default Argument Problem -----")

def add_item(item, item_list=[]):
    item_list.append(item)
    return item_list

print(add_item("Apple"))
print(add_item("Banana"))


print("\n----- 8. Correct Way to Handle Mutable Default Argument -----")

def add_item_safe(item, item_list=None):
    if item_list is None:
        item_list = []
    item_list.append(item)
    return item_list

print(add_item_safe("Apple"))
print(add_item_safe("Banana"))
```
