# 🧠 What is `map()` in Python?

`map()` is a built-in **higher-order function** in Python.

## 👉 Key Points

- Takes a **function** and an **iterable** as input  
- Applies the function to **each element** of the iterable  
- Returns a **map object (iterator)**  

## 🔹 Syntax

```python
map(function, iterable)

````python
# ==========================================================
# Python map() Function
# ==========================================================

print("\n========== 1. BASIC UNDERSTANDING ==========\n")

# Example 1: Square of numbers
nums = [4, 5, 6, 3, 9]

def square(n):
    return n * n

mapped_obj = map(square, nums)

print("Map object:", mapped_obj)
print("Type:", type(mapped_obj))

# Convert to list
print("Squares:", list(mapped_obj))

````

````python

print("\n========== 2. MAP OBJECT IS USED ONLY ONCE ==========\n")

nums = [1, 2, 3]
mapped_obj = map(square, nums)

print("First use:", list(mapped_obj))
print("Second use:", list(mapped_obj))   # Empty!

````

````python

print("\n========== 3. STRING EXAMPLE ==========\n")

names = ["Raj", "Yash", "Yashraj"]

def find_length(name):
    return len(name)

lengths = list(map(find_length, names))

print("Names:", names)
print("Lengths:", lengths)

````

````python

print("\n========== 4. RETURN MULTIPLE VALUES (TUPLES) ==========\n")

def name_with_length(name):
    return (name, len(name))

mapped_obj = map(name_with_length, names)
print(dict(list(mapped_obj)))

````

````python

print("\n========== 5. USING LAMBDA FUNCTION ==========\n")

nums = [5, 3, 8, 11, 6]

squares = list(map(lambda n: n**2, nums))

print("Squares using lambda:", squares)

````

````python

print("\n========== 6. CONDITIONAL LOGIC IN MAP ==========\n")

def odd_square(n):
    if n % 2 != 0:
        return n**2

result = list(map(odd_square, nums))
print("Odd squares (with None for even):", result)

````

````python
#   Using lambda Expression
nums = [5, 3, 8, 11, 6]
result = list(map(lambda n: n**2 if n % 2 != 0 else None, nums))
print("Odd squares (with None for even):", result)
````

````python

print("\n========== 7. HANDLE NONE CASE ==========\n")

def odd_square_else_same(n):
    if n % 2 != 0:
        return n**2
    else:
        return n

result = list(map(odd_square_else_same, nums))
print("Odd square, even same:", result)

````

````python

print("\n========== 8. DIFFERENT OPERATIONS ==========\n")

def square_or_cube(n):
    if n % 2 != 0:
        return n**2     # square for odd
    else:
        return n**3     # cube for even

result = list(map(square_or_cube, nums))
print("Square (odd), Cube (even):", result)

````

````python

print("\n========== 9. SAME USING LAMBDA ==========\n")

result = list(map(lambda n: n**2 if n % 2 != 0 else n**3, nums))
print("Lambda version:", result)

````

````python

print("\n========== 10. MULTIPLE ITERABLES ==========\n")

a = [1, 2, 3]
b = [4, 5, 6]

# Add corresponding elements
result = list(map(lambda x, y: x + y, a, b))
print("Addition of two lists:", result)

````

````python

print("\n========== 11. MAP WITH STRINGS ==========\n")

words = ["hello", "world", "python"]

# Convert to uppercase
upper_words = list(map(str.upper, words))
print("Uppercase:", upper_words)

````

````python

print("\n========== 12. MAP VS LIST COMPREHENSION ==========\n")

nums = [1, 2, 3, 4]

# Using map
map_result = list(map(lambda x: x * 2, nums))

# Using list comprehension
list_comp_result = [x * 2 for x in nums]

print("Map result:", map_result)
print("List comprehension result:", list_comp_result)

````

````python

print("\n========== 13. REAL WORLD EXAMPLE ==========\n")

prices = [100, 200, 300]

# Add 18% GST
final_prices = list(map(lambda p: p * 1.18, prices))

print("Original prices:", prices)
print("Final prices with GST:", final_prices)

````

````python

print("\n========== 14. IMPORTANT NOTES ==========\n")

print("""
1. map() returns a map object (iterator)
2. It is consumed only once
3. It is memory efficient
4. Works faster than loops in many cases
5. Use lambda for short logic
6. Use normal function for complex logic
""")
````

````python
print("\n========== 4. REAL-WORLD EXAMPLE (BUDGET FILTER) ==========\n")

# Dictionary of laptops with prices
laptops = {
    "HP": 50000,
    "Dell": 65000,
    "Lenovo": 45000,
    "Apple": 120000,
    "Asus": 55000
}

budget = 60000

# Function to filter laptops within budget
def within_budget(brand):
    return laptops[brand] <= budget

filtered_laptops = list(filter(within_budget, laptops))

print("Budget:", budget)
print("Available options:", filtered_laptops)


print("\n========== 5. SHOW NAME + PRICE ==========\n")

result = list(filter(lambda item: item[1] <= 60000, laptops.items()))
print(result)


print("\n========== 6. EMPTY RESULT CASE ==========\n")

budget = 30000

filtered_laptops = list(filter(lambda brand: laptops[brand] <= budget, laptops))

print("Low Budget:", budget)
print("Result:", filtered_laptops)   # Empty list

````
