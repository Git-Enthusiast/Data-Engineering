```python
"""
        ENCAPSULATION IN PYTHON 
============================================

This single .py file explains encapsulation in Python in a detailed, example-driven way.

What this file covers:
1) What encapsulation means
2) Why encapsulation is useful
3) Public, protected, and private members in Python
4) Name mangling
5) How to make private variables and private methods
6) Getter and setter methods
7) Pythonic way using @property
8) Validation and controlled access
9) Real-world examples
10) Tricky cases and interview-style notes

Important idea:
Encapsulation means wrapping data and the methods that work on that data into one unit
(usually a class) and controlling how that data is accessed or changed.

Python does not enforce strict privacy like Java or C++.
Instead, Python uses naming conventions and name mangling.

Run this file directly to see examples in action.
"""
```

```python

# ============================================================
# 1. BASIC IDEA OF ENCAPSULATION
# ============================================================

class Student:
    """
    A simple class to explain encapsulation.

    Here, we keep data and methods together.
    The object controls access to its own data through methods.
    """

    def __init__(self, name: str, age: int, marks: float):
        # Public attribute:
        # Can be accessed directly from outside.
        self.name = name

        # "Protected" attribute by convention:
        # Python does not truly block access, but by convention it means
        # "internal use, do not access directly from outside".
        self._age = age

        # Private attribute:
        # Double underscore triggers name mangling.
        # It becomes harder to access from outside the class.
        self.__marks = marks

    def display(self):
        """
        Public method to show student details.
        Notice that we are accessing the internal data through a class method.
        """
        print("Student name:", self.name)
        print("Student age :", self._age)
        print("Student marks:", self.__marks)

    def get_marks(self):
        """Getter method for private marks."""
        return self.__marks

    def set_marks(self, new_marks: float):
        """
        Setter method for marks with validation.
        This protects the object from invalid values.
        """
        if not isinstance(new_marks, (int, float)):
            print("Marks must be a number.")
            return

        if new_marks < 0 or new_marks > 100:
            print("Marks must be between 0 and 100.")
            return

        self.__marks = new_marks
        print("Marks updated successfully.")

    def update_age(self, new_age: int):
        """Another controlled update method."""
        if new_age < 0:
            print("Age cannot be negative.")
            return
        self._age = new_age
        print("Age updated successfully.")

    def __internal_message(self):
        """
        Private method.
        This is not meant to be called directly from outside.
        """
        print("This is a private internal method.")

    def use_private_method(self):
        """
        Public wrapper method that can call the private method internally.
        """
        self.__internal_message()

```

```python

# ============================================================
# 2. WHY ENCAPSULATION IS NEEDED
# ============================================================

class BankAccount:
    """
    Real-world example.

    A bank balance should not be changed directly by anyone.
    We should control deposit/withdraw through methods.
    """

    def __init__(self, account_holder: str, initial_balance: float = 0.0):
        self.account_holder = account_holder
        self.__balance = initial_balance

    def get_balance(self):
        """Read-only access to balance."""
        return self.__balance

    def deposit(self, amount: float):
        if amount <= 0:
            print("Deposit amount must be positive.")
            return

        self.__balance += amount
        print(f"Deposited {amount}. Current balance = {self.__balance}")

    def withdraw(self, amount: float):
        if amount <= 0:
            print("Withdraw amount must be positive.")
            return

        if amount > self.__balance:
            print("Insufficient balance.")
            return

        self.__balance -= amount
        print(f"Withdrawn {amount}. Current balance = {self.__balance}")

    def __validate_account(self):
        """Private helper method."""
        print("Account validated internally.")

```

```python

# ============================================================
# 3. ACCESS MODIFIERS IN PYTHON
# ============================================================

class AccessDemo:
    """
    Python has these common access levels by convention:
    - public      -> accessible everywhere
    - protected   -> internal/convention only (_name)
    - private     -> name-mangled (__name)
    """

    def __init__(self):
        self.public_var = "I am public"
        self._protected_var = "I am protected by convention"
        self.__private_var = "I am private via name mangling"

    def public_method(self):
        print("This is a public method.")

    def _protected_method(self):
        print("This is a protected method by convention.")

    def __private_method(self):
        print("This is a private method via name mangling.")

    def access_private_inside(self):
        # Private method can be accessed from inside the class.
        self.__private_method()

    def show_vars(self):
        print("public_var   :", self.public_var)
        print("_protected_var:", self._protected_var)
        print("__private_var :", self.__private_var)

```

```python

# ============================================================
# 4. NAME MANGLING EXPLAINED
# ============================================================

class ManglingExample:
    """
    In Python, __variable is internally rewritten as:
        _ClassName__variable

    This is called name mangling.
    """

    def __init__(self):
        self.__secret = "top secret"

    def show_secret_inside(self):
        print("Inside class, secret =", self.__secret)

```

```python

# ============================================================
# 5. PROPERTY-BASED ENCAPSULATION (PYTHONIC STYLE)
# ============================================================

class Temperature:
    """
    A clean Pythonic approach uses @property.

    Why use property?
    - It lets you control access like a getter/setter
    - It looks like normal attribute access
    - It hides validation logic behind attribute syntax
    """

    def __init__(self, celsius: float):
        self.__celsius = None
        self.celsius = celsius  # uses property setter

    @property
    def celsius(self):
        """Getter property."""
        return self.__celsius

    @celsius.setter
    def celsius(self, value):
        """Setter property with validation."""
        if not isinstance(value, (int, float)):
            raise TypeError("Temperature must be a number.")
        if value < -273.15:
            raise ValueError("Temperature cannot be below absolute zero.")
        self.__celsius = value

    @property
    def fahrenheit(self):
        """Read-only computed property."""
        return (self.__celsius * 9 / 5) + 32

    def __str__(self):
        return f"{self.__celsius}°C / {self.fahrenheit}°F"

```

```python

# ============================================================
# 6. ENCAPSULATION IN SUBCLASSING
# ============================================================

class Parent:
    def __init__(self):
        self.__private_value = "parent private"
        self._protected_value = "parent protected"

    def show_inside(self):
        print("Parent sees private:", self.__private_value)
        print("Parent sees protected:", self._protected_value)


class Child(Parent):
    def show_child_access(self):
        # self.__private_value does NOT refer to Parent's private variable.
        # It would be mangled as _Child__private_value if defined here.
        print("Child can access protected directly:", self._protected_value)

    def try_to_access_parent_private(self):
        # The next line would fail if uncommented:
        # print(self.__private_value)
        print("Child cannot directly access Parent.__private_value")

```

```python

# ============================================================
# 7. HELPER FUNCTIONS FOR DEMONSTRATION
# ============================================================

def separator(title: str):
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)


def show_object_dictionary(obj):
    """
    __dict__ shows internal attribute storage.
    This is useful to understand name mangling.
    """
    print(f"{obj.__class__.__name__}.__dict__ of instance:")
    print(obj.__dict__)

```

```python

# ============================================================
# 8. DEMONSTRATIONS
# ============================================================

if __name__ == "__main__":
    separator("1) BASIC ENCAPSULATION EXAMPLE USING Student")

    s = Student("Rajan", 22, 91.5)
    s.display()

    print("\nAccessing public member directly:")
    print(s.name)

    print("\nAccessing protected member directly (allowed, but not recommended):")
    print(s._age)

    print("\nTrying to access private member directly:")
    try:
        print(s.__marks)  # This will fail
    except AttributeError as e:
        print("Error:", e)

    print("\nAccess private data using getter:")
    print(s.get_marks())

    print("\nUpdate private data using setter:")
    s.set_marks(96)
    print("Updated marks:", s.get_marks())

    print("\nTry invalid marks:")
    s.set_marks(120)

    print("\nCall private method through public wrapper:")
    s.use_private_method()

    print("\nObject dictionary:")
    show_object_dictionary(s)

    separator("2) WHY ENCAPSULATION MATTERS USING BankAccount")

    acc = BankAccount("Rajan", 1000)
    print("Initial balance:", acc.get_balance())

    print("\nDeposit money:")
    acc.deposit(500)

    print("\nWithdraw money:")
    acc.withdraw(300)

    print("\nTry invalid withdrawal:")
    acc.withdraw(5000)

    print("\nTry invalid deposit:")
    acc.deposit(-10)

    print("\nTrying to access private balance directly:")
    try:
        print(acc.__balance)
    except AttributeError as e:
        print("Error:", e)

    print("\nObject dictionary:")
    show_object_dictionary(acc)

    separator("3) ACCESS MODIFIERS DEMO")

    a = AccessDemo()
    a.show_vars()

    print("\nPublic method call:")
    a.public_method()

    print("\nProtected method call (allowed, but convention says avoid):")
    a._protected_method()

    print("\nPrivate method call from inside class:")
    a.access_private_inside()

    print("\nTrying to access private method directly:")
    try:
        a.__private_method()
    except AttributeError as e:
        print("Error:", e)

    print("\nName-mangled access (possible, but not recommended):")
    print("Accessing private var using mangled name:", a._AccessDemo__private_var)
    a._AccessDemo__private_method()

    print("\nObject dictionary:")
    show_object_dictionary(a)

    separator("4) NAME MANGLING EXAMPLE")

    m = ManglingExample()
    m.show_secret_inside()

    print("\nTrying direct access to __secret:")
    try:
        print(m.__secret)
    except AttributeError as e:
        print("Error:", e)

    print("\nMangled access works:")
    print(m._ManglingExample__secret)

    print("\nObject dictionary:")
    show_object_dictionary(m)

    separator("5) PROPERTY-BASED ENCAPSULATION")

    t = Temperature(25)
    print("Temperature object:", t)
    print("Celsius:", t.celsius)
    print("Fahrenheit:", t.fahrenheit)

    print("\nUpdate with valid value:")
    t.celsius = 30
    print("Updated temperature:", t)

    print("\nTry invalid low temperature:")
    try:
        t.celsius = -300
    except ValueError as e:
        print("Error:", e)

    print("\nTry invalid type:")
    try:
        t.celsius = "hot"
    except TypeError as e:
        print("Error:", e)

    print("\nObject dictionary:")
    show_object_dictionary(t)

    separator("6) ENCAPSULATION WITH INHERITANCE")

    child = Child()
    child.show_inside()
    child.show_child_access()
    child.try_to_access_parent_private()

    print("\nTrying to access parent private member directly from child object:")
    try:
        print(child.__private_value)
    except AttributeError as e:
        print("Error:", e)

    print("\nBut mangled access is still technically possible:")
    print(child._Parent__private_value)

    separator("7) IMPORTANT NOTES / INTERVIEW POINTS")

    notes = [
        "Encapsulation is not the same as data hiding, but they are closely related.",
        "Encapsulation = bundling data + methods together.",
        "Data hiding = restricting direct access to internal data.",
        "Python uses conventions, not hard enforcement, for privacy.",
        "A single underscore _name means 'internal use'.",
        "A double underscore __name triggers name mangling.",
        "Name mangling makes accidental access harder, not impossible.",
        "Use getters/setters or @property for validation and controlled access.",
        "Encapsulation improves maintainability, readability, and safety."
    ]

    for i, note in enumerate(notes, start=1):
        print(f"{i}. {note}")

    separator("8) QUICK SUMMARY")

    print(
        "Encapsulation means keeping data and methods together in a class, "
        "and controlling how the data is accessed or changed. "
        "In Python, public members are freely accessible, protected members are "
        "conventionally internal, and private members use name mangling."
    )

    print("\nFinished running encapsulation demo.")
```
