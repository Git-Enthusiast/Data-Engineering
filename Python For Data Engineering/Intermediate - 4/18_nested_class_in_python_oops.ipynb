```python

# ============================================================
# 1) WHAT IS A NESTED CLASS?
# ============================================================
# A nested class is a class defined inside another class.
#
# Syntax idea:
#
# class OuterClass:
#     class InnerClass:
#         ...
#
# The outer class acts like a container.
# The inner class is usually used when a piece of functionality is tightly
# connected to the outer class and does not need to exist independently.
#
# In simple words:
# - Outer class = main class
# - Inner class = helper class inside the main class
#
# Nested classes are often used when the inner class logically belongs to the outer class.


def show_definition():
    print("=" * 80)
    print("1. DEFINITION OF NESTED CLASS")
    print("=" * 80)
    print("A nested class is a class defined inside another class.")
    print("The outer class contains the inner class.")
    print("The inner class is usually used for closely related functionality.")
    print()


show_definition()

```

```python

# ============================================================
# 2) REAL-WORLD ANALOGY
# ============================================================
# A university contains colleges.
# A college exists inside the university structure.
#
# That is a good mental model for nested classes:
# - University = outer class
# - College = inner class
#
# The college can exist in the context of the university.
# This does not mean the college is always useless outside that context.
# It simply means the design groups related things together.


class University:
    def __init__(self, name):
        self.name = name

    class College:
        def __init__(self, college_name, department):
            self.college_name = college_name
            self.department = department

        def display(self):
            print(f"College Name : {self.college_name}")
            print(f"Department   : {self.department}")


def show_university_example():
    print("=" * 80)
    print("2. REAL-WORLD ANALOGY: UNIVERSITY AND COLLEGE")
    print("=" * 80)

    uni = University("Central University")
    print("University Name:", uni.name)

    # Inner class object can be created using the outer class name.
    # Since College is nested inside University, it is accessed as University.College
    college = University.College("School of Engineering", "Computer Science")
    college.display()
    print()
    print("The inner class is grouped inside the outer class for logical organization.")
    print()


show_university_example()

```

```python

# ============================================================
# 3) HOW TO CREATE OBJECTS OF OUTER AND INNER CLASS
# ============================================================
# Important rule:
#
# Outer class object:
#   outer_obj = OuterClass()
#
# Inner class object:
#   inner_obj = OuterClass.InnerClass()
#
# Sometimes people say the inner class must be created through an outer class
# instance, but that is not always true in Python.
# If the inner class does NOT need the outer instance, you can create it using:
#   OuterClass.InnerClass()
#
# If the inner class needs outer object data, then you usually pass the outer
# object explicitly or create the inner object from a method of the outer class.


class OuterExample:
    def __init__(self, outer_value):
        self.outer_value = outer_value

    class InnerExample:
        def __init__(self, inner_value):
            self.inner_value = inner_value

        def display(self):
            print(f"Inner value = {self.inner_value}")


def show_object_creation():
    print("=" * 80)
    print("3. CREATING OUTER AND INNER CLASS OBJECTS")
    print("=" * 80)

    outer_obj = OuterExample("I am outer")
    print("Outer object value:", outer_obj.outer_value)

    inner_obj = OuterExample.InnerExample("I am inner")
    inner_obj.display()
    print()
    print("Outer object and inner object are created separately here.")
    print("The inner class is accessed through the outer class name.")
    print()


show_object_creation()

```

```python

# ============================================================
# 4) ACCESSING OUTER CLASS MEMBERS INSIDE INNER CLASS
# ============================================================
# If the inner class needs data from the outer class, the inner object usually
# receives the outer object reference as an argument.
#
# This is one of the most useful patterns in nested classes.
#
# Example:
#   inner_obj = outer_obj.InnerClass(outer_obj)
#
# Then inner class can use:
#   outer_obj.some_member
#
# This is how the inner class can access outer class attributes and methods.


class Company:
    def __init__(self, company_name, location):
        self.company_name = company_name
        self.location = location

    def display_company(self):
        print(f"Company Name: {self.company_name}")
        print(f"Location    : {self.location}")

    class Department:
        def __init__(self, outer_obj, dept_name):
            # outer_obj is the Company instance
            self.outer_obj = outer_obj
            self.dept_name = dept_name

        def display(self):
            print("Department Name:", self.dept_name)
            print("Accessing outer class members from inner class:")
            print("Company Name:", self.outer_obj.company_name)
            print("Location    :", self.outer_obj.location)
            self.outer_obj.display_company()


def show_outer_member_access_inside_inner():
    print("=" * 80)
    print("4. ACCESSING OUTER CLASS MEMBERS INSIDE INNER CLASS")
    print("=" * 80)

    company = Company("TechNova", "Bangalore")
    department = Company.Department(company, "Research and Development")
    department.display()
    print()
    print("The inner class can use the outer object reference to read outer data.")
    print()


show_outer_member_access_inside_inner()

```

```python

# ============================================================
# 5) STUDENT + DOB EXAMPLE
# ============================================================
# This is the classic nested-class example.
#
# Student is the outer class.
# DOB (Date of Birth) is the inner class.
#
# Student has main information:
# - name
# - roll number
# - DOB object
#
# DOB has its own information:
# - day
# - month
# - year
#
# The Student class creates a DOB object and stores it in self.dob.
# This is a very common way to model related data.


class Student:
    def __init__(self, name, roll, dd, mm, yy):
        self.name = name
        self.roll = roll
        self.dob = self.DOB(dd, mm, yy)

    def display(self):
        print(f"Name is {self.name} and roll is {self.roll}")
        self.dob.display()

    class DOB:
        def __init__(self, dd, mm, yy):
            self.date = dd
            self.month = mm
            self.year = yy

        def display(self):
            print(f"Date of birth is: {self.date}/{self.month}/{self.year}")


def show_student_dob_example():
    print("=" * 80)
    print("5. STUDENT + DOB NESTED CLASS EXAMPLE")
    print("=" * 80)

    s1 = Student("Ajay", 101, 26, 3, 1999)
    s1.display()
    print()
    print("Here Student is the outer class and DOB is the nested inner class.")
    print("The Student object contains a DOB object as an attribute.")
    print()


show_student_dob_example()

```

```python

# ============================================================
# 6) ACCESSING ONE CLASS MEMBERS INSIDE ANOTHER CLASS
# ============================================================
# This is slightly different from nested classes, but it is related.
#
# Sometimes one class holds an object of another class and uses its members.
# This is a common OOP design pattern.
#
# Example:
# - Employee stores employee data
# - Changes is a helper class with a static method
# - Changes.increment(obj) receives an Employee object
# - Then it directly accesses and modifies that object's attributes
#
# This shows how one class can work on another class's object.


class Employee:
    def __init__(self, eid, name, sal):
        self.emp_id = eid
        self.emp_name = name
        self.emp_salary = sal

    def display(self):
        print("Employee id   :", self.emp_id)
        print("Employee name :", self.emp_name)
        print("Employee salary:", self.emp_salary)


class Changes:
    @staticmethod
    def increment(obj):
        # obj is an Employee object.
        # We are directly reading and updating its member variable.
        obj.emp_salary = obj.emp_salary + 2000
        obj.display()


def show_employee_changes_example():
    print("=" * 80)
    print("6. ACCESSING MEMBERS OF ONE CLASS INSIDE ANOTHER CLASS")
    print("=" * 80)

    e1 = Employee(101, "Shantanu", 50000)
    print("Before increment:")
    e1.display()
    print()
    print("After increment through Changes.increment(e1):")
    Changes.increment(e1)
    print()
    print("This is not a nested class example, but it is a useful class-to-class interaction.")
    print("A class can accept another class object and use its members directly.")
    print()


show_employee_changes_example()

```

```python

# ============================================================
# 7) WHY NESTED CLASSES ARE USED
# ============================================================
# Nested classes are useful when:
# - one class is tightly related to another
# - you want better organization
# - you want to group helper functionality inside a main class
# - the inner class is not needed as a completely separate top-level class
#
# Advantages:
# - better code structure
# - logical grouping
# - improved readability
# - helps model real-world relationships
#
# Limitations:
# - can become confusing if overused
# - not every helper class should be nested
# - if the inner class is large or reused elsewhere, a separate class may be better


def show_why_nested_classes():
    print("=" * 80)
    print("7. WHY NESTED CLASSES ARE USED")
    print("=" * 80)
    print("1. To group related classes together")
    print("2. To model real-world containment relationships")
    print("3. To improve readability and organization")
    print("4. To keep helper classes close to the outer class")
    print()
    print("Nested classes should be used only when the relationship is strong and meaningful.")
    print()


show_why_nested_classes()

```

```python

# ============================================================
# 8) IMPORTANT RULES TO REMEMBER
# ============================================================
# Rule 1:
# The inner class can be defined inside the outer class.
#
# Rule 2:
# The inner class can be accessed as OuterClass.InnerClass.
#
# Rule 3:
# If the inner class needs the outer object, pass it explicitly.
#
# Rule 4:
# Nested class does not automatically mean it can access outer instance members.
# A nested class can access outer members only if you give it a reference.
#
# Rule 5:
# Nested classes are different from nested functions.
#
# Rule 6:
# Nesting is a design choice, not a requirement.


def show_rules():
    print("=" * 80)
    print("8. IMPORTANT RULES")
    print("=" * 80)
    rules = [
        "Inner class can be accessed using OuterClass.InnerClass.",
        "Inner class does not automatically know outer instance values.",
        "Pass the outer object if the inner class must use it.",
        "Nested class is a design choice, not mandatory.",
    ]
    for i, rule in enumerate(rules, start=1):
        print(f"{i}. {rule}")
    print()


show_rules()

```

```python

# ============================================================
# 9) COMMON CONFUSION: NESTED CLASS VS INNER OBJECT
# ============================================================
# People often confuse these two ideas:
#
# - Nested class: a class defined inside another class
# - Inner object/member object: an object stored as an attribute inside another
#   object, such as self.dob = self.DOB(...)
#
# In the Student example, DOB is a nested class, and s1.dob is an inner object
# created from that nested class.
#
# This is why the Student example is powerful: it shows both nesting and object
# composition.


def show_common_confusion():
    print("=" * 80)
    print("9. COMMON CONFUSION")
    print("=" * 80)
    print("Nested class = class inside another class")
    print("Inner object = object stored inside another object")
    print("Student.DOB is a nested class")
    print("s1.dob is an object created from that nested class")
    print()


show_common_confusion()

```
