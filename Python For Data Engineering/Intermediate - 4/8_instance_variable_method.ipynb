```python
# =========================================================
# INSTANCE VARIABLES & INSTANCE METHODS IN PYTHON
# =========================================================
# Instance Variables:
# - Variables that belong to an object (instance)
# - Each object has its own copy (separate memory)
#
# Instance Methods:
# - Methods that operate on instance variables
# - Must use 'self' parameter
# - 'self' refers to the current object

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Student:
    def __init__(self, name, marks):
        # Instance variables (unique for each object)
        self.name = name
        self.marks = marks

    # Instance method to display data
    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

    # Instance method to update data
    def update(self, new_name, new_marks):
        self.name = new_name
        self.marks = new_marks

```

```python

# =========================================================
# OBJECT CREATION
# =========================================================
s1 = Student("Rajan", 85)
s2 = Student("Amit", 90)

print("========== INITIAL DATA ==========")
s1.display()
s2.display()

```

```python

# =========================================================
# PROOF: INSTANCE VARIABLES ARE SEPARATE
# =========================================================
print("\n========== MODIFY ONE OBJECT ==========")

# Modify only s1
s1.marks = 95

print("After modifying s1:")
s1.display()   # changed
s2.display()   # unchanged

```

```python

# =========================================================
# ACCESSING INSTANCE VARIABLES
# =========================================================
print("\n========== ACCESSING VARIABLES ==========")

print("s1 name:", s1.name)
print("s2 marks:", s2.marks)

```

```python

# =========================================================
# ADDING INSTANCE VARIABLE DYNAMICALLY
# =========================================================
print("\n========== ADD NEW VARIABLE ==========")

s1.grade = "A"   # added only to s1

print("s1 grade:", s1.grade)

# s2 does not have 'grade'
print("Does s2 have grade?", hasattr(s2, "grade"))

```

```python

# =========================================================
# DELETING INSTANCE VARIABLE
# =========================================================
print("\n========== DELETE VARIABLE ==========")

print("Before deletion:", s1.__dict__)

del s1.grade   # delete attribute

print("After deletion:", s1.__dict__)

```

```python

# =========================================================
# USING INSTANCE METHODS
# =========================================================
print("\n========== USING INSTANCE METHODS ==========")

# Update s2 using method
s2.update("Neha", 92)

print("After update:")
s2.display()

```

```python

# =========================================================
# HOW 'self' WORKS (IMPORTANT CONCEPT)
# =========================================================
# When you call:
# s1.display()
# Python internally converts it to:
# Student.display(s1)
#
# So 'self' = s1 (current object)

```

```python

# =========================================================
# OBJECT NAMESPACE (__dict__)
# =========================================================
print("\n========== OBJECT DATA (__dict__) ==========")

print("s1 data:", s1.__dict__)
print("s2 data:", s2.__dict__)

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. Instance variables are different for each object.
# 2. Changing one object does NOT affect others.
# 3. 'self' is mandatory in instance methods.
# 4. Instance variables can be added/removed dynamically.
# 5. __dict__ helps in debugging object data.
```
