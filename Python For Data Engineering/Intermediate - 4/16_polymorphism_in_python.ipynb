```python
"""
===========================================
        POLYMORPHISM IN PYTHON
===========================================

Topics covered:
1. What polymorphism means
2. Real-life example
3. Polymorphism with operators
4. Polymorphism with built-in functions
5. Polymorphism with user-defined functions
6. Polymorphism with inheritance
7. Method overriding
8. Duck typing
9. Polymorphism with abstract-style thinking
10. Common mistakes
11. Interview/exam style notes

You can run this file directly. It prints examples section by section.

------------------------------------------------------------
CORE IDEA
------------------------------------------------------------
Polymorphism means "many forms".
In Python, the same name or same interface can behave differently depending on:
- the data type
- the object
- the class
- the method implementation
- the context

Example:
- 5 + 6 gives 11
- "5" + "6" gives "56"

Same operator (+), different behavior.
That is polymorphism.
"""

```

```python

# ============================================================
# 1) BASIC DEFINITION
# ============================================================

def show_basic_definition():
    print("=" * 70)
    print("1. BASIC DEFINITION OF POLYMORPHISM")
    print("=" * 70)
    print("Polymorphism = poly (many) + morph (forms)")
    print("It means one thing can take many forms.")
    print()
    print("In Python, polymorphism appears when the same:")
    print("- operator")
    print("- function")
    print("- method name")
    print("- object interface")
    print("works differently for different inputs or objects.")
    print()

show_basic_definition()
```

```python

# ============================================================
# 2) REAL-LIFE EXAMPLE
# ============================================================
# A person behaves differently in different situations.
# Same person, different behavior.

class Person:
    def behave(self, situation):
        if situation == "friend":
            return "Friendly, relaxed, casual behavior"
        elif situation == "teacher":
            return "Respectful, formal behavior"
        elif situation == "stranger":
            return "Cautious, polite behavior"
        else:
            return "Normal behavior"


def show_real_life_example():
    print("=" * 70)
    print("2. REAL-LIFE EXAMPLE")
    print("=" * 70)
    person = Person()
    print("With a friend   :", person.behave("friend"))
    print("With a teacher  :", person.behave("teacher"))
    print("With a stranger :", person.behave("stranger"))
    print()
    print("Same object, different behavior depending on situation.")
    print("That idea is similar to polymorphism.")
    print()

show_real_life_example()

# ============================================================
# 3) POLYMORPHISM WITH OPERATORS
# ============================================================
# The + operator behaves differently depending on the type.


def show_operator_polymorphism():
    print("=" * 70)
    print("3. POLYMORPHISM WITH OPERATORS")
    print("=" * 70)

    # Integer addition
    a = 10
    b = 20
    print("10 + 20 =", a + b)

    # Float addition
    x = 2.5
    y = 4.5
    print("2.5 + 4.5 =", x + y)

    # String concatenation
    first = "Hello"
    second = "World"
    print('"Hello" + "World" =', first + second)

    # List concatenation
    list1 = [1, 2]
    list2 = [3, 4]
    print("[1, 2] + [3, 4] =", list1 + list2)

    print()
    print("Same operator (+), different meanings:")
    print("- numbers: addition")
    print("- strings: concatenation")
    print("- lists: merging")
    print()

show_operator_polymorphism()
```

```python

# ============================================================
# 4) POLYMORPHISM WITH BUILT-IN FUNCTIONS
# ============================================================
# len() behaves differently for different object types.


def show_builtin_function_polymorphism():
    print("=" * 70)
    print("4. POLYMORPHISM WITH BUILT-IN FUNCTIONS")
    print("=" * 70)

    text = "Polymorphism"
    numbers = [10, 20, 30, 40]
    data = {"name": "Rajan", "age": 25, "city": "India"}
    tuple_data = (1, 2, 3)

    print('len("Polymorphism") =', len(text))
    print("len([10, 20, 30, 40]) =", len(numbers))
    print('len({"name":..., "age":..., "city":...}) =', len(data))
    print("len((1, 2, 3)) =", len(tuple_data))

    print()
    print("What does len() measure?")
    print("- string -> number of characters")
    print("- list   -> number of items")
    print("- dict   -> number of keys")
    print("- tuple  -> number of elements")
    print()

    # Another common example: sorted()
    print("sorted([3, 1, 2]) =", sorted([3, 1, 2]))
    print('sorted("cab") =', sorted("cab"))
    print("sorted({3, 1, 2}) =", sorted({3, 1, 2}))
    print()
    print("The same built-in function can work with many data types.")
    print()

show_builtin_function_polymorphism()
```

```python

# ============================================================
# 5) POLYMORPHISM WITH USER-DEFINED FUNCTIONS
# ============================================================
# A single function can work on different types if they support the required behavior.


def reverse_item(item):
    """
    Reverse different kinds of data.

    Works for:
    - strings
    - lists
    - tuples
    - custom objects that provide a reverse-like method or representation

    For unsupported types, it returns a helpful message.
    """
    try:
        return item[::-1]
    except TypeError:
        return "This object does not support slicing-based reverse."


class WordBox:
    def __init__(self, word):
        self.word = word

    def reverse(self):
        return self.word[::-1]


def show_user_defined_function_polymorphism():
    print("=" * 70)
    print("5. POLYMORPHISM WITH USER-DEFINED FUNCTIONS")
    print("=" * 70)

    print('reverse_item("python") =', reverse_item("python"))
    print("reverse_item([1, 2, 3, 4]) =", reverse_item([1, 2, 3, 4]))
    print("reverse_item((10, 20, 30)) =", reverse_item((10, 20, 30)))
    print("reverse_item(100) =", reverse_item(100))

    box = WordBox("hello")
    print("WordBox.reverse() =", box.reverse())

    print()
    print("A function becomes polymorphic when it can handle multiple kinds of input.")
    print("Sometimes this happens because the objects support the same operation.")
    print()

show_user_defined_function_polymorphism()
```

```python

# ============================================================
# 6) POLYMORPHISM THROUGH INHERITANCE
# ============================================================
# Base class defines common interface.
# Subclasses reuse and customize behavior.


class Vehicle:
    def __init__(self, name, color, price):
        self.name = name
        self.color = color
        self.price = price

    def get_details(self):
        return f"Name: {self.name}, Color: {self.color}, Price: {self.price}"

    def max_speed(self):
        return "Generic vehicle speed"

    def fuel_type(self):
        return "Generic fuel type"


class Car(Vehicle):
    def max_speed(self):
        return "Car max speed: 180 km/h"

    def fuel_type(self):
        return "Petrol / Diesel / Electric"


class Truck(Vehicle):
    def max_speed(self):
        return "Truck max speed: 100 km/h"

    def fuel_type(self):
        return "Diesel"


class Bike(Vehicle):
    def max_speed(self):
        return "Bike max speed: 120 km/h"

    def fuel_type(self):
        return "Petrol / Electric"


def show_inheritance_polymorphism():
    print("=" * 70)
    print("6. POLYMORPHISM THROUGH INHERITANCE")
    print("=" * 70)

    car = Car("Swift", "Red", 800000)
    truck = Truck("Tata Truck", "Blue", 2500000)
    bike = Bike("Apache", "Black", 150000)

    vehicles = [car, truck, bike]

    for vehicle in vehicles:
        print(vehicle.get_details())
        print(vehicle.max_speed())
        print(vehicle.fuel_type())
        print("-" * 50)

    print("The same method names behave differently in different subclasses.")
    print("That is polymorphism with inheritance.")
    print()

show_inheritance_polymorphism()
```

```python

# ============================================================
# 7) METHOD OVERRIDING
# ============================================================
# Child class replaces parent class method with its own version.


class Animal:
    def sound(self):
        return "Some generic animal sound"


class Dog(Animal):
    def sound(self):
        return "Bark"


class Cat(Animal):
    def sound(self):
        return "Meow"


class Cow(Animal):
    def sound(self):
        return "Moo"


def show_method_overriding():
    print("=" * 70)
    print("7. METHOD OVERRIDING")
    print("=" * 70)

    animals = [Dog(), Cat(), Cow(), Animal()]
    for animal in animals:
        print(type(animal).__name__, "->", animal.sound())

    print()
    print("Method overriding is one of the most important ways polymorphism appears in OOP.")
    print("Parent class has a method, child class provides a different implementation.")
    print()

show_method_overriding()
```

```python

# ============================================================
# 8) DYNAMIC POLYMORPHISM / DUCK TYPING
# ============================================================
# Python uses duck typing:
# "If it walks like a duck and quacks like a duck, treat it like a duck."
#
# This means Python often cares more about what an object CAN DO than what it IS.


class Bird:
    def fly(self):
        return "Bird is flying"


class Airplane:
    def fly(self):
        return "Airplane is flying"


class SuperHero:
    def fly(self):
        return "SuperHero is flying"


def make_it_fly(object_that_can_fly):
    return object_that_can_fly.fly()


def show_duck_typing():
    print("=" * 70)
    print("8. DUCK TYPING")
    print("=" * 70)

    bird = Bird()
    plane = Airplane()
    hero = SuperHero()

    print(make_it_fly(bird))
    print(make_it_fly(plane))
    print(make_it_fly(hero))

    print()
    print("Python does not always need a common parent class.")
    print("If the object has the required method, the function can use it.")
    print("That is duck typing, and it is closely related to polymorphism.")
    print()

show_duck_typing()
```

```python

# ============================================================
# 9) POLYMORPHISM WITH A COMMON INTERFACE
# ============================================================
# A common interface means different objects expose the same method name.
# The caller does not care which specific class is used.


class PDFReport:
    def generate(self):
        return "Generating PDF report"


class ExcelReport:
    def generate(self):
        return "Generating Excel report"


class HTMLReport:
    def generate(self):
        return "Generating HTML report"


def process_report(report):
    return report.generate()


def show_common_interface_polymorphism():
    print("=" * 70)
    print("9. POLYMORPHISM WITH A COMMON INTERFACE")
    print("=" * 70)

    reports = [PDFReport(), ExcelReport(), HTMLReport()]
    for report in reports:
        print(process_report(report))

    print()
    print("The function process_report() works with any object that has generate().")
    print("This is a very practical polymorphism pattern in real projects.")
    print()

show_common_interface_polymorphism()
```

```python

# ============================================================
# 10) POLYMORPHISM WITH ABSTRACT-STYLE THINKING
# ============================================================
# Even without using abc module, we can design classes with the same required method.


class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area()")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


def show_shape_polymorphism():
    print("=" * 70)
    print("10. SHAPE EXAMPLE")
    print("=" * 70)

    shapes = [Rectangle(10, 5), Circle(7), Triangle(10, 8)]
    for shape in shapes:
        print(type(shape).__name__, "area =", shape.area())

    print()
    print("Same method name: area()")
    print("Different formulas depending on the shape.")
    print()

show_shape_polymorphism()
```

```python

# ============================================================
# 11) FUNCTION OVERLOADING IN PYTHON?
# ============================================================
# Python does NOT support traditional function overloading like Java/C++
# in the same direct way.
#
# But we can simulate it using:
# - default arguments
# - variable-length arguments (*args, **kwargs)
# - type checking
# - functools.singledispatch (advanced)
#
# This is related to polymorphism because one function name can handle many forms.


def add(*args):
    """
    Simulate different forms of addition.

    Examples:
    add(2, 3)
    add(2, 3, 4)
    add("Hello", " ", "World")
    """

    # Less Optimized approach O(n2)
    ''' 
    
    result = args[0]
    for value in args[1:]:
        result = result + value
    return result

    ''' 
    # More Optimized apporach 
    if isinstance(args[0], str):
        return "".join(args)
    
    result = args[0]
    for value in args[1:]:
        result += value
    return result


def show_function_overloading_style():
    print("=" * 70)
    print("11. FUNCTION OVERLOADING STYLE USING *args")
    print("=" * 70)

    print("add(2, 3) =", add(2, 3))
    print("add(2, 3, 4) =", add(2, 3, 4))
    print('add("Hello", " ", "World") =', add("Hello", " ", "World"))

    print()
    print("Python does not use classical overloading in the same way as some languages.")
    print("But we can still build flexible functions that behave differently based on input.")
    print()

show_function_overloading_style()
```

```python

# ============================================================
# 12) EXAMPLE OF POLYMORPHIC PARAMETER
# ============================================================
# A function can accept any object that follows the expected behavior.


def area_of_shape(shape):
    return shape.area()


def show_polymorphic_parameter_example():
    print("=" * 70)
    print("12. POLYMORPHIC PARAMETER EXAMPLE")
    print("=" * 70)

    print("area_of_shape(Rectangle(4, 6)) =", area_of_shape(Rectangle(4, 6)))
    print("area_of_shape(Circle(3)) =", area_of_shape(Circle(3)))
    print("area_of_shape(Triangle(5, 2)) =", area_of_shape(Triangle(5, 2)))
    print()
    print("The function does not need to know the exact class in advance.")
    print("It only needs the object to support area().")
    print()

show_polymorphic_parameter_example()
```

```python

# ============================================================
# 13) COMMON MISTAKES
# ============================================================

def show_common_mistakes():
    print("=" * 70)
    print("13. COMMON MISTAKES")
    print("=" * 70)

    print("Mistake 1: Thinking polymorphism only means inheritance.")
    print("  - Inheritance is one way to get polymorphism.")
    print("  - Duck typing and operator behavior also show polymorphism.")
    print()

    print("Mistake 2: Confusing polymorphism with encapsulation.")
    print("  - Encapsulation hides data.")
    print("  - Polymorphism gives many forms of behavior.")
    print()

    print("Mistake 3: Assuming function overloading works exactly like Java/C++.")
    print("  - Python usually uses flexible arguments or other patterns instead.")
    print()

    print("Mistake 4: Forgetting that built-in functions can be polymorphic too.")
    print("  - len(), sorted(), type conversion, etc. work with many types.")
    print()

show_common_mistakes()
```
