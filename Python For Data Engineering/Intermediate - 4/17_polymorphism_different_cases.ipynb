```python

# ============================================================
# 1) POLYMORPHISM — CORE THEORY
# ============================================================
#
# "Polymorphism" means "many forms".
# In Python, polymorphism appears when the same operation, function name,
# or method name behaves differently based on the object or input.
#
# Examples:
# - + adds numbers
# - + concatenates strings
# - len() works for strings, lists, tuples, dictionaries, sets, and many
#   other container types
# - a method name like display() can mean different things in different classes
#
# In simple words:
#   Same name / same interface -> different behavior
#
# This is one of the biggest reasons Python code can be flexible and clean.


def show_polymorphism_introduction():
    print("=" * 80)
    print("1. POLYMORPHISM INTRODUCTION")
    print("=" * 80)
    print("Polymorphism means one interface, many forms.")
    print("The same operation can behave differently depending on the object.")
    print()


show_polymorphism_introduction()

```

```python

# ============================================================
# 2) MAGIC METHODS / DUNDER METHODS
# ============================================================
#
# Magic methods are special methods surrounded by double underscores.
# Example:
#   __str__
#   __len__
#   __add__
#   __gt__
#   __repr__
#   __eq__
#
# They are called "magic" because Python uses them automatically in many places.
# They are also called "dunder" methods because "double underscore" = dunder.
#
# These methods are where a lot of polymorphism becomes visible in Python.
#
# Important idea:
# Python does not literally modify the built-in function print() or len().
# Instead, print(), len(), +, >, and many other operations look for special
# methods in the object and call them internally.
#
# So when you write:
#   len(obj)
# Python internally tries to call:
#   obj.__len__()
#
# When you write:
#   str(obj)
# Python internally tries to call:
#   obj.__str__()
#
# When you write:
#   obj1 + obj2
# Python internally tries to call:
#   obj1.__add__(obj2)
#
# When you write:
#   obj1 > obj2
# Python internally tries to call:
#   obj1.__gt__(obj2)
#
# That is the engine behind operator overloading.

```

```python

# ============================================================
# 3) __str__ : CUSTOM STRING REPRESENTATION
# ============================================================
#
# Without __str__, printing a custom object often shows a default representation
# like <__main__.ClassName object at 0x...>
#
# That default text is not very readable for users.
#
# __str__ lets you control what should be displayed when print(obj) is used.
#
# Python will try:
#   print(obj)
# -> str(obj)
# -> obj.__str__()
#
# Use __str__ when you want a friendly, human-readable output.


class CartItem:
    def __init__(self, name, quantity, price_per_unit):
        self.name = name
        self.quantity = quantity
        self.price_per_unit = price_per_unit

    def total_price(self):
        return self.quantity * self.price_per_unit

    # __str__ controls the readable display of the object.
    def __str__(self):
        return (
            f"CartItem(name={self.name}, quantity={self.quantity}, "
            f"price_per_unit={self.price_per_unit}, total={self.total_price()})"
        )


def show_str_magic_method():
    print("=" * 80)
    print("2. __str__ MAGIC METHOD")
    print("=" * 80)

    item = CartItem("Notebook", 3, 50)

    # print(item) internally uses item.__str__()
    print("Printing CartItem object:")
    print(item)
    print()
    print("A custom __str__ gives a clean, useful display instead of a memory address.")
    print()


show_str_magic_method()

```

```python

# ============================================================
# 4) __len__ : CUSTOM LENGTH BEHAVIOR
# ============================================================
#
# len(obj) works by calling obj.__len__() internally.
#
# This is a classic polymorphism example because len() behaves differently for
# different built-in objects:
# - len("Python") -> number of characters
# - len([1, 2, 3]) -> number of elements
# - len({"a": 1, "b": 2}) -> number of keys
#
# We can also define __len__ in our own class so len() can work on it.
#
# In the video example, a Cart class uses internal lists (baskets) and __len__
# returns the total item count across them.


class Cart:
    def __init__(self):
        # Imagine a shopping cart with multiple internal baskets/sections.
        # Each basket is a list of item names.
        self.fruits = []
        self.groceries = []
        self.electronics = []

    def add_fruit(self, item):
        self.fruits.append(item)

    def add_grocery(self, item):
        self.groceries.append(item)

    def add_electronic(self, item):
        self.electronics.append(item)

    # __len__ tells Python what "length" means for this custom object.
    # Here, length means total number of items across all internal baskets.
    def __len__(self):
        return len(self.fruits) + len(self.groceries) + len(self.electronics)

    def __str__(self):
        return (
            f"Cart(fruits={self.fruits}, groceries={self.groceries}, "
            f"electronics={self.electronics})"
        )


def show_len_magic_method():
    print("=" * 80)
    print("3. __len__ MAGIC METHOD")
    print("=" * 80)

    cart = Cart()
    cart.add_fruit("Apple")
    cart.add_fruit("Banana")
    cart.add_grocery("Rice")
    cart.add_grocery("Salt")
    cart.add_electronic("Mouse")

    print(cart)
    print("len(cart) =", len(cart))
    print()
    print("Here len(cart) counts the total number of items in all baskets.")
    print("That means you are defining your own meaning of 'length'.")
    print()


show_len_magic_method()

```

```python

# ============================================================
# 5) OPERATOR OVERLOADING WITH __add__
# ============================================================
#
# Operator overloading means giving a custom meaning to an operator for your own
# class.
#
# The + operator is already polymorphic in Python:
# - numbers: addition
# - strings: concatenation
# - lists: merging
#
# For user-defined objects, Python does not know what + should mean unless you
# define __add__.
#
# When you write:
#   b1 + b2
# Python internally tries:
#   b1.__add__(b2)
#
# If __add__ exists, Python uses it.
# If not, Python raises TypeError.
#
# This is why the behavior feels flexible and "overloaded".


class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book(title={self.title}, pages={self.pages})"

    # __add__ defines what happens when two Book objects are added with +.
    # Here we choose a useful meaning:
    #   book1 + book2 = new Book containing combined page count
    # We also preserve the idea that a book object still represents a book-like
    # object, not just a raw number.
    def __add__(self, other):
        if not isinstance(other, Book):
            return NotImplemented

        total_pages = self.pages + other.pages
        combined_title = f"{self.title} + {other.title}"
        return Book(combined_title, total_pages)


def show_add_operator_overloading():
    print("=" * 80)
    print("4. OPERATOR OVERLOADING WITH __add__")
    print("=" * 80)

    b1 = Book("One Indian Girl", 300)
    b2 = Book("Making India Awesome", 200)
    b3 = Book("Half Girlfriend", 400)
    b4 = Book("Girl in 105 Room", 300)
    b5 = Book("Atomic Habits", 250)

    print("Individual books:")
    print(b1)
    print(b2)
    print(b3)
    print(b4)
    print(b5)
    print()

    # Add books one by one.
    # Each + uses __add__ and returns a NEW Book object.
    total_1 = b1 + b2
    total_2 = total_1 + b3
    total_3 = total_2 + b4
    total_4 = total_3 + b5

    print("After adding books step by step:")
    print(total_1)
    print(total_2)
    print(total_3)
    print(total_4)
    print()
    print("Final total pages =", total_4.pages)
    print()
    print("Important point:")
    print("The + operator is not hardcoded to one meaning.")
    print("Python asks the object how + should behave by calling __add__.")
    print()


show_add_operator_overloading()

```

```python

# ============================================================
# 6) HOW PYTHON DECIDES WHICH ADDITION METHOD RUNS
# ============================================================
#
# This answers an important question:
#   How does Python decide which addition method runs?
#
# The rule is roughly:
# 1. Python checks the left operand.
# 2. It tries left_operand.__add__(right_operand).
# 3. If that returns a valid result, Python uses it.
# 4. If that returns NotImplemented, Python may try the right operand's reverse
#    method: right_operand.__radd__(left_operand).
# 5. If neither works, TypeError is raised.
#
# This is one reason operator overloading feels flexible.
# Python is using a protocol-based lookup process.


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        if isinstance(other, int):
            return Vector(self.x + other, self.y + other)
        return NotImplemented

    # __radd__ is the reverse-add method.
    # It is useful when the left side does not know how to add the object.
    # Example:
    #   10 + vector
    # Python first tries int.__add__(vector)
    # If that fails, it tries vector.__radd__(10)
    def __radd__(self, other):
        if isinstance(other, int):
            return Vector(self.x + other, self.y + other)
        return NotImplemented


def show_addition_method_resolution():
    print("=" * 80)
    print("5. HOW PYTHON DECIDES WHICH ADDITION METHOD RUNS")
    print("=" * 80)

    v1 = Vector(2, 3)
    v2 = Vector(4, 5)

    print("v1 =", v1)
    print("v2 =", v2)
    print("v1 + v2 =", v1 + v2)
    print("v1 + 10 =", v1 + 10)
    print("10 + v1 =", 10 + v1)
    print()
    print("Python tries the left object's __add__ first.")
    print("If that does not work, reverse addition __radd__ may be used.")
    print()


show_addition_method_resolution()

```

```python

# ============================================================
# 7) POLYMORPHIC FUNCTIONS WITH COMMON INTERFACES
# ============================================================
#
# A polymorphic function is a function that works with different object types
# as long as those objects follow the expected interface.
#
# This is a very Pythonic style.
#
# Example:
#   car_details(car_object)
#
# The function may not care whether the object is Ferrari or BMW.
# It only cares that the object provides the methods it needs.


class Ferrari:
    def fuel_type(self):
        return "Petrol"

    def max_speed(self):
        return "Ferrari max speed: 340 km/h"

    def car_type(self):
        return "Sports car"


class BMW:
    def fuel_type(self):
        return "Diesel / Petrol / Electric"

    def max_speed(self):
        return "BMW max speed: 250 km/h"

    def car_type(self):
        return "Luxury car"


# This function is polymorphic because it can accept different car objects.
# It does not need to know the exact class in advance.
# It simply calls the methods expected from the object.
def car_details(car_obj):
    print("Car Type:", car_obj.car_type())
    print("Fuel Type:", car_obj.fuel_type())
    print("Speed Info:", car_obj.max_speed())


def show_polymorphic_function():
    print("=" * 80)
    print("6. POLYMORPHIC FUNCTION EXAMPLE")
    print("=" * 80)

    ferrari = Ferrari()
    bmw = BMW()

    print("Ferrari details:")
    car_details(ferrari)
    print("-" * 50)
    print("BMW details:")
    car_details(bmw)
    print()
    print("The same function works with different objects.")
    print("That is polymorphism through a common interface.")
    print()


show_polymorphic_function()

```

```python

# ============================================================
# 8) OPERATOR OVERLOADING WITH COMPARISON: __gt__
# ============================================================
#
# The > operator is also polymorphic.
# For built-in types, it compares numbers, strings, dates, etc.
# For custom objects, Python needs you to define comparison methods.
#
# __gt__ means "greater than".
#
# When you write:
#   h1 > h2
# Python internally tries:
#   h1.__gt__(h2)
#
# You can decide what "greater" means.
# It could be based on rating, price, area, speed, or any logical rule.


class Hotel:
    def __init__(self, name, rating, price_per_night):
        self.name = name
        self.rating = rating
        self.price_per_night = price_per_night

    def __str__(self):
        return (
            f"Hotel(name={self.name}, rating={self.rating}, "
            f"price_per_night={self.price_per_night})"
        )

    # Here we define "greater than" based on rating.
    # If ratings are equal, we can use price as a tie-breaker.
    def __gt__(self, other):
        if not isinstance(other, Hotel):
            return NotImplemented

        if self.rating != other.rating:
            return self.rating > other.rating
        return self.price_per_night > other.price_per_night


def show_greater_than_overloading():
    print("=" * 80)
    print("7. OPERATOR OVERLOADING WITH __gt__")
    print("=" * 80)

    h1 = Hotel("Grand Palace", 4.5, 8000)
    h2 = Hotel("Royal Stay", 4.2, 9000)
    h3 = Hotel("Elite Inn", 4.5, 7000)

    print(h1)
    print(h2)
    print(h3)
    print()

    print("h1 > h2 =", h1 > h2)
    print("h1 > h3 =", h1 > h3)
    print("h3 > h2 =", h3 > h2)
    print()
    print("In this example, 'greater than' is defined by hotel rating.")
    print("That is why custom comparison methods are so powerful.")
    print()


show_greater_than_overloading()

```

```python

# ============================================================
# 9) WHAT HAPPENS IF YOU DEFINE TWO IDENTICAL METHODS?
# ============================================================
#
# This is the important point about method overloading in Python.
#
# In languages like Java or C++, you can define multiple methods with the same
# name but different parameters. That is method overloading.
#
# Python does NOT support that natively.
# If you define the same method name twice in the same class, the last one wins.
# The earlier definition is overwritten.
#
# So Python does not automatically choose methods based on parameter count.
# That is why the answer to:
#   "Does Python handle method overloading like Java?"
# is:
#   No, not in the same native way.


class DemoDuplicateMethods:
    def show(self, value):
        return f"First show(): {value}"

    # This second definition overwrites the first one.
    # Python keeps only the latest method with the same name.
    def show(self, value, extra=None):
        return f"Second show(): {value}, extra={extra}"


def show_duplicate_method_behavior():
    print("=" * 80)
    print("8. WHAT HAPPENS IF YOU DEFINE TWO IDENTICAL METHODS?")
    print("=" * 80)

    demo = DemoDuplicateMethods()
    print(demo.show(10))
    print(demo.show(10, 20))
    print()
    print("The second method replaced the first one.")
    print("This is why Python does not support classical method overloading directly.")
    print()


show_duplicate_method_behavior()

```

```python

# ============================================================
# 10) HOW TO IMPLEMENT METHOD OVERLOADING IN PYTHON
# ============================================================
#
# Since Python does not have native method overloading, we simulate it using:
# - default parameters
# - *args
# - **kwargs
# - type checking
# - conditional logic
#
# This is often what interview questions mean by "method overloading in Python".
#
# The goal is to make one method handle multiple input styles gracefully.


class Calculator:
    # One method can behave differently depending on which arguments are provided.
    # Here we use default values.
    def add(self, a=None, b=None, c=None):
        # Case 1: only a is provided
        if a is not None and b is None and c is None:
            return a

        # Case 2: two values are provided
        if a is not None and b is not None and c is None:
            return a + b

        # Case 3: three values are provided
        if a is not None and b is not None and c is not None:
            return a + b + c

        return "No valid input provided"

    # Another flexible method using *args
    def multiply(self, *args):
        if not args:
            return "No values provided"

        result = 1
        for value in args:
            result *= value
        return result


class ShapeAreaCalculator:
    # Here area() acts like an overloaded method using default arguments.
    # If only one side is given, we treat it like a square.
    # If length and width are given, we treat it like a rectangle.
    def area(self, length=None, width=None):
        if length is not None and width is None:
            return length * length
        elif length is not None and width is not None:
            return length * width
        return "Invalid dimensions"


def show_method_overloading_simulation():
    print("=" * 80)
    print("9. HOW TO IMPLEMENT METHOD OVERLOADING IN PYTHON")
    print("=" * 80)

    calc = Calculator()
    print("calc.add(5) =", calc.add(5))
    print("calc.add(5, 10) =", calc.add(5, 10))
    print("calc.add(5, 10, 15) =", calc.add(5, 10, 15))
    print("calc.multiply(2, 3, 4) =", calc.multiply(2, 3, 4))
    print()

    area_calc = ShapeAreaCalculator()
    print("area_calc.area(6) -> square area =", area_calc.area(6))
    print("area_calc.area(6, 4) -> rectangle area =", area_calc.area(6, 4))
    print()
    print("This is how Python developers simulate overloading:")
    print("one method name + flexible arguments + condition checks")
    print()


show_method_overloading_simulation()

```

```python

# ============================================================
# 11) WHY THIS IS STILL POLYMORPHISM
# ============================================================
#
# Even in the method overloading simulation above, the behavior changes based on:
# - how many arguments are passed
# - what kind of object is passed
# - what method the object exposes
#
# So polymorphism is broader than just inheritance.
# It includes:
# - operator overloading
# - magic methods
# - duck typing
# - common interfaces
# - flexible function behavior


def show_polymorphism_summary_points():
    print("=" * 80)
    print("10. WHY THIS IS STILL POLYMORPHISM")
    print("=" * 80)
    points = [
        "Same operator can mean different things.",
        "Same function can work with different object types.",
        "Same method name can be used in multiple classes with different behavior.",
        "Custom objects can behave like built-in objects using magic methods.",
        "Python uses special method lookup rules to decide what to call.",
    ]
    for i, point in enumerate(points, start=1):
        print(f"{i}. {point}")
    print()


show_polymorphism_summary_points()

```
