```python

# ================================================================
# 1. SINGLE INHERITANCE
# ================================================================
# Definition:
# One child class inherits from exactly one parent class.
# This is the simplest form of inheritance.

class Parent:
    def __init__(self, x):
        # Constructor of parent
        self.x = x
        print("[Parent] Constructor executed")

    def show(self):
        # Parent method
        print("[Parent] show() -> x =", self.x)


class Child(Parent):
    def __init__(self, x, y):
        # Calling parent constructor using super()
        super().__init__(x)

        # Child-specific attribute
        self.y = y
        print("[Child] Constructor executed")

    def show(self):
        # Method overriding + calling parent method
        super().show()
        print("[Child] show() -> y =", self.y)


c = Child(10, 20)
c.show()

# EXPECTED OUTPUT:
# [Parent] Constructor executed
# [Child] Constructor executed
# [Parent] show() -> x = 10
# [Child] show() -> y = 20

```

```python

# ================================================================
# 2. MULTILEVEL INHERITANCE
# ================================================================
# Definition:
# A chain of inheritance (A -> B -> C)
# Each class builds on top of previous one

class A:
    def __init__(self):
        self.a = 1
        print("[A] constructor")

class B(A):
    def __init__(self):
        # Important: calling super to continue chain
        super().__init__()
        self.b = 2
        print("[B] constructor")

class C(B):
    def __init__(self):
        super().__init__()
        self.c = 3
        print("[C] constructor")

obj = C()
print(obj.a, obj.b, obj.c)

# OUTPUT:
# [A] constructor
# [B] constructor
# [C] constructor
# 1 2 3

# IMPORTANT EDGE CASE:
# If you break super(), chain breaks

class BrokenB(A):
    def __init__(self):
        self.b = 999  # super() NOT called

class BrokenC(BrokenB):
    def __init__(self):
        super().__init__()

bc = BrokenC()
print("Broken object:", bc.__dict__)

# OUTPUT:
# {'b': 999}   -> 'a' is missing (BUG!)

```

```python

# ================================================================
# 3. HIERARCHICAL INHERITANCE
# ================================================================
# Definition:
# One parent, multiple children

class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, marks):
        super().__init__(name)
        self.marks = marks

class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

s = Student("Rajan", 90)
e = Employee("Amit", 50000)

print(s.name, s.marks)
print(e.name, e.salary)

# OUTPUT:
# Rajan 90
# Amit 50000

# RULE:
# Child classes CANNOT access each other's properties
# s.salary -> ERROR

```

```python

# ================================================================
# 4. MULTIPLE INHERITANCE
# ================================================================
# Definition:
# One child inherits from multiple parents

class X:
    def show(self):
        print("[X] show")

class Y:
    def show(self):
        print("[Y] show")

class Z(X, Y):
    pass

z = Z()
z.show()

# OUTPUT:
# [X] show

# WHY?
# Because Python follows MRO (left to right)

```

```python

# ================================================================
# 5. MRO (METHOD RESOLUTION ORDER)
# ================================================================
# MRO defines HOW Python searches for methods/attributes

print(Z.mro())

# OUTPUT:
# [Z, X, Y, object]

# Meaning:
# 1. Check Z
# 2. Then X
# 3. Then Y
# 4. Then object

```

```python
class Test:
    def __init__(self):
        print("Object created")

Test()   # anonymous object

# output:
# object created

# 👉 Test() creates an instance of class Test without assigning it to a variable,
# so it is called an anonymous (unnamed) object.

# 👉 Test() = Anonymous object creation
```

```python

# ================================================================
# 6. DIAMOND PROBLEM (CRITICAL)
# ================================================================
# Classic multiple inheritance problem

class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")
        super().show()

class C(A):
    def show(self):
        print("C")
        super().show()

class D(B, C):
    def show(self):
        print("D")
        super().show()

print(D.mro())
D().show()

# OUTPUT:
# [D, B, C, A, object]
# D
# B
# C
# A

# KEY IDEA:
# Python ensures A is called ONLY ONCE
# (C3 Linearization)

```

```python

# ================================================================
# 7. super() DEEP UNDERSTANDING
# ================================================================
# super() does NOT mean "parent"
# It means "next class in MRO"

class P:
    def __init__(self):
        print("P")

class C(P):
    def __init__(self):
        print("C")
        super().__init__()

C()

# OUTPUT:
# C
# P

```

```python

# ================================================================
# 8. CONSTRUCTOR BEHAVIOR
# ================================================================

# Case 1: Child has NO constructor
class P1:
    def __init__(self):
        print("P1 constructor")

class C1(P1):
    pass

C1()

# OUTPUT: P1 constructor

# Case 2: Child overrides constructor
class P2:
    def __init__(self):
        print("P2 constructor")

class C2(P2):
    def __init__(self):
        print("C2 constructor")

C2()

# OUTPUT: C2 constructor

# Case 3: Proper chaining
class P3:
    def __init__(self):
        print("P3")

class C3(P3):
    def __init__(self):
        print("C3")
        super().__init__()

C3()

# OUTPUT:
# C3
# P3

```

```python

# ================================================================
# 9. METHOD OVERRIDING
# ================================================================

class Animal:
    def sound(self):
        print("Generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

Dog().sound()

# OUTPUT: Bark

```

```python

# ================================================================
# 10. ATTRIBUTE ACCESS RULES
# ================================================================

class Base:
    def __init__(self):
        self.a = 100

class Child(Base):
    def __init__(self):
        super().__init__()
        self.b = 200

obj = Child()
print(obj.a, obj.b)

# OUTPUT: 100 200

# RULES:
# ✔ Child can access parent
# ✖ Parent cannot access child
# ✖ Siblings cannot access each other

```

```python

# ================================================================
# 11. REAL-WORLD EXAMPLE (BANK)
# ================================================================

class BankUser:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amt):
        self.balance += amt

class Customer(BankUser):
    def withdraw(self, amt):
        if amt <= self.balance:
            self.balance -= amt

cust = Customer("Rajan", 10000)
cust.deposit(2000)
cust.withdraw(5000)
print(cust.balance)

# OUTPUT: 7000

```

```python

# ================================================================
# 12. DEBUGGING (__dict__)
# ================================================================

print(cust.__dict__)

# OUTPUT:
# {'name': 'Rajan', 'balance': 7000}

```

```python

# ================================================================
# 13. 🧠 TRICKY INTERVIEW QUESTIONS (MRO + super TRAPS)
# ================================================================

# Q1: What will be the output?
class A:
    def show(self): print("A")

class B(A):
    def show(self): print("B")

class C(A):
    def show(self): print("C")

class D(B, C):
    pass

print("Q-1")
D().show()

# ANSWER:
# B
# WHY?
# MRO: D -> B -> C -> A
# B has show(), so stops there


# Q2: What if super() is used?
class B(A):
    def show(self):
        print("B")
        super().show()

class C(A):
    def show(self):
        print("C")
        super().show()

class D(B, C):
    pass

print("Q-2")
D().show()

# ANSWER:
# B
# C
# A
# WHY?
# super() follows MRO chain


# Q3: Constructor trap
class A:
    def __init__(self): print("A")

class B(A):
    def __init__(self): print("B")

class C(B):
    def __init__(self):
        super().__init__()
        print("C")

print("Q-3")
C()

# ANSWER:
# B
# C
# (A is NOT called because B didn't call super())


# Q4: super() is NOT parent (important concept)
class A:
    def show(self): print("A")

class B(A):
    def show(self):
        print("B")
        super().show()

class C(B):
    def show(self):
        print("C")
        super().show()

print("Q-4")
C().show()

# OUTPUT:
# C
# B
# A

# KEY:
# super() means "next in MRO", not strictly parent

```

```python

# ================================================================
# 14. 📊 VISUAL DIAGRAMS (TEXT BASED FOR UNDERSTANDING)
# ================================================================

# ---------------- SINGLE INHERITANCE ----------------
# Parent
#   ↓
# Child

# ---------------- MULTILEVEL ----------------
# A → B → C

# ---------------- HIERARCHICAL ----------------
#        Parent
#       /      \
#   Child1   Child2

# ---------------- MULTIPLE ----------------
# Parent1   Parent2
#      \   /
#       Child

# ---------------- DIAMOND ----------------
#        A
#      /   \
#     B     C
#      \   /
#        D

# MRO FLOW FOR DIAMOND:
# D → B → C → A → object

# Execution Flow with super():
# D calls B
# B calls C
# C calls A

```

```python

# ================================================================
# 15. 🔍 MRO TRACE VISUALIZATION CODE
# ================================================================

class A:
    def process(self): print("A")

class B(A):
    def process(self):
        print("B")
        super().process()

class C(A):
    def process(self):
        print("C")
        super().process()

class D(B, C):
    def process(self):
        print("D")
        super().process()

print("MRO:", D.mro())
D().process()

# OUTPUT:
# MRO: [D, B, C, A, object]
# D
# B
# C
# A

# Think of it like a pipeline → each class passes control forward

```

```python
"""
MRO (Method Resolution Order) — Depth First Left to Right (Visualization Example)
===============================================================================


Rule (simplified understanding):
→ Depth First
→ Left to Right

(Actual Python uses C3 Linearization, but this example matches your diagram intuition.)

Hierarchy:

        O
      / | \
     A  B  C
      \ | / \
        X     Y
         \   /
           P

"""

# ============================================================
# BASE CLASS
# ============================================================

class O:
    pass


# ============================================================
# LEVEL 1
# ============================================================

class A(O):
    pass

class B(O):
    pass

class C(O):
    pass


# ============================================================
# LEVEL 2
# ============================================================

class X(A, B, C):
    pass

class Y(B, C):
    pass


# ============================================================
# LEVEL 3 (FINAL)
# ============================================================

class P(X, Y):
    pass


# ============================================================
# PRINT MROs
# ============================================================

def print_mro(cls):
    print(f"\nMRO of {cls.__name__}:")
    for c in cls.mro():
        print(c.__name__, end=" -> ")
    print("END")


if __name__ == "__main__":

    print("=" * 60)
    print("MRO DEMONSTRATION")
    print("=" * 60)

    print_mro(O)   # Expected: O -> object
    print_mro(A)   # Expected: A -> O -> object
    print_mro(B)   # Expected: B -> O -> object
    print_mro(C)   # Expected: C -> O -> object

    print_mro(X)   # Expected: X -> A -> B -> C -> O -> object
    print_mro(Y)   # Expected: Y -> B -> C -> O -> object

    print_mro(P)   # IMPORTANT CASE


# ============================================================
# INTERVIEW NOTE
# ============================================================

"""
Expected Output for P:

P -> X -> A -> Y -> B -> C -> O -> object

Explanation:
1. Start from P
2. Go to first parent X
3. Follow X's order: A → B → C
4. But Y must also be respected → comes before repeating B
5. Final order resolves conflicts properly

NOTE:
Python actually uses C3 Linearization (not pure DFS).
But for learning, "Depth First Left to Right" gives intuition.

Always verify using:
    print(P.mro())

"""
```
