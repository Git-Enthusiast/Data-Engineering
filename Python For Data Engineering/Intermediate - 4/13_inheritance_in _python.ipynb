```python

# =========================================================
# 1. WHAT IS INHERITANCE?
# =========================================================
"""
Inheritance is a mechanism in OOP where one class (child)
acquires properties (variables) and behaviors (methods)
of another class (parent).

Parent Class = Base Class
Child Class = Derived Class

Real-life analogy:
Father -> Son
Son inherits traits from father
"""

```

```python

# =========================================================
# 2. BASIC SYNTAX
# =========================================================

class Parent:
    pass

class Child(Parent):
    pass

# Child inherits everything from Parent

```

```python

# =========================================================
# 3. WHY WE NEED INHERITANCE (PROBLEM WITHOUT IT)
# =========================================================

"""
Without inheritance -> Code duplication
"""

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print(f"Employee: {self.name}, Salary: {self.salary}")


# Duplicate code (BAD PRACTICE)
class Manager:
    def __init__(self, name, salary, dept):
        self.name = name
        self.salary = salary
        self.dept = dept

    def display(self):
        print(f"Manager: {self.name}, Salary: {self.salary}, Dept: {self.dept}")

```

```python

# =========================================================
# 4. SOLUTION USING INHERITANCE
# =========================================================

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print(f"Employee: {self.name}, Salary: {self.salary}")


class Manager(Employee):  # Inheriting Employee
    def __init__(self, name, salary, dept):
        # calling parent constructor
        super().__init__(name, salary)
        self.dept = dept

    def display(self):
        print(f"Manager: {self.name}, Salary: {self.salary}, Dept: {self.dept}")


m1 = Manager("Rajan", 50000, "IT")
m1.display()

```

```python

# =========================================================
# 5. IMPORTANT RULES OF INHERITANCE
# =========================================================

"""
Rule 1:
Child CAN access parent properties

Rule 2:
Parent CANNOT access child properties

Rule 3:
Child can override parent methods

Rule 4:
If child has no constructor -> parent constructor runs
"""

```

```python

# =========================================================
# 6. PARENT -> CHILD ACCESS (VALID)
# =========================================================

class A:
    def show(self):
        print("This is parent method")


class B(A):
    pass

b = B()
b.show()  # inherited method

```

```python

# =========================================================
# 7. CHILD -> PARENT ACCESS (INVALID)
# =========================================================

class A:
    def show(self):
        print("Parent")


class B(A):
    def child_method(self):
        print("Child only")


a = A()

# a.child_method()  # ERROR -> AttributeError

```

```python

# =========================================================
# 8. CONSTRUCTOR IN INHERITANCE
# =========================================================

"""
Case 1: Child has NO constructor
-> Parent constructor is automatically called
"""

class Parent:
    def __init__(self):
        print("Parent constructor called")


class Child(Parent):
    pass

c = Child()  # Parent constructor runs

```

```python

# =========================================================
# 9. CONSTRUCTOR OVERRIDING
# =========================================================

"""
If child defines its own constructor ->
Parent constructor will NOT run automatically
"""

class Parent:
    def __init__(self):
        print("Parent constructor")


class Child(Parent):
    def __init__(self):
        print("Child constructor")


c = Child()  # Only child constructor runs

```

```python

# =========================================================
# 10. USING super() FUNCTION (VERY IMPORTANT)
# =========================================================

"""
super() is used to call parent class methods/constructor
"""

class Computer:
    def __init__(self, ram, storage):
        self.ram = ram
        self.storage = storage
        print("Computer constructor called")


class Mobile(Computer):
    def __init__(self, ram, storage):
        # Calling parent constructor
        super().__init__(ram, storage)
        self.model = "iPhone X"
        print("Mobile constructor called")


phone = Mobile("8GB", "512GB")

```

```python

# =========================================================
# 11. ACCESSING PARENT METHODS USING super()
# =========================================================

class Parent:
    def show(self):
        print("Parent show")


class Child(Parent):
    def show(self):
        super().show()  # calling parent method
        print("Child show")


c = Child()
c.show()

```

```python

# =========================================================
# 12. METHOD OVERRIDING
# =========================================================

"""
Child class can override parent method
"""

class Parent:
    def greet(self):
        print("Hello from Parent")


class Child(Parent):
    def greet(self):
        print("Hello from Child")


c = Child()
c.greet()  # Child method runs

```

```python

# =========================================================
# 13. ACCESSING OBJECT DATA (__dict__)
# =========================================================

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class CollegeStudent(Student):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course


s = CollegeStudent("Rajan", 21, "CSE")

print(s.__dict__)  # shows all attributes

```

```python

# =========================================================
# 14. BUILT-IN OBJECT CLASS
# =========================================================

"""
All classes in Python are derived from 'object' class
"""

class Demo(object):
    pass

```

```python

# =========================================================
# 16. REAL-WORLD EXAMPLE (BANK SYSTEM)
# =========================================================

class BankUser:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def show_balance(self):
        print(f"Balance: {self.balance}")


class Customer(BankUser):
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient funds")


class Employee(BankUser):
    def approve_loan(self):
        print("Loan approved")


cust = Customer("Rajan", 10000)
cust.deposit(2000)
cust.withdraw(5000)
cust.show_balance()

emp = Employee("Admin", 0)
emp.approve_loan()

```
