```python
# =========================================================
# BUILT-IN CLASS ATTRIBUTES IN PYTHON
# =========================================================
# Python automatically provides some special attributes
# for every class. These are mainly used for metadata,
# debugging, and introspection.

# Built-in attributes covered:
# 1. __doc__     -> Class documentation string
# 2. __name__    -> Class name
# 3. __module__  -> Module name where class is defined
# 4. __dict__    -> Namespace (all attributes of class)

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Student:
    """
    This is a Student class.
    It stores student name and age.
    """

    college = "ABC University"   # class attribute

    def __init__(self, name, age):
        self.name = name        # instance attribute
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

```

```python

# =========================================================
# ACCESSING BUILT-IN CLASS ATTRIBUTES
# =========================================================

print("========== BUILT-IN CLASS ATTRIBUTES ==========\n")

# 1. __doc__ -> Documentation string of class
print("1. __doc__ (Class Documentation):")
print(Student.__doc__)


# 2. __name__ -> Name of the class
print("\n2. __name__ (Class Name):")
print(Student.__name__)


# 3. __module__ -> Module where class is defined
print("\n3. __module__ (Module Name):")
print(Student.__module__)
# If this file is run directly → "__main__"
# If imported from another file → module name


# 4. __dict__ -> All attributes of class (in dictionary form)
print("\n4. __dict__ (Class Namespace):")
print(Student.__dict__)

```

```python

# =========================================================
# OBJECT CREATION
# =========================================================
s1 = Student("Rajan", 22)

print("\n========== OBJECT DATA ==========")
s1.display()


# =========================================================
# ACCESSING __dict__ OF OBJECT
# =========================================================
print("\nObject __dict__ (Instance Namespace):")
print(s1.__dict__)

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. __doc__ helps in documentation (used in help() function).
# 2. __name__ is useful in logging, debugging, reflection.
# 3. __module__ tells where class is defined.
# 4. __dict__ shows all attributes (very important for debugging).
# 5. Class __dict__ contains methods + class variables.
# 6. Object __dict__ contains only instance variables.


# =========================================================
# BONUS: USING help() FUNCTION
# =========================================================
print("\n========== HELP OUTPUT ==========")
help(Student)
```
