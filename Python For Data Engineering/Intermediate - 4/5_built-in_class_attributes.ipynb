```python
# =========================================================
# BUILT-IN CLASS FUNCTIONS IN PYTHON
# =========================================================
# These functions help us dynamically access and manipulate
# object attributes from outside the class.

# Functions covered:
# 1. getattr()  -> Get attribute value
# 2. setattr()  -> Set / Update attribute value
# 3. delattr()  -> Delete attribute
# 4. hasattr()  -> Check if attribute exists

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Student:
    def __init__(self, name, age):
        # instance attributes
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


# =========================================================
# OBJECT CREATION
# =========================================================
s1 = Student("Rajan", 22)

print("Initial Object Data:")
s1.display()
```

```python

# =========================================================
# 1. getattr() -> ACCESS ATTRIBUTE
# =========================================================
# Syntax: getattr(object, attribute_name, default_value)

print("\nUsing getattr():")

# Access existing attribute
name = getattr(s1, "name")
print("Name:", name)

# Access non-existing attribute (with default)
grade = getattr(s1, "grade", "Not Available")
print("Grade:", grade)

```

```python

# =========================================================
# 2. setattr() -> SET / UPDATE ATTRIBUTE
# =========================================================
# Syntax: setattr(object, attribute_name, value)

print("\nUsing setattr():")

# Update existing attribute
setattr(s1, "age", 23)

# Add new attribute dynamically
setattr(s1, "grade", "A")

print("After setattr():")
s1.display()
print("Grade:", s1.grade)

```

```python

# =========================================================
# 3. hasattr() -> CHECK ATTRIBUTE EXISTS OR NOT
# =========================================================
# Syntax: hasattr(object, attribute_name)

print("\nUsing hasattr():")

print("Has 'name'? ->", hasattr(s1, "name"))
print("Has 'marks'? ->", hasattr(s1, "marks"))

```

```python

# =========================================================
# 4. delattr() -> DELETE ATTRIBUTE
# =========================================================
# Syntax: delattr(object, attribute_name)

print("\nUsing delattr():")

# Before deleting
print("Before deletion:", s1.__dict__)

# Delete attribute
delattr(s1, "grade")

# After deleting
print("After deletion:", s1.__dict__)

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. These functions allow dynamic behavior (very useful in frameworks).
# 2. getattr() avoids errors if attribute may or may not exist.
# 3. setattr() can create new attributes at runtime.
# 4. delattr() permanently removes an attribute.
# 5. hasattr() helps in safe programming (checking before accessing).

# =========================================================
# BONUS: ACCESSING ALL ATTRIBUTES USING __dict__
# =========================================================
print("\nAll attributes using __dict__:")
print(s1.__dict__)
```
