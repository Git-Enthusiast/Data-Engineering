```python
# =========================================================
# CLASS VARIABLES & CLASS METHODS IN PYTHON
# =========================================================
# Class Variables:
# - Shared among ALL objects of a class
# - Defined inside class but outside methods
#
# Class Methods:
# - Bound to the class (NOT object)
# - Defined using @classmethod decorator
# - First parameter is 'cls' (class reference)

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Employee:
    # -----------------------------------------------------
    # CLASS VARIABLE (shared by all objects)
    # -----------------------------------------------------
    company_name = "TechCorp"
    employee_count = 0

    # -----------------------------------------------------
    # CONSTRUCTOR (INSTANCE VARIABLES)
    # -----------------------------------------------------
    def __init__(self, name, salary):
        self.name = name              # instance variable
        self.salary = salary          # instance variable

        # increase employee count whenever object is created
        Employee.employee_count += 1

    # -----------------------------------------------------
    # INSTANCE METHOD
    # -----------------------------------------------------
    def display(self):
        print(f"Name: {self.name}, Salary: {self.salary}, Company: {Employee.company_name}")

    # -----------------------------------------------------
    # CLASS METHOD
    # -----------------------------------------------------
    @classmethod
    def change_company(cls, new_name):
        # cls refers to the class (Employee)
        cls.company_name = new_name

    # Another class method example
    @classmethod
    def get_employee_count(cls):
        return cls.employee_count

    # -----------------------------------------------------
    # STATIC METHOD (for comparison)
    # -----------------------------------------------------
    @staticmethod
    def general_info():
        print("Employees are valuable assets of a company.")

```

```python

# =========================================================
# OBJECT CREATION
# =========================================================
e1 = Employee("Rajan", 50000)
e2 = Employee("Amit", 60000)

print("========== INITIAL DATA ==========")
e1.display()
e2.display()

```

```python

# =========================================================
# CLASS VARIABLE BEHAVIOR (SHARED DATA)
# =========================================================
print("\n========== CLASS VARIABLE SHARING ==========")

# Access via class
print("Company (via class):", Employee.company_name)

# Access via objects
print("Company (via e1):", e1.company_name)
print("Company (via e2):", e2.company_name)

```

```python

# =========================================================
# MODIFY CLASS VARIABLE USING CLASS (CORRECT WAY)
# =========================================================
print("\n========== MODIFY USING CLASS ==========")

Employee.company_name = "Infosys"

e1.display()
e2.display()

```

```python

# =========================================================
# MODIFY CLASS VARIABLE USING OBJECT (IMPORTANT TRAP)
# =========================================================
print("\n========== MODIFY USING OBJECT (TRAP) ==========")

# This creates a NEW instance variable, NOT changing class variable
e1.company_name = "TCS"

print("e1 company:", e1.company_name)   # TCS (instance variable)
print("e2 company:", e2.company_name)   # Infosys (class variable)
print("Class company:", Employee.company_name)  # Infosys

# Check namespaces
print("\ne1 __dict__:", e1.__dict__)
print("Employee __dict__:", Employee.__dict__)

```

```python

# =========================================================
# USING CLASS METHOD
# =========================================================
print("\n========== CLASS METHOD USAGE ==========")

# Change company using class method
Employee.change_company("Google")

e1.display()
e2.display()

# Change company using class variable directly
Employee.company_name = "Y combinator"

e1.display()
e2.display()

# Access employee count
print("Total Employees:", Employee.get_employee_count())

```

```python

# =========================================================
# CALLING CLASS METHOD USING OBJECT
# =========================================================
print("\n========== CALL VIA OBJECT ==========")

# Even if called via object, it still works on class
e1.change_company("Microsoft")

e1.display()
e2.display()


# =========================================================
# HOW cls WORKS
# =========================================================
# When you call:
# Employee.change_company("ABC")
# Python internally converts it to:
# Employee.change_company(Employee, "ABC")
#
# So 'cls' = Employee

```

```python

# =========================================================
# PRACTICAL USE CASE
# =========================================================
print("\n========== PRACTICAL USE ==========")

class Student:
    school_name = "ABC School"

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_school(cls, new_school):
        cls.school_name = new_school

    def display(self):
        print(f"Name: {self.name}, School: {Student.school_name}")


s1 = Student("Neha")
s2 = Student("Rahul")

s1.display()
s2.display()

# Change for ALL students
Student.change_school("XYZ School")

print("After changing school:")
s1.display()
s2.display()

```

```python

# =========================================================
# STATIC METHOD USAGE (FOR COMPARISON)
# =========================================================
print("\n========== STATIC METHOD ==========")

Employee.general_info()

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. Class variables are shared among all objects.
# 2. Use class name to modify class variables (best practice).
# 3. Using object to modify creates a NEW instance variable.
# 4. Class methods use 'cls' (class reference).
# 5. Class methods can modify class variables globally.
# 6. Can call class methods using class or object.
# 7. Static methods are independent (no self/cls).

```

```python

# =========================================================
# QUICK REVISION
# =========================================================
# Class Variable  -> Shared data
# Instance Variable -> Unique per object
# Class Method    -> Uses cls (class-level)
# Instance Method -> Uses self (object-level)
# Static Method   -> No self/cls
```
