# 🌟 SIMPLE UNDERSTANDING FIRST

## 🧠 What is a Class?
👉 **A class is a blueprint (template)**

Think like this:
> A class = design of a house

It defines:
* rooms
* doors
* windows

But…
👉 **It is NOT a real house yet.**

---

## 🏠 What is an Object?
👉 **An object is the real thing created from the class**

* **Class** = blueprint
* **Object** = actual house built using blueprint

---

## 🚗 Real-Life Example

**Class:** Car
It defines:
* color
* brand
* speed

**Object:**
* Red BMW
* Black Tesla

> 👉 *Same class → different objects*

---

## 🔥 CORE CONCEPT

| Concept | Meaning |
| :--- | :--- |
| **Class** | Blueprint |
| **Object** | Instance (real object) |
| **Attributes** | Variables (data) |
| **Methods** | Functions inside class |

````python

# =========================================================
# 1. SIMPLE CLASS AND OBJECT
# =========================================================

# Question:
# Create a class "Student" with name and age


class Student:
    # Constructor: used to initialize object data
    def __init__(self, name, age):
        # Attributes (data)
        self.name = name
        self.age = age


# Creating objects
s1 = Student("Rajan", 22)
s2 = Student("Amit", 21)

# Accessing data
print("Student 1:", s1.name, s1.age)
print("Student 2:", s2.name, s2.age)

````

````python

# =========================================================
# 2. CLASS WITH METHOD
# =========================================================

# Question:
# Add a method to display student details


class StudentWithMethod:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


s3 = StudentWithMethod("Rahul", 23)
s3.display()

````

````python

# =========================================================
# 3. REAL LIFE EXAMPLE: CAR
# =========================================================

# Question:
# Create a Car class with brand, color, speed


class Car:
    def __init__(self, brand, color, speed):
        self.brand = brand
        self.color = color
        self.speed = speed

    def drive(self):
        print(f"{self.brand} car is driving at {self.speed} km/h")

    def stop(self):
        print(f"{self.brand} car has stopped")


car1 = Car("BMW", "Red", 200)
car2 = Car("Audi", "Black", 180)

car1.drive()
car2.stop()

````

````python

# =========================================================
# 4. EVERYTHING IS OBJECT IN PYTHON
# =========================================================

# Even basic data types are objects

x = 10
print(type(x))  # int is also a class

lst = [1, 2, 3]
print(type(lst))  # list is also a class

````

````python

# =========================================================
# 5. BANK ACCOUNT EXAMPLE
# =========================================================

# Question:
# Create a BankAccount class with deposit and withdraw


class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. Remaining balance: {self.balance}")


acc1 = BankAccount("Rajan", 1000)
acc1.deposit(500)
acc1.withdraw(300)

````

````python

# =========================================================
# 1. WHAT IS A CONSTRUCTOR?
# =========================================================

"""
A CONSTRUCTOR is a special method in a class.

👉 It runs automatically when object is created
👉 Used to initialize object data

In Python:
Constructor = __init__ method

IMPORTANT:
- You NEVER call constructor manually
- Python calls it automatically
"""

````

````python

# =========================================================
# 2. BASIC CONSTRUCTOR (__init__)
# =========================================================

class Student:
    def __init__(self, name, age):
        """
        __init__ is called automatically when object is created
        self = current object
        """
        self.name = name
        self.age = age


# Object creation (constructor runs automatically)
s1 = Student("Rajan", 22)

print("Name:", s1.name)
print("Age:", s1.age)
print(f"Details: {s1.__dict__}")

````

````python

# =========================================================
# 3. UNDERSTANDING 'self' (VERY IMPORTANT)
# =========================================================

"""
self represents the CURRENT OBJECT

Example:
s1 = Student("Rajan", 22)

Internally Python does:
Student.__init__(s1, "Rajan", 22)

So:
self → s1

IMPORTANT RULE:
- self is always FIRST parameter
- Python automatically passes it
"""

# Example to visualize self
class Demo:
    def __init__(self):
        print("Constructor called for object:", self)


d1 = Demo()
d2 = Demo()

print(id(d1))
print(id(d2))

````

````python

# =========================================================
# 4. TYPES OF CONSTRUCTORS
# =========================================================

# ---------------------------------------------------------
# 4.1 DEFAULT CONSTRUCTOR
# ---------------------------------------------------------

"""
Constructor without parameters (except self)
"""

class DefaultExample:
    def __init__(self):
        print("Default constructor called")


obj1 = DefaultExample()

````

````python
# Even if you don’t write a constructor, Python provides a default one internally.
class Student:
    pass

s1 = Student()  # Works fine
````

````python

# ---------------------------------------------------------
# 4.3 PARAMETERIZED CONSTRUCTOR
# ---------------------------------------------------------

"""
Constructor with parameters
Used to pass values during object creation
"""

class ParameterizedExample:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"{self.name} scored {self.marks}")


p1 = ParameterizedExample("Rajan", 95)
p1.display()

````

### ⚠️ Important Note (Python-specific)

Unlike languages like Java or C++:

* Python has only one constructor method → `__init__()`
* **No constructor overloading** ❌

But you can simulate it using:
* Default values
* `*args`
* `**kwargs`

````python
class Student:
    def __init__(self, *args, **kwargs):
        print("Args:", args)
        print("Kwargs:", kwargs)

s1 = Student("Rajan", 22)
s2 = Student(name="Aman", age=21)
s3 = Student()
````

````python

# =========================================================
# 5. CONSTRUCTOR VS NORMAL METHOD
# =========================================================

"""
Constructor:
- Runs automatically
- Used for initialization

Normal Method:
- Called manually
"""

class Example:
    def __init__(self):
        print("Constructor executed")

    def normal_method(self):
        print("Normal method executed")


e = Example()          # constructor runs automatically
e.normal_method()      # manually called

````

````python

# =========================================================
# 6. REAL LIFE EXAMPLE: BANK ACCOUNT
# =========================================================

class BankAccount:
    def __init__(self, owner, balance):
        """
        Constructor initializes account details
        """
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. Balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. Balance: {self.balance}")


acc = BankAccount("Rajan", 1000)
acc.deposit(500)
acc.withdraw(300)

````

````python

# =========================================================
# 7. MULTIPLE OBJECTS (CONSTRUCTOR RUNS EACH TIME)
# =========================================================
class Student:
    def __init__(self, *args, **kwargs):
        self.args = args
        self.kwargs = kwargs

s4 = Student("Amit", 21)
s5 = Student("Neha", 20)

print(s4.args, s5.args)
print(s4.__dict__)
print(s5.__dict__)

````

## 1. What `__dict__` is

Every Python object has a special attribute `__dict__`.

It’s a dictionary that stores all instance attributes for that object.

The keys are the attribute names, and the values are the attribute values.
For example:

```python
self.args = args
self.kwargs = kwargs
```

`args` and `kwargs` are now instance attributes, so they go into `__dict__`.

---

### Memory Visualization:
        
**Class and Objects Memory Layout**
```text
          Student Class
   ---------------------------
   | __init__(self, *args, **kwargs) |
   ---------------------------
```

```python
s4 = Student("Amit", 21)
s5 = Student("Neha", 20)
```

**Object s4**
```text
s4
--------------------------
| args   | ('Amit', 21) |
| kwargs | {}           |
--------------------------
s4.__dict__ → {'args': ('Amit', 21), 'kwargs': {}}
```

**Object s5**
```text
s5
--------------------------
| args   | ('Neha', 20) |
| kwargs | {}           |
--------------------------
s5.__dict__ → {'args': ('Neha', 20), 'kwargs': {}}
```

---

### Key Points from Diagram

* Each object (`s4`, `s5`) has its own `__dict__` storing its attributes.
* Changes in one object do not affect the other.
* `__dict__` allows dynamic inspection and modification of an object’s attributes:

```python
s4.grade = "A+"
print(s4.__dict__)  # {'args': ('Amit', 21), 'kwargs': {}, 'grade': 'A+'}
```

* Class-level methods like `__init__` are shared, but instance attributes live in `__dict__` per object.

---

### Complete Memory Layout Diagram:

```text
        +--------------------+
        |   Student Class    |
        +--------------------+
        | __init__ method    |  <- shared by all objects
        +--------------------+
```

```python
s4 = Student("Amit", 21)
s5 = Student("Neha", 20)
```

**Memory Representation:**

```text
         +---------+             +------------------+
         |  s4     |             |  s4.__dict__     |
         +---------+             +------------------+
         | pointer | ----------> | 'args': ('Amit',21) |
         | to obj  |             | 'kwargs': {}        |
         +---------+             +------------------+

         +---------+             +------------------+
         |  s5     |             |  s5.__dict__     |
         +---------+             +------------------+
         | pointer | ----------> | 'args': ('Neha',20) |
         | to obj  |             | 'kwargs': {}        |
         +---------+             +------------------+
```

---

### Notes:
* Both `s4` and `s5` **share the `Student` class**, i.e., they use the same `__init__` method.
* Each object has its **own `__dict__`**, storing only the attributes assigned to that instance.
* Adding a new attribute to `s4`, e.g., `s4.grade = "A+"`, only affects `s4.__dict__`:

```text
         s4.__dict__ → {'args': ('Amit',21), 'kwargs': {}, 'grade': 'A+'}
         s5.__dict__ → {'args': ('Neha',20), 'kwargs': {}}
```

````python

# =========================================================
# 11. CAN WE HAVE MULTIPLE CONSTRUCTORS?
# =========================================================

"""
Python does NOT support multiple constructors directly.

But we can simulate using:
- Default arguments
"""

class MultiConstructor:
    def __init__(self, name=None, age=None):
        if name and age:
            print("Parameterized constructor")
        else:
            print("Default constructor")


m1 = MultiConstructor()
m2 = MultiConstructor("Rajan", 22)

````
