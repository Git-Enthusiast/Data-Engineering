# 🌟 SIMPLE UNDERSTANDING FIRST

## 🧠 What is a Class?
👉 **A class is a blueprint (template)**

Think like this:
> A class = design of a house

It defines:
* rooms
* doors
* windows

But…
👉 **It is NOT a real house yet.**

---

## 🏠 What is an Object?
👉 **An object is the real thing created from the class**

* **Class** = blueprint
* **Object** = actual house built using blueprint

---

## 🚗 Real-Life Example

**Class:** Car
It defines:
* color
* brand
* speed

**Object:**
* Red BMW
* Black Tesla

> 👉 *Same class → different objects*

---

## 🔥 CORE CONCEPT

| Concept | Meaning |
| :--- | :--- |
| **Class** | Blueprint |
| **Object** | Instance (real object) |
| **Attributes** | Variables (data) |
| **Methods** | Functions inside class |

```python
"""
===========================================================
            CLASS AND OBJECTS IN PYTHON 
===========================================================

This file explains:
1. What is a class
2. What is an object
3. Attributes and Methods
4. Real-life examples
5. Multiple Python examples
6. Everything explained in EASY language

-----------------------------------------------------------
TEXT DIAGRAM:

        CLASS (Blueprint)
              ↓
      -----------------
      |   Car Class   |
      -----------------
       |   |     |
    color brand speed

              ↓ (create)

        OBJECTS (Real Instances)
      ----------------------------
      car1 → Red BMW 200km/h
      car2 → Black Audi 180km/h
      ----------------------------

===========================================================
"""

```

```python

# =========================================================
# 1. SIMPLE CLASS AND OBJECT
# =========================================================

# Question:
# Create a class "Student" with name and age


class Student:
    # Constructor: used to initialize object data
    def __init__(self, name, age):
        # Attributes (data)
        self.name = name
        self.age = age


# Creating objects
s1 = Student("Rajan", 22)
s2 = Student("Amit", 21)

# Accessing data
print("Student 1:", s1.name, s1.age)
print("Student 2:", s2.name, s2.age)

```

```python

# =========================================================
# 2. CLASS WITH METHOD
# =========================================================

# Question:
# Add a method to display student details


class StudentWithMethod:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


s3 = StudentWithMethod("Rahul", 23)
s3.display()

```

```python

# =========================================================
# 3. REAL LIFE EXAMPLE: CAR
# =========================================================

# Question:
# Create a Car class with brand, color, speed


class Car:
    def __init__(self, brand, color, speed):
        self.brand = brand
        self.color = color
        self.speed = speed

    def drive(self):
        print(f"{self.brand} car is driving at {self.speed} km/h")

    def stop(self):
        print(f"{self.brand} car has stopped")


car1 = Car("BMW", "Red", 200)
car2 = Car("Audi", "Black", 180)

car1.drive()
car2.stop()

```

```python

# =========================================================
# 4. EVERYTHING IS OBJECT IN PYTHON
# =========================================================

# Even basic data types are objects

x = 10
print(type(x))  # int is also a class

lst = [1, 2, 3]
print(type(lst))  # list is also a class

```

```python

# =========================================================
# 5. BANK ACCOUNT EXAMPLE
# =========================================================

# Question:
# Create a BankAccount class with deposit and withdraw


class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. Remaining balance: {self.balance}")


acc1 = BankAccount("Rajan", 1000)
acc1.deposit(500)
acc1.withdraw(300)

```

```python

# =========================================================
# 6. DIFFERENCE BETWEEN CLASS AND OBJECT
# =========================================================

"""
CLASS:
- Blueprint
- No memory allocated (until object created)

OBJECT:
- Real instance
- Memory allocated
- Can access data and methods

Example:
Student → Class
s1 → Object
"""

```

```python

# =========================================================
# 7. MULTIPLE OBJECTS SAME CLASS
# =========================================================

# One class → many objects

s4 = Student("Neha", 20)
s5 = Student("Priya", 19)

print(s4.name)
print(s5.name)

```

```python

# =========================================================
# 1. WHAT IS A CONSTRUCTOR?
# =========================================================

"""
A CONSTRUCTOR is a special method in a class.

👉 It runs automatically when object is created
👉 Used to initialize object data

In Python:
Constructor = __init__ method

IMPORTANT:
- You NEVER call constructor manually
- Python calls it automatically
"""

```

```python

# =========================================================
# 2. BASIC CONSTRUCTOR (__init__)
# =========================================================

class Student:
    def __init__(self, name, age):
        """
        __init__ is called automatically when object is created
        self = current object
        """
        self.name = name
        self.age = age


# Object creation (constructor runs automatically)
s1 = Student("Rajan", 22)

print("Name:", s1.name)
print("Age:", s1.age)

```

```python

# =========================================================
# 3. UNDERSTANDING 'self' (VERY IMPORTANT)
# =========================================================

"""
self represents the CURRENT OBJECT

Example:
s1 = Student("Rajan", 22)

Internally Python does:
Student.__init__(s1, "Rajan", 22)

So:
self → s1

IMPORTANT RULE:
- self is always FIRST parameter
- Python automatically passes it
"""

# Example to visualize self
class Demo:
    def __init__(self):
        print("Constructor called for object:", self)


d1 = Demo()
d2 = Demo()

```

```python

# =========================================================
# 4. TYPES OF CONSTRUCTORS
# =========================================================

# ---------------------------------------------------------
# 4.1 DEFAULT CONSTRUCTOR
# ---------------------------------------------------------

"""
Constructor without parameters (except self)
"""

class DefaultExample:
    def __init__(self):
        print("Default constructor called")


obj1 = DefaultExample()

```

```python

# ---------------------------------------------------------
# 4.2 PARAMETERIZED CONSTRUCTOR
# ---------------------------------------------------------

"""
Constructor with parameters
Used to pass values during object creation
"""

class ParameterizedExample:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"{self.name} scored {self.marks}")


p1 = ParameterizedExample("Rajan", 95)
p1.display()

```

```python

# =========================================================
# 5. CONSTRUCTOR VS NORMAL METHOD
# =========================================================

"""
Constructor:
- Runs automatically
- Used for initialization

Normal Method:
- Called manually
"""

class Example:
    def __init__(self):
        print("Constructor executed")

    def normal_method(self):
        print("Normal method executed")


e = Example()          # constructor runs automatically
e.normal_method()      # manually called

```

```python

# =========================================================
# 6. REAL LIFE EXAMPLE: BANK ACCOUNT
# =========================================================

class BankAccount:
    def __init__(self, owner, balance):
        """
        Constructor initializes account details
        """
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. Balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. Balance: {self.balance}")


acc = BankAccount("Rajan", 1000)
acc.deposit(500)
acc.withdraw(300)

```

```python

# =========================================================
# 7. MULTIPLE OBJECTS (CONSTRUCTOR RUNS EACH TIME)
# =========================================================

s2 = Student("Amit", 21)
s3 = Student("Neha", 20)

print(s2.name, s3.name)

```

```python

# =========================================================
# 8. IMPORTANT POINTS ABOUT CONSTRUCTORS
# =========================================================

"""
1. Constructor name must be __init__
2. It runs automatically
3. Used to initialize data
4. self is mandatory
5. Can take parameters
"""

```

```python

# =========================================================
# 9. COMMON MISTAKES
# =========================================================

# ❌ Mistake 1: Forgetting self
class WrongExample:
    def __init__(self, name):
        # self missing in variable assignment
        # This creates local variable, not instance variable
        name = name   # WRONG


# ❌ Mistake 2: Not using __init__
class WrongExample2:
    def init(self):   # WRONG name
        print("This will NOT act as constructor")

```

```python

# =========================================================
# 10. ADVANCED: DEFAULT VALUES IN CONSTRUCTOR
# =========================================================

class AdvancedStudent:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age


a1 = AdvancedStudent()
a2 = AdvancedStudent("Rajan", 22)

print(a1.name, a1.age)
print(a2.name, a2.age)

```

```python

# =========================================================
# 11. CAN WE HAVE MULTIPLE CONSTRUCTORS?
# =========================================================

"""
Python does NOT support multiple constructors directly.

But we can simulate using:
- Default arguments
"""

class MultiConstructor:
    def __init__(self, name=None, age=None):
        if name and age:
            print("Parameterized constructor")
        else:
            print("Default constructor")


m1 = MultiConstructor()
m2 = MultiConstructor("Rajan", 22)

```

```python

# =========================================================
# 12. EVERYTHING IS OBJECT
# =========================================================

x = 10
print(type(x))  # int class

lst = [1, 2, 3]
print(type(lst))  # list class

```
