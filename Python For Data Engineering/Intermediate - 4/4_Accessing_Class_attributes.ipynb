```python
"""
                        Accessing Class Members in Python
"""

# =========================================================
# 1. Defining Class Members
# =========================================================
# Class members include:
# 1. Attributes (variables inside the class)
# 2. Methods (functions inside the class)

class Employee:
    def __init__(self, name, salary):
        self.name = name    # Instance attribute
        self.salary = salary  # Instance attribute

    def display(self):
        """Method to display employee details"""
        print(f"Employee Name: {self.name}, Salary: {self.salary}")

    def give_raise(self, amount):
        """Method to increase salary"""
        self.salary += amount
        print(f"{self.name}'s salary updated to {self.salary}")

print("\n[INFO] Class Employee defined with attributes and methods.\n")

# =========================================================
# 2. Creating an Object (Memory Allocation)
# =========================================================
emp = Employee("Rajan", 50000)  
# Memory is allocated for emp object and __init__ initializes attributes
print(f"[MEMORY] Object 'emp' created at memory location: {id(emp)}")
print(f"[MEMORY] emp.__dict__ = {emp.__dict__}\n")

# =========================================================
# 3. Accessing Attributes from Outside the Class
# =========================================================
print("[ACCESS ATTRIBUTES OUTSIDE CLASS]")
print(f"emp.name = {emp.name}")
print(f"emp.salary = {emp.salary}")

# Internal memory working:
# emp points to a memory location where 'name' and 'salary' are stored
# Accessing emp.name retrieves the value stored at that memory location

# =========================================================
# 4. Updating Attributes from Outside the Class
# =========================================================
print("\n[UPDATE ATTRIBUTES OUTSIDE CLASS]")
emp.salary = 60000  # Updating attribute value
emp.name = "Rajan Raj"
print(f"Updated emp.name = {emp.name}")
print(f"Updated emp.salary = {emp.salary}")
print(f"[MEMORY] emp.__dict__ = {emp.__dict__}\n")

# =========================================================
# 5. Accessing and Calling Methods from Outside the Class
# =========================================================
print("[CALLING METHODS OUTSIDE CLASS]")
emp.display()  # Calling method using object
emp.give_raise(5000)  # Method modifies object data
emp.display()

# Internal memory working:
# The object reference 'emp' points to memory of this instance
# Methods use 'self' to operate on the memory of the correct object
# Dot notation ensures Python knows which object's method to execute

# =========================================================
# 6. Creating Another Object to Show Independence
# =========================================================
emp2 = Employee("Neha", 45000)
print("\n[NEW OBJECT CREATED]")
print(f"emp2 memory location: {id(emp2)}")
emp2.display()
emp2.give_raise(3000)
emp2.display()

# Demonstrates that attributes and methods are **per-object**
print(f"\nemp.__dict__ = {emp.__dict__}")
print(f"emp2.__dict__ = {emp2.__dict__}")

# =========================================================
# 7. Summary
# =========================================================
print("""
[SUMMARY]
1. Class members include attributes (variables) and methods (functions).
2. Attributes can be accessed outside the class using object.attribute syntax.
3. Attributes can be updated outside the class by assigning a new value.
4. Methods can be called outside the class using object.method() syntax.
5. Each object has its own memory space; 'self' ensures methods operate on the correct object.
6. Dot notation allows Python to locate the object's memory location and retrieve or modify its members.
""")
```
