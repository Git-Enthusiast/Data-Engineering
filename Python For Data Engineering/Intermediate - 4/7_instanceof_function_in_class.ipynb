```python
# =========================================================
# isinstance() FUNCTION IN PYTHON
# =========================================================
# isinstance() is used to check whether an object belongs
# to a particular class (or multiple classes).
#
# Syntax:
# isinstance(object, class_or_tuple_of_classes)
#
# Returns:
# True  -> if object is instance of given class
# False -> otherwise

```

```python

# =========================================================
# CLASS DEFINITIONS
# =========================================================
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def bark(self):
        print("Dog barks")


class Cat(Animal):
    def meow(self):
        print("Cat meows")



# =========================================================
# OBJECT CREATION
# =========================================================
dog1 = Dog()
cat1 = Cat()

```

```python

# =========================================================
# BASIC USAGE OF isinstance()
# =========================================================
print("========== BASIC CHECK ==========")

print("dog1 is Dog? ->", isinstance(dog1, Dog))       # True
print("dog1 is Animal? ->", isinstance(dog1, Animal)) # True (Inheritance)
print("dog1 is Cat? ->", isinstance(dog1, Cat))       # False

print(type(dog1))
print(type(cat1))
```

```python

# =========================================================
# CHECKING MULTIPLE CLASSES (TUPLE)
# =========================================================
print("\n========== MULTIPLE CLASS CHECK ==========")

print("dog1 is Dog or Cat? ->", isinstance(dog1, (Dog, Cat)))
print("cat1 is Dog or Cat? ->", isinstance(cat1, (Dog, Cat)))

```

```python

# =========================================================
# WITH BUILT-IN TYPES
# =========================================================
print("\n========== BUILT-IN TYPE CHECK ==========")

x = 10
y = "Hello"
z = [1, 2, 3]

print("x is int? ->", isinstance(x, int))
print("y is str? ->", isinstance(y, str))
print("z is list? ->", isinstance(z, list))

```

```python

# =========================================================
# PRACTICAL USE CASE
# =========================================================
print("\n========== PRACTICAL EXAMPLE ==========")

def process_data(data):
    if isinstance(data, int):
        print("Integer detected:", data * 2)
    elif isinstance(data, str):
        print("String detected:", data.upper())
    elif isinstance(data, list):
        print("List detected:", len(data))
    else:
        print("Unknown type")

process_data(5)
process_data("python")
process_data([1, 2, 3])

```

```python

# =========================================================
# DIFFERENCE: type() vs isinstance()
# =========================================================
print("\n========== type() vs isinstance() ==========")

# type() checks exact type only
print("type(dog1) == Dog ->", type(dog1) == Dog)         # True
print("type(dog1) == Animal ->", type(dog1) == Animal)   # False

# isinstance() supports inheritance
print("isinstance(dog1, Animal) ->", isinstance(dog1, Animal))  # True

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. isinstance() supports inheritance (very important).
# 2. It is better than type() in most real-world cases.
# 3. Can check multiple classes using tuple.
# 4. Commonly used in validation, APIs, and frameworks.

```

```python

# =========================================================
# OBJECT DATA (__dict__)
# =========================================================
print("\n========== OBJECT DATA ==========")
print("dog1:", dog1.__dict__)
print("cat1:", cat1.__dict__)
```
