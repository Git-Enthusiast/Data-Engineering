```python
# =========================================================
# TYPES OF METHODS IN PYTHON (FOCUS: STATIC METHOD)
# =========================================================
# 1. Instance Method -> works with object data (uses self)
# 2. Class Method    -> works with class data (uses cls)
# 3. Static Method   -> works independently (no self/cls)

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Bank:
    bank_name = "ABC Bank"   # class variable

    def __init__(self, account_holder, balance):
        self.account_holder = account_holder  # instance variable
        self.balance = balance                # instance variable

    # -----------------------------------------------------
    # INSTANCE METHOD (brief)
    # -----------------------------------------------------
    def display_account(self):
        print(f"Holder: {self.account_holder}, Balance: {self.balance}")

    # -----------------------------------------------------
    # CLASS METHOD (brief)
    # -----------------------------------------------------
    @classmethod
    def change_bank_name(cls, new_name):
        cls.bank_name = new_name

    # =====================================================
    # STATIC METHODS (MAIN FOCUS)
    # =====================================================

    # -----------------------------------------------------
    # 1. SIMPLE STATIC METHOD
    # -----------------------------------------------------
    @staticmethod
    def bank_rules():
        print("Minimum balance must be 1000.")

    # -----------------------------------------------------
    # 2. CALCULATION EXAMPLE (IMPORTANT)
    # -----------------------------------------------------
    @staticmethod
    def calculate_simple_interest(principal, rate, time):
        # Formula: SI = (P * R * T) / 100
        return (principal * rate * time) / 100

    # -----------------------------------------------------
    # 3. VALIDATION EXAMPLE
    # -----------------------------------------------------
    @staticmethod
    def is_valid_amount(amount):
        return amount > 0

    # -----------------------------------------------------
    # 4. UTILITY FUNCTION EXAMPLE
    # -----------------------------------------------------
    @staticmethod
    def convert_to_inr(usd, rate=83):
        return usd * rate

    # -----------------------------------------------------
    # 5. LOGIC WITHOUT CLASS/INSTANCE DATA
    # -----------------------------------------------------
    @staticmethod
    def greet_user(name):
        return f"Welcome, {name}!"



# =========================================================
# OBJECT CREATION
# =========================================================
b1 = Bank("Rajan", 5000)

print("========== INSTANCE METHOD ==========")
b1.display_account()

```

```python

# =========================================================
# STATIC METHOD CALLING
# =========================================================
print("\n========== STATIC METHOD CALL ==========")

# Call using class (recommended)
Bank.bank_rules()

# Call using object (also possible)
b1.bank_rules()

```

```python

# =========================================================
# SIMPLE INTEREST CALCULATION
# =========================================================
print("\n========== SIMPLE INTEREST ==========")

si = Bank.calculate_simple_interest(10000, 5, 2)
print("Simple Interest:", si)

```

```python

# =========================================================
# VALIDATION USING STATIC METHOD
# =========================================================
print("\n========== VALIDATION ==========")

amount = 500

if Bank.is_valid_amount(amount):
    print("Valid amount")
else:
    print("Invalid amount")

```

```python

# =========================================================
# UTILITY FUNCTION USAGE
# =========================================================
print("\n========== UTILITY FUNCTION ==========")

usd = 100
inr = Bank.convert_to_inr(usd)
print(f"{usd} USD =", inr, "INR")

```

```python

# =========================================================
# GREETING EXAMPLE
# =========================================================
print("\n========== GREETING ==========")

msg = Bank.greet_user("Rajan")
print(msg)

```

```python

# =========================================================
# WHY STATIC METHOD?
# =========================================================
# 1. Does NOT use self (object)
# 2. Does NOT use cls (class)
# 3. Works with input data only
# 4. Used for utility/helper functions
# 5. Keeps code organized inside class


# =========================================================
# KEY DIFFERENCE (IMPORTANT)
# =========================================================
print("\n========== METHOD COMPARISON ==========")

print("Instance Method -> Uses self (object data)")
print("Class Method    -> Uses cls (class data)")
print("Static Method   -> No self/cls (independent)")

```

```python

# =========================================================
# HOW STATIC METHOD WORKS INTERNALLY
# =========================================================
# When you call:
# Bank.calculate_simple_interest(10000, 5, 2)
#
# Python DOES NOT pass:
# - self
# - cls
#
# It behaves like a normal function inside class.

```

```python

# =========================================================
# REAL-WORLD ANALOGY
# =========================================================
# Instance Method -> Personal data (your bank account)
# Class Method    -> Bank-wide data (bank name)
# Static Method   -> General utility (interest calculation)


# =========================================================
# COMMON MISTAKES
# =========================================================
# ❌ Using self inside static method
# ❌ Trying to access instance variables directly
# ❌ Forgetting @staticmethod decorator

```

```python

# =========================================================
# OBJECT & CLASS DATA CHECK
# =========================================================
print("\n========== DATA ==========")

print("Object data:", b1.__dict__)
print("Class data:", Bank.__dict__)

```

```python

# =========================================================
# QUICK REVISION
# =========================================================
# @staticmethod -> defines static method
# No self, no cls
# Called via ClassName.method()
# Best for utility/helper functions
```
