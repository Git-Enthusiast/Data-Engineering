```python

# ================================================================
# 🧠 0. BIG PICTURE (MENTAL MODEL)
# ================================================================
"""
                EVOLUTION OF PROGRAMMING

        Procedural → Functional → OOP

    WHY?
    Because problems became more complex

---------------------------------------------------------------

REAL-WORLD EVOLUTION:

Small scripts → Programs → Systems → Applications → Platforms

Small Script    # like adding two numbers
   ↓
Program (functions added)   # like calculator program using functions
   ↓
System (multiple modules + data)    # Like liberary management or banking software here OOPs is necessary
   ↓
Application (user-facing software)  # Large scale applicaion like whatsapp ,instagram
                                        (UI,frontend, backend, database) So OOPs must
   ↓
Platform (ecosystem of applications)    # like amazon, google, meta platform where multiple applicaion 
                                        # Applicaion interacts so distributed sytem Hence OOPs critical

---------------------------------------------------------------

KEY SHIFT:

Procedural → "HOW to do?"
Functional → "WHAT to do?"
OOP        → "WHO is doing it?"
---------------------------------------------------------------
"""

```

```python

# ================================================================
# 1. PROCEDURAL PROGRAMMING
# ================================================================
"""
DEFINITION:
Code is written as a sequence of steps (instructions).

FOCUS: LOGIC (Step-by-step execution)

---------------------------------------------------------------
DIAGRAM:

    START
      ↓
  Step 1
      ↓
  Step 2
      ↓
  Step 3
      ↓
    END

---------------------------------------------------------------
PROBLEM:
Data is separate from logic → leads to poor structure
---------------------------------------------------------------
"""

# Example
balance = 1000

def deposit(amount):
    global balance
    balance += amount

def withdraw(amount):
    global balance
    if amount <= balance:
        balance -= amount

deposit(500)
withdraw(200)

print("Procedural Balance:", balance)


"""
🔴 PROBLEMS:
1. Global variable (balance) → anyone can change
2. No control over data
3. Not scalable
4. Not real-world oriented

---------------------------------------------------------------
REAL PROBLEM DIAGRAM:

   balance  ← global variable
      ↑   ↑   ↑
      |   |   |
 deposit withdraw other functions

⚠️ Everyone can access and modify balance → DANGEROUS
---------------------------------------------------------------
"""

```

```python

# ================================================================
# 2. FUNCTIONAL PROGRAMMING
# ================================================================
"""
DEFINITION:
Everything is treated as a function.
Data is passed between functions.

FOCUS: FUNCTIONS

---------------------------------------------------------------
DIAGRAM:

    data → function → new data → function → new data

Example:

    balance → deposit() → new_balance → withdraw() → final_balance

---------------------------------------------------------------
KEY IDEA:
Avoid changing data → return new data
---------------------------------------------------------------
"""

def deposit_fp(balance, amount):
    return balance + amount

def withdraw_fp(balance, amount):
    if amount <= balance:
        return balance - amount
    return balance

balance_fp = 1000
balance_fp = deposit_fp(balance_fp, 500)
balance_fp = withdraw_fp(balance_fp, 200)

print("Functional Balance:", balance_fp)


"""
🟢 ADVANTAGES:
✔ Predictable
✔ No side effects
✔ Easy testing

---------------------------------------------------------------
🔴 LIMITATION (IMPORTANT):

DATA and LOGIC are STILL SEPARATE

You always pass data manually:

    balance = deposit(balance, amount)

---------------------------------------------------------------
VISUAL PROBLEM:

    balance ──► deposit() ──► balance ──► withdraw()

❗ Data is flowing everywhere → hard to manage in large systems
---------------------------------------------------------------
"""


# ================================================================
# 🔥 MOST IMPORTANT CONCEPT
# ================================================================
"""
FUNCTIONAL vs OOP

---------------------------------------------------------------

FUNCTIONAL PROGRAMMING:

    focus → FUNCTIONS

    deposit(balance, 100)
    withdraw(balance, 50)

    DATA is external

---------------------------------------------------------------

OOP:

    focus → OBJECT (DATA)

    account.deposit(100)
    account.withdraw(50)

    DATA is internal

---------------------------------------------------------------

VISUAL COMPARISON:

FUNCTIONAL:

    [balance] → function → function → function

OOP:

    [OBJECT]
     ├── data (balance)
     └── methods (deposit, withdraw)

---------------------------------------------------------------
INTERVIEW LINE:
"In functional programming, data flows through functions,
while in OOP, functions are attached to the data."
---------------------------------------------------------------
"""

```

```python

# ================================================================
# 3. OBJECT-ORIENTED PROGRAMMING (OOP)
# ================================================================
"""
DEFINITION:
OOP organizes code into objects (real-world entities).

FOCUS: DATA + BEHAVIOR TOGETHER

---------------------------------------------------------------
DIAGRAM:

        OBJECT (BankAccount)
        --------------------
        | balance (data)   |
        |------------------|
        | deposit()        |
        | withdraw()       |
        --------------------

---------------------------------------------------------------
KEY IDEA:
Data is protected inside object
---------------------------------------------------------------
"""

class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount

    def show(self):
        print("OOP Balance:", self.balance)


acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
acc.show()


"""
🟢 ADVANTAGES:

1. ENCAPSULATION (Data Protection)
   Only methods can access data

2. MODULARITY
   Code divided into classes

3. REUSABILITY
   Use inheritance

4. REAL-WORLD MODELING
   Code matches real-world structure

---------------------------------------------------------------
VISUAL SECURITY:

    acc.balance ❌ (should not be directly modified)

    acc.deposit() ✔ (controlled access)

---------------------------------------------------------------
"""

```

```python

# ================================================================
# 4. WHY MOVE FROM FUNCTIONAL → OOP?
# ================================================================
"""
Imagine a large system:

E-commerce system:
- Users
- Orders
- Payments
- Products

---------------------------------------------------------------
FUNCTIONAL APPROACH:

process_user(user)
process_order(order)
process_payment(payment)

❗ Data is scattered everywhere

---------------------------------------------------------------
OOP APPROACH:

class User
class Order
class Payment

Each class handles its own data

---------------------------------------------------------------
VISUAL:

FUNCTIONAL:
    data → function → data → function → data

OOP:
    [User Object]
    [Order Object]
    [Payment Object]

---------------------------------------------------------------
RESULT:
✔ Clean structure
✔ Easy debugging
✔ Scalable system
---------------------------------------------------------------
"""

```

```python

# ================================================================
# 5. REAL-WORLD COMPARISON
# ================================================================

"""
PROBLEM: Student System

---------------------------------------------------------------
FUNCTIONAL VERSION:
"""

def create_student(name, marks):
    return {"name": name, "marks": marks}

def add_marks(student, m):
    student["marks"] += m

s = create_student("Rajan", 80)
add_marks(s, 10)

print("Functional Student:", s)


"""
🔴 PROBLEM:
- Dictionary → no structure
- Keys can change → unsafe

---------------------------------------------------------------
OOP VERSION:
"""

class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def add_marks(self, m):
        self.marks += m

    def show(self):
        print("OOP Student:", self.name, self.marks)


s1 = Student("Rajan", 80)
s1.add_marks(10)
s1.show()


"""
🟢 ADVANTAGE:
✔ Structured
✔ Secure
✔ Extendable
"""

```

```python

# ================================================================
# 6. OOP CORE FEATURES
# ================================================================
"""
1. ENCAPSULATION → Data security
2. INHERITANCE → Reuse code
3. ABSTRACTION → Hide complexity
4. POLYMORPHISM → Same method, different behavior
"""


class PremiumAccount(BankAccount):

    def bonus(self):
        print("Bonus added!")


p = PremiumAccount(2000)
p.deposit(500)
p.bonus()
p.show()

```
