```python
# =========================================================
# GETTER & SETTER METHODS IN PYTHON
# (TYPES OF INSTANCE METHODS)
# =========================================================
# Instance methods can be categorized into:
# 1. Getter Methods -> used to get/retrieve data
# 2. Setter Methods -> used to set/update data
#
# These are used for ENCAPSULATION (data hiding + control)

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Employee:
    def __init__(self, name, salary):
        # Private-like variables (by convention)
        self._name = name
        self._salary = salary

    # -----------------------------------------------------
    # GETTER METHODS (Access data)
    # -----------------------------------------------------
    def get_name(self):
        return self._name

    def get_salary(self):
        return self._salary

    # -----------------------------------------------------
    # SETTER METHODS (Modify data)
    # -----------------------------------------------------
    def set_name(self, new_name):
        self._name = new_name

    def set_salary(self, new_salary):
        # validation example
        if new_salary > 0:
            self._salary = new_salary
        else:
            print("Invalid salary!")


    # -----------------------------------------------------
    # DISPLAY METHOD (NORMAL INSTANCE METHOD)
    # -----------------------------------------------------
    def display(self):
        print(f"Name: {self._name}, Salary: {self._salary}")



# =========================================================
# OBJECT CREATION
# =========================================================
emp1 = Employee("Rajan", 50000)

print("========== INITIAL DATA ==========")
emp1.display()

```

```python

# =========================================================
# USING GETTER METHODS
# =========================================================
print("\n========== GETTER METHODS ==========")

print("Name:", emp1.get_name())
print("Salary:", emp1.get_salary())

```

```python

# =========================================================
# USING SETTER METHODS
# =========================================================
print("\n========== SETTER METHODS ==========")

emp1.set_name("Amit")
emp1.set_salary(60000)

print("After update:")
emp1.display()

```

```python

# =========================================================
# VALIDATION USING SETTER
# =========================================================
print("\n========== VALIDATION ==========")

emp1.set_salary(-1000)   # invalid
emp1.display()


# =========================================================
# WHY USE GETTERS & SETTERS?
# =========================================================
# Direct access (BAD PRACTICE):
# emp1._salary = -5000   ❌ (no control)
#
# Using setter:
# emp1.set_salary(50000) ✅ (validation possible)

```

```python

# =========================================================
# PRACTICAL EXAMPLE WITH USER INPUT
# =========================================================
print("\n========== PRACTICAL EXAMPLE ==========")

class Student:
    def __init__(self):
        self._name = ""
        self._marks = 0

    # setter methods
    def set_data(self, name, marks):
        self._name = name
        self._marks = marks

    # getter methods
    def get_data(self):
        return self._name, self._marks


s1 = Student()

# simulate user input
name = "Neha"
marks = 92

s1.set_data(name, marks)

n, m = s1.get_data()
print(f"Name: {n}, Marks: {m}")

```

```python

# =========================================================
# MODERN APPROACH: PROPERTY DECORATOR (IMPORTANT)
# =========================================================
print("\n========== USING @property ==========")

class Person:
    def __init__(self, age):
        self._age = age

    # getter
    @property
    def age(self):
        return self._age

    # setter
    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("Invalid age!")

p1 = Person(25)

print("Age:", p1.age)   # getter (no parentheses)

p1.age = 30            # setter
print("Updated Age:", p1.age)

p1.age = -5            # invalid

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. Getter -> retrieve value
# 2. Setter -> modify value
# 3. Used for encapsulation (data protection)
# 4. Helps in validation & controlled access
# 5. Not mandatory in Python, but good practice
# 6. @property is modern & cleaner approach


# =========================================================
# QUICK REVISION
# =========================================================
# Getter Method -> get_data()
# Setter Method -> set_data()
# self          -> current object
# _variable     -> protected (by convention)
# @property     -> advanced getter/setter
```
