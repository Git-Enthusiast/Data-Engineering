```python
# ============================================================
#               CALLABLES IN PYTHON
# ============================================================
# A callable is any object that can be "called" using ()
# Examples:
#   - functions
#   - methods
#   - classes
#   - objects with __call__()
#
# Internally, Python checks:
#   Does this object have a __call__ method?
#   If YES → it is callable
# ============================================================

```

```python

# ============================================================
# 1) BASIC UNDERSTANDING OF CALLABLE
# ============================================================

def my_function():
    return "I am a function"

print("Function call:", my_function())  # Works

x = 10
# print(x())  # ❌ ERROR → int is not callable

print("Is function callable?", callable(my_function))  # True
print("Is int callable?", callable(x))                # False

```

```python

# ============================================================
# 2) EVERYTHING IS OBJECT IN PYTHON
# ============================================================
# Functions are also objects, but they implement __call__()

print("\nChecking __call__ existence:")
print(hasattr(my_function, "__call__"))  # True
print(hasattr(x, "__call__"))            # False

```

```python

# ============================================================
# 3) INTERNAL WORKING OF ()
# ============================================================
# When you do:
#   obj()
# Python internally does:
#   obj.__call__()

print("\nDirectly calling __call__:")
print(my_function.__call__())  # Same as my_function()

```

```python

# ============================================================
# 4) MAKING OBJECTS CALLABLE USING __call__
# ============================================================

class Adder:
    def __init__(self, value):
        self.value = value

    def __call__(self, x):
        print("Inside __call__ method")
        return self.value + x


adder = Adder(10)

print("\nIs object callable?", callable(adder))  # True

# Calling object like function
print("Result:", adder(5))  # Internally calls adder.__call__(5)

```

```python

# ============================================================
# 5) STEP-BY-STEP INTERNAL FLOW
# ============================================================
# adder(5) triggers:
#   Step 1: Python sees ()
#   Step 2: Looks for __call__ in Adder class
#   Step 3: Executes:
#           Adder.__call__(adder, 5)
#   Step 4: Returns result

```

```python

# ============================================================
# 6) MULTIPLE ARGUMENTS SUPPORT
# ============================================================

class Calculator:
    def __call__(self, a, b):
        return a + b

calc = Calculator()
print("\nCalculator result:", calc(3, 4))

```

```python

# ============================================================
# 7) STATEFUL CALLABLE OBJECT (VERY IMPORTANT)
# ============================================================
# Object remembers previous state unlike functions

class Counter:
    def __init__(self):
        self.count = 0

    def __call__(self):
        self.count += 1
        return self.count


counter = Counter()

print("\nStateful callable:")
print(counter())  # 1
print(counter())  # 2
print(counter())  # 3

```

```python

# ============================================================
# 8) REAL-WORLD USE CASE (FUNCTION-LIKE OBJECT)
# ============================================================

class Logger:
    def __init__(self, prefix):
        self.prefix = prefix

    def __call__(self, message):
        print(f"{self.prefix}: {message}")


log = Logger("INFO")

print("\nLogger example:")
log("System started")
log("Processing data")

```

```python

# ============================================================
# 9) CALLABLE CLASS (CLASS ITSELF IS CALLABLE)
# ============================================================
# Classes are callable because they create objects

class Person:
    def __init__(self, name):
        self.name = name

p = Person("Rajan")  # Person() is callable

print("\nClass callable?", callable(Person))  # True

```

```python

# ============================================================
# 10) DIFFERENCE: FUNCTION VS CALLABLE OBJECT
# ============================================================
# Function:
#   - Stateless (unless closure)
#
# Callable Object:
#   - Can maintain state
#   - Used in decorators, frameworks, ML pipelines

```

```python

# ============================================================
# 11) BONUS: USING CALLABLE IN VALIDATION
# ============================================================

def execute_if_callable(obj):
    if callable(obj):
        print("Object is callable → Executing")
        return obj()
    else:
        print("Not callable")


print("\nCallable validation:")
execute_if_callable(my_function)
execute_if_callable(100)

```

```python

# ============================================================
# 12) WHERE __call__ IS USED IN REAL WORLD
# ============================================================
# - Decorators
# - Middleware (Django, Flask)
# - Machine Learning pipelines
# - Custom function-like APIs
#
# Example:
#   torch.nn.Module uses __call__ internally
# ============================================================


# ============================================================
# FINAL SUMMARY
# ============================================================
# ✔ Callable = anything that can be called using ()
# ✔ Python internally calls __call__()
# ✔ Functions already have __call__
# ✔ Objects become callable by defining __call__
# ✔ callable(obj) checks this behavior
# ✔ Very useful for:
#     - stateful logic
#     - decorators
#     - advanced OOP design
# ============================================================
```
