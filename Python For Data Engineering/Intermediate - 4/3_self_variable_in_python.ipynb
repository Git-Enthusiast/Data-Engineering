```python
"""
                    self Vairable in Python
"""

# =========================================================
# PART 1: Introduction to 'self' and Object Creation
# =========================================================
print("\n[PART 1] Class Definition vs Object Creation")

class Employee:
    """
    Employee class demonstrates the use of 'self' to refer
    to the current object.
    """
    def __init__(self, name, salary):
        """
        Constructor method.
        'self' refers to the current instance being created.
        """
        self.name = name
        self.salary = salary
        print(f"[INIT] Object created at memory location: {id(self)}")

# Defining the class does NOT allocate memory
print("[INFO] Class defined, no memory allocated for objects yet.\n")
```

```python

# =========================================================
# PART 2: Object Creation and Memory Allocation
# =========================================================
emp1 = Employee("Rajan", 50000)
emp2 = Employee("Neha", 60000)

print("\n[INFO] Memory addresses of objects:")
print(f"emp1: {id(emp1)}")
print(f"emp2: {id(emp2)}\n")
```

```python

# =========================================================
# PART 3: Role of 'self' in Methods
# =========================================================
print("[PART 3] Role of 'self' in methods")

class EmployeeDetails:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print(f"Employee Name: {self.name}, Salary: {self.salary}, Memory: {id(self)}")

    def give_raise(self, amount):
        self.salary += amount
        print(f"[UPDATE] {self.name}'s salary updated to {self.salary}")

emp_a = EmployeeDetails("Amit", 45000)
emp_b = EmployeeDetails("Neha", 47000)

emp_a.display()
emp_b.display()

emp_a.give_raise(5000)
emp_a.display()
emp_b.display()
```

```python

# =========================================================
# PART 4: Dot Notation
# =========================================================
print("\n[PART 4] Dot Notation Examples")
print(f"emp_a.name: {emp_a.name}")
print(f"emp_b.salary: {emp_b.salary}")

emp_b.salary = 52000
print(f"[UPDATED] emp_b.salary: {emp_b.salary}")
```

```python

# =========================================================
# PART 5: Memory Representation and Multiple Objects
# =========================================================
print("\n[PART 5] Memory Representation of Objects")
print(f"emp_a.__dict__: {emp_a.__dict__}")
print(f"emp_b.__dict__: {emp_b.__dict__}")

emp_c = EmployeeDetails("Suresh", 48000)
emp_c.display()
```

```python

# =========================================================
# PART 6: Advanced Example - Company with Multiple Employees
# =========================================================
class Company:
    def __init__(self):
        self.employees = []

    def add_employee(self, emp):
        self.employees.append(emp)
        print(f"[INFO] Added {emp.name} to company.")

    def display_all(self):
        print("\n[ALL EMPLOYEES DATA]")
        for emp in self.employees:
            emp.display()

my_company = Company()
my_company.add_employee(emp_a)
my_company.add_employee(emp_b)
my_company.add_employee(emp_c)

my_company.display_all()
```

```python

# =========================================================
# PART 7: Step-by-Step Internal Memory Working
# =========================================================
print("\n[PART 7] Internal Memory Flow")

class DemoMemory:
    def __init__(self, value):
        self.value = value
        print(f"[INIT] Object at memory location {id(self)} initialized with value = {self.value}")

# Step 1 & 2: Memory allocation and reference
obj1 = DemoMemory(100)
obj2 = DemoMemory(200)

print(f"Memory reference of obj1: {id(obj1)}")
print(f"Memory reference of obj2: {id(obj2)}\n")

# Step 3: Reference automatically passed to constructor
print(f"Inside obj1: self points to {id(obj1)}")
print(f"Inside obj2: self points to {id(obj2)}\n")

# Step 4: Constructor initializes variables at memory reference
obj1.value = 150
print(f"Updated obj1.value = {obj1.value}, Memory: {id(obj1)}")
print(f"obj2.value remains = {obj2.value}, Memory: {id(obj2)}\n")

# Step 5: Dot notation accesses correct memory
print(f"obj1.value via dot notation: {obj1.value}")
print(f"obj2.value via dot notation: {obj2.value}\n")

# Step 6: Multiple objects maintain independent memory
obj3 = DemoMemory(300)
print(f"obj1.__dict__ = {obj1.__dict__}")
print(f"obj2.__dict__ = {obj2.__dict__}")
print(f"obj3.__dict__ = {obj3.__dict__}\n")

# Step 7: Summary
print("""
[SUMMARY]
1. When an object is created, Python allocates memory for it.
2. Memory reference is returned and stored in the object variable.
3. Reference is automatically passed as 'self' to __init__.
4. Constructor initializes variables at object's memory.
5. Dot notation uses 'self' to access object's memory.
6. Multiple objects maintain separate memory and data.
7. 'self' ensures each object retains its identity and data.
""")
```
