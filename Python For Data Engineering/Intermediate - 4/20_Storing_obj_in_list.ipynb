```python
# ============================================================
# Python OOP: Storing Multiple Objects in a List (Movie Database)
# ============================================================

# 1) CLASS DEFINITION
class Movie:
    def __init__(self, movie_title, runtime, hero):
        self.movie_title = movie_title
        self.runtime = runtime
        self.hero = hero

    # 2) METHOD TO DISPLAY MOVIE DETAILS
    def display_movie_details(self):
        print("\n----- Movie Details -----")
        print(f"Title   : {self.movie_title}")
        print(f"Runtime : {self.runtime} minutes")
        print(f"Hero    : {self.hero}")
        print("-------------------------\n")


# ============================================================
# MAIN PROGRAM
# ============================================================
if __name__ == "__main__":

    # 3) LIST TO STORE MULTIPLE MOVIE OBJECTS
    list_of_movies = []

    print("🎬 Welcome to Movie Database 🎬")

    # 4) TAKING USER INPUT + CREATING OBJECTS
    while True:
        movie_title = input("\nEnter Movie Title: ")
        runtime = int(input("Enter Runtime (in minutes): "))
        hero = input("Enter Lead Actor Name: ")

        # Create Movie object
        movie_obj = Movie(movie_title, runtime, hero)

        # 5) STORE OBJECT IN LIST
        list_of_movies.append(movie_obj)

        # 6) ASK USER TO CONTINUE
        choice = input("Do you want to add another movie? (yes/no): ").lower()
        if choice != "yes":
            break

    # 7) DISPLAY ALL STORED MOVIES
    print("\n📚 All Stored Movies:")
    for movie in list_of_movies:
        movie.display_movie_details()
```
