```python
# =========================================================
# ALL TYPES OF VARIABLES IN PYTHON (OOP)
# =========================================================
# 1. Instance Variables
# 2. Class Variables (also called Static Variables in Python)
# 3. Local Variables
#
# NOTE:
# Python does NOT have a separate keyword called "static variable"
# like Java/C++. In Python:
# 👉 Class Variable = Static Variable

```

```python

# =========================================================
# CLASS DEFINITION
# =========================================================
class Student:

    # -----------------------------------------------------
    # CLASS VARIABLE (STATIC VARIABLE)
    # -----------------------------------------------------
    school_name = "ABC School"     # shared among all objects
    student_count = 0              # also static variable

    # -----------------------------------------------------
    # CONSTRUCTOR (INSTANCE VARIABLES)
    # -----------------------------------------------------
    def __init__(self, name, marks):
        # Instance variables (unique for each object)
        self.name = name
        self.marks = marks

        # Update class variable
        Student.student_count += 1

    # -----------------------------------------------------
    # INSTANCE METHOD
    # -----------------------------------------------------
    def display(self):
        # Local variable (exists only inside method)
        result = "Pass" if self.marks >= 40 else "Fail"

        print(f"Name: {self.name}, Marks: {self.marks}, Result: {result}")
        print(f"School: {Student.school_name}")

    # -----------------------------------------------------
    # METHOD TO UPDATE INSTANCE VARIABLE
    # -----------------------------------------------------
    def update_marks(self, new_marks):
        self.marks = new_marks


# =========================================================
# OBJECT CREATION
# =========================================================
s1 = Student("Rajan", 85)
s2 = Student("Amit", 35)

print("========== INITIAL DATA ==========")
s1.display()
s2.display()

```

```python

# =========================================================
# INSTANCE VARIABLES (OBJECT-SPECIFIC)
# =========================================================
print("\n========== INSTANCE VARIABLES ==========")

print("s1 marks:", s1.marks)
print("s2 marks:", s2.marks)

# Modify only s1
s1.marks = 95

print("After modifying s1:")
print("s1 marks:", s1.marks)  # changed
print("s2 marks:", s2.marks)  # unchanged

```

```python

# =========================================================
# CLASS VARIABLES (STATIC VARIABLES)
# =========================================================
print("\n========== CLASS / STATIC VARIABLES ==========")

# Access via class
print("School (class):", Student.school_name)

# Access via objects
print("School (s1):", s1.school_name)
print("School (s2):", s2.school_name)

```

```python

# =========================================================
# MODIFY CLASS VARIABLE (CORRECT WAY)
# =========================================================
print("\n========== MODIFY CLASS VARIABLE ==========")

Student.school_name = "XYZ School"

s1.display()
s2.display()

```

```python

# =========================================================
# MODIFY USING OBJECT (IMPORTANT TRAP)
# =========================================================
print("\n========== MODIFY VIA OBJECT (TRAP) ==========")

# This creates a NEW instance variable (NOT modifying class variable)
s1.school_name = "My Personal School"

print("s1 school:", s1.school_name)   # instance variable
print("s2 school:", s2.school_name)   # class variable
print("Class school:", Student.school_name)

```

```python

# =========================================================
# DYNAMIC INSTANCE VARIABLES
# =========================================================
print("\n========== DYNAMIC INSTANCE VARIABLE ==========")

s1.grade = "A"   # added only to s1

print("s1 grade:", s1.grade)
print("Does s2 have grade?", hasattr(s2, "grade"))

```

```python

# =========================================================
# LOCAL VARIABLES
# =========================================================
print("\n========== LOCAL VARIABLES ==========")

def demo():
    x = 10   # local variable
    print("Inside function:", x)

demo()

# print(x)  # ❌ ERROR (not accessible outside function)

```

```python

# =========================================================
# STATIC VARIABLE CONCEPT (IMPORTANT)
# =========================================================
print("\n========== STATIC VARIABLE CONCEPT ==========")

# In Python:
# Static Variable = Class Variable

print("Student Count:", Student.student_count)

# Create new object → affects class variable
s3 = Student("Neha", 90)

print("Updated Student Count:", Student.student_count)

```

```python

# =========================================================
# MEMORY VIEW (__dict__)
# =========================================================
print("\n========== MEMORY VIEW ==========")

print("s1 data:", s1.__dict__)          # instance data
print("s2 data:", s2.__dict__)
print("Class data:", Student.__dict__) # class data

```

```python

# =========================================================
# COMPARISON TABLE (PRINTED)
# =========================================================
print("\n========== VARIABLE TYPES SUMMARY ==========")

print("Instance Variable -> Unique per object (self.var)")
print("Class/Static Var  -> Shared across objects (Class.var)")
print("Local Variable    -> Inside function/method only")

```

```python

# =========================================================
# IMPORTANT NOTES
# =========================================================
# 1. Instance variables belong to objects.
# 2. Class variables (static) are shared.
# 3. Python does not have separate 'static' keyword.
# 4. Modifying class variable via object creates new instance variable.
# 5. Local variables exist temporarily.


# =========================================================
# QUICK REVISION
# =========================================================
# Instance Variable -> self.var
# Class Variable    -> ClassName.var
# Static Variable   -> SAME as Class Variable
# Local Variable    -> inside function only
# __dict__          -> shows stored data
```
