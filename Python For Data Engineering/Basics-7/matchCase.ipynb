```python
"""
========================================================
    Python match-case (Structural Pattern Matching)
========================================================

Introduced in Python 3.10

match-case is NOT just a switch-case alternative.
It is a POWERFUL pattern matching system.

It supports:
✔ Value matching
✔ Multiple values (OR pattern)
✔ Variable binding
✔ Guards (conditions)
✔ Sequence matching (list, tuple)
✔ Mapping matching (dictionary)
✔ Nested patterns
✔ Class/Object matching (advanced)

--------------------------------------------------------
WHY match-case?
--------------------------------------------------------
Cleaner than if-elif
More readable for complex structures
Handles structured data easily

--------------------------------------------------------
SYNTAX:
--------------------------------------------------------

match <expression>:
    case <pattern1>:
        code
    case <pattern2>:
        code
    case _:
        default

--------------------------------------------------------
IMPORTANT:
--------------------------------------------------------
- Works only in Python 3.10+
- No 'break' needed
- First matching case executes
- '_' is default (like else)
"""
```

```python

# =======================================================
# 1. BASIC MATCH-CASE (VALUE MATCHING)
# =======================================================

choice = 2

match choice:
    case 1:
        print("Option 1")
    case 2:
        print("Option 2")
    case 3:
        print("Option 3")
    case _:
        print("Invalid option")

# NOTE:
# It checks EXACT value match

```

```python

# =======================================================
# 2. MATCH WITH STRINGS + MULTIPLE VALUES (OR)
# =======================================================

day = "sunday"

match day:
    case "monday":
        print("Start of week")
    case "saturday" | "sunday":
        print("Weekend")
    case _:
        print("Invalid day")

# "|" means OR condition

```

```python

# =======================================================
# 3. VARIABLE BINDING (IMPORTANT CONCEPT)
# =======================================================

value = 100

match value:
    case x:
        print("Captured value:", x)

# ⚠ IMPORTANT:
# This ALWAYS matches because x captures ANY value
# So avoid placing it before other cases

```

```python

# =======================================================
# 4. GUARDS (CONDITIONS INSIDE CASE)
# =======================================================

age = 25

match age:
    case x if x < 18:
        print("Minor")
    case x if 18 <= x < 60:
        print("Adult")
    case _:
        print("Senior")

# Guards = extra condition check

```

```python

# =======================================================
# 5. TUPLE MATCHING (PATTERN UNPACKING)
# =======================================================

point = (0, 5)

match point:
    case (0, 0):
        print("Origin")
    case (0, y):
        print("Y-axis at", y)
    case (x, 0):
        print("X-axis at", x)
    case (x, y):
        print("General point:", x, y)

# Pattern unpacking happens automatically

```

```python

# =======================================================
# 6. LIST MATCHING
# =======================================================

numbers = [1, 2, 3]

match numbers:
    case [1, 2, 3]:
        print("Exact match")
    case [1, x, y]:
        print("Starts with 1:", x, y)
    case _:
        print("No match")

# NOTE:
# Length must match unless using *

```

```python

# =======================================================
# 7. USING * (WILDCARD FOR MULTIPLE VALUES)
# =======================================================

nums = [1, 2, 3, 4, 5]

match nums:
    case [1, *rest]:
        print("Starts with 1, rest:", rest)

# *rest captures remaining values

```

```python

# =======================================================
# 8. DICTIONARY MATCHING (VERY IMPORTANT)
# =======================================================

student = {
    "name": "Rajan",
    "role": "student",
    "course": "CSE"
}

match student:
    case {"name": name, "role": "student"}:
        print(name, "is a student")
    case {"name": name, "role": "teacher"}:
        print(name, "is a teacher")
    case _:
        print("Unknown role")

# NOTE:
# Extra keys are allowed

```

```python

# =======================================================
# 9. NESTED PATTERN MATCHING
# =======================================================

data = (1, [10, 20])

match data:
    case (1, [x, y]):
        print("Nested match:", x, y)

# Very powerful for structured data

```

```python

# =======================================================
# 10. MATCH WITH TYPES (TYPE CHECKING)
# =======================================================

value = 10

match value:
    case int():
        print("Integer")
    case str():
        print("String")
    case _:
        print("Unknown type")

# Works like isinstance()

```

```python

# =======================================================
# 11. MATCH WITH CLASSES (ADVANCED)
# =======================================================

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Rajan", 22)

match p:
    case Person(name="Rajan", age=22):
        print("Exact match person")
    case Person(name=n, age=a):
        print("Person:", n, a)

# Matches object attributes

```

```python

# =======================================================
# 12. MENU-DRIVEN PROGRAM (REAL USE CASE)
# =======================================================

choice = 3

match choice:
    case 1:
        print("Add")
    case 2:
        print("Delete")
    case 3:
        print("Update")
    case 4:
        print("View")
    case _:
        print("Invalid choice")

```

```python

# =======================================================
# 13. MATCH-CASE vs IF-ELIF
# =======================================================

num = 10

# if-elif
if num == 10:
    print("Ten (if)")
elif num == 20:
    print("Twenty (if)")
else:
    print("Other (if)")

# match-case
match num:
    case 10:
        print("Ten (match)")
    case 20:
        print("Twenty (match)")
    case _:
        print("Other (match)")

```

```python

# =======================================================
# 14. IMPORTANT INTERNAL BEHAVIOR
# =======================================================

"""
1. Pattern matching happens TOP → DOWN
2. First match executes, others ignored
3. No fall-through (unlike C/C++)
4. No break needed
5. case _ is default
"""

```

```python

# =======================================================
# 15. COMMON MISTAKES (VERY IMPORTANT)
# =======================================================

"""
❌ Mistake 1: Using variable before specific cases

match x:
    case y:        # catches everything
        ...
    case 10:       # NEVER reached


❌ Mistake 2: Expecting exact dictionary match
Extra keys are allowed!


❌ Mistake 3: Thinking it's just switch-case
NO → It is pattern matching system
"""

```

```python

# =======================================================
# 16. WHEN TO USE MATCH-CASE
# =======================================================

"""
✔ Multiple conditions on same variable
✔ Structured data (tuple, list, dict)
✔ Cleaner than if-elif chains
✔ Parsing JSON-like data
✔ Menu-driven programs
"""

```

```python

# =======================================================
# 17. WHEN NOT TO USE
# =======================================================

"""
❌ Complex boolean logic → use if-elif
❌ Single condition → use if
❌ Very dynamic conditions → use if
"""

```

```python
# =======================================================
# 20. DYNAMIC INPUT USING eval() + MATCH-CASE
# =======================================================

"""
This section demonstrates how to:
✔ Take ANY input from user
✔ Convert it into actual Python datatype using eval()
✔ Use match-case to identify and process it

eval() converts string input into real Python object:
"10"      → int
"10.5"    → float
"[1,2,3]" → list
"(1,2)"   → tuple
"{'a':1}" → dict

⚠ WARNING:
eval() can be dangerous if used with untrusted input
(used here only for learning purpose)
"""

# Take input from user
user_input = input("\nEnter any Python data (int, float, list, tuple, dict): ")

try:
    data = eval(user_input)  # convert string → actual datatype

    match data:

        # -------------------------
        # INTEGER
        # -------------------------
        case int() as x:
            print("Type: Integer")
            print("Value:", x)
            print("Square:", x * x)

        # -------------------------
        # FLOAT
        # -------------------------
        case float() as x:
            print("Type: Float")
            print("Value:", x)
            print("Rounded:", round(x, 2))

        # -------------------------
        # STRING
        # -------------------------
        case str() as x:
            print("Type: String")
            print("Uppercase:", x.upper())

        # -------------------------
        # LIST
        # -------------------------
        case list() as x:
            print("Type: List")
            print("Length:", len(x))
            print("First element:", x[0] if x else "Empty list")

        # -------------------------
        # TUPLE
        # -------------------------
        case tuple() as x:
            print("Type: Tuple")
            print("Length:", len(x))
            print("Elements:", x)

        # -------------------------
        # DICTIONARY
        # -------------------------
        case dict() as x:
            print("Type: Dictionary")
            print("Keys:", list(x.keys()))
            print("Values:", list(x.values()))

        # -------------------------
        # BOOLEAN
        # -------------------------
        case bool() as x:
            print("Type: Boolean")
            print("Value:", x)

        # -------------------------
        # NONE TYPE
        # -------------------------
        case None:
            print("Type: NoneType")

        # -------------------------
        # DEFAULT
        # -------------------------
        case _:
            print("Unknown data type")

except Exception as e:
    print("Invalid input! Error:", e)

```

```python

# =======================================================
# 21. ADVANCED PATTERN MATCHING WITH USER INPUT
# =======================================================

"""
Now let's go deeper:
We not only detect type, but also structure inside it
"""

user_input = input("\nEnter structured data (list/tuple/dict): ")

try:
    data = eval(user_input)

    match data:

        # LIST PATTERNS
        case [1, 2, 3]:
            print("Exact list [1,2,3]")

        case [1, *rest]:
            print("List starts with 1, rest:", rest)

        case [x, y]:
            print("List with exactly 2 elements:", x, y)

        # TUPLE PATTERNS
        case (0, 0):
            print("Origin point")

        case (x, 0):
            print("X-axis point:", x)

        case (0, y):
            print("Y-axis point:", y)

        # DICTIONARY PATTERNS
        case {"name": name, "age": age}:
            print(f"Person -> Name: {name}, Age: {age}")

        case {"role": "admin"}:
            print("Admin detected")

        # DEFAULT
        case _:
            print("No structured pattern matched")

except Exception as e:
    print("Invalid structured input!", e)

```

```python

# =======================================================
# 23. SAFE ALTERNATIVE TO eval() (BEST PRACTICE)
# =======================================================

import json

print("\nUsing SAFE parsing (JSON):")

safe_input = input("Enter JSON data (e.g. [1,2,3] or {\"a\":1}): ")

try:
    data = json.loads(safe_input)

    match data:
        case list() as x:
            print("Safe List:", x)
        case dict() as x:
            print("Safe Dict:", x)
        case _:
            print("Other JSON type")

except Exception as e:
    print("Invalid JSON!", e)
```
