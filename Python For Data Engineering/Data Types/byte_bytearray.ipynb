## 1️⃣ What are `bytes` and `bytearray`?

Both `bytes` and `bytearray` are **sequences of integers** in the range **0–255** (i.e., individual bytes). They are used to handle **raw binary data** in Python.

| Property | `bytes` | `bytearray` |
|---|---|---|
| **Type** | `bytes` | `bytearray` |
| **Mutable?** | ❌ No (Immutable) | ✅ Yes (Mutable) |
| **Hashable?** | ✅ Yes (can be dict key) | ❌ No |
| **Syntax** | `b'hello'` or `bytes(...)` | `bytearray(...)` |
| **Element type** | Each element is an `int` (0–255) | Each element is an `int` (0–255) |
| **Use case** | Read-only binary data | Editable binary data |

> 💡 Think of `bytes` as an **immutable list of integers (0–255)** and `bytearray` as a **mutable list of integers (0–255)**.

> ⚠️ **Important:** `bytes` and `bytearray` are **NOT strings**. They store raw binary data. Strings store Unicode text.

````python
# ============================================
# 1️⃣ What are bytes and bytearray?
# ============================================

# bytes → immutable sequence of integers (0–255)
b = b'Hello'
print("bytes object:", b)
print("Type:", type(b))

# bytearray → mutable sequence of integers (0–255)
ba = bytearray(b'Hello')
print("\nbytearray object:", ba)
print("Type:", type(ba))

# Each element is an integer, not a character
print("\nFirst element of bytes:", b[0])       # 72 (ASCII of 'H')
print("First element of bytearray:", ba[0])    # 72
print("Type of element:", type(b[0]))           # <class 'int'>
````

## 2️⃣ Creating `bytes` Objects

There are multiple ways to create a `bytes` object:

| Method | Syntax | Example |
|---|---|---|
| Literal | `b'...'` | `b'Hello'` |
| From list of ints | `bytes([...])` | `bytes([72, 101, 108])` |
| From string (encoding) | `bytes(str, encoding)` | `bytes('Hello', 'utf-8')` |
| Using `.encode()` | `str.encode(encoding)` | `'Hello'.encode('utf-8')` |
| Zero-filled | `bytes(n)` | `bytes(5)` → 5 zero bytes |
| From iterable | `bytes(iterable)` | `bytes(range(5))` |
| Empty | `bytes()` or `b''` | Empty bytes object |

> 💡 The bytes literal `b'...'` can only contain **ASCII characters**. For non-ASCII, use hex escapes like `b'\xf0\x9f'`.

````python
# ============================================
# 2️⃣ Creating bytes Objects
# ============================================

# Method 1: Literal (only ASCII characters allowed)
b1 = b'Hello'
print("Literal:", b1)                    # b'Hello'

# Method 2: From a list of integers (each must be 0–255)
b2 = bytes([72, 101, 108, 108, 111])     # ASCII values of H, e, l, l, o
print("From list:", b2)                   # b'Hello'

# Method 3: From a string with encoding
b3 = bytes('Hello', 'utf-8')
print("From string:", b3)                 # b'Hello'

# Method 4: Using str.encode()
b4 = 'Hello'.encode('utf-8')
print("Using encode():", b4)              # b'Hello'

# Method 5: Zero-filled bytes (creates n zero bytes)
b5 = bytes(5)
print("Zero-filled:", b5)                 # b'\x00\x00\x00\x00\x00'
print("List form:", list(b5))             # [0, 0, 0, 0, 0]

# Method 6: From range (iterable)
b6 = bytes(range(5))
print("From range:", b6)                  # b'\x00\x01\x02\x03\x04'
print("List form:", list(b6))             # [0, 1, 2, 3, 4]

# Method 7: Empty bytes
b7 = bytes()
b8 = b''
print("Empty bytes:", b7, "| Length:", len(b7))  # b'' | Length: 0
````

## 3️⃣ Creating `bytearray` Objects

The `bytearray` constructor mirrors `bytes`, but returns a **mutable** object:

| Method | Syntax | Example |
|---|---|---|
| From list of ints | `bytearray([...])` | `bytearray([10, 20, 30])` |
| From string (encoding) | `bytearray(str, encoding)` | `bytearray('Hi', 'utf-8')` |
| Zero-filled | `bytearray(n)` | `bytearray(5)` → 5 zero bytes |
| From bytes | `bytearray(b'...')` | `bytearray(b'Hello')` |
| From iterable | `bytearray(iterable)` | `bytearray(range(5))` |
| Empty | `bytearray()` | Empty mutable bytes |

> ⚠️ There is **no bytearray literal** syntax. You must always use the `bytearray()` constructor.

````python
# ============================================
# 3️⃣ Creating bytearray Objects
# ============================================

# Method 1: From a list of integers
ba1 = bytearray([10, 20, 30, 40, 50])
print("From list:", ba1)                   # bytearray(b'\n\x14\x1e(2')
print("Type:", type(ba1))                   # <class 'bytearray'>

# Method 2: From string with encoding
ba2 = bytearray('Hello', 'utf-8')
print("From string:", ba2)                  # bytearray(b'Hello')

# Method 3: Zero-filled bytearray
ba3 = bytearray(5)
print("Zero-filled:", ba3)                  # bytearray(b'\x00\x00\x00\x00\x00')
print("List form:", list(ba3))              # [0, 0, 0, 0, 0]

# Method 4: From bytes object
ba4 = bytearray(b'Python')
print("From bytes:", ba4)                   # bytearray(b'Python')

# Method 5: From range
ba5 = bytearray(range(65, 70))              # ASCII: A=65, B=66, ...
print("From range:", ba5)                   # bytearray(b'ABCDE')

# Method 6: Empty bytearray
ba6 = bytearray()
print("Empty bytearray:", ba6, "| Length:", len(ba6))  # bytearray(b'') | Length: 0
````

## 4️⃣ Immutability of `bytes` vs Mutability of `bytearray`

This is the **key difference** between `bytes` and `bytearray`.

| Operation | `bytes` | `bytearray` |
|---|---|---|
| Read element | ✅ `b[0]` | ✅ `ba[0]` |
| Modify element | ❌ `TypeError` | ✅ `ba[0] = 100` |
| Append | ❌ Not possible | ✅ `ba.append(200)` |
| Delete element | ❌ Not possible | ✅ `del ba[0]` |
| Extend | ❌ Not possible | ✅ `ba.extend([1, 2])` |

> 💡 `bytes` is to `bytearray` what `str` is to `list` — one is **frozen** (immutable), the other is **editable** (mutable).

> ⚠️ **Interview Tip:** Because `bytes` is immutable and hashable, it **can be used as a dictionary key or in a set**. `bytearray` **cannot**.

````python
# ============================================
# 4️⃣ Immutability of bytes vs Mutability of bytearray
# ============================================

b = bytes([10, 20, 30, 40, 50])
ba = bytearray([10, 20, 30, 40, 50])

# --- Reading works for both ---
print("bytes[0]:", b[0])       # 10
print("bytearray[0]:", ba[0]) # 10

# --- bytes is IMMUTABLE → cannot modify ---
print("\n--- Attempting to modify bytes ---")
try:
    b[0] = 100
except TypeError as e:
    print("Error:", e)  # 'bytes' object does not support item assignment

# --- bytearray is MUTABLE → can modify ---
print("\n--- Modifying bytearray ---")
print("Before modification:", ba)    # bytearray(b'\n\x14\x1e(2')
ba[0] = 100
print("After ba[0] = 100:", ba)      # bytearray(b'd\x14\x1e(2')
print("List form:", list(ba))         # [100, 20, 30, 40, 50]

# --- bytearray supports append, extend, delete ---
ba.append(200)
print("After append(200):", list(ba))  # [100, 20, 30, 40, 50, 200]

del ba[0]
print("After del ba[0]:", list(ba))    # [20, 30, 40, 50, 200]

# --- bytes can be dict key, bytearray cannot ---
my_dict = {b'key': 'value'}       # ✅ Works
print("\nbytes as dict key:", my_dict)

try:
    my_dict = {ba: 'value'}       # ❌ TypeError
except TypeError as e:
    print("bytearray as dict key Error:", e)
````

## 5️⃣ Range Limitation (0–255)

Every element in `bytes` and `bytearray` must be an **integer between 0 and 255** (inclusive). This is because each element represents a single **byte** (8 bits), and 8 bits can represent values from `0` (`00000000`) to `255` (`11111111`).

| Value | Valid? | Reason |
|---|---|---|
| `0` | ✅ | Minimum byte value |
| `255` | ✅ | Maximum byte value (2⁸ - 1) |
| `256` | ❌ | Exceeds 8-bit range → `ValueError` |
| `-1` | ❌ | Negative values not allowed → `ValueError` |
| `3.14` | ❌ | Must be integer → `TypeError` |

> 💡 **Why 0–255?** A byte = 8 bits. Maximum value = 2⁸ − 1 = 255.

````python
# ============================================
# 5️⃣ Range Limitation (0–255)
# ============================================

# ✅ Valid byte values
valid_bytes = bytes([0, 100, 255])
print("Valid bytes:", valid_bytes)
print("List form:", list(valid_bytes))    # [0, 100, 255]

# ❌ Value > 255 → ValueError
try:
    invalid = bytes([256])
except ValueError as e:
    print("\nError (256):", e)  # bytes must be in range(0, 256)

# ❌ Negative value → ValueError
try:
    invalid = bytes([-1])
except ValueError as e:
    print("Error (-1):", e)

# ❌ Float value → TypeError
try:
    invalid = bytes([3.14])
except TypeError as e:
    print("Error (3.14):", e)

# Same rules apply for bytearray
try:
    invalid_ba = bytearray([300])
except ValueError as e:
    print("bytearray Error (300):", e)
````

## 6️⃣ Accessing Elements — Indexing & Slicing

Both `bytes` and `bytearray` support indexing and slicing, just like lists and strings.

| Operation | Returns | Example |
|---|---|---|
| `b[i]` | Single **int** (0–255) | `b[0]` → `72` |
| `b[i:j]` | New `bytes`/`bytearray` object | `b[1:3]` → `b'el'` |
| `b[::-1]` | Reversed `bytes`/`bytearray` | `b[::-1]` → `b'olleH'` |

> ⚠️ **Tricky Detail:** Indexing returns an **int**, but slicing returns a **bytes/bytearray object**. This is different from strings where both return strings!

````python
# ============================================
# 6️⃣ Accessing Elements — Indexing & Slicing
# ============================================

b = b'Hello'
ba = bytearray(b'Hello')

# --- Indexing returns an int ---
print("b[0]:", b[0])             # 72 (ASCII of 'H')
print("b[-1]:", b[-1])           # 111 (ASCII of 'o')
print("Type of b[0]:", type(b[0]))  # <class 'int'>

# --- Slicing returns bytes/bytearray ---
print("\nb[1:4]:", b[1:4])       # b'ell'
print("b[:3]:", b[:3])           # b'Hel'
print("b[2:]:", b[2:])           # b'llo'
print("Type of b[1:4]:", type(b[1:4]))  # <class 'bytes'>

# --- Reversing ---
print("\nReversed bytes:", b[::-1])       # b'olleH'
print("Reversed bytearray:", ba[::-1])    # bytearray(b'olleH')

# --- Step slicing ---
print("\nEvery 2nd byte:", b[::2])        # b'Hlo'

# ⚠️ Key difference from strings:
s = "Hello"
print("\n--- String vs bytes indexing ---")
print("String s[0]:", s[0], "→ type:", type(s[0]))  # 'H' → str
print("Bytes  b[0]:", b[0], "→ type:", type(b[0]))  # 72  → int
````

## 7️⃣ Iterating through `bytes` and `bytearray`

When you iterate over `bytes` or `bytearray`, each element is an **integer** (0–255).

| Method | Use Case |
|---|---|
| `for byte in b` | Access each byte value (int) |
| `enumerate(b)` | Get both **index** and **value** |
| `chr(byte)` | Convert byte int back to character |

````python
# ============================================
# 7️⃣ Iterating through bytes and bytearray
# ============================================

b = bytes([72, 101, 108, 108, 111])  # b'Hello'
ba = bytearray([10, 20, 30, 40, 50])

# --- Iterating bytes → each element is an int ---
print("--- Iterating bytes ---")
for byte in b:
    print(byte, end=' ')   # 72 101 108 108 111
print()

# --- Iterating bytearray ---
print("\n--- Iterating bytearray ---")
for byte in ba:
    print(byte, end=' ')   # 10 20 30 40 50
print()

# --- Using enumerate ---
print("\n--- Using enumerate ---")
for index, byte in enumerate(b):
    print(f"Index {index}: {byte} → '{chr(byte)}'")

# --- Convert bytes to characters while iterating ---
print("\n--- Bytes to characters ---")
for byte in b:
    print(chr(byte), end='')  # Hello
print()
````

## 8️⃣ Common Operations — Concatenation, Repetition, Membership, Length

| Operator | Symbol | Description |
|---|---|---|
| **Concatenation** | `+` | Joins two `bytes`/`bytearray` objects → creates a **new** object |
| **Repetition** | `*` | Repeats the object `n` times |
| **Membership** | `in` / `not in` | Checks if an integer value exists |
| **Length** | `len()` | Returns number of bytes |
| **Comparison** | `==`, `<`, `>` | Lexicographic comparison of byte values |

> ⚠️ You **cannot** concatenate `bytes` with `bytearray` directly — they must be the same type!

````python
# ============================================
# 8️⃣ Common Operations
# ============================================

b1 = b'Hello'
b2 = b' World'
ba1 = bytearray(b'Hello')
ba2 = bytearray(b' World')

# --- Concatenation (+) ---
print("bytes + bytes:", b1 + b2)          # b'Hello World'
print("bytearray + bytearray:", ba1 + ba2) # bytearray(b'Hello World')

# ❌ Cannot mix types
# print(b1 + ba1)  # TypeError: can't concat bytes to bytearray

# --- Repetition (*) ---
print("\nbytes * 3:", b'Ha' * 3)          # b'HaHaHa'
print("bytearray * 2:", bytearray(b'Go') * 2)  # bytearray(b'GoGo')

# --- Membership (in / not in) ---
# Check if a byte VALUE (int) exists
print("\n72 in b'Hello':", 72 in b1)      # True (72 = 'H')
print("200 in b'Hello':", 200 in b1)      # False

# Can also check for sub-sequence
print("b'ell' in b'Hello':", b'ell' in b1)  # True

# --- Length ---
print("\nlen(b'Hello'):", len(b1))         # 5
print("len(bytearray):", len(ba1))         # 5

# --- Comparison ---
print("\nb'abc' == b'abc':", b'abc' == b'abc')  # True
print("b'abc' < b'abd':", b'abc' < b'abd')      # True
````

## 9️⃣ Important Methods — `bytes` & `bytearray`

### Methods shared by both `bytes` and `bytearray`
| Method | Description |
|---|---|
| `.count(value)` | Count occurrences of a byte or subsequence |
| `.index(value)` | Find index of first occurrence (raises `ValueError` if not found) |
| `.find(value)` | Find index of first occurrence (returns `-1` if not found) |
| `.replace(old, new)` | Replace occurrences (returns **new** object) |
| `.startswith(prefix)` | Check if starts with given bytes |
| `.endswith(suffix)` | Check if ends with given bytes |
| `.join(iterable)` | Join bytes objects using separator |
| `.split(sep)` | Split into list of bytes objects |
| `.hex()` | Return hex string representation |
| `.decode(encoding)` | Decode bytes to string |

### Methods **only** available on `bytearray` (mutation methods)
| Method | Description |
|---|---|
| `.append(int)` | Add a single byte at the end |
| `.extend(iterable)` | Add multiple bytes at the end |
| `.insert(index, int)` | Insert a byte at a position |
| `.pop(index)` | Remove and return byte at index |
| `.remove(value)` | Remove first occurrence of value |
| `.reverse()` | Reverse in place |
| `.clear()` | Remove all bytes |

````python
# ============================================
# 9️⃣ Important Methods
# ============================================

b = b'Hello World Hello'
ba = bytearray(b'Hello World')

# --- Shared Methods ---
print("count('l'):", b.count(ord('l')))      # 4
print("count(b'Hello'):", b.count(b'Hello'))  # 2
print("index(b'World'):", b.index(b'World'))  # 6
print("find(b'World'):", b.find(b'World'))    # 6
print("find(b'xyz'):", b.find(b'xyz'))        # -1 (not found)

# replace → returns NEW object
print("replace:", b.replace(b'Hello', b'Hi')) # b'Hi World Hi'

print("startswith:", b.startswith(b'Hello'))   # True
print("endswith:", b.endswith(b'Hello'))       # True

# hex() → hexadecimal string representation
print("hex():", b'ABC'.hex())                  # 414243

# split and join
print("split:", b'a-b-c'.split(b'-'))          # [b'a', b'b', b'c']
print("join:", b'-'.join([b'a', b'b', b'c']))  # b'a-b-c'
````

````python
# ============================================
# 9️⃣ bytearray-only Methods (Mutation)
# ============================================

ba = bytearray(b'Hello')
print("Original:", list(ba))        # [72, 101, 108, 108, 111]

# append(int) — add one byte at end
ba.append(33)                       # 33 = '!'
print("After append(33):", ba)      # bytearray(b'Hello!')

# extend(iterable) — add multiple bytes
ba.extend([32, 72, 105])            # 32=' ', 72='H', 105='i'
print("After extend:", ba)          # bytearray(b'Hello! Hi')

# insert(index, int) — insert at position
ba.insert(0, 42)                    # 42 = '*'
print("After insert(0, 42):", ba)   # bytearray(b'*Hello! Hi')

# pop(index) — remove and return
popped = ba.pop(0)
print("Popped:", popped, "→", chr(popped))  # 42 → '*'
print("After pop:", ba)

# remove(value) — remove first occurrence
ba.remove(33)                       # remove '!'
print("After remove(33):", ba)

# reverse() — reverse in place
ba.reverse()
print("After reverse:", ba)

# clear() — empty the bytearray
ba.clear()
print("After clear:", ba)           # bytearray(b'')
````

## 🔟 Converting Between `bytes`, `bytearray`, and Other Types

| Conversion | Method | Example |
|---|---|---|
| `bytes` → `bytearray` | `bytearray(b)` | `bytearray(b'Hi')` |
| `bytearray` → `bytes` | `bytes(ba)` | `bytes(ba)` |
| `bytes` → `list` | `list(b)` | `list(b'Hi')` → `[72, 105]` |
| `list` → `bytes` | `bytes(lst)` | `bytes([72, 105])` → `b'Hi'` |
| `bytes` → `str` | `b.decode('utf-8')` | `b'Hi'.decode()` → `'Hi'` |
| `str` → `bytes` | `s.encode('utf-8')` | `'Hi'.encode()` → `b'Hi'` |
| `bytes` → `hex string` | `b.hex()` | `b'Hi'.hex()` → `'4869'` |
| `hex string` → `bytes` | `bytes.fromhex(s)` | `bytes.fromhex('4869')` → `b'Hi'` |

````python
# ============================================
# 🔟 Converting Between Types
# ============================================

# bytes ↔ bytearray
b = b'Hello'
ba = bytearray(b)           # bytes → bytearray
b_back = bytes(ba)          # bytearray → bytes
print("bytes → bytearray:", ba)
print("bytearray → bytes:", b_back)

# bytes ↔ list
lst = list(b)               # bytes → list of ints
b_from_list = bytes(lst)    # list → bytes
print("\nbytes → list:", lst)          # [72, 101, 108, 108, 111]
print("list → bytes:", b_from_list)    # b'Hello'

# bytes ↔ string (Encoding / Decoding)
s = b.decode('utf-8')       # bytes → string
b_from_str = s.encode('utf-8')  # string → bytes
print("\nbytes → string:", s)          # Hello
print("string → bytes:", b_from_str)   # b'Hello'

# bytes ↔ hex string
hex_str = b'Hello'.hex()                # bytes → hex string
b_from_hex = bytes.fromhex(hex_str)     # hex string → bytes
print("\nbytes → hex:", hex_str)        # 48656c6c6f
print("hex → bytes:", b_from_hex)       # b'Hello'

# Check default encoding
import sys
print("\nDefault encoding:", sys.getdefaultencoding())  # utf-8
````

## 1️⃣1️⃣ Encoding & Decoding Strings (Unicode ↔ Bytes)

In Python, **strings** are **Unicode text** and **bytes** are **raw binary data**. To convert between them:

| Direction | Method | Description |
|---|---|---|
| `str` → `bytes` | `.encode(encoding)` | Converts Unicode text into bytes using the specified encoding |
| `bytes` → `str` | `.decode(encoding)` | Converts raw bytes back into Unicode text |

### Common Encodings
| Encoding | Description |
|---|---|
| `utf-8` | Most common; 1–4 bytes per character; backward-compatible with ASCII |
| `ascii` | 1 byte per character; only supports 0–127 |
| `latin-1` | 1 byte per character; supports 0–255 |
| `utf-16` | 2 or 4 bytes per character |

> 💡 **Default encoding in Python is `utf-8`**. ASCII characters (English letters, digits) use 1 byte in UTF-8. Emojis and special characters use 2–4 bytes.

````python
# ============================================
# 1️⃣1️⃣ Encoding & Decoding
# ============================================

# --- Encoding: str → bytes ---
text = "Hello"
encoded = text.encode('utf-8')
print("Encoded:", encoded)           # b'Hello'
print("Type:", type(encoded))         # <class 'bytes'>
print("Byte values:", list(encoded))  # [72, 101, 108, 108, 111]

# --- Decoding: bytes → str ---
decoded = encoded.decode('utf-8')
print("\nDecoded:", decoded)          # Hello
print("Type:", type(decoded))         # <class 'str'>

# --- Non-ASCII characters use more bytes ---
hindi = "नमस्ते"
hindi_bytes = hindi.encode('utf-8')
print("\nHindi string:", hindi)
print("Encoded bytes:", hindi_bytes)
print("Number of bytes:", len(hindi_bytes))  # 18 (each Devanagari char = 3 bytes)
print("Number of chars:", len(hindi))        # 6

# --- Emoji (uses 4 bytes in UTF-8) ---
emoji = "🙂"
emoji_bytes = emoji.encode('utf-8')
print("\nEmoji:", emoji)
print("Encoded bytes:", emoji_bytes)
print("Number of bytes:", len(emoji_bytes))  # 4
print("Byte values:", list(emoji_bytes))     # [240, 159, 152, 130]

# --- Encoding error handling ---
try:
    "🙂".encode('ascii')
except UnicodeEncodeError as e:
    print("\nASCII encode error:", e)
````

## 1️⃣2️⃣ The Full Conversion Flow — String → Unicode → UTF-8 Bytes → Bits

Understanding how Python handles text internally:

```
Python str ("rajan")
       ↓
Unicode Code Points (U+0072, U+0061, U+006A, U+0061, U+006E)
       ↓
Decimal Values (114, 97, 106, 97, 110)
       ↓
UTF-8 Bytes (b'rajan')
       ↓
Binary Bits ('01110010', '01100001', '01101010', '01100001', '01101110')
```

### Key Takeaways
1. Python `str` = Unicode text object
2. Unicode code points are integers (shown as hex for humans: `U+0072`)
3. UTF-8 encodes Unicode to bytes (1–4 bytes per character)
4. Each byte = 8 bits in memory
5. ASCII chars: 1 code point = 1 byte; Emoji: 1 code point = 4 bytes

````python
# ============================================
# 1️⃣2️⃣ Full Conversion Flow
# ============================================

name = "rajan"
print("Python str object:", name)

# Step 1: Unicode Code Points (decimal)
unicode_points = [ord(c) for c in name]
print("Unicode code points (decimal):", unicode_points)  # [114, 97, 106, 97, 110]

# Step 2: Decimal → Binary
binary_repr = [format(cp, '08b') for cp in unicode_points]
print("Binary per character:", binary_repr)

# Step 3: Encode to UTF-8 Bytes
data = name.encode('utf-8')
print("UTF-8 bytes object:", data)
print("List of bytes (decimal):", list(data))

# Step 4: Bytes → Binary in memory
bytes_binary = [format(b, '08b') for b in data]
print("Binary bits in memory:", bytes_binary)

# Step 5: Full Flow Visualization
print("\n--- Full Conversion Flow for 'rajan' ---")
for c, cp, b in zip(name, unicode_points, bytes_binary):
    print(f"'{c}' → Unicode U+{cp:04X} → Byte {cp} → Binary {b}")

# Step 6: Non-ASCII example (Emoji)
emoji = "🙂"
print(f"\n--- Full Conversion Flow for '{emoji}' ---")
print("Unicode code point:", hex(ord(emoji)))  # U+1F642
emoji_bytes = emoji.encode('utf-8')
emoji_binary = [format(b, '08b') for b in emoji_bytes]
for b_dec, b_bin in zip(list(emoji_bytes), emoji_binary):
    print(f"Byte {b_dec} → Binary {b_bin}")
````

## 1️⃣3️⃣ `bytes` vs `bytearray` vs `str` — Comparison

| Feature | `str` | `bytes` | `bytearray` |
|---|---|---|---|
| **Content** | Unicode text | Raw binary data | Raw binary data |
| **Element type** | `str` (character) | `int` (0–255) | `int` (0–255) |
| **Mutable?** | ❌ No | ❌ No | ✅ Yes |
| **Hashable?** | ✅ Yes | ✅ Yes | ❌ No |
| **Literal** | `'hello'` | `b'hello'` | None (use constructor) |
| **Indexing returns** | `str` | `int` | `int` |
| **Slicing returns** | `str` | `bytes` | `bytearray` |
| **encoding needed?** | No (already Unicode) | Yes (to convert from str) | Yes (to convert from str) |
| **Use case** | Text processing | Reading files, network data | Modifying binary data |

````python
# ============================================
# 1️⃣3️⃣ bytes vs bytearray vs str — Side by Side
# ============================================

s = "Hello"           # str
b = b"Hello"          # bytes
ba = bytearray(b"Hello")  # bytearray

print("--- Types ---")
print("str  :", type(s))      # <class 'str'>
print("bytes:", type(b))      # <class 'bytes'>
print("bytearray:", type(ba)) # <class 'bytearray'>

print("\n--- Indexing (element type) ---")
print("str[0]  :", s[0], "→", type(s[0]))    # 'H' → str
print("bytes[0]:", b[0], "→", type(b[0]))     # 72  → int
print("bytearray[0]:", ba[0], "→", type(ba[0])) # 72  → int

print("\n--- Slicing (return type) ---")
print("str[0:2]  :", s[0:2], "→", type(s[0:2]))    # 'He' → str
print("bytes[0:2]:", b[0:2], "→", type(b[0:2]))     # b'He' → bytes
print("bytearray[0:2]:", ba[0:2], "→", type(ba[0:2])) # bytearray(b'He') → bytearray

print("\n--- Mutability ---")
# str and bytes → immutable
# bytearray → mutable
ba[0] = 74  # 'J'
print("bytearray after ba[0]=74:", ba)  # bytearray(b'Jello')
````

## 1️⃣4️⃣ Real-World Use Cases

| Use Case | Type | Why |
|---|---|---|
| Reading binary files (images, PDFs) | `bytes` | File content is raw binary |
| Network socket communication | `bytes` | Data sent over network is bytes |
| Cryptographic hashing | `bytes` | Hash functions work on bytes |
| Modifying binary file content | `bytearray` | Need to edit bytes in-place |
| Building network packets | `bytearray` | Construct data piece by piece |
| Database BLOBs | `bytes` | Store/retrieve binary data |
| Base64 encoding/decoding | `bytes` | Encoding produces bytes |

````python
# ============================================
# 1️⃣4️⃣ Real-World Use Cases
# ============================================

# --- Example 1: Hashing with hashlib ---
import hashlib

message = "Hello World"
hash_bytes = hashlib.sha256(message.encode('utf-8')).digest()
print("SHA-256 hash (bytes):", hash_bytes)
print("SHA-256 hash (hex):", hash_bytes.hex())

# --- Example 2: Base64 encoding ---
import base64

original = b"Python bytes and bytearray!"
encoded = base64.b64encode(original)
decoded = base64.b64decode(encoded)

print("\nBase64 encoded:", encoded)
print("Base64 decoded:", decoded)

# --- Example 3: Building a byte buffer with bytearray ---
buffer = bytearray()
buffer.append(0x48)     # 'H'
buffer.append(0x69)     # 'i'
buffer.extend(b'!!!')
print("\nBuffer:", buffer)           # bytearray(b'Hi!!!')
print("As string:", buffer.decode('utf-8'))  # Hi!!!
````

## 1️⃣5️⃣ Common Mistakes & Edge Cases (Interview Focused)

| Mistake | Why It Fails |
|---|---|
| `bytes([300])` | Value 300 exceeds 0–255 range → `ValueError` |
| `b[0] = 100` | `bytes` is immutable → `TypeError` |
| `b'Hello' + bytearray(b'!')` | Cannot concatenate different types → `TypeError` |
| `bytes('Hello')` | Missing encoding argument → `TypeError` |
| `b'नमस्ते'` | Non-ASCII in bytes literal → `SyntaxError` |
| Confusing `b[0]` with `b[0:1]` | `b[0]` = int, `b[0:1]` = bytes |
| `bytearray` as dict key | `bytearray` is unhashable → `TypeError` |

````python
# ============================================
# 1️⃣5️⃣ Common Mistakes & Edge Cases
# ============================================

# Mistake 1: Missing encoding for bytes()
try:
    b = bytes('Hello')  # ❌ Missing encoding
except TypeError as e:
    print("Mistake 1:", e)  # string argument without an encoding

# Correct way:
b = bytes('Hello', 'utf-8')  # ✅
print("Correct:", b)

# Mistake 2: Confusing indexing vs slicing
b = b'Hello'
print("\nb[0]  :", b[0], "→ type:", type(b[0]))     # 72 → int
print("b[0:1]:", b[0:1], "→ type:", type(b[0:1]))   # b'H' → bytes
# ⚠️ b[0] ≠ b[0:1] in type!

# Mistake 3: Comparing bytes with str
print("\nb'Hello' == 'Hello':", b'Hello' == 'Hello')  # False!
# They are different types, always False

# Mistake 4: bytes(n) creates n zero bytes, NOT bytes of the number n
print("\nbytes(5):", bytes(5))       # b'\x00\x00\x00\x00\x00' (5 zero bytes)
print("bytes([5]):", bytes([5]))     # b'\x05' (byte with value 5)

# Mistake 5: Concatenating bytes + bytearray
try:
    result = b'Hello' + bytearray(b' World')
    # Actually this works! bytes + bytearray → bytes
    print("\nbytes + bytearray:", result, type(result))
except TypeError as e:
    print("Error:", e)

# But bytearray + bytes → bytearray
result2 = bytearray(b'Hello') + b' World'
print("bytearray + bytes:", result2, type(result2))
````

## ✅ Final Summary Table — Python `bytes` & `bytearray`

| Concept | Key Point |
|---|---|
| **`bytes`** | Immutable sequence of integers (0–255); raw binary data |
| **`bytearray`** | Mutable sequence of integers (0–255); editable binary data |
| **Element range** | Every element must be `0 ≤ value ≤ 255` |
| **Indexing** | Returns `int`, not a character |
| **Slicing** | Returns new `bytes` or `bytearray` |
| **Immutability** | `bytes` = frozen (like `str`/`tuple`); `bytearray` = mutable (like `list`) |
| **Hashable** | `bytes` ✅ (can be dict key); `bytearray` ❌ |
| **Encoding** | `str.encode('utf-8')` → `bytes`; `bytes.decode('utf-8')` → `str` |
| **Literal** | `bytes`: `b'Hello'`; `bytearray`: no literal (use constructor) |
| **Methods** | Both share string-like methods; `bytearray` adds `append`, `extend`, `pop`, `insert`, `remove`, `reverse`, `clear` |
| **Use case** | File I/O, networking, cryptography, protocol handling |
| **vs `str`** | `str` = Unicode text; `bytes`/`bytearray` = raw binary data |
| **Common trap** | `bytes(5)` → 5 zero bytes, NOT the number 5 as bytes |
| **Common trap** | `b[0]` returns `int`, `b[0:1]` returns `bytes` |
