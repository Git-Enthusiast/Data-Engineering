```python
"""
====================================================================
                LIST vs TUPLE IN PYTHON (DETAILED GUIDE)
====================================================================

This file explains the complete difference between List and Tuple.
This is one of the MOST common Python interview questions.

A good interview answer should cover:
✔ Syntax
✔ Mutability
✔ Memory
✔ Performance
✔ Methods
✔ Packing & Unpacking
✔ Use Cases
✔ Practical Examples
✔ Comprehension

====================================================================
"""

# ================================================================
# 1️⃣  INTRODUCTION
# ================================================================

"""
List and Tuple are both built-in data structures in Python
used to store multiple items in a single variable.

Both:
✔ Allow duplicate values
✔ Preserve insertion order
✔ Allow different data types
✔ Support indexing and slicing

But they differ in some very important ways.
"""

# Example:
sample_list = [1, "Python", 3.14, True]
sample_tuple = (1, "Python", 3.14, True)

print("List Example:", sample_list)
print("Tuple Example:", sample_tuple)


# ================================================================
# 2️⃣  SYNTAX DIFFERENCE
# ================================================================

"""
List  → Created using square brackets []
Tuple → Created using parentheses ()

NOTE:
Parentheses are optional for tuples,
but recommended for clarity.
"""

my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

# Tuple without parentheses
tuple_without_brackets = 10, 20, 30

# Single element tuple must have comma
single_element_tuple = (5,)
not_tuple = (5)

print(type(single_element_tuple))  # <class 'tuple'>
print(type(not_tuple))             # <class 'int'>


# ================================================================
# 3️⃣  MUTABILITY (MOST IMPORTANT DIFFERENCE)
# ================================================================

"""
List → Mutable
Meaning: We can change, add, or remove elements after creation.

Tuple → Immutable
Meaning: Once created, elements cannot be changed.
"""

# ---- List is mutable ----
numbers_list = [1, 2, 3]
numbers_list[0] = 100
numbers_list.append(4)

print("Modified List:", numbers_list)
# Output: [100, 2, 3, 4]

# ---- Tuple is immutable ----
numbers_tuple = (1, 2, 3)

# numbers_tuple[0] = 100
# ❌ TypeError: 'tuple' object does not support item assignment


# ================================================================
# 4️⃣  MEMORY CONSUMPTION
# ================================================================

"""
Tuples consume LESS memory than lists.

Reason:
Tuples are immutable, so Python can optimize their storage.
Lists require extra memory because they are dynamic.
"""

import sys

list_memory = sys.getsizeof([1, 2, 3, 4, 5])
tuple_memory = sys.getsizeof((1, 2, 3, 4, 5))

print("List Memory:", list_memory)      # Example: 104 (may vary)
print("Tuple Memory:", tuple_memory)    # Example: 88 (may vary)

# Observation:
# Tuple generally uses less memory.


# ================================================================
# 5️⃣  PERFORMANCE
# ================================================================

"""
Tuples are slightly faster than lists.

Reason:
✔ Smaller memory footprint
✔ Immutable structure
✔ Less overhead
"""

import timeit

list_time = timeit.timeit(stmt="[1,2,3,4,5]", number=1000000)
tuple_time = timeit.timeit(stmt="(1,2,3,4,5)", number=1000000)

print("List Creation Time:", list_time)
print("Tuple Creation Time:", tuple_time)

# Tuple creation is usually faster.


# ================================================================
# 6️⃣  METHODS COMPARISON
# ================================================================

"""
List has MANY built-in methods.
Tuple has VERY LIMITED methods.
"""

sample_list = [1, 2, 3]
sample_tuple = (1, 2, 3)

# ---- List Methods ----
sample_list.append(4)
sample_list.remove(2)
sample_list.insert(1, 99)
sample_list.pop()

print("List After Operations:", sample_list)

# ---- Tuple Methods ----
print("Tuple Count:", sample_tuple.count(1))
print("Tuple Index:", sample_tuple.index(2))

# Tuple supports only:
# ✔ count()
# ✔ index()


# ================================================================
# 7️⃣  PACKING AND UNPACKING
# ================================================================

"""
Packing:
Multiple values assigned to a single tuple.

list also supports Packing so u can combine and make
a collection of element and pack together

Unpacking:
Assign tuple elements to multiple variables.

list doesn't support unpacking
"""

# Packing
packed = 10, 20, 30
print("Packed Tuple:", packed)

# Unpacking
a, b, c = packed
print("Unpacked Values:", a, b, c)

# Extended unpacking
x, *y = (1, 2, 3, 4, 5)
print("First:", x)
print("Remaining:", y)


# ================================================================
# 8️⃣  PRACTICAL USE CASES
# ================================================================

"""
Use Tuple When:
✔ Data should not change (constants)
✔ Storing fixed configuration values
✔ Returning multiple values from functions
✔ Dictionary keys (tuples can be keys, lists cannot)

Use List When:
✔ Data changes frequently
✔ Need insert/delete operations
✔ Dynamic applications
"""

# Example: Tuple as dictionary key
location = {(10, 20): "Point A"}
print(location[(10, 20)])


# ================================================================
# 9️⃣  COMPARISON TABLE (INTERVIEW READY)
# ================================================================

"""
+----------------------+------------------+------------------+
| Feature              | List             | Tuple            |
+----------------------+------------------+------------------+
| Syntax               | []               | ()               |
| Mutability           | Mutable          | Immutable        |
| Memory Usage         | More             | Less             |
| Performance          | Slightly slower  | Slightly faster  |
| Built-in Methods     | Many             | Few (2 only)     |
| Dictionary Key       | ❌ Not allowed   | ✔ Allowed       |
| Use Case             | Dynamic Data     | Fixed Data       |
| Comprehension        | list supports    | Tuple does not support |
+----------------------+------------------+------------------+
"""


# ================================================================
# 🔟 FINAL INTERVIEW ANSWER (SHORT SUMMARY)
# ================================================================

"""
If interviewer asks:

"What is the difference between List and Tuple?"

Answer:

List is a mutable data structure created using square brackets,
while Tuple is immutable and created using parentheses.

Tuples consume less memory and are slightly faster,
whereas lists provide more flexibility with many built-in methods.

We use tuples when data should remain constant,
and lists when data needs frequent modification.
"""

print("\n========== END OF DEMONSTRATION ==========")
```
