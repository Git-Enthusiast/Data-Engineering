## 1Ô∏è‚É£ Creating  Strings

**Strings** in Python are sequences of characters enclosed in quotes.

| Type | Syntax | Use |
|---|---|---|
| Single quotes | `'Hello'` | Simple strings |
| Double quotes | `"Hello"` | Simple strings |
| Triple quotes | `"""..."""` | Multiline strings |
| Mixed quotes | `'Python "Programming"'` | Embed quotes inside string |

> üí° Strings in Python are **immutable** ‚Äî once created, they cannot be changed in place.

````python
# ============================================
# 1Ô∏è‚É£ Creating Strings
# ============================================

# Single quotes
str1 = "Hello"

# Double quotes
str2 = 'World'

# Triple quotes ‚Üí used for multiline strings
str3 = """This is a 
multiline string"""

# Mix of quotes ‚Üí embed double quotes inside single-quoted string
str4 = 'Python "Programming"'

print(str1, str2, str3, str4)
````

## 3Ô∏è‚É£ Indexing

Each character in a string has a positional index.

- **Positive index**: left to right ‚Üí `0, 1, 2, ...`
- **Negative index**: right to left ‚Üí `-1, -2, -3, ...`

```
H  e  l  l  o
0  1  2  3  4   ‚Üê positive
-5 -4 -3 -2 -1  ‚Üê negative
```

````python
# ============================================
# 3Ô∏è‚É£ Indexing
# ============================================

str1 = "Hello"

# Positive indexing ‚Äî counts from the left (starts at 0)
print(str1[0])   # H ‚Üí first character
print(str1[1])   # e ‚Üí second character

# Negative indexing ‚Äî counts from the right (starts at -1)
print(str1[-1])  # o ‚Üí last character
print(str1[-2])  # l ‚Üí second from last
````

## 4Ô∏è‚É£ Slicing

Extract a **substring** using slice notation.

**Syntax:** `string[start : stop : step]`

| Parameter | Meaning |
|---|---|
| `start` | Begin index (inclusive). Default = `0` |
| `stop` | End index (**exclusive**). Default = end of string |
| `step` | Step size. Default = `1`. Use `-1` to reverse |

> üí° `str[::-1]` is the standard Python way to **reverse** a string.

````python
# ============================================
# 4Ô∏è‚É£ Slicing ‚Äî string[start:stop:step]
# ============================================

str1 = "Hello"

print(str1[1:4])   # ell  ‚Üí index 1 to 3 (4 excluded)
print(str1[:3])    # Hel  ‚Üí start defaults to 0, up to index 2
print(str1[2:])    # llo  ‚Üí from index 2 to end
print(str1[::2])   # Hlo  ‚Üí every 2nd character
print(str1[::-1])  # olleH ‚Üí reversed (step = -1)
print(str1[::])    # Hello -> default start 0 end up to last step 1

str5 = "Hello This is reference String for slicing"
print(str5[-1:-4:-1])   # start with -1 goto -4 (-4 excluded) with step -1
print(str5[-2:6:2])     # Empty string
print(str5[-1::-1])     # it will reverse the string
print(str5[:-5:-1])     # as step is negative so start -1
print(str5[-1::-1])     # reverse the string
print(str5[-3::-11])    # start with -3 go to negative till last step -11
print(str5[-1:-1:-1])   # will not print anything start = end
print(str5[:-4:-1])     # as step is negative so default value of 
                        # start is -negative if step is posative 
                        # default value of start is 0
````

## 7Ô∏è‚É£ Membership ‚Äî `in` / `not in`

Check whether a **substring** exists inside a string.

| Operator | Returns `True` when... |
|---|---|
| `x in s` | `x` is found in `s` |
| `x not in s` | `x` is NOT found in `s` |

````python
# ============================================
# 7Ô∏è‚É£ Membership ‚Äî in / not in
# ============================================

s1 = "Hello"

# 'in' ‚Äî True if substring is found inside the string
print("H" in s1)      # True  ‚Üí 'H' exists in 'Hello'
print("ell" in s1)    # True  ‚Üí 'ell' is a substring of 'Hello'

# 'not in' ‚Äî True if substring does NOT exist
print("z" not in s1)  # True  ‚Üí 'z' is NOT in 'Hello'
````

# 8.1 String Replacement in Python

This notebook covers how to replace text within strings using Python's built-in `str.replace()` method and the `re.sub()` function from the `re` (regular expressions) module.

---

### 1Ô∏è‚É£ Basic `str.replace()`
The `replace()` method substitutes occurrences of a specified substring with another. 
* **Immutability:** Strings in Python are immutable. `replace()` does not change the original string; it returns a new, modified copy.

### 2Ô∏è‚É£ `str.replace()` with `count` Parameter
By default, `replace()` swaps out *all* occurrences. You can pass an integer as the third argument (`count`) to limit the number of replacements made, starting from the left.

### 3Ô∏è‚É£ Case-Sensitive Replace
The standard `str.replace()` method is strictly case-sensitive. If you ask it to replace "python", it will ignore "Python" or "PYTHON".

---

### 4Ô∏è‚É£ Advanced Replacement: `re.sub()` and Word Boundaries (`\b`)
When you need more complex matching, you can use `re.sub(pattern, replacement, string)`. 
* **Word Boundaries (`\b`):** This is a regex token that ensures you only match a whole word, not part of another word. 

* *Example:* If you want to replace "cat", `\b` ensures you don't accidentally replace the "cat" inside "concatenate" or "catalog".

### 5Ô∏è‚É£ `re.sub()` with `count` Parameter
Just like `str.replace()`, `re.sub()` also accepts a `count` parameter to limit how many regex matches are replaced.

### 6Ô∏è‚É£ Case-Insensitive `re.sub()`
You can make your regular expression replacement case-insensitive by passing `flags=re.IGNORECASE` to `re.sub()`. This will replace all variations of a word regardless of their capitalization.

---

### ‚ö° Summary
* ‚úî **`str.replace()`** is simple but case-sensitive and literal.
* ‚úî **`count`** parameter limits the number of replacements in both methods.
* ‚úî **`re.sub()`** allows for powerful, regex-based replacements.
* ‚úî **`\b`** matches word boundaries so you only replace whole words.
* ‚úî **`flags=re.IGNORECASE`** enables case-insensitive replacements in regex.

````python
import re  # Required for re.sub()

# ==========================================
# 1Ô∏è‚É£ Basic str.replace()
# ==========================================
text = "C is a programming language and C is easy"

print("--- 1. Basic replace() ---")
print("Original String:", text)

new_text = text.replace("C", "Python")
print("After replace():", new_text)
print("Original string still unchanged:", text)

# ==========================================
# 2Ô∏è‚É£ str.replace() with count parameter
# ==========================================
print("\n--- 2. replace() with count=1 ---")
limited_replace = text.replace("C", "Python", 1)
print(limited_replace)  # Only the first "C" is replaced

# ==========================================
# 3Ô∏è‚É£ Case-sensitive replace
# ==========================================
print("\n--- 3. Case-sensitive replace() ---")
text2 = "Python is Fun. python is powerful."
print(text2.replace("python", "Java"))  # only lowercase replaced

# ==========================================
# 4Ô∏è‚É£ re.sub() with regex and word boundaries (\b)
# ==========================================
print("\n--- 4. re.sub() with word boundaries (\\b) ---")
sentence = "The cat sat on the catalog of categories."

# Replace whole word 'cat' with 'dog'
replaced_sentence = re.sub(r'\bcat\b', 'dog', sentence)
print(replaced_sentence)
# Output: 'The dog sat on the catalog of categories.'

# ==========================================
# 5Ô∏è‚É£ re.sub() with count parameter
# ==========================================
print("\n--- 5. re.sub() with count=2 ---")
sentence_cats = "cat cat cat catalog cat"

# Replace only first 2 occurrences of 'cat'
limited_replacement = re.sub(r'\bcat\b', 'dog', sentence_cats, count=2)
print(limited_replacement)
# Output: 'dog dog cat catalog cat'

# ==========================================
# 6Ô∏è‚É£ Case-insensitive re.sub()
# ==========================================
print("\n--- 6. Case-insensitive re.sub() ---")
sentence_mixed = "Cat cAt CAT cat concatenate"

# Replace all variations of 'cat' (case-insensitive) with 'dog'
case_insensitive = re.sub(r'\bcat\b', 'dog', sentence_mixed, flags=re.IGNORECASE)
print(case_insensitive)
# Output: 'dog dog dog dog concatenate'
````

### In-Depth: `strip()`, `lstrip()`, and `rstrip()`
These methods are incredibly common when cleaning up data (e.g., removing trailing spaces from user input).

**Syntax:** `.strip([chars])`
- If `chars` is **not provided**, it removes all leading and trailing **whitespace** (spaces, tabs `\t`, newlines `\n`).
- If `chars` **is provided**, it removes all combinations of the characters specified in the string from the ends.

| Method | Description |
|---|---|
| `.strip()` | Removes characters from **both** ends (left and right) |
| `.lstrip()` | Removes characters from the **left** end only |
| `.rstrip()` | Removes characters from the **right** end only |

> ‚ö†Ô∏è **Tricky Interview Detail:** The `chars` argument is not a prefix/suffix. It is a **set of characters**. Python will keep stripping from the ends as long as it encounters *any* character from that set.

````python
# 1. Default Behavior: Stripping Whitespace
s = " \t  Hello World \n  "

print(f"Original: '{s}'")
print(f"strip():  '{s.strip()}'")    # Both sides cleaned
print(f"lstrip(): '{s.lstrip()}'")   # Only left side cleaned
print(f"rstrip(): '{s.rstrip()}'")   # Only right side cleaned
````

````python
# ------------------------------
# 1Ô∏è‚É£ Using lstrip carefully
# ------------------------------
# ‚ö†Ô∏è Important: lstrip() does NOT remove a fixed string.
# It removes all characters in the argument from the left,
# until the first character that is not in that set.
# This is called the "first mismatch" behavior.

url = "https://www.example.com"

# Using lstrip with a string
print("Using lstrip():", url.lstrip("https://www."))  

# Output might be surprising: 'example.com'
# Why it works:
# lstrip("https://www.") treats the argument as a set of characters:
# {'h', 't', 'p', 's', ':', '/', 'w', '.'}
# Python removes characters from the left until it hits a character not in this set ('e' here)
# ‚ö†Ô∏è Not safe for all URLs, might remove unwanted characters at start.

# ------------------------------
# 2Ô∏è‚É£ Correct way: remove a fixed prefix
# ------------------------------
# For Python 3.9+, use removeprefix() to remove exact string
clean_url = url.removeprefix("https://www.")
print("Using removeprefix():", clean_url)  # Output: example.com
# ‚úÖ Safe, exact removal of "https://www."

# ------------------------------
# 3Ô∏è‚É£ Example with a list of URLs
# ------------------------------
urls = [
    "https://www.google.com",
    "https://www.github.com",
    "https://www.stackoverflow.com"
]

# Remove prefix from all URLs
clean_urls = [u.removeprefix("https://www.") for u in urls]
print("Clean URLs:", clean_urls)
# Output: ['google.com', 'github.com', 'stackoverflow.com']

# ------------------------------
# ‚ö° Summary:
# ------------------------------
# Method              Removes                        Notes
# lstrip("abc")       Any combination of 'a','b','c' from the left   Stops at first mismatch
# removeprefix("abc") Exact substring "abc" from the left            Safe for URLs
````

````python
# ------------------------------
# Cleaning URLs: strip() vs removeprefix()/slicing
# ------------------------------

urls = ['http://www.cricbuzz.com', 'http://www.youtube.com', 'http://www.facebook.com']

print("=== Using strip() (not safe) ===")
for url in urls:
    # ‚ö†Ô∏è strip() here removes ANY of the characters in 'http.com' from both ends
    cleaned_url = url.strip('http.com')
    print(cleaned_url)
# Output may be surprising due to 'first mismatch' behavior

print("\n=== Using removeprefix() (Python 3.9+) ===")
for url in urls:
    cleaned_url = url.removeprefix("http://www.")
    print(cleaned_url)
# ‚úÖ Safe removal of the exact prefix

print("\n=== Using slicing with startswith() (safe for older Python versions) ===")
for url in urls:
    if url.startswith("http://www."):
        cleaned_url = url[len("http://www."):]  # remove exact prefix
    else:
        cleaned_url = url
    print(cleaned_url)
````

````python
# =============================================================================
# Python string method: rstrip()
# =============================================================================
# THEORY:
# The rstrip() method in Python returns a copy of the string with trailing 
# characters removed (characters at the end/right side of the string). 
# By default, it removes whitespace characters like space, newline (\n), tab (\t), etc.
# You can also specify a set of characters to remove from the end of the string.
# 
# SYNTAX:
#   string.rstrip([chars])
# PARAMETERS:
#   chars (optional) - a string specifying the set of characters to be removed. 
#                      If omitted, trailing whitespaces are removed by default.
#
# RETURNS:
#   A new string with specified trailing characters removed.
# =============================================================================

# Example 1: Removing trailing spaces (default behavior)
text1 = "   Hello, World!   "
print("Original:", repr(text1))
print("After rstrip():", repr(text1.rstrip()))
# Output: '   Hello, World!'

# Example 2: Removing specific trailing characters
text2 = "Hello, World!!!"
print("\nOriginal:", repr(text2))
print("After rstrip('!'):", repr(text2.rstrip('!')))
# Output: 'Hello, World'

# Example 3: Removing multiple specified characters
text3 = "bananaaa"
print("\nOriginal:", repr(text3))
print("After rstrip('a'):", repr(text3.rstrip('a')))
# Output: 'ban'

# Example 4: Trailing newline characters
text4 = "Line with newline\n\n"
print("\nOriginal:", repr(text4))
print("After rstrip():", repr(text4.rstrip()))
# Output: 'Line with newline'

# Example 5: Trailing combination of characters
text5 = "Python....!!!"
print("\nOriginal:", repr(text5))
print("After rstrip('.!'):", repr(text5.rstrip('.!')))
# Output: 'Python'

# Example 6: String with no trailing characters to remove
text6 = "CleanString"
print("\nOriginal:", repr(text6))
print("After rstrip():", repr(text6.rstrip()))
# Output: 'CleanString' (unchanged)

# =============================================================================
# NOTES:
# 1. rstrip() only removes trailing (right-end) characters.
#    To remove leading characters, use lstrip().
#    To remove both ends, use strip().
# 2. The chars argument treats its input as a set of characters,
#    not as a substring.
# =============================================================================
````

# String Padding Methods in Python

This notebook demonstrates two useful built-in string methods in Python for padding text and numbers: `center()` and `zfill()`. 

Remember that strings in Python are **immutable**, meaning these methods do not change the original string, but instead return a new, padded copy.

---

### 1Ô∏è‚É£ The `center()` Method
The `center()` method aligns a string to the center by adding padding characters to both the left and right sides.

* **Syntax:** `string.center(width, fillchar)`
    * **`width` (mandatory):** The total length of the resulting string.
    * **`fillchar` (optional):** The character used to fill the extra space. If omitted, it defaults to a space (` `).

### 2Ô∏è‚É£ The `zfill()` Method
The `zfill()` (zero-fill) method adds leading zeros to a numeric string until it reaches the specified length. It is particularly smart about negative numbers, placing the zeros *after* the negative sign.

* **Syntax:** `string.zfill(width)`
    * **`width` (mandatory):** The total length of the resulting string.

````python
# ==========================================
# 1Ô∏è‚É£ Using center() with default padding
# ==========================================
text = "Hello"
centered_text = text.center(11)  # total width = 11

print("--- 1. center() with default space ---")
print(f"Original: '{text}'")
print(f"Centered: '{centered_text}'")  # Output: '   Hello   '


# ==========================================
# 2Ô∏è‚É£ Using center() with custom fill character
# ==========================================
centered_text2 = text.center(11, "*")

print("\n--- 2.1 center() with '*' ---")
print(f"Centered: '{centered_text2}'")  # Output: '***Hello***'

centered_text_2 = text.center((len(text)+7), "!")
print("\n--- 2.2 center() with '!' ---")
print(f"Centered: '{centered_text_2}'")  # Output: '!!!Hello!!!!'


# ==========================================
# 3Ô∏è‚É£ Using zfill() for positive numbers
# ==========================================
num_str = "42"
padded_num = num_str.zfill(5)

print("\n--- 3. zfill() with positive numbers ---")
print(f"Original: '{num_str}'")
print(f"Padded:   '{padded_num}'")  # Output: '00042'


# ==========================================
# 4Ô∏è‚É£ Using zfill() for negative numbers
# ==========================================
neg_num = "-42"
padded_neg = neg_num.zfill(5)

print("\n--- 4. zfill() with negative numbers ---")
print(f"Original: '{neg_num}'")
print(f"Padded:   '{padded_neg}'")  # Output: '-0042' (Note the sign placement)
````

### `split(separator, maxsplit)`
- Breaks a string into a **list** of substrings using a separator.
- Default separator = space `" "`.
- Returns a **list**. Original string is unchanged.

### `separator.join(iterable)`
- Combines elements of an iterable (list/tuple of strings) into a **single string**.

- Returns a **string**.

```

```

````python
# ============================================
# 9.1 split()
# ============================================
"""

split() - Converts a string into a list of substrings based on a separator.
   - Purpose: Count words or extract elements from a string.
   - Syntax: str.split(separator=None, maxsplit=-1)
   - Separator: Delimiter to split the string; default is whitespace.
   - maxsplit: Optional; limits the number of splits.
   - Returns: A list of strings.

"""

# --------------------------------------------
# Example 1: split() with custom separator
# --------------------------------------------
s = "apple,banana,cherry"

# Split at every comma
lst = s.split(",")
print("Split by comma:", lst)          # ['apple', 'banana', 'cherry']
print("Type of result:", type(lst))    # <class 'list'>

# --------------------------------------------
# Example 2: split() with default whitespace
# --------------------------------------------
s2 = "one two three"
words = s2.split()   # Default splits by whitespace
print("Split by whitespace:", words)  # ['one', 'two', 'three']

# --------------------------------------------
# Example 3: split() with maxsplit
# --------------------------------------------
s3 = "red,green,blue,yellow"
# Split only at the first comma (maxsplit=1)
limited_split = s3.split(",", maxsplit=1)
print("Split with maxsplit=1:", limited_split)  # ['red', 'green,blue,yellow']

# Split only at the first two commas (maxsplit=2)
limited_split2 = s3.split(",", maxsplit=2)
print("Split with maxsplit=2:", limited_split2) # ['red', 'green', 'blue,yellow']

# --------------------------------------------
# Example 4: join() to convert list to string
# --------------------------------------------
# Join with a custom separator
joined = "-".join(lst)
print("Joined with '-':", joined)     # apple-banana-cherry

# --------------------------------------------
# Example 5: join() with a set of strings
# --------------------------------------------
words_set = {"apple", "banana", "cherry"}
joined_set = ", ".join(words_set)
print("Joined set (unordered):", joined_set)             # Order may vary

# --------------------------------------------
# Example 6: join() with a dictionary
# --------------------------------------------
d = {"name": "Alice", "age": 25, "city": "Paris"}
joined_dict_keys = " | ".join(d)
print("Joined dictionary keys:", joined_dict_keys)       # name | age | city


# Joining numbers ‚Äî convert to str first
nums = [1, 2, 3]
joined_nums = ",".join(str(n) for n in nums)
print("Joined numbers:", joined_nums)  # 1,2,3

````

## üîü substring searching & counting: `count()`, `find()`, `rfind()`, `index()`

These built-in string methods are frequently used to locate substrings or count their occurrences. They share a similar syntax for checking within specific bounds:

**Common Syntax Pattern:** `.method(substring, start, end)`
- `substring`: the string to look for
- `start` (optional): the starting index (inclusive). Default is `0`.
- `end` (optional): the ending index (**exclusive**). Default is the end of the string.

### Key Rules and Tricky Behavior (Interview Focused)
1. **Bounds behave like slicing `[start:end]`**: the `start` index is included, but the `end` index is excluded.
2. **Negative Indices**: Supported exactly like slicing.
3. **Non-Overlapping**: Python's `count()`, `find()`, and `index()` do **NOT** count or find overlapping substrings.
4. **Step is inherently forward**: You cannot search backward by providing negative steps (though `rfind` and `rindex` search from the right).

### 10.1 `count()`
Returns the number of non-overlapping occurrences of a substring.

````python
s = "banana"

# 1. Default bounds (searches entire string)
print("Default bounds:", s.count("ana"))   # 1 -> NOTE: 'ana' appears twice, but OVERLAPPING matches are NOT counted.
print("Default bounds 'a':", s.count("a")) # 3

# 2. Specific start value
print("Start at index 2:", s.count("n", 2))  # 2 -> looks in 'nana'

# 3. Specific start and end value (end is EXCLUDED)
print("Start 1, End 4:", s.count("a", 1, 4)) # 1 -> looks in 'ana'. Index 4 is excluded.

# 4. Tricky Interview Cases: Negative indices
print("Negative indices (-4 to -1):", s.count("a", -4, -1)) # 1 -> looks in 'nan' ('b a n a n a' -> -4 is 'n', -1 is 'a' excluded)
print("Invalid bounds (start > end):", s.count("a", -2, -6)) # 0 -> bounds are empty slice, python safely returns 0.
print("Count empty string:", s.count(""))  # 7 -> (Length of string + 1) empty strings exist between characters.
````

### 10.2 `find()` and `rfind()`
- `find()` returns the index of the **first** occurrence of the substring.
- `rfind()` returns the index of the **last** occurrence (searching from the right).
- **Crucial Benefit:** If the substring is not found, they safely return `-1` (does not throw an error).

````python
s = "hello world"

# 1. Default bounds
print("find 'o':", s.find("o"))   # 4 -> first 'o' is at index 4
print("rfind 'o':", s.rfind("o")) # 7 -> last 'o' is at index 7
print("find 'x':", s.find("x"))   # -1 -> not found, safe!

# 2. Specific start value
print("find 'o' from index 5:", s.find("o", 5)) # 7 -> begins searching from the space onward

# 3. Specific start and end value
print("find 'o' between 0 and 5:", s.find("o", 0, 5)) # 4 -> looks in 'hello'
print("find 'o' between 0 and 4:", s.find("o", 0, 4)) # -1 -> looks in 'hell' ('o' at index 4 is excluded!)

# 4. Tricky Interview Cases
# Searching for an empty string
print("find empty string:", s.find(""))     # 0 -> always matches at the very beginning
print("rfind empty string:", s.rfind(""))   # 11 -> matches at the very end (length of string)
````

### 10.3 `index()` and `rindex()`
Behave exactly like `find()` and `rfind()`, but **raise a `ValueError`** if the substring is not found. 
> üí° **Best Practice:** Use `.find()` when you are *checking* if a substring exists. Use `.index()` only when you logically *know* and *expect* the substring to be there, and it failing means there's a bug in your logic.

````python
s = "python programming"

print("index 'pro':", s.index("pro"))    # 7
print("rindex 'o':", s.rindex("o"))      # 9

# s.index("java") # ‚ùå ValueError: substring not found

# Tricky Interview Case: Handling the error
try:
    idx = s.index("java")
except ValueError:
    print("Handling ValueError nicely: Substring not found!")
````

````python
# ---------------------------------------------------------------
# Python String Methods Tutorial: isupper(), islower(), istitle()
# ---------------------------------------------------------------

# THEORY:
# 1. isupper(): Returns True if all alphabetic characters are uppercase.
#    Non-alphabetic characters (digits, symbols) do not affect the result.

# 2. islower(): Returns True if all alphabetic characters are lowercase.
#    Any uppercase character will make it False. Non-alphabetic characters are ignored.

# 3. istitle(): Returns True if string is in title case (first letter of each word uppercase,
#    rest lowercase). Special symbols at start of word will cause False.

# -----------------------------
# EXAMPLES
# -----------------------------

# Example 1: isupper()
text1 = "HELLO WORLD"
text2 = "HELLO world!"
text_3 = "12HELLO #WORLD"
print(text1.isupper())  # True -> all letters uppercase
print(text2.isupper())  # False -> 'world' is lowercase
print(text_3.isupper())  # True -> all letters uppercase as it only cares about alphabet

# Example 2: islower()
text3 = "hello world"
text_3 = "hello @world"
text4 = "Hello World"
print(text3.islower())  # True -> all letters lowercase
print(text_3.islower())  # True -> all letters lowercase as it only cares about alphabet
print(text4.islower())  # False -> 'H' and 'W' are uppercase

# Example 3: istitle()
text5 = "Hello World"
text6 = "Hello world"
text7 = "123Hello World"
text8 = "Hello @World"
print(text5.istitle())  # True -> both words start with uppercase, rest lowercase
print(text6.istitle())  # False -> 'world' starts with lowercase
print(text7.istitle())  # True -> numbers ignored, 'Hello' and 'World' title cased
print(text8.istitle())  # False -> '@World' starts with special symbol, not letter
````

````python
# ---------------------------------------------------------------
# Python String Methods Tutorial: isidentifier(), isprintable(), isspace()
# ---------------------------------------------------------------

# THEORY:
# 1. isidentifier(): Returns True if the string is a valid Python identifier.
#    - Must start with a letter (a-z, A-Z) or underscore (_)
#    - Can contain letters, digits, or underscores
#    - Cannot contain spaces or special symbols
# 2. isprintable(): Returns True if all characters in the string are printable, including letters, digits, symbols, and spaces.
#    - Returns False for non-printable characters like escape sequences (\n, \t)
#    - Returns True for empty string
# 3. isspace(): Returns True if all characters in the string are whitespace characters (space, tab, newline).
#    - Returns False if any character is not whitespace
#    - Returns False for empty string

# -----------------------------
# EXAMPLES
# -----------------------------

# Example 1: isidentifier()
id1 = "variable_name"
id2 = "2variable"
id3 = "var$name"
id4 = "_hiddenVar"
print(id1.isidentifier())  # True -> valid identifier
print(id2.isidentifier())  # False -> starts with digit
print(id3.isidentifier())  # False -> contains special symbol $
print(id4.isidentifier())  # True -> starts with underscore, valid

# Example 2: isprintable()
text1 = "Hello World!"  # printable
text2 = "Hello\nWorld"  # contains newline (\n) -> not printable
text3 = ""               # empty string -> printable
text4 = "123456"         # numbers -> printable
print(text1.isprintable())  # True
print(text2.isprintable())  # False
print(text3.isprintable())  # True
print(text4.isprintable())  # True

# Example 3: isspace()
space1 = "    "       # spaces only
space2 = "\t\n"       # tab and newline -> whitespace
space3 = "Hello "     # contains non-space
space4 = ""           # empty string
print(space1.isspace())  # True
print(space2.isspace())  # True
print(space3.isspace())  # False
print(space4.isspace())  # False
````

## 1Ô∏è‚É£4Ô∏è‚É£ `ord()` and `chr()` Functions + `sum()` with `ord()`

Every character in Python has a **unique number** called its **Unicode code point** (ASCII for basic English characters).

Python gives us two built-in functions to convert between characters and their numeric codes:

| Function | What it does | Input | Output | Example |
|---|---|---|---|---|
| `ord(char)` | Character ‚Üí Number | A **single** character | `int` (Unicode code point) | `ord('A')` ‚Üí `65` |
| `chr(num)` | Number ‚Üí Character | An `int` (0 to 1,114,111) | `str` (single character) | `chr(65)` ‚Üí `'A'` |

### üîÑ They are the **reverse** of each other:
```
ord('A')  ‚Üí  65
chr(65)   ‚Üí  'A'

chr(ord('A'))  ‚Üí  'A'   (round trip)
ord(chr(65))   ‚Üí  65    (round trip)
```

> **üí° Remember:**  
> - `ord()` = **ord**inal ‚Üí gives the ordinal (position number) of a character  
> - `chr()` = **chr**aracter ‚Üí gives the character at a position number

### üìã Common ASCII / Unicode Values

| Characters | Range | `ord()` Values |
|---|---|---|
| `'0'` to `'9'` | Digits | 48 ‚Äì 57 |
| `'A'` to `'Z'` | Uppercase letters | 65 ‚Äì 90 |
| `'a'` to `'z'` | Lowercase letters | 97 ‚Äì 122 |
| `' '` (space) | Space character | 32 |
| `'\n'` | Newline | 10 |
| `'\t'` | Tab | 9 |

> **üí° Key Pattern:** Uppercase (`65-90`) come **before** lowercase (`97-122`).  
> The difference is always **32**: `ord('a') - ord('A')` = `97 - 65` = **32**

````python
# ============================================
# ord() ‚Äî Character to Unicode Number
# ============================================

# --- Letters ---
print("ord('A') =", ord('A'))   # 65
print("ord('Z') =", ord('Z'))   # 90
print("ord('a') =", ord('a'))   # 97
print("ord('z') =", ord('z'))   # 122
print()

# --- Digits (as characters, not numbers!) ---
print("ord('0') =", ord('0'))   # 48
print("ord('9') =", ord('9'))   # 57
print()

# --- Special characters ---
print("ord(' ')  =", ord(' '))   # 32 (space)
print("ord('@')  =", ord('@'))   # 64
print()

# --- Unicode (beyond ASCII) ---
print("ord('‚Ç¨')  =", ord('‚Ç¨'))   # 8364
print("ord('‚Çπ')  =", ord('‚Çπ'))   # 8377 (Indian Rupee)
print("ord('üòä') =", ord('üòä'))  # 128522
````

````python
# ============================================
# chr() ‚Äî Unicode Number to Character
# ============================================

print("chr(65)  =", chr(65))    # A
print("chr(90)  =", chr(90))    # Z
print("chr(97)  =", chr(97))    # a
print("chr(122) =", chr(122))   # z
print()

print("chr(48)  =", chr(48))    # '0'
print("chr(57)  =", chr(57))    # '9'
print()

print("chr(32)  =", repr(chr(32)))   # ' ' (space)
print("chr(8377)=", chr(8377))        # ‚Çπ
print()

# --- Round trip: they are inverses ---
print("chr(ord('P')) =", chr(ord('P')))   # P
print("ord(chr(80)) =", ord(chr(80)))     # 80
````

### üõ†Ô∏è Practical Uses of `ord()` and `chr()`

| Use Case | How |
|---|---|
| **Check if uppercase** | `65 <= ord(ch) <= 90` |
| **Check if lowercase** | `97 <= ord(ch) <= 122` |
| **Check if digit** | `48 <= ord(ch) <= 57` |
| **Uppercase ‚Üí lowercase** | `chr(ord(ch) + 32)` |
| **Lowercase ‚Üí uppercase** | `chr(ord(ch) - 32)` |
| **Caesar cipher** | Shift each char by N positions |
| **Generate char sequences** | Loop with `chr(i)` |

````python
# ============================================
# Practical Example 1: Manual Case Conversion
# (without using .lower() / .upper() methods)
# ============================================

ch = 'H'
# Uppercase to lowercase: add 32
if 65 <= ord(ch) <= 90:
    lower_ch = chr(ord(ch) + 32)
    print(f"'{ch}' (ord={ord(ch)}) ‚Üí '{lower_ch}' (ord={ord(lower_ch)})")

# Lowercase to uppercase: subtract 32
ch2 = 'p'
if 97 <= ord(ch2) <= 122:
    upper_ch = chr(ord(ch2) - 32)
    print(f"'{ch2}' (ord={ord(ch2)}) ‚Üí '{upper_ch}' (ord={ord(upper_ch)})")

print()
# Convert entire string manually
text = "Hello World"
manual_upper = ""
for ch in text:
    if 97 <= ord(ch) <= 122:
        manual_upper += chr(ord(ch) - 32)
    else:
        manual_upper += ch
print(f"Original: '{text}'  ‚Üí  Manual upper: '{manual_upper}'")
````

````python
# ============================================
# Practical Example 2: Caesar Cipher (Encryption)
# Shift each letter by N positions
# ============================================

def caesar_encrypt(text, shift):
    result = ""
    for ch in text:
        if ch.isalpha():
            base = ord('A') if ch.isupper() else ord('a')
            shifted = chr((ord(ch) - base + shift) % 26 + base)
            result += shifted
        else:
            result += ch
    return result

message = "Hello Python!"
encrypted = caesar_encrypt(message, 3)
decrypted = caesar_encrypt(encrypted, -3)

print(f"Original : {message}")
print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
````

````python
# ============================================
# Practical Example 3: Classify Characters
# WITHOUT using .isdigit(), .isupper() etc.
# ============================================

def classify_char(ch):
    code = ord(ch)
    if 48 <= code <= 57:    return "Digit"
    elif 65 <= code <= 90:  return "Uppercase Letter"
    elif 97 <= code <= 122: return "Lowercase Letter"
    elif code == 32:        return "Space"
    else:                   return "Special Character"

test = "Hi 42! @_@"
print(f"Classifying '{test}':")
print("-" * 40)
for ch in test:
    print(f"  '{ch}'  ‚Üí  ord={ord(ch):>5}  ‚Üí  {classify_char(ch)}")
````

```python
sum(ord(c) for c in "abc")   # 97 + 98 + 99 = 294
```
Useful for simple checksums, hash values, and algorithm problems.

````python
# ============================================
# sum() with ord()
# ============================================

s = "abc"
print(f"String: '{s}'")
for ch in s:
    print(f"  ord('{ch}') = {ord(ch)}")

total = sum(ord(c) for c in s)
print(f"Sum: {' + '.join(str(ord(c)) for c in s)} = {total}")
print()

word = "Python"
total2 = sum(ord(c) for c in word)
print(f"Sum of '{word}' = {total2}")
print()

# chr() ‚Äî reverse of ord()
print("chr(97) =", chr(97))   # a
print()
# ‚ö†Ô∏è sum('abc') ‚Üí TypeError! Always use sum(ord(c) for c in s)
print("‚ö†Ô∏è sum('abc') would raise TypeError")
````

### ‚ö†Ô∏è Common Mistakes with `ord()` and `chr()`

| Mistake | Error | Fix |
|---|---|---|
| `ord('AB')` | `TypeError` ‚Äî must be **one character** | Use `ord('A')` |
| `ord('')` | `TypeError` ‚Äî empty string | Ensure non-empty |
| `chr(-1)` | `ValueError` ‚Äî negative | Use range 0 to 1,114,111 |
| Confusing `'5'` with `5` | `ord('5')` = 53, not 5! | Char `'5'` ‚â† int `5` |

````python
# ============================================
# Common Mistakes with ord() and chr()
# ============================================

# ‚úÖ Correct: single character
print("ord('A')  =", ord('A'))    # 65

# ‚ùå ord('AB') ‚Üí TypeError (must be single character)
# ‚ùå ord('')   ‚Üí TypeError (cannot be empty)
print()

# ‚úÖ chr() valid range: 0 to 1,114,111
print("chr(0)       =", repr(chr(0)))       # null
print("chr(1114111) =", chr(1114111))       # max Unicode
print()

# ‚ö†Ô∏è Character '5' vs Integer 5
print("ord('5')  =", ord('5'))   # 53 (not 5!)
print("chr(53)   =", chr(53))    # '5'
````

## 1Ô∏è‚É£6Ô∏è‚É£ Immutability

Strings in Python are **immutable** ‚Äî you **cannot** change individual characters after creation.

```python
s = "Python"
s[0] = "J"   # ‚ùå TypeError: 'str' object does not support item assignment
```

**Workaround**: Create a **new string** by combining parts.

> This also applies to all string methods like `.upper()`, `.replace()`, `.title()` ‚Äî they all return **new strings** and never modify the original.

````python
# ============================================
# 1Ô∏è‚É£6Ô∏è‚É£ Immutability
# ============================================

s = "Python"

# ‚ùå This would raise a TypeError ‚Äî strings cannot be changed in place
# s[0] = "J"  # TypeError: 'str' object does not support item assignment

# ‚úÖ Workaround: build a new string
# s[1:] ‚Üí takes substring starting from index 1 ‚Üí 'ython'
# 'J' + s[1:] ‚Üí 'J' + 'ython' = 'Jython'
s2 = "J" + s[1:]
print(s2)    # Jython
print(s)     # Python ‚Üí original is UNCHANGED
````

## 1Ô∏è‚É£8Ô∏è‚É£ String Formatting ‚Äî f-string & format()

Two modern ways to embed variables inside strings:

| Method | Syntax | Notes |
|---|---|---|
| **f-string** | `f"Hello {name}"` | Fastest, most readable. Python 3.6+ |
| **format()** | `"Hello {}".format(name)` | Works in Python 2.6+ |

````python
# ============================================
# 1Ô∏è‚É£8Ô∏è‚É£ String Formatting
# ============================================

name = "Rajan"
age = 22

# --- f-string (f"") ‚Äî place variable directly inside {} ---
# Prefix the string with 'f' and use {variable_name} inside
print(f"My name is {name} and I am {age} years old")
# My name is Rajan and I am 22 years old

# --- format() method ‚Äî use {} as placeholder, pass values to format() ---
print("My name is {} and I am {} years old".format(name, age))
# My name is Rajan and I am 22 years old

# f-string supports expressions directly inside {}
print(f"Next year I will be {age + 1}")  # Next year I will be 23
````

````python
"""
C-STYLE STRING FORMATTING IN PYTHON
-----------------------------------

C-style string formatting is an older method of formatting strings in Python, similar
to how it is done in C. It uses the percent (%) operator to inject values into strings.

Key Points:
1. Uses '%' operator for formatting.
2. Format specifiers define data types:
   - %s -> string
   - %d -> integer
   - %f -> float
3. Order of values must match the order of format specifiers.
4. Dictionaries can be used for unordered data with keys matching specifiers.
5. While still functional, it's less commonly used today in favor of `format()` or f-strings.
"""

# Example 1: Basic C-style formatting with strings and integers
name = "Alice"
age = 25
print("Hello %s, you are %d years old." % (name, age))

# Example 2: Formatting floats with precision
pi = 3.14159265
print("Value of pi: %.2f" % pi)  # two decimal places

# Example 3: Multiple values
x = 10
y = 20
print("Coordinates: x=%d, y=%d" % (x, y))

# Example 4: Using a dictionary for unordered data
data = {"name": "Bob", "score": 95}
print("Student: %(name)s, Score: %(score)d" % data)

# Example 5: Combining different data types
item = "apple"
quantity = 4
price = 0.75
print("You bought %d %s(s) at $%.2f each." % (quantity, item, price))
````

## 2Ô∏è‚É£0Ô∏è‚É£ Reversing a String

The cleanest way to reverse a string in Python is using **slicing** with `step = -1`.

```python
s[::-1]  # reads string from end to start
```

This is also useful for palindrome checks.

````python
# ============================================
# 2Ô∏è‚É£0Ô∏è‚É£ Reversing a String
# ============================================
# ============================================================
# Reversing a String in Python
# This notebook explains multiple ways to reverse a string.
# ============================================================


# ------------------------------------------------------------
# Method 1Ô∏è‚É£ : Using a WHILE Loop
# ------------------------------------------------------------
# Logic:
# - Create an empty string.
# - Start from the last index of the string.
# - Keep decreasing the index.
# - Append each character to the new string.

name = "codeyug"

reversed_string = ""
i = len(name) - 1   # last index

while i >= 0:
    reversed_string += name[i]
    i -= 1

print("Using while loop:", reversed_string)


# ------------------------------------------------------------
# Method 2Ô∏è‚É£ : Using a FOR Loop
# ------------------------------------------------------------
# Logic:
# - Create an empty string.
# - Traverse the string normally.
# - Add each character to the FRONT of the new string.
# - This reverses the order automatically.

name = "codeyug"

reversed_string = ""

for char in name:
    reversed_string = char + reversed_string

print("Using for loop:", reversed_string)


# ------------------------------------------------------------
# Method 3Ô∏è‚É£ : Using SLICING (Preferred & Simplest Method)
# ------------------------------------------------------------
# String slicing syntax:
# string[start:stop:step]
#
# If step = -1 ‚Üí it moves backward.
# name[::-1] means:
# - start from beginning
# - go till end
# - move with step -1 (reverse direction)
#
# Strings are immutable in Python,
# so slicing creates a new reversed string.

name = "codeyug"

reversed_string = name[::-1]

print("Using slicing:", reversed_string)

# ------------------------------------------------------------
# Palindrome Check using Reversal
# ------------------------------------------------------------
# A palindrome reads the same forward and backward.
# Example: madam, level, racecar

word = "madam"

# If original word == reversed word ‚Üí it's a palindrome
print("Is 'madam' a palindrome?", word == word[::-1])   # True
````

## 2Ô∏è‚É£1Ô∏è‚É£ Combined Example ‚Äî split, len, join together

This example shows how multiple string methods can be **chained** together in a real-world scenario:
1. Split a comma-separated string into a list
2. Use list comprehension to calculate the length of each word
3. Join the list back into a string with a new separator

````python
# ============================================
# 2Ô∏è‚É£1Ô∏è‚É£ Combined Example ‚Äî split, len, join
# ============================================

s = "apple,banana,cherry"

# Step 1: split() ‚Üí break string into a list by comma
fruits = s.split(",")
print(fruits)              # ['apple', 'banana', 'cherry']

# Step 2: List comprehension ‚Üí count length of each fruit name
fruits_count = [len(f) for f in fruits]
print(fruits_count)        # [5, 6, 6]

# Step 3: join() ‚Üí combine list back with ' & ' as separator
joined = " & ".join(fruits)
print(joined)              # apple & banana & cherry
````

# 2Ô∏è‚É£2Ô∏è‚É£ String Comparison in Python

This notebook explains how strings are compared in Python.

Python compares strings:
1. **Character by character**
2. **If characters differ**, their Unicode values are compared
3. **The character with a lower Unicode value** is considered smaller

### Comparison Operators:
* `==` ‚Üí Equal to
* `!=` ‚Üí Not equal to
* `>`  ‚Üí Greater than
* `<`  ‚Üí Less than
* `>=` ‚Üí Greater than or equal to
* `<=` ‚Üí Less than or equal to

---


````python
# ------------------------------------------------------------
# 1Ô∏è‚É£ Basic String Comparisons
# ------------------------------------------------------------

a = "apple"
b = "banana"
c = "apple"

print("a == c:", a == c)   # True
print("a != b:", a != b)   # True
print("a > b:", a > b)     # False
print("a < b:", a < b)     # True
print("a >= c:", a >= c)   # True
print("a <= b:", a <= b)   # True

````

````python
# ------------------------------------------------------------
# 2Ô∏è‚É£ Character-by-Character Comparison
# ------------------------------------------------------------
# Python compares first character of both strings.
# If they are equal ‚Üí move to next character.
# If different ‚Üí compare Unicode values.

x = "cat"
y = "car"

print("\nComparing 'cat' and 'car':")
print("x > y:", x > y)  # True because 't' > 'r'
````

````python
# ------------------------------------------------------------
# 3Ô∏è‚É£ Unicode Value Comparison
# ------------------------------------------------------------
# If characters are different:
# Python compares their Unicode values using ord()

print("\nUnicode values:")
print("Unicode of 'a':", ord('a'))  # 97
print("Unicode of 'b':", ord('b'))  # 98
print("Unicode of 'A':", ord('A'))  # 65
print("Unicode of 'a':", ord('a'))  # 97

# Since Unicode of 'A' (65) < 'a' (97)
print("\n'A' < 'a':", 'A' < 'a')  # True
````

````python
# ------------------------------------------------------------
# 4Ô∏è‚É£ Case Sensitivity in String Comparison
# ------------------------------------------------------------
# String comparison is case-sensitive.

print("\nCase-sensitive comparison:")
print("'apple' == 'Apple':", "apple" == "Apple")
print("'apple' > 'Apple':", "apple" > "Apple")
````
