```python
# ============================================
# Python Strings ‚Äì Complete Operations
# ============================================

# 1Ô∏è‚É£ Creating Strings
str1 = "Hello"
str2 = 'World'
str3 = """This is a 
multiline string"""
str4 = 'Python "Programming"'

print(str1, str2, str3, str4)

# 2Ô∏è‚É£ LENGTH
print(len(str1))  # 5

# 3Ô∏è‚É£ INDEXING
print(str1[0])   # 'H' ‚Üí first character
print(str1[-1])  # 'o' ‚Üí last character

# 4Ô∏è‚É£ SLICING
print(str1[1:4])   # 'ell'
print(str1[:3])    # 'Hel'
print(str1[2:])    # 'llo'
print(str1[::2])   # 'Hlo' ‚Üí every 2nd char
print(str1[::-1])  # 'olleH' ‚Üí reversed string

# 5Ô∏è‚É£ CONCATENATION
s1 = "Hello"
s2 = "World"
s3 = s1 + " " + s2
print(s3)  # 'Hello World'

# 6Ô∏è‚É£ REPEAT
print(s1 * 3)  # 'HelloHelloHello'

# 7Ô∏è‚É£ MEMBERSHIP
print("H" in s1)   # True
print("z" not in s1)  # True

# 8Ô∏è‚É£ STRING METHODS

# Changing case
print(s1.upper())   # 'HELLO'
print(s2.lower())   # 'world'
print(s1.capitalize())  # 'Hello'
print(s2.title())       # 'World'  -> converts the first character of every word to uppercase

''' 
s2.title() does not modify the original string, because strings are immutable in Python.

If you print string, it will still be:

"world"
üîπ To Actually Change the String

You must store the result:

s2 = "world"
s2 = string.title()
print(s2)

Now output will be:

World

üîπ Why?

Strings in Python cannot be changed directly (immutable), so methods like:

.title()

.upper()

.lower()

.replace()

always return a new string, they don‚Äôt modify the original one.

''' 

# Strip whitespace
s = "   Python   "
print(s.strip())    # 'Python'
print(s.lstrip())   # 'Python   '
print(s.rstrip())   # '   Python'

# Replace
s = "I love Python"
print(s.replace("Python", "Java"))  # 'I love Java'

# Split and Join
s = "apple,banana,cherry"
lst = s.split(",")  # ['apple','banana','cherry']
print(lst)
joined = "-".join(lst)
print(joined)       # 'apple-banana-cherry'

''' 
üîπ Python Split & Join Cheat Sheet
1Ô∏è‚É£ split()

Purpose: Break a string into a list of strings using a separator.

Syntax:

string.split(separator, maxsplit)

separator ‚Üí string where you want to split (default: space " ")

maxsplit ‚Üí optional, maximum number of splits

Examples:

s = "apple,banana,cherry"
lst = s.split(",")
print(lst)  # ['apple', 'banana', 'cherry']
print(type(lst))  # <class 'list'>
s = "one two three"
lst = s.split()  # default separator is space
print(lst)  # ['one', 'two', 'three']

‚úÖ Key Points:

Output is always a list

Original string is not modified

Works only on strings

2Ô∏è‚É£ join()

Purpose: Combine elements of an iterable of strings into a single string.

Syntax:

separator.join(iterable_of_strings)

separator ‚Üí string placed between elements

iterable_of_strings ‚Üí list, tuple, etc. (all elements must be strings)

Examples:

lst = ["apple", "banana", "cherry"]
s = ",".join(lst)
print(s)  # 'apple,banana,cherry'
print(type(s))  # <class 'str'>
tup = ("1", "2", "3")
s = "-".join(tup)
print(s)  # '1-2-3'
nums = [1, 2, 3]
s = ",".join(str(n) for n in nums)  # convert numbers to strings
print(s)  # '1,2,3'

‚úÖ Key Points:

Input must be strings, otherwise TypeError

Output is always a string

Can join lists, tuples, sets (order may vary for sets)

3Ô∏è‚É£ Split ‚Üî Join Conversion
s = "apple,banana,cherry"

# Split string into list
lst = s.split(",")  # ['apple', 'banana', 'cherry']

# Join list back into string
joined_s = ",".join(lst)  # 'apple,banana,cherry'

üîπ Split ‚Üí List ‚Üí Join ‚Üí String
üîπ .split() ‚Üí list, .join() ‚Üí string

4Ô∏è‚É£ Quick Tips for Interviews

Remember immutability of strings: split/join returns a new object, original string unchanged

Joining numbers: always convert to string first (str(n))

Default separators: split() ‚Üí " ", join() ‚Üí whatever separator you choose

Tuple input is fine, but output is still a string

'''

# Count occurrences
s = "banana"
print(s.count("a"))  # 3

# Find / index
print(s.find("a"))    # 1 ‚Üí first occurrence
print(s.rfind("a"))   # 5 ‚Üí last occurrence
print(s.index("b"))   # 0 ‚Üí first occurrence

"""

1Ô∏è‚É£ .find()

Returns the first index of the substring if it exists.

Returns -1 if the substring is not found.

‚úÖ Safe to use when you‚Äôre unsure if the substring exists.

s = "banana"

print(s.find("a"))  # 1 ‚Üí first occurrence
print(s.find("x"))  # -1 ‚Üí substring not found
üîπ 2Ô∏è‚É£ .index()

Returns the first index of the substring if it exists.

Raises ValueError if the substring is not found.

‚ö†Ô∏è Use this only when you are sure the substring exists, or wrap it in a try-except.

s = "banana"

print(s.index("a"))  # 1 ‚Üí first occurrence
print(s.index("x"))  # ‚ùå ValueError: substring not found

"""

# Startswith / Endswith
print(s.startswith("b"))  # True
print(s.endswith("a"))    # True

# Check alphanumeric
print("abc123".isalnum())  # True       to be isalnum there should be atleast either number or alphabet 
print("123".isalnum())     # True
print("abc 123".isalnum()) # False
print("123".isdigit())     # True
print("abc".isalpha())     # True
print("   ".isspace())     # True

# 9Ô∏è‚É£ MIN and MAX
s = "banana"
print(min(s))  # 'a'
print(max(s))  # 'n'

# 1Ô∏è‚É£0Ô∏è‚É£ SUM (not directly for string, use ord)
s = "abc"
print(sum(ord(c) for c in s))  # 97+98+99=294

# 1Ô∏è‚É£1Ô∏è‚É£ ITERATION
for ch in "Hello":
    print(ch)

for index, ch in enumerate("World", start=1):
    print(index, ch)

# 1Ô∏è‚É£2Ô∏è‚É£ IMMUTABILITY
s = "Python"
# s[0] = "J"  # ‚ùå TypeError ‚Üí strings cannot be changed
# To change, create new string
s2 = "J" + s[1:]    # here s[1:] means taking substring starting from index 1 of s string and added to the J
print(s2)  # 'Jython'

# 1Ô∏è‚É£3Ô∏è‚É£ NESTED / MULTILINE STRINGS
multi = """Line1
Line2
Line3"""
print(multi)

# 1Ô∏è‚É£4Ô∏è‚É£ STRING FORMATTING
name = "Rajan"
age = 22
# Using f-string
print(f"My name is {name} and I am {age} years old")
# Using format()
print("My name is {} and I am {} years old".format(name, age))

# 1Ô∏è‚É£5Ô∏è‚É£ ESCAPE CHARACTERS
print("Hello\nWorld")    # new line
print("Hello\tWorld")    # tab
print("It\'s Python")    # single quote inside string

# 1Ô∏è‚É£6Ô∏è‚É£ REVERSING STRING
s = "Python"
rev = s[::-1]
print(rev)  # 'nohtyP'

# 1Ô∏è‚É£7Ô∏è‚É£ COMBINED EXAMPLES
s = "apple,banana,cherry"
# split, count, join
fruits = s.split(",")            # ['apple','banana','cherry']
print(fruits)
fruits_count = [len(f) for f in fruits]  # [5,6,6]
print(fruits_count)
joined = " & ".join(fruits)      # 'apple & banana & cherry'
print(joined)
```
