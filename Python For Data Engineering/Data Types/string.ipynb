```python
# ============================================
# Python Strings – Complete Operations
# ============================================

# 1️⃣ Creating Strings
str1 = "Hello"
str2 = 'World'
str3 = """This is a 
multiline string"""
str4 = 'Python "Programming"'

print(str1, str2, str3, str4)

# 2️⃣ LENGTH
print(len(str1))  # 5

# 3️⃣ INDEXING
print(str1[0])   # 'H' → first character
print(str1[-1])  # 'o' → last character

# 4️⃣ SLICING
print(str1[1:4])   # 'ell'
print(str1[:3])    # 'Hel'
print(str1[2:])    # 'llo'
print(str1[::2])   # 'Hlo' → every 2nd char
print(str1[::-1])  # 'olleH' → reversed string

# 5️⃣ CONCATENATION
s1 = "Hello"
s2 = "World"
s3 = s1 + " " + s2
print(s3)  # 'Hello World'

# 6️⃣ REPEAT
print(s1 * 3)  # 'HelloHelloHello'

# 7️⃣ MEMBERSHIP
print("H" in s1)   # True
print("z" not in s1)  # True

# 8️⃣ STRING METHODS

# Changing case
print(s1.upper())   # 'HELLO'
print(s2.lower())   # 'world'
print(s1.capitalize())  # 'Hello'
print(s2.title())       # 'World'

# Strip whitespace
s = "   Python   "
print(s.strip())    # 'Python'
print(s.lstrip())   # 'Python   '
print(s.rstrip())   # '   Python'

# Replace
s = "I love Python"
print(s.replace("Python", "Java"))  # 'I love Java'

# Split and Join
s = "apple,banana,cherry"
lst = s.split(",")  # ['apple','banana','cherry']
print(lst)
joined = "-".join(lst)
print(joined)       # 'apple-banana-cherry'

# Count occurrences
s = "banana"
print(s.count("a"))  # 3

# Find / index
print(s.find("a"))    # 1 → first occurrence
print(s.rfind("a"))   # 5 → last occurrence
print(s.index("b"))   # 0 → first occurrence

# Startswith / Endswith
print(s.startswith("b"))  # True
print(s.endswith("a"))    # True

# Check alphanumeric
print("abc123".isalnum())  # True
print("abc 123".isalnum()) # False
print("123".isdigit())     # True
print("abc".isalpha())     # True
print("   ".isspace())     # True

# 9️⃣ MIN and MAX
s = "banana"
print(min(s))  # 'a'
print(max(s))  # 'n'

# 1️⃣0️⃣ SUM (not directly for string, use ord)
s = "abc"
print(sum(ord(c) for c in s))  # 97+98+99=294

# 1️⃣1️⃣ ITERATION
for ch in "Hello":
    print(ch)

for index, ch in enumerate("World", start=1):
    print(index, ch)

# 1️⃣2️⃣ IMMUTABILITY
s = "Python"
# s[0] = "J"  # ❌ TypeError → strings cannot be changed
# To change, create new string
s2 = "J" + s[1:]
print(s2)  # 'Jython'

# 1️⃣3️⃣ NESTED / MULTILINE STRINGS
multi = """Line1
Line2
Line3"""
print(multi)

# 1️⃣4️⃣ STRING FORMATTING
name = "Rajan"
age = 22
# Using f-string
print(f"My name is {name} and I am {age} years old")
# Using format()
print("My name is {} and I am {} years old".format(name, age))

# 1️⃣5️⃣ ESCAPE CHARACTERS
print("Hello\nWorld")    # new line
print("Hello\tWorld")    # tab
print("It\'s Python")    # single quote inside string

# 1️⃣6️⃣ REVERSING STRING
s = "Python"
rev = s[::-1]
print(rev)  # 'nohtyP'

# 1️⃣7️⃣ COMBINED EXAMPLES
s = "apple,banana,cherry"
# split, count, join
fruits = s.split(",")            # ['apple','banana','cherry']
print(fruits)
fruits_count = [len(f) for f in fruits]  # [5,6,6]
print(fruits_count)
joined = " & ".join(fruits)      # 'apple & banana & cherry'
print(joined)
```
