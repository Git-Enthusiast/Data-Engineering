# ðŸ”¢ Python Integer (`int`) Data Type

## ðŸ“Œ Introduction
This notebook provides a complete explanation of the Integer data type in Python.

An **Integer (`int`)** represents:
- Whole numbers
- No decimal point
- Can be positive, negative, or zero

**Examples:** `10`, `-25`, `0`

**Definition:**
Integers are whole numbers (positive or negative) without any decimal value.

````python
# BASIC INTEGER EXAMPLES
a = 10
b = -25
c = 0

print("Basic Integers:")
print(a, type(a))
print(b, type(b))
print(c, type(c))
````

---
## ðŸ“Œ Important Characteristics of Integer in Python

### 1ï¸âƒ£ No Length Limitation
Unlike many programming languages (like C/C++/Java), Python integers have **NO fixed size limit**.
- âœ” Limited only by system memory.
- âœ” You can store extremely large numbers.

### 2ï¸âƒ£ Immutable Nature
Integers in Python are **IMMUTABLE**.
**Meaning:** Once an integer object is created, it cannot be changed.
When you modify an integer (e.g. `x = x + 1`), Python does NOT modify the original number. Instead, it creates a new integer object and assigns it to `x`.

### 3ï¸âƒ£ Leading Zeros Rule
Decimal integers **CANNOT** start with leading zeros.
- âŒ **Invalid:** `x = 0123   (SyntaxError)`
- âœ” **Valid:** `x = 123`

*However, leading zeros are allowed in Binary (`0b`), Octal (`0o`), and Hexadecimal (`0x`).*

### 4ï¸âƒ£ Boolean Behavior
In Python:
- `0` â†’ `False`
- Non-zero (e.g., `1`, `-5`) â†’ `True`

````python
# IMMUTABILITY DEMONSTRATION
num = 10
print("Before change:", num, "| Memory ID:", id(num))

num = num + 1
print("After change: ", num, "| Memory ID:", id(num))

# Notice: The memory id changes â†’ A new object is created!
````

````python
# BOOLEAN BEHAVIOR
print("Boolean Behavior:")
if 5:
    print("5 evaluates to True")
if not 0:
    print("0 evaluates to False")
````

````python
# VERY LARGE INTEGER (NO LIMIT)
big_number = 999999999999999999999999999999999999999999999999999999
print("Very Large Integer:")
print(big_number)
````

## ðŸ“Œ Four Types of Integers (Number Systems)

Python supports integers in different number systems. Internally, Python stores all integers in binary format, but it allows you to write and represent them in four different bases.

| Number System | Base | Prefix | Allowed Digits | Example |
|---------------|------|--------|----------------|---------|
| **Decimal** | 10 | (None) | `0` to `9` | `100` |
| **Binary** | 2 | `0b` or `0B` | `0` and `1` | `0b1010` |
| **Octal** | 8 | `0o` or `0O` | `0` to `7` | `0o12` |
| **Hexadecimal**| 16 | `0x` or `0X` | `0`-`9`, `a`-`f` (or `A`-`F`) | `0xA` |

````python
# FOUR TYPES OF INTEGERS
decimal_num = 100          # Decimal (Base 10)
binary_num = 0b1010        # Binary (Base 2) -> equals 10
octal_num = 0o12           # Octal (Base 8) -> equals 10
hex_num = 0xA              # Hexadecimal (Base 16) -> equals 10

print("Different Number Systems (Python prints them as Decimal by default):")
print(f"{decimal_num =}")
print(f"{binary_num = }")
print(f"{octal_num = }")
print(f"{hex_num = }")
````

## ðŸ“Œ Conversion Between Number Systems

Python provides built-in functions to convert numbers from decimal to other bases.

| Target Base | Function | Output Format | Example | Output |
|-------------|----------|---------------|---------|--------|
| **Binary** | `bin(n)` | String with `0b` prefix | `bin(10)` | `'0b1010'` |
| **Octal** | `oct(n)` | String with `0o` prefix | `oct(10)` | `'0o12'` |
| **Hex** | `hex(n)` | String with `0x` prefix | `hex(10)` | `'0xa'` |

> **Rule:** These functions (`bin()`, `oct()`, `hex()`) return a **String**, not an integer object! If you need to convert them back to a decimal integer, use the `int(string, base)` function.

````python
# CONVERTING DECIMAL TO OTHER BASES
num = 255

print(f"Decimal: {num}")
print(f"To Binary:      {bin(num)}  (Type: {type(bin(num))})")
print(f"To Octal:       {oct(num)}  (Type: {type(oct(num))})")
print(f"To Hexadecimal: {hex(num)}  (Type: {type(hex(num))})")
````

````python
# CONVERTING OTHER BASES BACK TO DECIMAL
# You can simply pass the literal directly, or use int() with strings

print("Converting back to Decimal integer:")
print(f"From Binary String '0b11111111': {int('0b11111111', 2)}")
print(f"From Octal String '0o377':       {int('0o377', 8)}")
print(f"From Hex String '0xff':          {int('0xff', 16)}")
````

## ðŸ“Œ Arithmetic Operations Supported

| Operator | Name | Example |
|----------|------|---------|
| `+` | Addition | `10 + 5 = 15` |
| `-` | Subtraction | `10 - 5 = 5` |
| `*` | Multiplication | `10 * 5 = 50` |
| `/` | Division | `10 / 3 = 3.3333` *(Results in Float)* |
| `//`| Floor Division| `10 // 3 = 3` *(Results in Integer, rounds down)* |
| `%` | Modulus | `10 % 3 = 1` *(Remainder)* |
| `**`| Power | `10 ** 3 = 1000` |

````python
# ARITHMETIC OPERATIONS
x = 15
y = 4

print("Arithmetic Operations:")
print(f"{x} + {y} = {x + y}")
print(f"{x} - {y} = {x - y}")
print(f"{x} * {y} = {x * y}")
print(f"{x} / {y} = {x / y} (Float Division)")
print(f"{x} // {y} = {x // y} (Floor Division)")
print(f"{x} % {y} = {x % y} (Modulus)")
print(f"{x} ** {y} = {x ** y} (Power)")
````

## ðŸ“Œ Type Conversion (Type Casting to `int`)

Use the `int()` function to convert to an integer.
- `int("25")` â†’ converts string to integer
- `int(9.8)` â†’ removes decimal part (truncates towards zero)

````python
# TYPE CONVERSION
str_value = "50"
converted = int(str_value)
print("String to int:", converted, type(converted))

float_value = 9.99
int_from_float = int(float_value)
print("Float to int (decimal truncated):", int_from_float)
````
