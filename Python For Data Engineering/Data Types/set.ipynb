## 1️⃣ What is a Set?

A **set** is an **unordered**, **mutable** collection of **unique**, **hashable** elements in Python.

| Property | Details |
|---|---|
| **Unordered** | Elements have no fixed position — order may vary each run |
| **Mutable** | You can add/remove elements after creation |
| **No Duplicates** | Automatically removes duplicate values |
| **Hashable Elements** | Only immutable types allowed (`int`, `float`, `str`, `tuple`) |
| **No Indexing/Slicing** | Cannot access elements by index |
| **Hash Table** | Internally implemented using a hash table → O(1) average lookups |

> 💡 **When to use sets?** When you need **fast membership testing**, **removing duplicates**, or performing **mathematical set operations** (union, intersection, etc.).

```python
# ============================================
# 1️⃣ What is a Set?
# ============================================

# --- UNORDERED COLLECTION ---
s1 = {10, 20, 30, 40}
print("Original set:", s1)       # Order is NOT guaranteed

for item in s1:
    print(item, end=" ")
print()   # Output order may vary each run

# --- NO INDEXING / NO SLICING ---
s2 = {1, 2, 3}
try:
    print(s2[0])
except TypeError as e:
    print("Indexing Error:", e)
# Output: TypeError: 'set' object is not subscriptable

# --- MUTABLE (Can be modified after creation) ---
s3 = {1, 2}
print("Before add:", s3)
s3.add(3)
print("After add:", s3)           # {1, 2, 3}

# --- STORES UNIQUE ELEMENTS ONLY ---
s4 = {1, 2, 2, 3, 3, 3}
print("Duplicates removed automatically:", s4)  # {1, 2, 3}

# --- ELEMENTS MUST BE HASHABLE ---
valid_set = {10, 3.14, "hello", (1, 2)}
print("Valid set:", valid_set)

try:
    invalid_set = {[1, 2, 3]}     # list is mutable → unhashable
except TypeError as e:
    print("Unhashable Error:", e)

# --- FAST MEMBERSHIP TESTING (Hash Table → O(1)) ---
s5 = {100, 200, 300}
print(200 in s5)   # True
print(999 in s5)   # False
```

## 2️⃣ Creating Sets

| Method | Syntax | Example |
|---|---|---|
| Curly Braces | `{val1, val2}` | `s = {1, 2, 3}` |
| `set()` constructor | `set(iterable)` | `s = set([1, 2, 3])` |
| From String | `set(string)` | `s = set("hello")` → unique chars |
| Empty Set | `set()` | `s = set()` |

> ⚠️ **Common Trap:** `{}` creates an **empty dictionary**, NOT an empty set! Use `set()` for an empty set.

```python
# ============================================
# 2️⃣ Creating Sets
# ============================================

# Method 1: Using curly braces
set1 = {10, 20, 30}
print(set1)                    # {10, 20, 30}

# Method 2: Using set() constructor (from a list)
set2 = set([1, 2, 3, 3])      # duplicates removed
print(set2)                    # {1, 2, 3}

# Method 3: From string (each unique character)
set3 = set("hello")
print(set3)                    # {'h', 'e', 'l', 'o'}

# ⚠️ Empty set vs empty dict
empty_set = set()              # ✅ Correct way to create empty set
print(type(empty_set))         # <class 'set'>

empty_dict = {}                # ❌ This is a dict, NOT a set
print(type(empty_dict))        # <class 'dict'>
```

## 3️⃣ Set Characteristics

> 💡 **Key Rules for Sets:**
> 1. **Unordered** — No guaranteed element order
> 2. **No Indexing** — `s[0]` will raise `TypeError`
> 3. **No Slicing** — `s[1:3]` is not allowed
> 4. **No Duplicates** — Duplicate values are automatically removed
> 5. **Fast Lookup** — `in` operator runs in O(1) average time

| Feature | Set | List |
|---|---|---|
| Ordered? | ❌ No | ✅ Yes |
| Duplicates? | ❌ No | ✅ Yes |
| Indexing? | ❌ No | ✅ Yes |

## 4️⃣ Adding Elements

| Method | Description |
|---|---|
| `.add(element)` | Adds a **single** element to the set. No effect if element already exists. |
| `.update(iterable)` | Adds **multiple** elements from any iterable (list, tuple, set, string). |

> 💡 `add()` takes exactly **one** argument. To add multiple elements at once, use `update()`.

```python
# ============================================
# 4️⃣ Adding Elements
# ============================================

s = {1, 2}
print(s)

# add() — adds ONE element
s.add(3)
print(s)           # {1, 2, 3}

# update() — adds MULTIPLE elements from an iterable
s.update([4, 5])
print(s)           # {1, 2, 3, 4, 5}

# update() with multiple iterables
s.update([6], (7,))
print(s)           # {1, 2, 3, 4, 5, 6, 7}

s.update((23,(1,2)))    # Valid u can add element which is immutable
print(s)

#   s.update((23,[45,67]))
#   TypeError: cannot use 'list' as a set element (unhashable type: 'list')
```

## 5️⃣ Removing Elements

| Method | Description | If element missing? |
|---|---|---|
| `.remove(x)` | Removes element `x` | ❌ Raises `KeyError` |
| `.discard(x)` | Removes element `x` | ✅ No error (silent) |
| `.pop()` | Removes and returns an **arbitrary** element | ❌ `KeyError` if set is empty |
| `.clear()` | Removes **all** elements | Returns `set()` |
| `del()` | Delete set from memory | Accessing `set_name` will raise NameError |

> 💡 **Best Practice:** Use `discard()` when you're not sure if the element exists — it won't crash your program.

```python
# ============================================
# 5️⃣ Removing Elements
# ============================================

s = {1, 2, 3, 4}

# remove() — raises KeyError if not found
s.remove(2)
print("After remove(2):", s)    # {1, 3, 4}

# discard() — NO error if not found (safe)
s.discard(10)                    # 10 not in set → no error
print("After discard(10):", s)   # {1, 3, 4}

# pop() — removes and returns an arbitrary element
popped = s.pop()
print("Popped:", popped)
print("After pop:", s)

# clear() — empties the entire set
s.clear()
print("After clear:", s)         # set()

# del → it will delete set from memory
del s
#   print(s)    # will lead to Name Error
```

## 6️⃣ Set Operations (VERY IMPORTANT)

Sets support **mathematical set operations** — extremely useful in data engineering.

| Operation | Operator | Method | Description |
|---|---|---|---|
| **Union** | `A \| B` | `A.union(B)` | All elements from both sets |
| **Intersection** | `A & B` | `A.intersection(B)` | Common elements only |
| **Difference** | `A - B` | `A.difference(B)` | Elements in A but NOT in B |
| **Symmetric Difference** | `A ^ B` | `A.symmetric_difference(B)` | Elements in A or B, but NOT both |

> 💡 **Operators (`|`, `&`, `-`, `^`)** require both operands to be sets.  
> **Methods (`.union()`, etc.)** accept any iterable (list, tuple, etc.).

```python
# ============================================
# 6️⃣ Set Operations
# ============================================

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# --- Union: All elements from both ---
print("Union:", A | B)              # {1, 2, 3, 4, 5, 6}
print("Union (method):", A.union(B))

# --- Intersection: Common elements ---
print("Intersection:", A & B)              # {3, 4}
print("Intersection (method):", A.intersection(B))

# --- Difference: In A but NOT in B ---
print("Difference A-B:", A - B)     # {1, 2}
print("Difference B-A:", B - A)     # {5, 6}

# --- Symmetric Difference: In A or B, but NOT both ---
print("Symmetric Difference:", A ^ B)  # {1, 2, 5, 6}

# Original Set doesn't change
print(A)
print(B)
```

## 7️⃣ In-Place Update Operators

These operators **modify the original set** instead of creating a new one.

| Operator | Method Equivalent | Description |
|---|---|---|
| `A \|= B` | `A.update(B)` | Add all elements of B to A |
| `A &= B` | `A.intersection_update(B)` | Keep only common elements |
| `A -= B` | `A.difference_update(B)` | Remove elements found in B |
| `A ^= B` | `A.symmetric_difference_update(B)` | Keep only non-common elements |

```python
# ============================================
# 7️⃣ In-Place Update Operators
# ============================================

C = {1, 2, 3}
C |= {3, 4}           # Union update → adds 4
print("After |= :", C)              # {1, 2, 3, 4}

C &= {1, 3, 4}        # Intersection update → keeps only {1, 3, 4}
print("After &= :", C)              # {1, 3, 4}

C -= {4}               # Difference update → removes 4
print("After -= :", C)              # {1, 3}

C ^= {2, 10}           # Symmetric difference update
print("After ^= :", C)              # {1, 2, 3, 10}
```

## 8️⃣ Subset, Superset & Disjoint

| Method | Operator | Description |
|---|---|---|
| `A.issubset(B)` | `A <= B` | Is every element of A also in B? |
| `A < B` | — | Is A a **proper** subset of B? (A ≠ B) |
| `A.issuperset(B)` | `A >= B` | Does A contain every element of B? |
| `A > B` | — | Is A a **proper** superset of B? (A ≠ B) |
| `A.isdisjoint(B)` | — | Do A and B share **no** common elements? |

```python
# ============================================
# 8️⃣ Subset, Superset & Disjoint
# ============================================

X = {1, 2}
Y = {1, 2, 3}

# Subset — is X contained in Y?
print("X subset of Y:", X.issubset(Y))    # True
print("Y superset of X:", Y.issuperset(X)) # True

# Operators
print("X <= Y:", X <= Y)    # True  (subset)
print("Y >= X:", Y >= X)    # True  (superset)
print("X < Y (proper subset):", X < Y)  # True (X ≠ Y)

# Disjoint — no common elements?
A = {1, 2}
B = {3, 4}
C = {2, 5}
print("A disjoint B:", A.isdisjoint(B))   # True  (no overlap)
print("A disjoint C:", A.isdisjoint(C))   # False (2 is common)
```

## 9️⃣ Set Comprehension

Just like list comprehensions, sets can be created concisely in a single line.

**Syntax:** `{expression for item in iterable if condition}`

> 💡 The result is always a **set** — duplicates are removed automatically.

```python
# ============================================
# 9️⃣ Set Comprehension
# ============================================

# Basic set comprehension
squares = {x**2 for x in range(6)}
print("Squares:", squares)          # {0, 1, 4, 9, 16, 25}

# With condition
even_sq = {x**2 for x in range(6) if x % 2 == 0}
print("Even squares:", even_sq)     # {0, 4, 16}

# From a string — unique vowels
word = "hello world"
vowels = {ch for ch in word if ch in "aeiou"}
print("Unique vowels:", vowels)     # {'o', 'e'}
```

## 🔟 Frozenset (Immutable Set)

A `frozenset` is an **immutable** version of a set — once created, you **cannot** add or remove elements.

| Feature | `set` | `frozenset` |
|---|---|---|
| Mutable? | ✅ Yes | ❌ No |
| Can be a dict key? | ❌ No | ✅ Yes |
| Can be inside a set? | ❌ No | ✅ Yes |
| Supports set operations? | ✅ Yes | ✅ Yes |

> 💡 Use `frozenset` when you need a **hashable** set (e.g., as a dictionary key or inside another set).

```python
# ============================================
# 🔟 Frozenset (Immutable Set)
# ============================================

fs = frozenset([1, 2, 3])
print(fs)                     # frozenset({1, 2, 3})

# ❌ Cannot modify
try:
    fs.add(4)
except AttributeError as e:
    print("Cannot modify frozenset:", e)

# ✅ Supports set operations (returns new frozenset)
fs2 = frozenset([3, 4, 5])
print(fs & fs2)               # frozenset({3})

# ✅ Can be used as a dictionary key
d = {fs: "my_set"}
print(d)
print("frozenset as dict key works!")
```

## 1️⃣1️⃣ Hashability Rule

All elements in a set **must be hashable** (i.e., immutable).

| Type | Hashable? | Can be in set? |
|---|---|---|
| `int`, `float`, `bool` | ✅ Yes | ✅ Yes |
| `str` | ✅ Yes | ✅ Yes |
| `tuple` (with hashable elements) | ✅ Yes | ✅ Yes |
| `frozenset` | ✅ Yes | ✅ Yes |
| `list` | ❌ No | ❌ No |
| `dict` | ❌ No | ❌ No |
| `set` | ❌ No | ❌ No |

> ⚠️ A `tuple` containing a `list` is **NOT hashable**: `(1, [2, 3])` cannot be in a set.

```python
# ============================================
# 1️⃣1️⃣ Hashability Rule
# ============================================

# ✅ Allowed types (all immutable/hashable)
valid_set = {1, 3.14, "hello", (1, 2)}
print("Valid set:", valid_set)

# ❌ List is NOT hashable
try:
    bad = {[1, 2, 3]}
except TypeError as e:
    print("Error (list):", e)

# ❌ Dict is NOT hashable
try:
    bad = {{"a": 1}}
except TypeError as e:
    print("Error (dict):", e)

# ❌ Tuple containing a list is NOT hashable
try:
    bad = {(1, [2, 3])}
except TypeError as e:
    print("Error (tuple with list):", e)
```

## 1️⃣2️⃣ Internal Working — Hash Table & Time Complexity

Python sets are implemented using a **Hash Table**.

**How it works:**
1. Python computes the **hash value** of the element using `hash()`
2. The hash value determines the **bucket** where the element is stored
3. For membership testing, Python hashes the value and checks the bucket directly

### ⏱️ Time Complexity

| Operation | Average | Worst Case |
|---|---|---|
| `add(x)` | O(1) | O(n) |
| `remove(x)` | O(1) | O(n) |
| `x in s` | O(1) | O(n) |
| `union` | O(len(A)+len(B)) | — |
| `intersection` | O(min(len(A),len(B))) | — |

> 💡 Worst case O(n) happens only due to **hash collisions** — extremely rare in practice.

```python
# ============================================
# 1️⃣2️⃣ Internal Working — Hash Table
# ============================================

# Python uses hash() to compute bucket position
print("hash(42) =", hash(42))             # integer hash
print("hash('hello') =", hash("hello"))   # string hash
print("hash((1,2)) =", hash((1, 2)))      # tuple hash

# Note: hash values may differ across Python sessions
# (due to hash randomization for security)

# This is WHY membership testing is O(1):
# 1) Compute hash(element)
# 2) Jump directly to that bucket
# 3) Check if element is there → constant time!
```

## 1️⃣3️⃣ Copying Sets — Shallow vs Deep Copy

| Method | Description |
|---|---|
| `=` (assignment) | Both variables point to the **same** set in memory |
| `.copy()` | Creates a **shallow copy** (new set object) |
| `set(s)` | Also creates a shallow copy |

> 💡 Since set elements must be **hashable** (immutable), shallow copy is sufficient — there are no nested mutable objects to worry about (unlike lists/dicts).

```python
# ============================================
# 1️⃣3️⃣ Copying Sets
# ============================================

original = {1, 2, 3}

# ❌ Assignment — both point to SAME object
alias = original
alias.add(99)
print("original:", original)    # {1, 2, 3, 99} ← also changed!
print("alias:", alias)          # {1, 2, 3, 99}
print("Same object?", original is alias)  # True

# ✅ copy() — creates a NEW independent set
copied = original.copy()
copied.add(50)
print("\noriginal after copy mod:", original)  # unchanged
print("copied:", copied)
print("Same object?", original is copied)  # False
```

## 1️⃣4️⃣ Iterating Over Sets

Sets are **iterable** — you can loop through them using `for`. However, since sets are **unordered**, the iteration order is not guaranteed.

| Method | Use Case |
|---|---|
| `for item in set` | Access each element |
| `enumerate(set)` | Get index + value (index is arbitrary, not meaningful) |

> ⚠️ **Never modify a set while iterating** — this raises `RuntimeError`.

```python
# ============================================
# 1️⃣4️⃣ Iterating Over Sets
# ============================================

fruits = {"apple", "banana", "cherry"}

# Simple for loop (order not guaranteed)
print("--- for loop ---")
for fruit in fruits:
    print(fruit, end=" ")
print()

# enumerate (index is just a counter, not position)
print("\n--- enumerate ---")
for i, fruit in enumerate(fruits):
    print(i, fruit)
```

```python
# ------------------------------------------------------------
# ❌ Example 1: Modifying Set During Iteration (Will Raise Error)
# ------------------------------------------------------------

print("Example 1: Modifying set during iteration\n")

s = {1, 2, 3, 4}

try:
    for item in s:
        if item == 2:
            s.remove(item)   # ❌ Not Allowed
except RuntimeError as e:
    print("Error occurred:", e)

"""
Output:
Error occurred: Set changed size during iteration
"""

print("\n" + "-"*60 + "\n")


# ------------------------------------------------------------
# ✅ Example 2: Safe Method - Iterate Over a Copy
# ------------------------------------------------------------

print("Example 2: Using s.copy()\n")

s = {1, 2, 3, 4}

for item in s.copy():   # ✔ Iterate over copy
    if item == 2:
        s.remove(item)

print("Updated Set:", s)

"""
Output:
Updated Set: {1, 3, 4}
"""

print("\n" + "-"*60 + "\n")


# ------------------------------------------------------------
# ✅ Example 3: Safe Method - Using Set Comprehension (Recommended)
# ------------------------------------------------------------

print("Example 3: Using Set Comprehension\n")

s = {1, 2, 3, 4}

# Remove element 2 safely
s = {x for x in s if x != 2}

print("Updated Set:", s)

"""
Output:
Updated Set: {1, 3, 4}
"""

print("\n" + "-"*60 + "\n")


# ------------------------------------------------------------
# ✅ Example 4: Collect Items First, Then Remove
# ------------------------------------------------------------

print("Example 4: Collect First, Then Remove\n")

s = {1, 2, 3, 4}
to_remove = []

for item in s:
    if item == 2:
        to_remove.append(item)

for item in to_remove:
    s.remove(item)

print("Updated Set:", s)

"""
Output:
Updated Set: {1, 3, 4}
"""


# ------------------------------------------------------------
# 🔥 Important Concept Explanation
# ------------------------------------------------------------

"""
Why does this error happen?

- Python creates an iterator when the loop starts.
- The iterator expects the set size to remain constant.
- If you add() or remove() elements, the internal hash table changes.
- This invalidates the iterator.
- Python raises RuntimeError to prevent unpredictable behavior.

Golden Rule:
👉 Never modify (add/remove) a set while iterating over it.
"""

print("\nProgram Completed Successfully ✅")
```

## 1️⃣6️⃣ Real-World Use Cases

| Use Case | How Sets Help |
|---|---|
| Remove duplicates from a list | `list(set(my_list))` |
| Find common elements between two lists | `set(list1) & set(list2)` |
| Find duplicates in a list | Track seen elements |
| Fast membership testing | `if item in my_set` → O(1) |
| Data validation | Check if required fields are present |

```python
# ============================================
# 1️⃣6️⃣ Real-World Use Cases
# ============================================

# 1) Remove duplicates from a list
nums = [1, 2, 2, 3, 4, 4]
unique = list(set(nums))
print("Unique:", unique)            # [1, 2, 3, 4]

# 2) Find duplicates in a list
seen = set()
duplicates = set()
for n in nums:
    if n in seen:
        duplicates.add(n)
    else:
        seen.add(n)
print("Duplicates:", duplicates)     # {2, 4}

# 3) Fast intersection check between two lists
list1 = ["apple", "banana"]
list2 = ["banana", "cherry"]
print("Common:", set(list1) & set(list2))  # {'banana'}

# 4) Data validation — check missing required fields
required = {"name", "email", "age"}
provided = {"name", "age"}
missing = required - provided
print("Missing fields:", missing)    # {'email'}
```

## 1️⃣7️⃣ Common Mistakes & Edge Cases

| Mistake | Why It's Wrong |
|---|---|
| `{}` for empty set | Creates a `dict`, not a `set`. Use `set()` |
| Modifying set during iteration | Raises `RuntimeError` |
| Expecting ordered output | Sets are unordered |
| Adding unhashable elements | `list`, `dict`, `set` cannot be added |
| `True` and `1` treated as same | `{True, 1}` → `{True}` (same hash) |
| `False` and `0` treated as same | `{False, 0}` → `{False}` (same hash) |

```python
# ============================================
# 1️⃣7️⃣ Common Mistakes & Edge Cases
# ============================================

# ❌ Modifying set during iteration → RuntimeError
try:
    s = {1, 2, 3}
    for i in s:
        s.add(10)
except RuntimeError as e:
    print("RuntimeError:", e)

# ⚠️ True/1 and False/0 are considered EQUAL in sets
# Because hash(1) == hash(True) and 1 == True
s1 = {True, 1, 2, 3}     # 1 is duplicate of True
print(s1)                  # {True, 2, 3}

s2 = {False, 0, 2, 3}    # 0 is duplicate of False
print(s2)                  # {False, 2, 3}

print("hash(1) == hash(True):", hash(1) == hash(True))  # True
```

## 1️⃣8️⃣ Interview Questions

**Q1: What is the difference between `set` and `list`?**
- List: ordered, allows duplicates, supports indexing
- Set: unordered, no duplicates, no indexing, O(1) lookup

**Q2: Why must set elements be hashable?**
- Sets use a hash table internally. Each element's hash determines its bucket. Mutable objects can change → hash would change → breaks the hash table.

**Q3: Difference between `remove()` and `discard()`?**
- `remove()` raises `KeyError` if element not found
- `discard()` does nothing if element not found (safe)

**Q4: What is a `frozenset` and when would you use it?**
- An immutable set. Use it when you need a set as a dictionary key or as an element of another set.

**Q5: How to find duplicates in a list using sets?**
- Iterate through the list, track seen items in a set. If an item is already in `seen`, it's a duplicate.

**Q6: Why does `{True, 1}` give `{True}`?**
- In Python, `True == 1` and `hash(True) == hash(1)`. So the set treats them as the same element.

**Q7: What is the time complexity of `in` operator for sets vs lists?**
- Set: O(1) average (hash table lookup)
- List: O(n) (linear search)

**Q8: Can you have a set of sets?**
- No. Sets are unhashable. Use `frozenset` instead: `{frozenset({1,2}), frozenset({3,4})}`
