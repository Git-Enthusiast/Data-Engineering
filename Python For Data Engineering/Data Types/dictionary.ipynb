## 1️⃣ What is a Dictionary?

A **dictionary** in Python is a mutable collection of **key-value pairs**.

| Property | Details |
|---|---|
| **Key-Value Pairs** | Each item is stored as a `key: value` pair |
| **Unordered $\rightarrow$ Ordered** | Since Python 3.7, dictionaries **maintain insertion order** |
| **Mutable** | You can add, remove, and update key-value pairs |
| **Keys Must Be Hashable** | Keys must be immutable types (`str`, `int`, `tuple`) |
| **No Duplicate Keys** | Keys must be unique (the last assigned value overwrites) |
| **Fast Lookups** | Implemented using a Hash Table $\rightarrow$ O(1) average lookup time |

```python
# ============================================
# 1️⃣ What is a Dictionary?
# ============================================

# A fast mapping of keys to values
person = {"name": "Alice", "age": 25, "role": "Engineer"}
print("Dictionary:", person)
print("Type:", type(person))

# Fast O(1) lookups by key
print("Name is:", person["name"])
```

## 2️⃣ Creating Dictionaries

| Method | Syntax | Example |
|---|---|---|
| Empty | `{}` or `dict()` | `d = {}` |
| Curly Braces | `{k: v}` | `d = {'a': 1, 'b': 2}` |
| `dict()` constructor | `dict(k=v)` | `d = dict(a=1, b=2)` |
| List of Tuples | `dict([(k, v)])` | `d = dict([('a', 1), ('b', 2)])` |
| Using `zip()` | `dict(zip(keys, values))` | `d = dict(zip(['a', 'b'], [1, 2]))` |

```python
# ============================================
# 2️⃣ Creating Dictionaries
# ============================================

# Method 1: Empty Dictionary
empty_dict = {}
print("Empty:", empty_dict)

# Method 2: Curly Braces
dict1 = {'name': 'Alice', 'age': 25}
print("Braces:", dict1)

# Method 3: dict() Constructor (keys must be valid identifiers)
dict2 = dict(name='Bob', age=30)
print("Constructor:", dict2)

# Method 4: List of Tuple Pairs
dict3 = dict([('id', 101), ('status', 'active')])
print("From List:", dict3)

# Method 5: Using zip() — VERY useful!
keys_list = ["x", "y", "z"]
values_list = [10, 20, 30]
dict4 = dict(zip(keys_list, values_list))
print("From zip:", dict4)
```

## 3️⃣ Dictionary Characteristics

> 💡 **Key Rules:**
> 1. **Keys Must Be Hashable (Immutable):** `int`, `float`, `str`, `tuple`, `frozenset` are allowed. `list`, `dict`, `set` are **not allowed**.
> 2. **Values Can Be Anything:** Mutable or immutable (`list`, `dict`, functions, classes, anything).
> 3. **Unique Keys:** Duplicate keys overwrite existing ones.

```python
# ============================================
# 3️⃣ Dictionary Characteristics
# ============================================

# ✔ Valid Immutable Keys
valid_dict = {
    1: "int key",
    3.14: "float key",
    "string_key": "str key",
    (1, 2): "tuple key"    # Typles are immutable, so they work
}
print("Valid keys:", valid_dict.keys())

# ❌ Invalid Mutable Keys
try:
    invalid_dict = {[1, 2]: "List is mutable!"}
except TypeError as e:
    print("KeyError (list):", e)

# ⚠️ Duplicate Keys (Last One Wins)
d_dupe = {'a': 1, 'b': 2, 'a': 99}
print("Duplicates overwritten:", d_dupe)  # {'a': 99, 'b': 2}
```

## 4️⃣ Accessing Elements — `[]` vs `.get()`

| Method | If Key Exists | If Key Does NOT Exist |
|---|---|---|
| `dict[key]` | Returns value | ❌ Raises `KeyError` |
| `dict.get(key)` | Returns value | ✅ Returns `None` (silent) |
| `dict.get(key, default)`| Returns value | ✅ Returns `default` value |

> 💡 **Best Practice:** Use `.get()` if you aren't 100% sure the key exists. It prevents program crashes!

```python
# ============================================
# 4️⃣ Accessing Elements
# ============================================

person = {'name': 'Alice', 'role': 'Admin'}

# 1. Bracket notation []
print("Name:", person['name'])

# Bracket notation crashes if key is missing
try:
    print(person['salary'])
except KeyError as e:
    print("KeyError:", e)

# 2. Safely accessing with .get()
print("get() method:", person.get('role'))

# If missing, get() returns None instead of crashing
print("Missing get():", person.get('salary'))

# You can provide a custom default value
print("Custom default:", person.get('salary', 'Not Provided'))
```

## 5️⃣ Adding and Updating Elements

| Method | Effect |
|---|---|
| `d[key] = value` | Adds a new key or overwrites an existing one |
| `d.update(other_dict)` | Merges another dictionary (or key-value iterable) into `d` |

```python
# ============================================
# 5️⃣ Adding and Updating Elements
# ============================================

user = {'name': 'Bob', 'age': 30}
print("Original:", user)

# ADD a new key
user['city'] = 'New York'

# UPDATE an existing key
user['age'] = 31
print("After add/update:", user)

# update() method — Multiple adds/updates at once
user.update({'role': 'Manager', 'name': 'Robert', 'country': 'USA'})
print("After .update():", user)
```

## 6️⃣ Removing Elements

| Method | Description |
|---|---|
| `.pop(key)` | Removes and returns the value. ❌ `KeyError` if missing |
| `.pop(key, default)`| Safely returns `default` if missing (no error) |
| `.popitem()` | Removes and returns the **last inserted** pair `(key, value)` |
| `del d[key]` | Deletes key. ❌ `KeyError` if missing. Can delete entire dict |
| `.clear()` | Empties the dictionary `{}` |

```python
# ============================================
# 6️⃣ Removing Elements
# ============================================

d = {'name': 'Alice', 'age': 25, 'city': 'London', 'job': 'Dev'}

# pop() — removes key and returns its value
age = d.pop('age')
print(f"Popped 'age': {age} $\rightarrow$ {d}")

# Safe pop with default (if key doesn't exist)
salary = d.pop('salary', 'Unknown')
print(f"Safe pop 'salary': {salary}")

# popitem() — removes last inserted item (LIFO)
last_item = d.popitem()
print(f"Popped last item: {last_item} $\rightarrow$ {d}")

# del keyword
del d['city']
print("After del 'city':", d)

# clear() — empties entire dict
d.clear()
print("After clear():", d)
```

## 7️⃣ Dictionary Views — `keys()`, `values()`, `items()`

These methods return **View Objects** (`dict_keys`, `dict_values`, `dict_items`). 

> 💡 **Dynamic Views:** Views act as a "window" to the dictionary. If the dictionary changes, the view updates automatically instantly!

```python
# ============================================
# 7️⃣ Dictionary Views
# ============================================

d = {'a': 1, 'b': 2, 'c': 3}

keys_view = d.keys()
vals_view = d.values()
items_view = d.items()

print("Keys view:  ", keys_view)
print("Values view:", vals_view)
print("Items view: ", items_view)

print("\n--- Dynamic Nature of Views ---")
d['d'] = 4
print("Added 'd': 4 to dictionary.")

# Notice that the views automatically contain 'd' now, 
# even though we didn't recreate the views!
print("Keys view updated:  ", keys_view)
print("Values view updated:", vals_view)
```

## 8️⃣ Membership Testing — `in` / `not in`

By default, `in` checks for existence among the **KEYS** of the dictionary (extremely fast, O(1)).
To check if a **VALUE** exists, you must explicitly use `in d.values()` (slow, O(n)).

```python
# ============================================
# 8️⃣ Membership Testing
# ============================================

d = {'name': 'Alice', 'role': 'Admin'}

# Checking KEYS (Default O(1))
print("'name' in d?", 'name' in d)         # True
print("'salary' in d?", 'salary' in d)     # False

# ❌ Common Mistake: checking value without .values()
print("\n'Admin' in d?", 'Admin' in d)    # False! (It only looks at keys)

# ✅ Checking VALUES (O(n) linear search)
print("'Admin' in d.values()?", 'Admin' in d.values()) # True
```

## 9️⃣ Iterating Over Dictionaries

| Loop syntax | What it iterates over |
|---|---|
| `for k in d:` | Keys (default behavior) |
| `for k in d.keys():` | Keys (explicit behavior) |
| `for v in d.values():`| Values only |
| `for k, v in d.items():`| Both Key and Value simultaneously (Most common) |

```python
# ============================================
# 9️⃣ Iterating Over Dictionaries
# ============================================

d = {'a': 10, 'b': 20, 'c': 30}

# 1. Iterate over keys
print("Keys:")
for key in d:
    print(f"  {key}")

# 2. Iterate over values
print("Values:")
for val in d.values():
    print(f"  {val}")

# 3. Iterate over items (Key-Value pairs) — Best Practice!
print("Items:")
for key, val in d.items():
    print(f"  {key} $\rightarrow$ {val}")
```

## 🔟 Dictionary Comprehension

A concise way to create dictionaries from iterables, similar to list comprehensions.

**Syntax:** `{key_expr: value_expr for item in iterable if condition}`

```python
# ============================================
# 🔟 Dictionary Comprehension
# ============================================

# Basic: Create a dictionary of squares
squares = {x: x**2 for x in range(1, 6)}
print("Squares:", squares)

# Filtering: Only even squares
even_squares = {x: x**2 for x in range(1, 6) if x % 2 == 0}
print("Even Squares:", even_squares)

# Transforming an existing dictionary (invert keys/values)
d = {'x': 1, 'y': 2, 'z': 3}
inverted = {val: key for key, val in d.items()}
print("Inverted dict:", inverted)

# Creating from two lists using zip()
keys = ['A', 'B', 'C']
vals = [90, 80, 85]
grades = {k: v for k, v in zip(keys, vals)}
print("From zip:", grades)
```

## 1️⃣1️⃣ `setdefault()` and `fromkeys()`

### `d.setdefault(key, default)`
Finds the key and returns its value. **If it doesn't exist**, it creates `key` with the `default` value and returns it. (Extremely useful for grouping data).

### `dict.fromkeys(iterable, default_value)`
Creates a **new** dictionary using elements from the iterable as keys, all assigned to the same `default_value` (usually `None`).

```python
# ============================================
# 1️⃣1️⃣ setdefault() and fromkeys()
# ============================================

d = {'a': 1}

# 1. setdefault() with existing key
val = d.setdefault('a', 0)     # Key exists → returns 1
print("Existing key:", val, d)

# 2. setdefault() with missing key
val2 = d.setdefault('b', 0)    # Key missing → adds 'b':0 and returns 0
print("Missing key:", val2, d)

# 3. Grouping using setdefault() (Common Trick)
words = ['apple', 'bat', 'bar', 'cat']
groups = {}
for w in words:
    first_char = w[0]
    # If letter not in groups, add it with [] as default, then append word
    groups.setdefault(first_char, []).append(w)
print("\nGrouped by first letter:", groups)

# 4. dict.fromkeys()
keys = ['name', 'age', 'role']
new_d = dict.fromkeys(keys, "Unknown")
print("\nFrom keys:", new_d)
```

## 1️⃣2️⃣ Merging Dictionaries (`|` Operator & `**` Unpacking)

How to combine two dictionaries `d1` and `d2` into a new dictionary. **In all cases, if keys overlap, the right-side dictionary (`d2`) overwrites the left-side (`d1`).**

| Method | Syntax | Available In |
|---|---|---|
| **Merge Operator** | `d = d1 \| d2` | Python 3.9+ |
| **Update Operator**| `d1 \|= d2` | Python 3.9+ (In-place update) |
| **Kwargs Unpacking**| `d = {**d1, **d2}` | Python 3.5+ |
| **`.update()`** | `d1.update(d2)` | Python 2.0+ (In-place update) |

```python
# ============================================
# 1️⃣2️⃣ Merging Dictionaries
# ============================================

d1 = {'a': 1, 'b': 2}
d2 = {'b': 99, 'c': 4}  # Overlapping 'b'

# 1. Unpacking (Python 3.5+) — Creates a new dict
merged_unpack = {**d1, **d2}
print("Unpacking **:", merged_unpack)   # 'b' becomes 99

# 2. Merge Operator (Python 3.9+) — Creates a new dict
merged_pipe = d1 | d2
print("Pipe operator |:", merged_pipe)

# 3. In-Place Update Operator (Python 3.9+)
d3 = {'x': 10}
d3 |= {'y': 20}
print("In-place |=:", d3)
```

## 1️⃣3️⃣ Copying Dictionaries (Shallow vs Deep Copy)

> ⚠️ **Tricky Interview Detail!**

- **Assignment `=`:** Does NOT copy. Both variables point to the **same** object in memory.
- **Shallow Copy `.copy()`:** Creates a new dictionary, BUT nested objects (like lists inside the dict) are still shared references!
- **Deep Copy `copy.deepcopy()`:** Creates a completely independent clone, including all nested objects.

```python
# ============================================
# 1️⃣3️⃣ Copying Dictionaries
# ============================================
import copy

original = {
    'name': 'Alice',
    'skills': ['Python', 'SQL']   # Notice the nested list!
}

# Assignment (Reference)
ref = original

# Shallow Copy
shallow = original.copy()

# Deep Copy
deep = copy.deepcopy(original)

# --- Let's make changes ---
# 1. Change top-level key in shallow (Only affects shallow)
shallow['name'] = 'Bob'

# 2. Append to nested list in shallow (Affects original AND shallow!)
shallow['skills'].append('AWS')

# 3. Append to nested list in deep (Affects ONLY deep!)
deep['skills'].append('Java')

print("Original: ", original['name'], "| Skills:", original['skills'])
print("Shallow:  ", shallow['name'], "  | Skills:", shallow['skills'])
print("Deep:     ", deep['name'], "| Skills:", deep['skills'])
```

## 1️⃣4️⃣ Nested Dictionaries

Dictionaries can contain other dictionaries as values. This is essential for structured data like JSON APIs.

```python
# ============================================
# 1️⃣4️⃣ Nested Dictionaries
# ============================================

employees = {
    'E101': {'name': 'Alice', 'role': 'Admin'},
    'E102': {'name': 'Bob', 'role': 'Dev'}
}

# Accessing nested values (chaining brackets)
print("Bob's Role:", employees['E102']['role'])

# Safely accessing nested values using get() to avoid KeyErrors
# If E103 exists, get its 'name', otherwise return 'Unknown'
name = employees.get('E103', {}).get('name', 'Unknown')
print("E103 Name:", name)

# Updating nested dictionary
employees['E101']['role'] = 'Super Admin'
print("Alice's updated role:", employees['E101']['role'])
```

## 1️⃣5️⃣ Built-in Functions

Unless explicitly specified by `.values()`, built-in functions apply to the **KEYS** of the dictionary.

| Function | Description |
|---|---|
| `len(d)` | Number of key-value pairs |
| `max(d)` / `min(d)`| Maximum / Minimum key (sorted alphabetically/numerically) |
| `sorted(d)` | Returns a sorted **list of keys** |
| `sum(d)` | Returns the sum of keys (if keys are numbers) |
| `any(d)` / `all(d)`| Checks truthiness of the keys |

```python
# ============================================
# 1️⃣5️⃣ Built-in Functions
# ============================================

d = {'banana': 3, 'apple': 1, 'orange': 2}

print("Length:", len(d))                 # 3
print("Max Key:", max(d))                # 'orange' (alphabetically last)
print("Min Key:", min(d))                # 'apple'  (alphabetically first)

# sorted() returns a LIST of keys
print("Sorted Keys:", sorted(d))         # ['apple', 'banana', 'orange']

# To target values instead:
print("Max Value:", max(d.values()))     # 3
print("Sorted Values:", sorted(d.values()))  # [1, 2, 3]

# 💡 Common Trick: Sort dictionary BY VALUE (Ascending)
sorted_by_val = dict(sorted(d.items(), key=lambda item: item[1]))
print("Dict sorted by value:", sorted_by_val)
```

## 1️⃣6️⃣ Advanced Dictionaries (`collections` module)

For complex use cases, Python provides specialized dictionaries in the `collections` module.

| Type | Description |
|---|---|
| `defaultdict` | Automatically provides a default value for missing keys (removes need for `.setdefault()` or `.get()`). |
| `Counter` | Extremely fast way to count frequencies of items in an iterable. |
| `OrderedDict` | Remembers order of insertion (Before Python 3.7, normal `dict` did not). Less used now, but still useful for LRU caches. |

```python
# ============================================
# 1️⃣6️⃣ Advanced Dictionaries
# ============================================
from collections import defaultdict, Counter

# 1. defaultdict: Automatically sets default value (list) for missing keys
dd = defaultdict(list)
dd['fruits'].append('apple')
dd['fruits'].append('banana')
print("DefaultDict:", dd['fruits'])
print("Missing key query yields default:", dd['vegetables'])  # []

# 2. Counter: The fastest way to count frequencies
word = "mississippi"
char_counts = Counter(word)
print("\nCounter object:", dict(char_counts))
print("Most common character:", char_counts.most_common(1))  # [('s', 4)]
```

## 1️⃣7️⃣ Real-World Use Cases

Dictionaries are arguably Python's most important data structure.
Typical use cases:
- **Caching/Memoization:** Storing expensive function results (keys = inputs, values = outputs).
- **JSON APIs:** Almost all web API responses map directly to Python dictionaries.
- **Switch/Case Replacements:** Using a dictionary of functions instead of long `if/elif` blocks.

```python
# ============================================
# 1️⃣7️⃣ Real-World Use Cases
# ============================================

# Use Case: Replacing 'Switch/Case' statements
def add(a, b): return a + b
def sub(a, b): return a - b
def mul(a, b): return a * b

operations = {
    '+': add,
    '-': sub,
    '*': mul
}

op_symbol = '+'
x, y = 10, 5

# Lookup the function by symbol, then call it with x and y
if op_symbol in operations:
    result = operations[op_symbol](x, y)
    print(f"{x} {op_symbol} {y} = {result}")  # 10 + 5 = 15
```

## 1️⃣8️⃣ Common Mistakes & Edge Cases

| Mistake | Why It's Wrong |
|---|---|
| Using `list` as a key | Lists are mutable $\rightarrow$ Unhashable $\rightarrow$ `TypeError` |
| Testing values using `in d` | `in` only checks keys. Use `in d.values()` to check values! |
| Mutating dict during literal iteration | Raises `RuntimeError`. Loop over `list(d.keys())` if you need to mutate. |
| `d[True] = 1`, `d[1] = 2` | `True` and `1` hash to the SAME value! `d[1]` overwrites `d[True]`. |

```python
# ============================================
# 1️⃣8️⃣ Common Mistakes & Edge Cases
# ============================================

# ❌ 1. Modifying dictionary size during iteration
try:
    d = {'a': 1, 'b': 2}
    for k in d:
        del d[k]   # Trying to delete while looping
except RuntimeError as e:
    print("RuntimeError:", e)

# ✅ Fix: Iterate over a copy of the keys
d2 = {'a': 1, 'b': 2}
for k in list(d2.keys()):
    del d2[k]
print("Proper deletion worked, dict is:", d2)

# ⚠️ 2. True/1 and False/0 collision
tricky_dict = {
    True: "Yes!",
    1: "Wow!",
    1.0: "Overwritten!"
}
# All three keys hash to the exact same bucket! They overwrite each other.
print("Tricky dict:", tricky_dict)  # {True: 'Overwritten!'}
```

## 1️⃣9️⃣ Interview Questions

**Q1: What gets stored in the dictionary when keys collide (e.g., `True` and `1`)?**
- Because `1 == True` in Python, they evaluate to the same hash. The **first** key object (`True`) is kept, but the **last** value (`"Overwritten!"`) overwrites previous values.

**Q2: Are dictionaries ordered in Python?**
- Before Python 3.7: **No**. Order was random.
- Python 3.7+: **Yes**. They maintain insertion order. (If strict legacy compatibility is needed, use `collections.OrderedDict`).

**Q3: What's the difference between `.copy()` and `copy.deepcopy()`?**
- `.copy()` is a shallow copy — outer dictionary is new, but nested objects are shared.
- `deepcopy()` creates an entire recursive clone so no lists/dicts are shared.

**Q4: What's the Time Complexity of dictionaries vs lists?**
- Dictionary lookups (`d[key]` or `key in d`) are **O(1)** average.
- List lookups (`item in lst`) are **O(n)** linear search.

**Q5: What makes an object hashable?**
- It must have a hash value that never changes during its lifetime (immutable). `int`, `str`, `tuple`, `frozenset` are hashable. `list`, `dict`, `set` are not.
