## 1️⃣ What is a Frozenset?

A `frozenset` is an **immutable** version of a Python `set`. Once created, its elements **cannot be changed** — no adding, removing, or updating.

| Property | Details |
|---|---|
| **Immutable** | Cannot add, remove, or modify elements after creation |
| **Unordered** | Elements have no fixed position — order may vary |
| **No Duplicates** | Automatically removes duplicate values |
| **Hashable** | Can be used as a dictionary key or element of another set |
| **No Indexing/Slicing** | Cannot access elements by position |
| **Supports Set Operations** | Union, intersection, difference, etc. (returns new frozenset) |

> 💡 **Key Insight:** `frozenset` is to `set` what `tuple` is to `list` — an immutable counterpart.

```
Mutable      →  Immutable
list         →  tuple
set          →  frozenset
dict         →  (no built-in equivalent)
```

````python
# ============================================
# 1️⃣ What is a Frozenset?
# ============================================

# A frozenset is immutable — once created, it cannot be changed
fs = frozenset([1, 2, 3])
print("Frozenset:", fs)
print("Type:", type(fs))

# It stores unique elements only
fs2 = frozenset([1, 1, 2, 2, 3])
print("Duplicates removed:", fs2)   # frozenset({1, 2, 3})

# It is hashable — unlike regular sets
print("Hash:", hash(fs))             # works! (sets would raise TypeError)
````

## 2️⃣ Creating Frozensets

Frozensets are created using the `frozenset()` constructor. There is **no literal syntax** (unlike `{}` for sets).

| Method | Syntax | Example |
|---|---|---|
| From List | `frozenset(list)` | `frozenset([1, 2, 3])` |
| From Tuple | `frozenset(tuple)` | `frozenset((1, 2, 3))` |
| From Set | `frozenset(set)` | `frozenset({1, 2, 3})` |
| From String | `frozenset(string)` | `frozenset("hello")` → unique chars |
| From Range | `frozenset(range)` | `frozenset(range(5))` |
| Empty | `frozenset()` | Creates an empty frozenset |

> ⚠️ **No curly brace syntax:** You **cannot** write `frozen{1, 2, 3}`. Always use `frozenset()`.

````python
# ============================================
# 2️⃣ Creating Frozensets
# ============================================

# From a list
fs1 = frozenset([1, 2, 3, 3])
print("From list:", fs1)             # frozenset({1, 2, 3})

# From a tuple
fs2 = frozenset((10, 20, 30))
print("From tuple:", fs2)            # frozenset({10, 20, 30})

# From a set
fs3 = frozenset({4, 5, 6})
print("From set:", fs3)              # frozenset({4, 5, 6})

# From a string (each unique character)
fs4 = frozenset("hello")
print("From string:", fs4)           # frozenset({'h', 'e', 'l', 'o'})

# From range
fs5 = frozenset(range(5))
print("From range:", fs5)            # frozenset({0, 1, 2, 3, 4})

# Empty frozenset
fs_empty = frozenset()
print("Empty:", fs_empty)            # frozenset()
print("Type:", type(fs_empty))       # <class 'frozenset'>
````

## 3️⃣ Frozenset vs Set — Key Differences

| Feature | `set` | `frozenset` |
|---|---|---|
| **Mutable?** | ✅ Yes | ❌ No |
| **Hashable?** | ❌ No | ✅ Yes |
| **Can be dict key?** | ❌ No | ✅ Yes |
| **Can be in another set?** | ❌ No | ✅ Yes |
| **Literal syntax?** | ✅ `{1, 2}` | ❌ Must use `frozenset()` |
| **`add()` / `remove()`?** | ✅ Yes | ❌ No |
| **`update()` / `discard()`?** | ✅ Yes | ❌ No |
| **Set operations?** | ✅ Yes | ✅ Yes |
| **Membership (`in`)?** | ✅ O(1) | ✅ O(1) |
| **Internal structure** | Hash Table | Hash Table |

> 💡 **Rule of thumb:** If your set data should **never change**, use `frozenset`. If you need to add/remove elements, use `set`.

## 4️⃣ Immutability in Action

Since frozensets are immutable, **none of the modifying methods** available on sets exist on frozensets.

| Method | Available on `set`? | Available on `frozenset`? |
|---|---|---|
| `add()` | ✅ | ❌ |
| `remove()` | ✅ | ❌ |
| `discard()` | ✅ | ❌ |
| `pop()` | ✅ | ❌ |
| `clear()` | ✅ | ❌ |
| `update()` | ✅ | ❌ |
| `union()` | ✅ | ✅ (returns new) |
| `intersection()` | ✅ | ✅ (returns new) |
| `copy()` | ✅ | ✅ |

````python
# ============================================
# 4️⃣ Immutability in Action
# ============================================

fs = frozenset([1, 2, 3])

# ❌ Cannot add elements
try:
    fs.add(4)
except AttributeError as e:
    print("add() Error:", e)

# ❌ Cannot remove elements
try:
    fs.remove(1)
except AttributeError as e:
    print("remove() Error:", e)

# ❌ Cannot discard elements
try:
    fs.discard(1)
except AttributeError as e:
    print("discard() Error:", e)

# ❌ Cannot pop elements
try:
    fs.pop()
except AttributeError as e:
    print("pop() Error:", e)

# ❌ Cannot clear
try:
    fs.clear()
except AttributeError as e:
    print("clear() Error:", e)

# ✅ The frozenset remains unchanged
print("\nFrozenset unchanged:", fs)
````

## 5️⃣ Frozenset Operations (Union, Intersection, Difference, Symmetric Difference)

Frozensets support **all mathematical set operations**. The result is always a **new frozenset** — the originals are never modified.

| Operation | Operator | Method | Description |
|---|---|---|---|
| **Union** | `A \| B` | `A.union(B)` | All elements from both |
| **Intersection** | `A & B` | `A.intersection(B)` | Common elements only |
| **Difference** | `A - B` | `A.difference(B)` | In A but NOT in B |
| **Symmetric Diff** | `A ^ B` | `A.symmetric_difference(B)` | In A or B, but NOT both |

````python
# ============================================
# 5️⃣ Frozenset Operations
# ============================================

A = frozenset([1, 2, 3])
B = frozenset([3, 4, 5])

# Union — all elements from both
print("Union:", A | B)                    # frozenset({1, 2, 3, 4, 5})
print("Union (method):", A.union(B))

# Intersection — common elements
print("Intersection:", A & B)              # frozenset({3})
print("Intersection (method):", A.intersection(B))

# Difference — in A but not in B
print("Difference A-B:", A - B)            # frozenset({1, 2})
print("Difference B-A:", B - A)            # frozenset({4, 5})

# Symmetric Difference — in A or B, but not both
print("Symmetric Diff:", A ^ B)            # frozenset({1, 2, 4, 5})

# All results are frozensets (originals unchanged)
print("\nOriginal A:", A)
print("Original B:", B)
````

## 6️⃣ Membership Testing — `in` / `not in`

Just like sets, frozensets support **O(1) average** membership testing using `in` and `not in`.

> 💡 This is because frozensets also use a **hash table** internally.

````python
# ============================================
# 6️⃣ Membership Testing
# ============================================

fs = frozenset([10, 20, 30, 40, 50])

print(20 in fs)        # True
print(99 in fs)        # False
print(99 not in fs)    # True

# Fast lookup — O(1) average time
# Same performance as regular sets
````

## 7️⃣ Subset, Superset & Disjoint

| Method | Operator | Description |
|---|---|---|
| `A.issubset(B)` | `A <= B` | Is every element of A also in B? |
| `A < B` | — | Is A a **proper** subset? (A ≠ B) |
| `A.issuperset(B)` | `A >= B` | Does A contain every element of B? |
| `A > B` | — | Is A a **proper** superset? (A ≠ B) |
| `A.isdisjoint(B)` | — | Do A and B share **no** common elements? |

````python
# ============================================
# 7️⃣ Subset, Superset & Disjoint
# ============================================

A = frozenset([1, 2])
B = frozenset([1, 2, 3, 4])
C = frozenset([5, 6])

# Subset
print("A subset of B:", A.issubset(B))     # True
print("A <= B:", A <= B)                    # True
print("A < B (proper):", A < B)             # True

# Superset
print("B superset of A:", B.issuperset(A))  # True
print("B >= A:", B >= A)                    # True

# Disjoint — no common elements
print("A disjoint C:", A.isdisjoint(C))     # True
print("A disjoint B:", A.isdisjoint(B))     # False
````

## 8️⃣ Iterating Over Frozensets

Frozensets are **iterable** — you can loop through them just like sets.

> ⚠️ Since frozensets are **unordered**, iteration order is not guaranteed.

````python
# ============================================
# 8️⃣ Iterating Over Frozensets
# ============================================

fs = frozenset(["apple", "banana", "cherry"])

# Simple for loop
print("--- for loop ---")
for item in fs:
    print(item)

# With enumerate
print("\n--- enumerate ---")
for i, item in enumerate(fs, start=1):
    print(f"{i}. {item}")
````

## 9️⃣ Frozenset as Dictionary Key

Since `frozenset` is **hashable**, it can be used as a **dictionary key** — something regular `set` **cannot** do.

This is one of the **primary reasons** `frozenset` exists.

> 💡 **Use case:** Mapping combinations of items to values — e.g., grouping permissions, ingredient sets, etc.

````python
# ============================================
# 9️⃣ Frozenset as Dictionary Key
# ============================================

# ❌ Regular set CANNOT be a dict key
try:
    d = {{1, 2}: "value"}
except TypeError as e:
    print("Set as key Error:", e)

# ✅ Frozenset CAN be a dict key
permissions = {
    frozenset(["read", "write"]): "Editor",
    frozenset(["read"]): "Viewer",
    frozenset(["read", "write", "admin"]): "Admin"
}

# Lookup by frozenset key
user_perms = frozenset(["read", "write"])
print("Role:", permissions[user_perms])    # Editor

print("\nAll roles:")
for perms, role in permissions.items():
    print(f"  {role}: {set(perms)}")
````

## 🔟 Frozenset Inside a Set (Nested Sets)

Regular sets **cannot** be elements of another set (because they are unhashable). But frozensets **can**!

This allows you to create **sets of sets** — useful for graph algorithms, combinations, etc.

```
❌  { {1,2}, {3,4} }           → TypeError (set is unhashable)
✅  { frozenset({1,2}), frozenset({3,4}) }  → works!
```

````python
# ============================================
# 🔟 Frozenset Inside a Set (Nested Sets)
# ============================================

# ❌ Set inside a set — NOT allowed
try:
    nested = {{1, 2}, {3, 4}}
except TypeError as e:
    print("Set in set Error:", e)

# ✅ Frozenset inside a set — works!
nested = {frozenset([1, 2]), frozenset([3, 4]), frozenset([1, 2])}
print("Nested set:", nested)       # duplicate frozenset({1,2}) removed

# ✅ Frozenset inside a frozenset
deep = frozenset([frozenset([1, 2]), frozenset([3, 4])])
print("Deep frozen:", deep)
````

## 1️⃣1️⃣ Mixed Operations (frozenset + set)

You can perform set operations between a `frozenset` and a regular `set`. The **return type** depends on which is on the **left side**:

| Expression | Return Type |
|---|---|
| `frozenset \| set` | `frozenset` |
| `set \| frozenset` | `set` |
| `frozenset.union(set)` | `frozenset` |
| `set.union(frozenset)` | `set` |

> 💡 The **left operand's type** determines the result type.

````python
# ============================================
# 1️⃣1️⃣ Mixed Operations (frozenset + set)
# ============================================

fs = frozenset([1, 2, 3])
s = {3, 4, 5}

# frozenset on LEFT → result is frozenset
result1 = fs | s
print("fs | s:", result1, type(result1).__name__)

# set on LEFT → result is set
result2 = s | fs
print("s | fs:", result2, type(result2).__name__)

# Same with methods
result3 = fs.union(s)
print("fs.union(s):", result3, type(result3).__name__)

result4 = s.union(fs)
print("s.union(fs):", result4, type(result4).__name__)

# Intersection mixed
result5 = fs & s
print("\nfs & s:", result5, type(result5).__name__)
````

## 1️⃣3️⃣ Converting Between Set and Frozenset

| Conversion | How |
|---|---|
| Set → Frozenset | `frozenset(my_set)` |
| Frozenset → Set | `set(my_frozenset)` |

> 💡 These always create a **new object** — the original is unchanged.

````python
# ============================================
# 1️⃣3️⃣ Converting Between Set and Frozenset
# ============================================

# Set → Frozenset
my_set = {1, 2, 3}
my_fs = frozenset(my_set)
print("Set to Frozenset:", my_fs, type(my_fs).__name__)

# Frozenset → Set
back_to_set = set(my_fs)
print("Frozenset to Set:", back_to_set, type(back_to_set).__name__)

# Modifying the new set doesn't affect the frozenset
back_to_set.add(99)
print("\nModified set:", back_to_set)
print("Original frozenset:", my_fs)  # unchanged
````

## 1️⃣4️⃣ Copying Frozensets

Since frozensets are **immutable**, copying behaves differently than with mutable objects:

| Method | What Happens |
|---|---|
| `=` assignment | Both names point to the **same** object (but this is fine — it's immutable!) |
| `.copy()` | Returns the **same** object (Python optimizes since it can't change) |
| `frozenset(fs)` | Also returns the **same** object |

> 💡 Python doesn't waste memory copying immutable objects — it just reuses the same one.

````python
# ============================================
# 1️⃣4️⃣ Copying Frozensets
# ============================================

fs = frozenset([1, 2, 3])

# Assignment
alias = fs
print(alias)
print("fs is alias:", fs is alias)          # True (same object)

# .copy() — Python returns the SAME object (optimization)
copied = fs.copy()
print(copied)
print("fs is copied:", fs is copied)        # True!

# frozenset() constructor — also same object
constructed = frozenset(fs)
print(constructed)
print("fs is constructed:", fs is constructed)  # True!

# This is safe because frozenset is immutable
# — no one can ever change it, so sharing is fine
````

## 1️⃣5️⃣ Internal Working — Hash Table & Time Complexity

Like `set`, `frozenset` is implemented using a **hash table**.

### ⏱️ Time Complexity

| Operation | Average |
|---|---|
| `x in fs` | O(1) |
| `len(fs)` | O(1) |
| `union` | O(len(A) + len(B)) |
| `intersection` | O(min(len(A), len(B))) |
| `difference` | O(len(A)) |
| `issubset` | O(len(A)) |

> 💡 Since frozensets are immutable and hashable, Python can **cache their hash value** — making them slightly more efficient as dict keys compared to computing hashes repeatedly.

````python
# ============================================
# 1️⃣5️⃣ Internal Working
# ============================================

fs = frozenset([10, 20, 30])

# hash() works on frozenset (unlike regular set)
print("hash(frozenset):", hash(fs))

# hash() does NOT work on regular set
try:
    hash({10, 20, 30})
except TypeError as e:
    print("hash(set) Error:", e)

# The hash is computed once and cached internally
# This makes frozensets very efficient as dict keys
````

## 1️⃣6️⃣ Real-World Use Cases

| Use Case | Why Frozenset? |
|---|---|
| **Dictionary keys** | When you need a set-like key (e.g., permission groups, tag combinations) |
| **Caching/Memoization** | Frozensets can be used as cache keys for function arguments |
| **Graph algorithms** | Represent edges as frozenset pairs: `frozenset({A, B})` |
| **Config constants** | Define immutable sets of allowed values |
| **Deduplication of sets** | Put frozensets in a set to remove duplicate groups |

````python
# ============================================
# 1️⃣6️⃣ Real-World Use Cases
# ============================================

# 1) Undirected graph edges — order doesn't matter
edges = {
    frozenset({"A", "B"}),
    frozenset({"B", "C"}),
    frozenset({"A", "B"}),   # duplicate — auto removed
}
print("Unique edges:", edges)

# 2) Immutable allowed values
VALID_STATUSES = frozenset(["active", "inactive", "pending"])
status = "active"
print(f"'{status}' valid?", status in VALID_STATUSES)

# 3) Deduplication of groups
groups = [
    frozenset([1, 2, 3]),
    frozenset([3, 2, 1]),   # same as above (order doesn't matter)
    frozenset([4, 5, 6]),
]
unique_groups = set(groups)
print("Unique groups:", unique_groups)
````

## 1️⃣7️⃣ Common Mistakes & Edge Cases

| Mistake | Why It's Wrong |
|---|---|
| Trying `fs.add()` / `fs.remove()` | Frozensets are immutable — these methods don't exist |
| Using `{frozen}` syntax | No literal syntax for frozenset — must use `frozenset()` |
| `frozenset({})` | Creates `frozenset()` (empty), not an error — but `{}` is a dict |
| Expecting consistent order | Frozensets are unordered |
| `True` and `1` are same | `frozenset([True, 1])` → `frozenset({True})` |

````python
# ============================================
# 1️⃣7️⃣ Common Mistakes & Edge Cases
# ============================================

# Edge case: True/1 and False/0 are same
fs1 = frozenset([True, 1, 2, 3])
print("True and 1:", fs1)        # frozenset({True, 2, 3})

fs2 = frozenset([False, 0, 2, 3])
print("False and 0:", fs2)       # frozenset({False, 2, 3})

# Edge case: frozenset from empty dict
fs3 = frozenset({})
print("From empty dict:", fs3)    # frozenset() — empty

# Edge case: frozenset from dict (uses KEYS only)
fs4 = frozenset({"a": 1, "b": 2})
print("From dict:", fs4)          # frozenset({'a', 'b'}) — keys!

# Edge case: single element
fs5 = frozenset([42])            # list with one element
print("Single element:", fs5)

# ❌ Common mistake: passing non-iterable
try:
    frozenset(42)                 # int is not iterable
except TypeError as e:
    print("Error:", e)
````

## 1️⃣8️⃣ Interview Questions

**Q1: What is a frozenset and how is it different from a set?**
- A `frozenset` is an **immutable** version of a set. It cannot be modified after creation. Unlike `set`, it is **hashable** and can be used as a dictionary key or element of another set.

**Q2: Why does Python need frozenset when we already have set?**
- Because sometimes we need an **immutable, hashable** collection of unique elements — for example, as dictionary keys, cache keys, or elements of other sets.

**Q3: Can you have a set of sets in Python?**
- No. Sets are unhashable. But you can have a **set of frozensets**: `{frozenset({1,2}), frozenset({3,4})}`

**Q4: What happens when you do `frozenset | set`?**
- It returns a **frozenset** (left operand's type wins). `set | frozenset` returns a **set**.

**Q5: Does `.copy()` on a frozenset create a new object?**
- No! Python optimizes by returning the **same** object, since frozensets are immutable and cannot change.

**Q6: What is the relationship between frozenset and tuple?**
- Both are immutable. `tuple` is an immutable version of `list`, while `frozenset` is an immutable version of `set`. However, tuples are ordered and allow duplicates; frozensets are unordered and unique.

**Q7: What methods are available on frozenset but NOT on set?**
- None. Frozenset has a **subset** of set's methods — it has all the read-only and operation methods, but none of the mutating ones (`add`, `remove`, `discard`, `pop`, `clear`, `update`).

**Q8: Can a frozenset contain mutable objects?**
- No. Like sets, frozensets require all elements to be **hashable** (immutable). Lists, dicts, and regular sets cannot be elements.
