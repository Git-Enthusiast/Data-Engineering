# 🔬 Python Object Model — "Everything is an Object"

This is **one of the most important Python concepts**.

---

## 1️⃣ What Does "Everything is an Object" Mean?

In Python:
- Numbers are objects
- Strings are objects
- Functions are objects
- Classes are objects
- Even `None`, `True`, `False` are objects
- Even classes themselves are objects

Every object has:
- A **type**
- A **memory address**
- **Methods**
- **Attributes**

````python
# Proof: Everything is an object
print(type(10))       # <class 'int'>
print(type("hello"))  # <class 'str'>
print(type(True))     # <class 'bool'>
print(type(None))     # <class 'NoneType'>
````

---
## 2️⃣ What is an Object Internally?

In Python, every object has **three properties**:

| Property | What It Is | How to Check |
|----------|------------|-------------|
| **Identity** | Memory address | `id(x)` |
| **Type** | What kind of object | `type(x)` |
| **Value** | The data it holds | `print(x)` |

````python
# Three properties of every object
x = 10

print("Identity (id):", id(x))    # Memory address
print("Type:", type(x))           # <class 'int'>
print("Value:", x)                # 10
````

---
## 3️⃣ Identity (`id()`) & Integer Caching

Python **caches small integers** from **-5 to 256** for optimization.

| Range | `a is b` Result | Reason |
|-------|-----------------|--------|
| `-5` to `256` | `True` | Cached/shared objects |
| Outside range | Might be `False` | New objects may be created |

````python
# Integer caching
a = 100
b = 100
print("a is b (100):", a is b)    # True — cached

a = 1000
b = 1000
print("a is b (1000):", a is b)   # Might be False — outside cache range
````

## 4️⃣ Python Memory Structure (Simplified)

Every Python object internally contains:

```
┌─────────────────────────────┐
│  Reference Count            │
│  Type Pointer               │
│  Actual Value Data          │
└─────────────────────────────┘
```

In CPython (C structure):
```c
typedef struct {
    PyObject_HEAD
} PyObject;
```

Every object begins with:
- **Reference count** (how many variables point to it)
- **Pointer to its type**

## 5️⃣ Everything Inherits from `object`

The base of **everything** in Python is the `object` class.

```
object
   ↑
  int
   ↑
 bool
```

````python
# MRO (Method Resolution Order) — proves inheritance from object
print("int MRO:", int.__mro__)
print("str MRO:", str.__mro__)
print("bool MRO:", bool.__mro__)
````

---
## 6️⃣ Functions Are Objects Too (First-Class Functions)

Functions can:
- Be **passed as arguments**
- Be **returned** from functions
- Be **stored in variables**

This is called **First-Class Functions**.

````python
# Functions are objects
def greet():
    print("Hello")

print("type(greet):", type(greet))   # <class 'function'>

# Assign function to variable
x = greet
x()         # Calls greet() through variable x
````

---
## 7️⃣ Classes Are Also Objects

Classes are created by `type`. And even `type` itself is an instance of `type`!

```
type is instance of type  (mind-blowing! 🤯)
```

````python
# Classes are objects
class Person:
    pass

print("type(Person):", type(Person))   # <class 'type'>
print("type(type):", type(type))       # <class 'type'> — type is instance of itself!
````

---
## 8️⃣ The Meta-Level

- `int` is an object → created by `type`
- `type` is also an object → and is instance of **itself**

### Complete Hierarchy:
```
object           ← base class of everything
   ↑
 type            ← metaclass that creates classes
   ↑
int, str, list   ← actual types/classes
```

````python
# Meta-level: type of built-in types
print("type(int):", type(int))     # <class 'type'>
print("type(str):", type(str))     # <class 'type'>
print("type(list):", type(list))   # <class 'type'>
print("type(type):", type(type))   # <class 'type'>
````

---
## 9️⃣ Mutability in Object Model

Objects can be **immutable** or **mutable**:

| Immutable (changing creates new object) | Mutable (same object modified) |
|----------------------------------------|-------------------------------|
| `int` | `list` |
| `float` | `dict` |
| `bool` | `set` |
| `str` | |
| `tuple` | |

````python
# Mutable example: list (same ID after modification)
x = [1, 2]
print("Before append — id:", id(x))

x.append(3)
print("After append  — id:", id(x))   # Same ID! Object was modified in-place.

print()

# Immutable example: int (different ID after "modification")
y = 10
print("Before change — id:", id(y))

y = y + 1
print("After change  — id:", id(y))   # Different ID! New object created.
````

## 🔟 Variable Is NOT the Object

Variables are just **references (labels/pointers)** to objects.

```python
x = 10     # x → points to object 10
x = 20     # x → now points to object 20 (old object may be garbage collected)
```

> The variable does NOT hold the value — it **points** to the object that holds the value.

````python
# Variables are references
x = 10
print("x = 10 → id:", id(x))

x = 20
print("x = 20 → id:", id(x))   # Different id — x now points to a new object
````

---
## 1️⃣1️⃣ Reference Counting

Python uses **Reference Counting + Garbage Collection** for memory management.

Every object tracks **how many references** point to it.

When ref count becomes `0` → **object destroyed**.

````python
# Reference counting
import sys

x = 10
print("Reference count of x:", sys.getrefcount(x))
# Note: getrefcount() itself creates a temporary reference, so count is +1
````

---
## 1️⃣2️⃣ Dynamic Typing

In Python, the **variable has no fixed type** — the **object** has the type.

```python
x = 10       # x refers to an int object
x = "hello"  # now x refers to a str object
```

This is called **Duck Typing + Dynamic Typing**.

````python
# Dynamic typing
x = 10
print("x =", x, "→ type:", type(x))

x = "hello"
print("x =", x, "→ type:", type(x))

# Same variable, different types — because the OBJECT has the type, not the variable.
````

---
## 1️⃣3️⃣ Attribute Access

Since everything is an object, even numbers have **methods and attributes**.

Use `dir()` to see all methods of any object.

````python
# Attribute access — even numbers have methods
print("bit_length of 10:", (10).bit_length())

# See all methods of int
print("\nAll methods of int object:")
print(dir(10))
````

## 1️⃣4️⃣ Why Python Chose This Model?

| Feature | Java | Python |
|---------|------|--------|
| Primitives vs Objects | `int ≠ Object` | `int IS object` |
| Design | Split (primitives + wrappers) | Unified |
| Consistency | Two systems to learn | Everything behaves similarly |

Python chose this model for:
- **Uniformity**
- **Simplicity**
- **Consistency**
- Cleaner design — no primitives vs objects difference
