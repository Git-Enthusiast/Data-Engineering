üîπ SCENARIO 1: Python REUSES IMMUTABLE OBJECTS (int, str)
Example 1Ô∏è‚É£ Integer Reuse (Object Interning)


```python
a = 10
b = 10

print(a == b)
print(id(a))
print(id(b))

```

‚úÖ Output (example):
True
140724812345456
140724812345456

üß† What happened?

10 is an immutable integer

Python reuses the same memory

Both a and b point to the same object

üìå Python does this for performance (called interning)

Example 2Ô∏è‚É£ Integer Change ‚Üí NEW Object


```python
a = 10
print("Before:", id(a))

a = a + 5
print("After :", id(a))

```

‚úÖ Output:
Before: 140724812345456
After : 140724812345616

üß† Explanation:

a + 5 does NOT modify 10

Python creates new integer 15

a now points to a new memory

‚úî Old 10 still exists if referenced somewhere else

üîπ SCENARIO 2: Python MUST NOT REUSE MUTABLE OBJECTS
Example 5Ô∏è‚É£ List ‚Üí NEVER Reused Automatically


```python
l1 = [1, 2, 3]
l2 = [1, 2, 3]

print(l1 == l2)
print(id(l1))
print(id(l2))

```

‚úÖ Output:
True
140724910345280
140724910346112

üß† Explanation:

Values are same

But lists are mutable

Python creates separate memory

Reuse would cause data corruption

Example 6Ô∏è‚É£ Why Python CANNOT Reuse Mutable Objects


```python
l1 = [1, 2, 3]
l2 = [1, 2, 3]

l1.append(100)

print("l1:", l1)
print("l2:", l2)

```

‚úÖ Output:
l1: [1, 2, 3, 100]
l2: [1, 2, 3]

üß† Imagine if Python reused memory:

‚ùå l2 would also change
‚ùå Programs would break
‚ùå Unpredictable bugs

So Python never reuses mutable objects automatically

üîπ SCENARIO 3: MUTABLE OBJECTS + REFERENCE = SIDE EFFECTS
Example 7Ô∏è‚É£ Same List Reference


```python
a = [10, 20, 30]
b = a

print(id(a))
print(id(b))

b.append(40)
print("a:", a)
print("b:", b)

```

‚úÖ Output:
140724920445600
140724920445600
a: [10, 20, 30, 40]
b: [10, 20, 30, 40]

üß† What happened?

a and b refer to same object

Mutation through b affects a

This is intentional behavior

üìå Python did NOT reuse memory automatically
üìå YOU explicitly reused it using b = a

Example 8Ô∏è‚É£ Safe Copy of Mutable Object


```python
a = [1, 2, 3]
b = a.copy()

b.append(99)

print("a:", a)
print("b:", b)

```

‚úÖ Output:
a: [1, 2, 3]
b: [1, 2, 3, 99]

üß† Explanation:

copy() creates new memory

Prevents side effects

üîπ SCENARIO 4: Function Calls (REAL-WORLD BUG SOURCE)
Example 9Ô∏è‚É£ Immutable in Function (SAFE)


```python
def add_one(x):
    x = x + 1
    print("Inside function:", id(x))

n = 10
print("Before function:", id(n))
add_one(n)
print("After function :", id(n))

```

‚úÖ Output:
Before function: 140724812345456
Inside function: 140724812345520
After function : 140724812345456

üß† Conclusion:

Immutable objects are safe

Function creates new object

Original value unchanged

Example üîü Mutable in Function (DANGEROUS)


```python
def add_item(lst):
    lst.append(100)

nums = [1, 2, 3]
add_item(nums)

print(nums)

```

‚úÖ Output:
[1, 2, 3, 100]

üß† What happened?

Function modified the same list

Original list changed

This is a common interview + real bug

üî• FINAL GOLDEN RULES (REMEMBER THIS)
üü¢ Immutable

Python may reuse memory

Modification ‚Üí new object

Safe in functions

üî¥ Mutable

Never reused automatically

Modified in-place
