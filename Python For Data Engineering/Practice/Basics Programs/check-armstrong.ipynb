# ðŸ”¢ What is an Armstrong Number?

## An Armstrong number (also called a Narcissistic number) is a number that is equal to the sum of its own digits raised to the power of the number of digits.

---

## ðŸ“Œ Definition

If a number has **n digits**, then:

\[
Number=âˆ‘(each digit)n
\]

---

## âœ… Example 1: 153

**Number = 153**  
**Digits = 3**

Now calculate:

\[
1^3 + 5^3 + 3^3
\]

\[
= 1 + 125 + 27
\]

\[
= 153
\]

âœ” Since result = original number  
ðŸ‘‰ **153 is an Armstrong number**

---

## âœ… Example 2: 9474

**Digits = 4**

\[
9^4 + 4^4 + 7^4 + 4^4
\]

\[
= 6561 + 256 + 2401 + 256
\]

\[
= 9474
\]

âœ” **9474 is an Armstrong number**

---

## âŒ Example: 123

**Digits = 3**

\[
1^3 + 2^3 + 3^3
\]

\[
= 1 + 8 + 27
\]

\[
= 36
\]

\[
36 != 123
\]

âŒ **123 is NOT an Armstrong number**


```python
# ðŸ§® Method 1: Using loops (manual calculation, no string conversion)

def count_digit(number):
    """Count number of digits in the number"""
    if number == 0:
        return 1
    number = abs(number)
    count = 0
    while number > 0:
        number //= 10
        count += 1
    return count

def calc_Power(digit, power):
    """Manually calculate digit^power"""
    total = 1
    for _ in range(power):
        total *= digit
    return total

def check_armstrong_loop(my_number):
    """Check Armstrong number using loop and manual calculation"""
    number = my_number
    total_sum = 0
    digits_count = count_digit(my_number)  # calculate once
    
    while number > 0:
        digit = number % 10
        total_sum += calc_Power(digit, digits_count)
        number //= 10
    
    return my_number == total_sum

# âœ… Test Method 1
test_numbers = [153, 9474, 123, 0]
for num in test_numbers:
    print(f"{num} -> {check_armstrong_loop(num)}")
```

```python
# ðŸ Method 2: Pythonic method using string conversion

def check_armstrong_pythonic(number):
    """Check Armstrong number using string conversion"""
    digits = str(abs(number))
    n = len(digits)
    total_sum = sum(int(d)**n for d in digits)
    return total_sum == number

# âœ… Test Method 2
for num in test_numbers:
    print(f"{num} -> {check_armstrong_pythonic(num)}")
```

# ðŸ”¹ Key Points

1. Armstrong number = sum of digits raised to the power of number of digits.  
2. Loop method:
   - Manual digit extraction using `% 10` and `// 10`  
   - Manual power calculation (good for learning fundamentals)
3. Pythonic method:
   - Convert number to string  
   - Use list comprehension and sum  
   - Cleaner, shorter, but less fundamental
4. Works for **0** and positive numbers; negative numbers are usually not considered Armstrong numbers.
