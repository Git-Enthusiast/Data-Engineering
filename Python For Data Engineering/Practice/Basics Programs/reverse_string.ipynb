#    Reverse a string using two pointer method

"""
We can not modify the string as string is immutable in python 
so first we convert string to list and then reverse the list
and back then using .join() we will convert list back to String

"""

```python
def reverse_list(my_list):
    left = 0
    right = len(my_list)-1
    while(left<right):
        temp = my_list[left]
        my_list[left] = my_list[right]
        my_list[right] = temp
        left += 1
        right -= 1
    return my_list

my_string = input("Please Enter a String to reverse: \n")
my_string = my_string.replace(" ","").lower()
my_list = list(my_string)
rev_String = ""
print(f"The Original String is: {my_string}")
print(f"The reversed String is: {rev_String.join(reverse_list(my_list))}")
```

```python
def reverse_string(my_string):
    my_list = list(my_string)   # convert to list (mutable)
    
    left = 0
    right = len(my_list) - 1
    
    while left < right:
        my_list[left], my_list[right] = my_list[right], my_list[left]
        left += 1
        right -= 1
    
    return ''.join(my_list)   # convert back to string

my_string = input("Please Enter a String to reverse") 
my_string = my_string.replace(" ","").lower() 
print(f"The Original String is: {my_string}")
print(f"The reversed String is: {reverse_string(my_string)}")
```
