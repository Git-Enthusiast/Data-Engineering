# PERIMETERS OF COMMON SHAPES

```python
import math

# 1️⃣ PERIMETER OF CIRCLE
# A circle is a shape with all points equidistant from the center.
# Radius = r
# Circumference / Perimeter = 2 * π * r
r = float(input("Enter radius of circle: "))
perimeter_circle = 2 * math.pi * r
print("Perimeter (Circumference) of Circle =", perimeter_circle)

# ------------------------------------------
```

```python
import math

# 2️⃣ PERIMETER OF EQUILATERAL TRIANGLE
# All three sides are equal.
# Side = a
# Perimeter = 3 * a
a = float(input("Enter side of equilateral triangle: "))
perimeter_eq_triangle = 3 * a
print("Perimeter of Equilateral Triangle =", perimeter_eq_triangle)

# ------------------------------------------
```

```python
import math

# 3️⃣ PERIMETER OF PARALLELOGRAM
# Opposite sides are equal
# Length = l, Width = w
# Perimeter = 2 * (l + w)
l = float(input("Enter length of parallelogram: "))
w = float(input("Enter width of parallelogram: "))
perimeter_parallelogram = 2 * (l + w)
print("Perimeter of Parallelogram =", perimeter_parallelogram)

# ------------------------------------------
```

```python
import math

# 4️⃣ PERIMETER OF RECTANGLE
# Opposite sides are equal
# Length = l, Width = w
# Perimeter = 2 * (l + w)
l_rect = float(input("Enter length of rectangle: "))
w_rect = float(input("Enter width of rectangle: "))
perimeter_rectangle = 2 * (l_rect + w_rect)
print("Perimeter of Rectangle =", perimeter_rectangle)

# ------------------------------------------
```

```python
import math

# 5️⃣ PERIMETER OF SQUARE
# All sides are equal
# Side = s
# Perimeter = 4 * s
s = float(input("Enter side of square: "))
perimeter_square = 4 * s
print("Perimeter of Square =", perimeter_square)

# ------------------------------------------
```

```python
import math

# 6️⃣ PERIMETER OF RHOMBUS
# All sides are equal
# Side = a
# Perimeter = 4 * a
a_rhombus = float(input("Enter side of rhombus: "))
perimeter_rhombus = 4 * a_rhombus
print("Perimeter of Rhombus =", perimeter_rhombus)
```
