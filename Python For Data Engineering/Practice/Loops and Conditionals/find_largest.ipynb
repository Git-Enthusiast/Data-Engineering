# Take integer inputs till the user enters 0 and print the largest number from all.

```python
all_input = []

while True:
    try:
        number = int(input("Please Enter integer (0 to exit): "))
        
        if number == 0:
            break
        
        all_input.append(number)

    except ValueError:
        print("Invalid Input. Please Enter Integer Only.")

if all_input:
    print(f"The largest among all is: {max(all_input)}")
else:
    print("No numbers were entered.")
```

```python
largest = float('-inf')

while True:
    try:
        number = int(input("Please Enter integer (0 to exit): "))
        
        if number == 0:
            break
        
        if number > largest:
            largest = number

    except ValueError:
        print("Invalid Input. Please Enter Integer Only.")

if largest != float('-inf'):
    print(f"The largest among all is: {largest}")
else:
    print("No numbers were entered.")
```
