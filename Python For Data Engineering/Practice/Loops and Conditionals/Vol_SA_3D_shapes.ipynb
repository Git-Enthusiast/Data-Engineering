# VOLUME AND SURFACE AREA OF COMMON 3D SHAPES

```python

import math

# ==============================
# 1️⃣ Volume of Cone
# Formula: (1/3) * π * r^2 * h
# r = radius, h = height
r_cone = float(input("Enter radius of cone: "))
h_cone = float(input("Enter height of cone: "))
volume_cone = (1/3) * math.pi * r_cone**2 * h_cone
print("Volume of Cone =", volume_cone)
```

```python
import math

# ==============================
# 2️⃣ Volume of Prism
# Formula: base area * height
# For simplicity, rectangular prism: base = l * w
l_prism = float(input("Enter length of prism base: "))
w_prism = float(input("Enter width of prism base: "))
h_prism = float(input("Enter height of prism: "))
volume_prism = l_prism * w_prism * h_prism
print("Volume of Prism =", volume_prism)
```

```python
import math

# ==============================
# 3️⃣ Volume of Cylinder
# Formula: π * r^2 * h
r_cylinder = float(input("Enter radius of cylinder: "))
h_cylinder = float(input("Enter height of cylinder: "))
volume_cylinder = math.pi * r_cylinder**2 * h_cylinder
print("Volume of Cylinder =", volume_cylinder)
```

```python
import math

# ==============================
# 4️⃣ Volume of Sphere
# Formula: (4/3) * π * r^3
r_sphere = float(input("Enter radius of sphere: "))
volume_sphere = (4/3) * math.pi * r_sphere**3
print("Volume of Sphere =", volume_sphere)
```

```python
import math

# ==============================
# 5️⃣ Volume of Pyramid
# Formula: (1/3) * base area * height
# Assume square base: side = a
a_pyramid = float(input("Enter side of square base of pyramid: "))
h_pyramid = float(input("Enter height of pyramid: "))
volume_pyramid = (1/3) * (a_pyramid**2) * h_pyramid
print("Volume of Pyramid =", volume_pyramid)
```

```python
import math

# ==============================
# 6️⃣ Curved Surface Area of Cylinder
# Formula: 2 * π * r * h
csa_cylinder = 2 * math.pi * r_cylinder * h_cylinder
print("Curved Surface Area of Cylinder =", csa_cylinder)
```

```python
import math

# ==============================
# 7️⃣ Total Surface Area of Cube
# Formula: 6 * side^2
side_cube = float(input("Enter side of cube: "))
tsa_cube = 6 * side_cube**2
print("Total Surface Area of Cube =", tsa_cube)
```
