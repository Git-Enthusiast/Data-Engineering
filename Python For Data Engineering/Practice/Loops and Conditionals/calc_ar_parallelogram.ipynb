```python
# ==========================================
# AREA OF PARALLELOGRAM
# ==========================================

# A parallelogram is a quadrilateral where:
# - Opposite sides are parallel and equal in length.
# - Opposite angles are equal.

# Let:
# base = length of the base side
# height = perpendicular distance from base to opposite side

# Formula for area:
# Area = base * height

# ==========================================
# Python Program
# ==========================================

# Input from user
base = float(input("Enter base of parallelogram: "))
height = float(input("Enter height of parallelogram: "))

# Calculate area
area = base * height

# Output
print("Area of Parallelogram =", area)
```

```python
import math

# Let a and b be adjacent sides, and theta the angle between them in degrees
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
theta = float(input("Enter angle between sides in degrees: "))

# Convert angle to radians
theta_rad = math.radians(theta)

# Area = a * b * sin(theta)
area = a * b * math.sin(theta_rad)

print("Area of Parallelogram =", area)
```
