```python
"""
Time Complexity in DSA (Python)
================================

This file is a detailed beginner-friendly theory + examples note on time complexity.
It is written as a Python script so you can read it like notes, and also run it to
see example outputs.


------------------------------------------------------------
1) What is Time Complexity?
------------------------------------------------------------

Time complexity tells us how the running time of an algorithm grows as the input size grows.

Important idea:
We do NOT usually measure exact seconds.
Instead, we study how the number of operations increases when n increases.

Example:
- If input size doubles, does the code take about twice the time?
- Or four times the time?
- Or only a little more?

That growth pattern is what complexity describes.

Why this matters:
- A solution may work for n = 100, but fail for n = 1,000,000.
- In interviews and real systems, choosing an efficient approach is crucial.

------------------------------------------------------------
2) Big-O Notation
------------------------------------------------------------

Big-O notation gives an upper bound on how an algorithm grows.
It describes the worst-case growth trend.

Common Big-O classes:
- O(1)      constant time
- O(log n)  logarithmic time
- O(n)      linear time
- O(n log n) linearithmic time
- O(n^2)    quadratic time
- O(2^n)    exponential time
- O(n!)     factorial time

The smaller the growth, the better it usually is for large inputs.

------------------------------------------------------------
3) Constants are ignored
------------------------------------------------------------

Example:
- 3n + 10  is still O(n)
- 1000n + 500000 is still O(n)
- n/2 is also O(n)

Why?
Because Big-O focuses on the dominant growth pattern, not small details.

For large n:
- n dominates constants
- n^2 dominates n
- n dominates log n

------------------------------------------------------------
4) Best, Average, and Worst Case
------------------------------------------------------------

Some algorithms have different performance depending on input.

Example: linear search
- Best case: O(1)   -> item found at first position
- Average case: O(n)
- Worst case: O(n)  -> item is last or not present

Example: binary search
- Best case: O(1)   -> middle element is target
- Average case: O(log n)
- Worst case: O(log n)

In interviews, worst case is usually the most important unless asked otherwise.

------------------------------------------------------------
5) Rules to Estimate Time Complexity
------------------------------------------------------------

Rule 1: Sequential statements add up
-------------------------------------

If two blocks run one after another:

    for i in range(n):
        ...
    for j in range(n):
        ...

This becomes O(n) + O(n) = O(n)
Because constants and same-order terms are simplified.

Rule 2: Nested loops multiply
-----------------------------

    for i in range(n):
        for j in range(n):
            ...

This becomes O(n * n) = O(n^2)

Rule 3: Ignore constant factors
--------------------------------

    for i in range(2 * n):
        ...

Still O(n)

Rule 4: Reduce recursive problems carefully
-------------------------------------------

Recursion often needs recurrence relations.
Examples:
- T(n) = T(n-1) + O(1) -> O(n)
- T(n) = 2T(n/2) + O(n) -> O(n log n)

Rule 5: Look at the dominant term only
---------------------------------------

    O(n^2 + n + 50) -> O(n^2)
    O(n log n + n)   -> O(n log n)

------------------------------------------------------------
6) Common Time Complexities with Intuition
------------------------------------------------------------

O(1) Constant time
-------------------
The time does not depend on n.
Example: accessing an array element by index.

O(log n) Logarithmic time
-------------------------
The input gets smaller very quickly.
Example: binary search.
Each step cuts the problem roughly in half.

O(n) Linear time
-----------------
The work grows directly with n.
Example: scanning a list once.

O(n log n) Linearithmic time
----------------------------
Common in efficient sorting algorithms.
Example: merge sort, heap sort.

O(n^2) Quadratic time
---------------------
Usually happens with nested loops over the same input.
Example: bubble sort, comparing every pair.

O(2^n) Exponential time
----------------------
Work doubles with each new element.
Often appears in brute-force recursion.
Example: generating all subsets.

O(n!) Factorial time
-------------------
Extremely slow for large n.
Example: generating all permutations.

------------------------------------------------------------
7) Example: O(1)
------------------------------------------------------------

Accessing one element from a list by index is constant time.

    arr = [10, 20, 30, 40]
    x = arr[2]   # always one direct access

No matter whether the list has 4 elements or 4 million elements,
getting arr[2] is still O(1).

Why?
Because Python can compute the memory location directly.

------------------------------------------------------------
8) Example: O(n)
------------------------------------------------------------

If we check every element once, complexity is linear.

    def print_all(items):
        for item in items:
            print(item)

If items has n elements, the loop runs n times.
So the time complexity is O(n).

Another example:

    def count_even(nums):
        count = 0
        for num in nums:
            if num % 2 == 0:
                count += 1
        return count

Again, one pass through the list => O(n).

------------------------------------------------------------
9) Example: O(n^2)
------------------------------------------------------------

Nested loops over the same list often create quadratic time.

    def print_pairs(nums):
        for i in range(len(nums)):
            for j in range(len(nums)):
                print(nums[i], nums[j])

If the outer loop runs n times and the inner loop also runs n times,
total operations are roughly n * n = n^2.
So this is O(n^2).

Important note:
If the inner loop runs fewer times each iteration, the complexity may differ.
You must inspect the exact code.

------------------------------------------------------------
10) Example: O(log n)
------------------------------------------------------------

Binary search is the classic logarithmic algorithm.
It works only on sorted data.

Idea:
- Compare target with middle element
- If target is smaller, search the left half
- If target is larger, search the right half
- Repeat until found or range becomes empty

Each step halves the search space.
That means:

n -> n/2 -> n/4 -> n/8 -> ... -> 1

The number of steps is about log2(n).
So binary search is O(log n).

------------------------------------------------------------
11) Example: O(n log n)
------------------------------------------------------------

Merge sort is a classic O(n log n) algorithm.

Why?
- The array is divided into halves repeatedly -> log n levels
- At each level, all n elements are processed during merging

So total work:
- n work per level
- log n levels
- total = n log n

This pattern appears in several efficient algorithms.

------------------------------------------------------------
12) Example: O(2^n)
------------------------------------------------------------

Generating all subsets of a set is exponential.

For each element, there are 2 choices:
- include it
- exclude it

If there are n elements, total subsets = 2^n.

That is why subset generation grows very fast.
For n = 10, subsets = 1024
For n = 20, subsets = 1,048,576

------------------------------------------------------------
13) Example: O(n!)
------------------------------------------------------------

Generating all permutations is factorial time.

For n items:
- first position: n choices
- second position: n-1 choices
- third position: n-2 choices
- ...

Total permutations = n!

This becomes huge very quickly.
For example:
- 5! = 120
- 10! = 3,628,800

Factorial time is usually too slow for large inputs.

------------------------------------------------------------
14) Python examples with explanations
------------------------------------------------------------
"""

from __future__ import annotations


# ------------------------------------------------------------
# O(1) example
# ------------------------------------------------------------
def get_first_element(arr: list[int]) -> int | None:
    """
    Return the first element of the list.

    Time complexity: O(1)
    Reason:
    - We access one fixed position directly.
    - The input size does not change the number of operations.
    """
    if not arr:
        return None
    return arr[0]


# ------------------------------------------------------------
# O(n) example
# ------------------------------------------------------------
def find_max(nums: list[int]) -> int | None:
    """
    Return the maximum value in a list.

    Time complexity: O(n)
    Reason:
    - We scan every element once.
    - Each element may need a comparison.
    """
    if not nums:
        return None

    maximum = nums[0]
    for num in nums[1:]:
        if num > maximum:
            maximum = num
    return maximum


# ------------------------------------------------------------
# O(n^2) example
# ------------------------------------------------------------
def print_all_pairs(nums: list[int]) -> list[tuple[int, int]]:
    """
    Return all ordered pairs from the list.

    Time complexity: O(n^2)
    Reason:
    - The outer loop runs n times.
    - The inner loop runs n times for each outer iteration.
    - Total work = n * n = n^2.
    """
    pairs = []
    for i in range(len(nums)):
        for j in range(len(nums)):
            pairs.append((nums[i], nums[j]))
    return pairs


# ------------------------------------------------------------
# O(log n) example
# ------------------------------------------------------------
def binary_search(arr: list[int], target: int) -> int:
    """
    Perform binary search on a sorted list.

    Returns the index of target if found, else -1.

    Time complexity: O(log n)
    Reason:
    - Each comparison cuts the search space in half.
    - We do not scan all elements.
    """
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


# ------------------------------------------------------------
# O(n log n) example explanation function
# ------------------------------------------------------------
def explain_merge_sort_complexity() -> str:
    """
    Explain why merge sort is O(n log n).

    This is a theory helper, not a sorting implementation.
    """
    return (
        "Merge sort has O(n log n) complexity because:\n"
        "1) The array is split into halves repeatedly -> about log n levels.\n"
        "2) At each level, all n elements are merged once -> O(n) work.\n"
        "3) Total = O(n) * O(log n) = O(n log n)."
    )


# ------------------------------------------------------------
# Exponential example: generate subsets
# ------------------------------------------------------------
def generate_subsets(nums: list[int]) -> list[list[int]]:
    """
    Generate all subsets of nums.

    Time complexity: O(2^n)
    Reason:
    - For each element, we have two choices:
      include or exclude.
    - Total subsets = 2^n.
    """
    subsets: list[list[int]] = [[]]

    for num in nums:
        new_subsets = []
        for subset in subsets:
            new_subsets.append(subset + [num])  
                                                                                                                                   
        subsets.extend(new_subsets)
    return subsets


# ------------------------------------------------------------
# Factorial example: generate permutations
# ------------------------------------------------------------
def generate_permutations(nums: list[int]) -> list[list[int]]:
    """
    Generate all permutations using recursion.

    Time complexity: O(n!)
    Reason:
    - The number of possible arrangements is n!.
    - Each additional element increases the number of possibilities sharply.
    """
    if len(nums) <= 1:
        return [nums[:]]

    result = []
    for i in range(len(nums)):
        current = nums[i]
        remaining = nums[:i] + nums[i + 1 :]

        for perm in generate_permutations(remaining):
            result.append([current] + perm)

    return result


# ------------------------------------------------------------
# Recursion example: simple linear recursion
# ------------------------------------------------------------
def sum_natural_numbers(n: int) -> int:
    """
    Return sum of first n natural numbers using recursion.

    Time complexity: O(n)
    Reason:
    - The function calls itself n times.
    - Each call does O(1) work.
    - Total = O(n)
    """
    if n == 1:
        return 1
    return n + sum_natural_numbers(n - 1)


# ------------------------------------------------------------
# Utility: compare growth intuition
# ------------------------------------------------------------
def show_complexity_table() -> None:
    """Print a simple summary table of common complexities."""
    print("\nCOMMON TIME COMPLEXITIES")
    print("-" * 60)
    print(f"{'Complexity':<12} | {'Example intuition'}")
    print("-" * 60)
    print(f"{'O(1)':<12} | Direct access, one operation")
    print(f"{'O(log n)':<12} | Halving the search space")
    print(f"{'O(n)':<12} | One full scan")
    print(f"{'O(n log n)':<12} | Efficient divide-and-conquer sorting")
    print(f"{'O(n^2)':<12} | Nested loops over same input")
    print(f"{'O(2^n)':<12} | All subsets / branching recursion")
    print(f"{'O(n!)':<12} | All permutations")
    print("-" * 60)


# ------------------------------------------------------------
# Practical guidance for reading code complexity
# ------------------------------------------------------------
def how_to_analyze_code() -> str:
    """
    Return a compact step-by-step method to analyze time complexity.
    """
    return (
        "How to analyze time complexity:\n"
        "1) Count loops.\n"
        "2) Check if loops are nested or sequential.\n"
        "3) See whether the loop size depends on n, n/2, n^2, etc.\n"
        "4) Look at recursion carefully.\n"
        "5) Keep the dominant term only.\n"
        "6) Ignore constants and low-order terms."
    )


# ------------------------------------------------------------
# Main demo
# ------------------------------------------------------------
def main() -> None:
    print("INTRODUCTION TO TIME COMPLEXITY IN DSA (PYTHON)")
    print("=" * 60)

    show_complexity_table()

    print("\nTHEORY SUMMARY")
    print("-" * 60)
    print(how_to_analyze_code())

    print("\nEXAMPLES")
    print("-" * 60)

    arr = [10, 20, 30, 40, 50]
    print("Array:", arr)
    print("First element (O(1)):", get_first_element(arr))
    print("Maximum (O(n)):", find_max(arr))
    print("Binary search for 30 (O(log n)):", binary_search(arr, 30))

    nums_small = [1, 2, 3]
    pairs = print_all_pairs(nums_small)
    print("All pairs from [1, 2, 3] (O(n^2)):", pairs)

    print("\nMerge sort complexity explanation:")
    print(explain_merge_sort_complexity())

    print("\nSubsets of [1, 2, 3] (O(2^n)):")
    print(generate_subsets(nums_small))

    print("\nPermutations of [1, 2, 3] (O(n!)):")
    print(generate_permutations(nums_small))

    print("\nRecursion example sum_natural_numbers(5) (O(n)):")
    print(sum_natural_numbers(5))

    print("\nIMPORTANT REMINDER")
    print("-" * 60)
    print("Always analyze the actual code, not just the presence of loops.")
    print("Some nested loops are not O(n^2) if the inner loop shrinks.")
    print("Some recursive code is not slow if memoization is used.")


if __name__ == "__main__":
    main()
```
